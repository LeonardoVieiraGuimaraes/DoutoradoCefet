## Algoritmo do Mountain Clustering - Passo 3
- Seleciona os centros dos grupos sequencialmente. O primeiro centro do grupo $c_1$ é determinado pela seleção do ponto com maior densidade.
- Para obter o centro do grupo seguinte, elimina-se o efeito do primeiro grupo revisando a função de montanha: uma nova função é formada subtraindo uma Gaussiana centrada em $c_1$:

- Após a subtração, o centro do segundo grupo é selecionado como o ponto de maior valor da nova função de montanha.
- O processo continua até encontrar o número suficiente de grupos.

## Agrupamento Subtrativo
- Proposto como algoritmo de agrupamento, não apenas inicialização.
- Similar ao Mountain Clustering, mas usa dados como centros candidatos ao invés de vértices do grid.
- Mountain Clustering tem custo computacional exponencial com a dimensão do problema, pois avalia a função montanha para cada ponto do grid.
- Agrupamento subtrativo resolve isso usando pontos de dados como candidatos, tornando o custo proporcional ao tamanho do problema.

## Algoritmo do Agrupamento Subtrativo
1. Calcule a medida de densidade no ponto $x_i$: $D_i = um_{j=1}^n xpeft(-rac{x_i - x_j^2}{(r_a/2)^2}
ight)$, onde $r_a$ é o raio de vizinhança.
2. Um ponto terá alta densidade se tiver muitos vizinhos.
3. O centro do primeiro grupo $x_{c_1}$ é o ponto com maior densidade $D_{c_1}$.
4. Revise a densidade de cada ponto: $D_i = D_i - D_{c_1} xpeft(-rac{x_i - x_{c_1}^2}{(r_b/2)^2}
ight)$, onde $r_b$ define a vizinhança para redução da densidade.
5. Pontos próximos ao centro do grupo terão densidade reduzida.
6. Após revisar, selecione o próximo centro como o ponto de maior densidade.
7. Repita até obter o número desejado de grupos.
