## Agrupamento por Partição Crisp
Dado um conjunto de dados $X = x_1, ... x_i, ..., x_n$, onde $x_i = (x_{i1}, ... x_{i2}, ..., x_{ip}) n athbb{R}^p$, encontrar um conjunto de $c$ grupos, $C = C_1, ..., C_k, ..., C_c$ ($c eq n$) tal que:
- $C_k 
eq mptyset$, $k = 1, ..., c$
- $igcup_{k=1}^c C_k = X$
- $C_k ap C_q = mptyset$, $k, q = 1, ..., c$, $k 
eq q$

## Exemplo - C-Means
- Seleciona-se $c$ pontos aleatórios (e.g. $c = 3$) sobre o espaço (centróides ou protótipos);
- Para cada entidade, computa-se o centróide mais próximo e rotula-se esse objeto como associado a tal centróide;
- Recalcula-se a posição dos centróides com base na posição de seus objetos associados;
- Para recalcular a posição de um centróide $c$, considera-se a distância média de seus objetos relacionados;
- Ao final, obtêm-se as coordenadas dos centróides que particionam o espaço.

## Algoritmo do C-Means
1. Inicialize os centros dos grupos.
2. Determine a matriz de pertinência;
3. Calcule a função de custo (distância).
4. Atualize os centros dos grupos.
5. Repita os passos 2, 3 e 4 até violar alguma condição de parada (número de iterações, nenhuma entidade mudar de grupo, porcentagem de entidades que alteram de grupo ser menor que um determinado limiar, etc.)

## Agrupamento por Partição Fuzzy
- Permite lidar com incertezas que surgem em diversas situações.
- As partições resultantes fazem uma entidade corresponder a um grupo com grau de pertinência no intervalo [0, 1], onde maior valor indica maior pertinência.
- Quanto mais próximo (ou distante) estiver uma entidade de um protótipo de um grupo, maior (ou menor) será o seu grau de pertinência a esse grupo.

## Fuzzy C-Means
Dado um conjunto de $n$ padrões $X = x_1, x_2, ..., x_n$ onde $x_i$ é um padrão $d$-dimensional
Um agrupamento fuzzy é uma coleção de $k$ grupos, $c_1, c_2, ..., c_k$ e uma matriz partição $U = u_{i,j} n [0, 1]$, para $i = 1, ..., n$ e $j = 1, ..., k$ onde cada elemento $u_{i,j}$ representa o grau de pertinência do padrão $i$ no grupo $c_j$.
A soma dos graus de pertinência associados a um padrão deve ser igual a 1: $um_{j=1}^k u_{i,j} = 1$ onde $x_i$ é um padrão $d$-dimensional.
Cada grupo contém pelo menos um ponto com grau de pertinência maior que zero e não contém todos os pontos com grau de pertinência unitário: $0 < um_{i=1}^n u_{i,j} < 1$.
