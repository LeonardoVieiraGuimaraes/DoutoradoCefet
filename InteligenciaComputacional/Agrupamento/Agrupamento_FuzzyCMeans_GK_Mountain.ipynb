## Fuzzy C-Means
O Fuzzy C-Means otimiza a função de custo:
$J = um_{j=1}^k um_{i=1}^n u_{ij}^m c_j - x_i^2$
onde $m n [1, nfty)$ é o expoente de ponderação que determina a influência dos pesos.
O cálculo dos centros é pela média de todos os padrões ponderados pelo grau de pertinência:
$c_j = rac{um_{i=1}^n u_{ij}^m x_i}{um_{i=1}^n u_{ij}^m}$
Atualização da matriz de partição:
$u_{ij} = rac{1}{um_{k=1}^c (rac{d_{ij}}{d_{kj}})^{2/(m-1)}}$
O valor da pertinência é alto se o ponto $i$ está próximo do centro $j$.
Se $x_i = c_j$, então $u_{ij} = 1$ e $u_{il} = 0$ para $l 
eq j$.

## Algoritmo do Fuzzy C-Means
1. Inicialize a matriz de pertinência $U$ com valores aleatórios entre 0 e 1, de forma a satisfazer $um_{i=1}^c u_{ij} = 1$, $orall j = 1, ..., n$;
2. Calcule o centro de cada grupo $c_i$ utilizando a matriz de partição;
3. Calcule a função de custo;
4. Finalize caso o critério de convergência seja atingido;
5. Atualize a matriz de partição $U$. Vá para o passo 2.

## Vantagens e Desvantagens do Fuzzy C-Means
**Vantagens:**
- Algoritmo de cluster fuzzy mais utilizado;
- Implementação simples;
- Computacionalmente rápido e eficiente;
- Convergência mais rápida que o C-Means.
**Desvantagens:**
- Número de clusters é parâmetro de entrada e afeta os resultados;
- Resultados dependem da posição dos centros;
- Má inicialização pode levar a mínimos locais;
- Recomenda-se executar o algoritmo várias vezes com diferentes centros iniciais.

## Gustafson-Kessel
Extensão do Fuzzy C-Means que utiliza medida de distância adaptativa, permitindo grupos de formas geométricas diferentes.
Função objetivo (distância de Mahalanobis):
$d_{A_i}^2 = [x_k - v_i]^T A_i [x_k - v_i]$
Matriz de covariância:
$A_i = p_i et(F_i)^{1/m} F_i^{-1}$, onde $F_i = rac{um_{k=1}^n u_{ik}^q (x_k - v_i)(x_k - v_i)^T}{um_{k=1}^n u_{ik}^q}$

## Algoritmo do Gustafson-Kessel
1. Calcule os centros: $v_i = rac{um_{k=1}^n u_{ik}^q x_k}{um_{k=1}^n u_{ik}^q}$;
2. Calcule as matrizes de covariância: $F_i = rac{um_{k=1}^n u_{ik}^q (x_k - v_i)(x_k - v_i)^T}{um_{k=1}^n u_{ik}^q}$;
3. Calcule as distâncias: $d_{ik} = [x_k - v_i]^T p_i et(F_i)^{1/m} F_i^{-1} [x_k - v_i]$;
4. Atualize a matriz de partição: $u_{ik} = rac{1}{um_{j=1}^c (rac{d_{ik}}{d_{jk}})^{1/(q-1)}}$;
5. Finalize caso o critério de convergência seja atingido, caso contrário vá para o passo 2.

## Mountain Clustering
Mountain Clustering (Yager, 1994) encontra centros de grupos com base em medida de densidade (Mountain Function).
Particiona o hipervolume dos dados por um grid, cujas interseções são centros candidatos.
O grid pode ser igualmente espaçado ou não; grid mais fino implica mais centros candidatos.

## Algoritmo do Mountain Clustering
1. Cria um grid no espaço de dados, onde as interseções constituem possíveis centros dos grupos, denotados por $V$;
2. Constrói a Mountain Function que representa a densidade de dados. A altura da função em $v n V$ é:

A medida de densidade é inversamente proporcional à distância entre os pontos $x_i$ e o ponto $v$;
A constante $igma$ determina a altura e suavidade da função de montanha.
