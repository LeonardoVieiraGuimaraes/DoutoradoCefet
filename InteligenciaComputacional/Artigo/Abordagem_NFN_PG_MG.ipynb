Sistemas Neuro-Fuzzy possuem o tratamento de incerteza e a interpretabilidade dos Sistemas Fuzzy e habilidade de aprendizado das Redes Neurais.
Existe um crescente interesse no desenvolvimento de novos algoritmos de aprendizado que possibilitem a construção autônoma do conjunto de regras de Sistemas.
Técnicas de Computação Evolucionária destacam-se pela sua capacidade de adaptação e aprendizado baseado em busca global.

### Estrutura da Rede NFN
Programação Genética Multi-Gene (PG-MG) para criação da base regras em redes Neuro-Fuzzy do tipo Neo-Fuzzy-Neuron (NFN)
Cada indivíduo da PG-MG representa uma base de regras.
As regras são representadas por funções de pertinência Gaussianas criadas e/ou modificadas pela PG-MG.
O ajuste dos parâmetros do consequente, é realizado por um método baseado no Gradiente Descendente.

### NFN-PG-MG - Fluxograma
Problemas de previsão e identificação de sistemas.
Comparação com modelos alternativos: RNA MLP, ANFIS, NFN.
Divisão do conjunto de dados: treinamento 60%; validação 20%; teste 20%.
Cada experimento foi executado 10 vezes e o melhor resultado foi utilizado para comparação.
O desempenho dos sistemas foi avaliado pelo RMSE (Root Mean Square Error):
$$\text{RMSE} = \sqrt{\frac{1}{k} \sum_{t=1}^{k} (y_t - \hat{y}_t)^2}$$

### Previsão de Temperatura
Previsão de temperatura em três regiões geográficas com padrões climáticos distintos.
Considera-se os dados referentes à temperatura média mensal em Ottawa, Vale da Morte e Lisboa.
O objetivo é prever a temperatura média mensal um passo à frente, utilizando os cinco primeiros valores defasados da série como entrada:
$$y_t = f(y_{t-1}, y_{t-2}, y_{t-3}, y_{t-4}, y_{t-5})$$
Dados normalizados entre 0 e 1.

### Previsão da Série Temporal de Mackey-Glass
A série temporal clássica de Mackey-Glass é gerada pela seguinte equação diferencial com atraso:
$$\frac{dx(t)}{dt} = \frac{0.2x(t-\tau)}{1 + x(t-\tau)^{10}} - 0.1x(t)$$
na qual $x(0) = 1.2$ e $\tau = 17$.
O objetivo é prever o valor $x_{t+6}$ para qualquer valor de $t$, baseado nas entradas $[x_{t-18}, x_{t-12}, x_{t-6}, x_t]$.
O conjunto de dados é composto de 3500 amostras, com 4 variáveis de entrada e 1 de saída, das quais 3000 foram coletadas com $t \in [201, 3200]$ e 500 com $t \in [5001, 5500]$.

- **NFN-PG-MG:** 0,05548425
- **NFN:** 0,06019720
- **MLP (3 camadas):** 0,11302678
- **MLP (2 camadas):** 0,19026832

### Considerações Finais
Este trabalho introduziu um novo algoritmo baseado em Programação Genética Multi-Gene para construção da base de regras em redes Neo-Fuzzy-Neuron, com ajuste de pesos por um método baseado no Gradiente.
Experimentos computacionais em problemas de previsão sugerem que a abordagem seja promissora.
O NFN-PG-MG se demonstrou eficiente quando comparado a modelos alternativos, sendo sempre superior a pelo menos dois dos três modelos utilizados para comparação.
Em relação ao NFN tradicional, os resultados, quando não superiores, são bastante próximos e competitivos.
A Programação Genética se demonstrou uma boa técnica para modelagem de redes NFN para problemas de previsão e identificação de sistemas.
Além da acurácia, destaca-se na abordagem proposta a construção autônoma e otimizada da base de regras de acordo com o conjunto de dados.
