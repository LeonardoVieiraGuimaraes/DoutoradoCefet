# Inteligência Computacional
## Redes Neurais Artificiais

### Neurônios Biológicos
O cérebro humano consiste de aproximadamente $10^{11}$ elementos de processamento chamados neurônios.
Eles se comunicam por meio de uma rede de conexões formada por axônios e sinapses (cada neurônio possui em média $10^4$ sinapses).
A comunicação ocorre essencialmente por impulsos elétricos e reações químicas.

O sistema nervoso humano é um sistema de controle em três estágios: receptores, redes neurais de processamento e transmissores.

Potencial de ação: um neurônio ativa sua saída se a soma das contribuições excitatórias e inibitórias ultrapassar um limiar de disparo.

### Conexionismo
O desenvolvimento de redes neurais se apoia no paradigma do conexionismo para produzir máquinas inteligentes.
Máquinas inteligentes são criadas a partir da reprodução e interconexão dos elementos de processamento do cérebro dos animais.

Uma rede neural é um processador massivamente paralelo distribuído, constituído de unidades de processamento simples, com propensão natural para armazenar conhecimento experimental e torná-lo disponível para uso.

Ela se assemelha ao cérebro em dois aspectos:
- O conhecimento é adquirido pela rede a partir do ambiente, via aprendizado.
- Forças de conexão entre neurônios (pesos sinápticos) armazenam o conhecimento adquirido.

(Haykin, 2001.)

### Modelos Matemáticos de Neurônios
Um neurônio artificial é representado por um modelo matemático simplificado do processamento biológico.
O modelo MCP (McCulloch e Pitts, 1943) é a primeira definição formal e ainda é usado em muitas aplicações.

As entradas $x_i$ são binárias ([0, 1] ou [-1, 1]), e os valores $w_i$ são pesos multiplicativos conectando a entrada ao núcleo do neurônio.

A regra de disparo $y(k)$ é:
$$
y(k) = Begin{cases}
1, & 	ext{se } um_{i=1}^n w_i x_i(k) eq D \
0, & 	ext{se } um_{i=1}^n w_i x_i(k) < D
nd{cases}
$$
Onde $D$ é o limiar de disparo.
#### Hiperplano de Decisão do Perceptron
Em um perceptron elementar, ou seja, uma rede formada por um único neurônio, o espaço de entrada $n$-dimensional é dividido por um hiperplano em duas regiões de decisão.
Este hiperplano é definido pela equação:
$$
um_{i=1}^n x_i w_i - 	heta = 0
$$
Onde $x_i$ são as entradas, $w_i$ os pesos e $	heta$ o limiar.

Para o caso particular de duas entradas, $x_1$ e $x_2$, a fronteira de decisão é uma linha reta:
$$
w_1 x_1 + w_2 x_2 - 	heta = 0
$$
A região acima da linha corresponde a uma classe, e abaixo da linha corresponde à outra classe.

O modelo MCP é importante historicamente, mas faz simplificações que limitam sua aplicação.
Redes neurais artificiais empregam modelos de neurônios mais complexos, mas baseados na mesma ideia:
- Múltiplas entradas com conexões sinápticas (pesos numéricos)
- Núcleo do neurônio executando transformação nas entradas ponderadas
- Saída única

O processamento geral:
$$
y(k) = f(w^T x - D) = feft(um_{i=1}^n w_i x_i - D
ight) = f(net)
$$
Onde $w = [w_1, w_2, ..., w_n]^T$ e $x = [x_1, x_2, ..., x_n]^T$ (vetores coluna).

A função $f(.)$ é chamada de função de ativação do neurônio.
O argumento da função de ativação, $net$, é o produto escalar entre pesos e entradas, análogo ao potencial de membrana nos neurônios biológicos.
O limiar de ativação $D$ pode ser embutido em $net$ por uma entrada fixa unitária e peso $w_0$.

- **Função linear:** Usada em camadas de saída de regressão.
- **Função sigmoide:** Saída entre 0 e 1, comum em classificação binária.
- **Função logística:** Semelhante à sigmoide, mais achatada em torno de zero.
- **Tangente hiperbólica:** Saída entre -1 e 1, média centrada em zero.
- **Softmax:** Usada na saída para classificação multiclasse, gera distribuição de probabilidade.

- **ReLU (Rectified Linear Unit):** Saída igual à entrada se positiva, zero se negativa.
- **Leaky ReLU:** Permite saída negativa com slope menor.
- **PReLU:** Slope negativo aprendido durante o treinamento.
- **ELU:** Inclinação exponencial para valores negativos.
- **SELU:** Autorregularização para redes profundas.
- **GELU:** Baseada em distribuição gaussiana, usada em redes transformadoras.

### Arquiteturas de Redes Neurais
Uma RNA é uma topologia interconectada de neurônios artificiais, com neurônios de entrada, internos e de saída.
A organização e conexão dos neurônios depende da arquitetura da rede.

#### Redes de Camada Simples
Camada de entrada (nós de distribuição) conectada a uma camada de saída (neurônios de processamento).
Fluxo unidirecional dos dados.

Vetores de entrada e saída: $x = [x_1, ..., x_m]^T$, $y = [y_1, ..., y_n]^T$
Pesos $w_{ij}$ conectam o neurônio de destino $i$ à entrada $j$.

Regra de ativação para o neurônio $i$: $net_i = um_{j=1}^m w_{ij} x_j$
Saída: $y_i = f(w_i^T x)$

#### Redes de Múltiplas Camadas
Possuem uma ou mais camadas internas (hidden layers).
Fluxo unidirecional de sinal.

Processamento em $k$ camadas:
$$
h_1 = amma_1[W_1 x]
h_2 = amma_2[W_2 h_1]
...
y = amma_k[W_k h_{k-1}]
$$

#### Redes Recorrentes
Obtidas pela realimentação dos neurônios de saída aos de entrada.
Tempo de atraso introduzido pelo operador de atraso $elta$.

Mapeamento:
$$
y_{t+elta} = amma[W y_t]
$$
A entrada $x$ é necessária apenas para inicialização: $y(0) = x(0)$.

### Processamento Neural
O cálculo das saídas para um vetor de entrada $x$ é chamado de Recall (recordação).
O processamento neural pode executar diferentes tarefas:

#### Memórias Associativas
Padrões podem ser armazenados na RNA.
Auto-associação: rede associa padrão de entrada a padrão similar armazenado.
Hetero-associação: associações entre pares de padrões.

#### Classificação
Conjunto de padrões de entrada dividido em classes/categorias.
Processamento neural responde pela pertinência do padrão a uma das categorias aprendidas.
Classes normalmente codificadas por vetores de dados, saídas binárias.

#### Aproximação de Funções
RNA possui capacidade de aproximar funções e mapeamentos arbitrários.
Capacidade de generalização: interpolação de padrões de saída a partir de padrões de entrada desconhecidos.

### Aprendizado
Processo de ajuste dos parâmetros livres da rede por estímulo do ambiente externo.
Tipos de aprendizado:
- Supervisionado
- Não supervisionado
- Semi-supervisionado
- Por reforço

#### Aprendizado Supervisionado
Regra de ajuste:
$$
w_i(p+1) = w_i(p) + lpha e(t) x_i(t)
$$
Onde $w_i(p)$ é o peso da entrada $i$, $lpha$ é a taxa de aprendizado, $e(t)$ é a medida de erro e $x_i(t)$ a entrada $i$.

Superfície de erro:
- Linear: redes com unidades lineares, superfície quadrática, único mínimo.
- Não-linear: redes com unidades não lineares, superfície irregular, vários mínimos locais e mínimo global.

### Perceptron
Proposto por Franklin Rosenblatt em 1958, primeiro modelo neural com capacidade de aprendizado.
Máquina adaptativa capaz de classificar padrões de entrada pela modificação das conexões.
Baseado no neurônio MCP, com combinador linear de entrada seguido de função de ativação de limiar estrito.

Classifica padrões de entrada em duas classes possíveis, $C_1$ e $C_2$.

No perceptron elementar (um único neurônio), o espaço de entrada $n$-dimensional é dividido por um hiperplano:
$$
um_{i=1}^n x_i w_i - 	heta = 0
$$
Para duas entradas, a fronteira de decisão é uma linha reta.
