# A Fast Clustering Algorithm for Evolving Fuzzy Classifier Based on Samples Mean
## eFCM

### Initialization
O eFCM inicia com dois clusters, sendo o primeiro centrado na primeira amostra e o segundo inicializado aleatoriamente. Esta etapa é realizada apenas para a primeira amostra.

### Membership Degree
Para cada amostra recebida pelo classificador, é computado seu grau de pertinência para cada cluster existente.

### Classification
Encontra o cluster mais ativo e classifica a amostra de acordo com a classe desse cluster.

### Supervision
Verifica se a amostra foi classificada corretamente e decide se um novo cluster deve ser criado ou se um cluster existente pode ser atualizado.

### Creation of Cluster
Um novo cluster e, consequentemente, uma nova classe é criada sempre que a classe da amostra atual não corresponde a nenhuma das classes existentes.

### Cluster’s Center Update
Se a amostra atual for classificada corretamente, o centro do cluster da classe da amostra é atualizado. Caso contrário, se a amostra for classificada incorretamente, atualiza-se o cluster ao qual a amostra pertence e não aquele em que foi classificada. Os centros são atualizados com base no valor médio das amostras.
