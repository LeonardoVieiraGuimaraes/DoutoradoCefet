## Características e Contribuições do eFCM
- Não necessita de treinamento prévio, o conhecimento é iniciado a partir da primeira amostra processada;
- Os dados podem ser processados conforme disponíveis ou normalizados, sem perda de capacidade preditiva;
- A classificação é baseada no grau de pertinência de cada amostra com os clusters existentes;
- Cada cluster é formado por conjuntos de amostras similares e está associado a uma única classe, formando um cluster para cada classe de saída apresentada ao modelo;
- A atualização do centro do cluster é realizada com base na média dos valores das amostras;
- O eFMC é livre de parâmetros, totalmente autônomo: não há necessidade de seleção de parâmetros pelo usuário;
- Outro fator importante é o baixo custo computacional do eFMC, que realiza apenas operações matemáticas básicas como soma, subtração, multiplicação, divisão e comparação.
