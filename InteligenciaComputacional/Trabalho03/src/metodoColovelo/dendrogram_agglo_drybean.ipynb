# Dendrograma para o Dataset Dry Bean (Agglomerative)

Este notebook permite analisar o número ideal de clusters para o dataset Dry Bean utilizando o dendrograma do Agglomerative Clustering.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from scipy.cluster.hierarchy import dendrogram, linkage

# Carregamento e pré-processamento do dataset Dry Bean
bean_df = pd.read_excel('data/DryBeanDataset/Dry_Bean_Dataset.xlsx')
bean_df = bean_df.dropna()
for col in bean_df.select_dtypes(include='object').columns:
    bean_df[col] = LabelEncoder().fit_transform(bean_df[col].astype(str))
X = bean_df.drop('Class', axis=1).values
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Dendrograma
plt.figure(figsize=(10, 5))
linked = linkage(X, method='ward')
dendrogram(linked, truncate_mode='level', p=10)
plt.title('Dendrograma - Agglomerative (Dry Bean)')
plt.xlabel('Amostras')
plt.ylabel('Distância')
plt.savefig('img/agglo_drybean_dendrogram.png')
plt.show()
```
