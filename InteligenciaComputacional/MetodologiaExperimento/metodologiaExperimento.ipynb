> **Importante:** Avaliar o custo dos diferentes tipos de erro é fundamental para definir a melhor estratégia de classificação.

# Métricas em Classificação Binária: VP, VN, FP, FN

Nos problemas de classificação binária existe uma **classe alvo** (classe positiva), cujo valor se deseja predizer. A outra classe é chamada de **classe negativa**.

Assim, temos:

- **VP (Verdadeiro Positivo):** objeto da classe positiva classificado como positivo.  
  Exemplo: um spam classificado como spam.
- **VN (Verdadeiro Negativo):** objeto da classe negativa classificado como negativo.  
  Exemplo: uma mensagem normal (ham) classificada como ham.
- **FP (Falso Positivo):** objeto da classe negativa classificado como positivo.  
  Exemplo: uma mensagem normal (ham) classificada como spam.
- **FN (Falso Negativo):** objeto da classe positiva classificado como negativo.  
  Exemplo: um spam classificado como ham.

Esses conceitos são fundamentais para calcular métricas como acurácia, precisão, recall e F1-score.

# Matriz de Confusão e Métricas de Avaliação

## Matriz de Confusão

|                 | Classe Positiva (Prevista) | Classe Negativa (Prevista) |
|-----------------|:-------------------------:|:-------------------------:|
| **Classe Positiva (Real)**   | VP (Verdadeiro Positivo)      | FN (Falso Negativo)         |
| **Classe Negativa (Real)**   | FP (Falso Positivo)           | VN (Verdadeiro Negativo)    |

## Principais Métricas

- **Taxa de Verdadeiros Positivos (TVP / Sensibilidade / Recall):** percentual de objetos positivos classificados corretamente.
  
  TVP = VP / (VP + FN)

- **Taxa de Falsos Positivos (TFP):** percentual de objetos negativos classificados como positivos.
  
  TFP = FP / (FP + VN)

- **Acurácia (ACC):** número de classificações corretas dividido pelo número total de classificações.
  
  ACC = (VP + VN) / (VP + FP + VN + FN)

- **Erro (E):** taxa de classificações incorretas.
  
  E = 1 − ACC

# Precisão, Revocação e F1 Score

- **Precisão (Pr):** acurácia das classificações positivas. Mede a capacidade do algoritmo de encontrar somente as amostras relevantes (não atribuir um rótulo positivo a um negativo). Indica a qualidade ou exatidão do algoritmo.
  
  Pr = VP / (VP + FP)

- **Revocação (Re) / Sensibilidade:** capacidade de encontrar todas as amostras positivas. Também conhecida como sensibilidade, mede a completude do algoritmo.
  
  Re = VP / (VP + FN)

- **F1 Score:** média harmônica entre precisão e revocação. Varia de 0 a 1 e equilibra qualidade e completude.
  
  F1 = 2 * (Pr * Re) / (Pr + Re)

Essas métricas são essenciais para avaliar classificadores, especialmente em bases desbalanceadas.

# Curva ROC, TVP, TFP, Especificidade e AUC

A curva ROC (Receiver Operating Characteristic) compara os benefícios de uma classificação correta (Taxa de Verdadeiros Positivos - TVP ou Sensibilidade) com o custo de uma classificação incorreta (Taxa de Falsos Positivos - TFP, também conhecida como 1 - Especificidade).

- **TVP (Sensibilidade):** Proporção de positivos corretamente classificados.
- **TFP:** Proporção de negativos classificados incorretamente como positivos.
- **Especificidade:** Proporção de negativos corretamente classificados (Especificidade = VN / (VN + FP)).
- **AUC (Area Under the ROC Curve):** Mede a área sob a curva ROC. Quanto mais próximo de 1, melhor o desempenho do classificador.

A curva ROC é útil para comparar classificadores e avaliar o desempenho em diferentes limiares de decisão.

```python
# Plotando a curva ROC e calculando a AUC
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Exemplo de dados simulados (substitua pelos seus dados reais)
y_true = np.array([0, 0, 1, 1])  # rótulos reais (0 = negativo, 1 = positivo)
y_scores = np.array([0.1, 0.4, 0.35, 0.8])  # probabilidades ou scores do classificador

# Calcula as taxas para a curva ROC
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

# Especificidade = 1 - FPR
especificidade = 1 - fpr

plt.figure(figsize=(6, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'Curva ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--', label='Aleatório')
plt.xlabel('Taxa de Falsos Positivos (TFP)')
plt.ylabel('Taxa de Verdadeiros Positivos (TVP)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# Para obter 1 - AUC (área acima da curva):
print(f'AUC: {roc_auc:.2f}')
print(f'1 - AUC: {1 - roc_auc:.2f}')
```

# Curva ROC: Conceitos e Interpretação

A curva ROC (Receiver Operating Characteristic) é uma ferramenta gráfica para avaliar o desempenho de classificadores binários. Ela mostra a relação entre:

- **Taxa de Verdadeiros Positivos (TVP ou Sensibilidade):** Proporção de positivos corretamente classificados. TVP = VP / (VP + FN)
- **Taxa de Falsos Positivos (TFP):** Proporção de negativos classificados incorretamente como positivos. TFP = FP / (FP + VN)
- **Especificidade:** Proporção de negativos corretamente classificados. Especificidade = VN / (VN + FP) = 1 - TFP
- **AUC (Área sob a Curva ROC):** Mede a capacidade do classificador em separar as classes. Quanto mais próximo de 1, melhor.

A curva ROC permite comparar classificadores e analisar o desempenho em diferentes limiares de decisão.

```python
# Exemplo prático: plotando a curva ROC e calculando a AUC
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Exemplo de dados simulados (substitua pelos seus dados reais)
y_true = np.array([0, 0, 1, 1])  # rótulos verdadeiros (0 = negativo, 1 = positivo)
y_scores = np.array([0.1, 0.4, 0.35, 0.8])  # probabilidades ou scores do classificador

# Calcula as taxas para a curva ROC
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 6))
plt.plot(fpr, tpr, color='navy', lw=2, label=f'Curva ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--', label='Aleatório')
plt.xlabel('Taxa de Falsos Positivos (TFP)')
plt.ylabel('Taxa de Verdadeiros Positivos (TVP)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

print(f'AUC: {roc_auc:.2f}')
```

O boxplot é uma ferramenta gráfica útil para visualizar a distribuição de uma variável em relação às classes de um problema de classificação binária. Ele permite comparar rapidamente os valores de uma feature para as duas classes (0 e 1), identificando diferenças, outliers e possíveis separações entre as classes.

# Classificação Multiclasse - Matriz de Confusão

A matriz de confusão para problemas multiclasse permite visualizar o desempenho do classificador em cada uma das classes, mostrando quantas amostras de cada classe real foram classificadas em cada classe prevista. É uma ferramenta fundamental para análise detalhada de erros e acertos em problemas com mais de duas classes.

```python
# Exemplo prático: Matriz de Confusão para Classificação Multiclasse
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Gerando dados simulados
np.random.seed(42)
y_true = np.random.choice([0, 1, 2], size=50)  # classes reais
# Simulando previsões com alguns erros
y_pred = y_true.copy()
indices_erro = np.random.choice(range(50), size=10, replace=False)
y_pred[indices_erro] = np.random.choice([0, 1, 2], size=10)

# Calculando a matriz de confusão
cm = confusion_matrix(y_true, y_pred)

# Exibindo a matriz de confusão
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Classe 0", "Classe 1", "Classe 2"])
disp.plot(cmap=plt.cm.Blues)
plt.title('Matriz de Confusão - Multiclasse')
plt.show()
```

# Classificação Multiclasse - Curva ROC

A curva ROC também pode ser utilizada em problemas de classificação multiclasse. Para isso, cada classe é considerada como "positiva" em relação às demais (estratégia one-vs-rest). Assim, é possível visualizar a separação de cada classe em relação às outras e calcular o AUC para cada uma delas.

```python
# Exemplo prático: Curva ROC para Classificação Multiclasse
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc

# Gerando dados simulados multiclasse
X, y = make_classification(n_samples=200, n_classes=3, n_informative=5, n_clusters_per_class=1, n_features=10, random_state=42)

# Binarizando as classes para one-vs-rest
classes = np.unique(y)
y_bin = label_binarize(y, classes=classes)

# Separando treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.3, random_state=42)

# Treinando um classificador
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)
y_score = clf.decision_function(X_test)

# Plotando a curva ROC para cada classe
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(len(classes)):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(7, 6))
for i in range(len(classes)):
    plt.plot(fpr[i], tpr[i], lw=2, label=f'Classe {i} (AUC = {roc_auc[i]:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--', label='Aleatório')
plt.xlabel('Taxa de Falsos Positivos (TFP)')
plt.ylabel('Taxa de Verdadeiros Positivos (TVP)')
plt.title('Curva ROC - Classificação Multiclasse (One-vs-Rest)')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

# Classificação Multiclasse - Curva Precisão-Revocação

A curva precisão-revocação é uma alternativa à curva ROC, especialmente útil em bases desbalanceadas. Para problemas multiclasse, a estratégia one-vs-rest é utilizada, permitindo visualizar a relação entre precisão e revocação para cada classe individualmente.

```python
# Exemplo prático: Curva Precisão-Revocação para Classificação Multiclasse
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_recall_curve, average_precision_score

# Gerando dados simulados multiclasse
X, y = make_classification(n_samples=200, n_classes=3, n_informative=5, n_clusters_per_class=1, n_features=10, random_state=42)

# Binarizando as classes para one-vs-rest
classes = np.unique(y)
y_bin = label_binarize(y, classes=classes)

# Separando treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.3, random_state=42)

# Treinando um classificador
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)
y_score = clf.decision_function(X_test)

# Plotando a curva precisão-revocação para cada classe
plt.figure(figsize=(7, 6))
for i in range(len(classes)):
    precision, recall, _ = precision_recall_curve(y_test[:, i], y_score[:, i])
    ap = average_precision_score(y_test[:, i], y_score[:, i])
    plt.plot(recall, precision, lw=2, label=f'Classe {i} (AP = {ap:.2f})')
plt.xlabel('Revocação')
plt.ylabel('Precisão')
plt.title('Curva Precisão-Revocação - Classificação Multiclasse (One-vs-Rest)')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

# Regressão - MSE e MAE

Em problemas de regressão, duas métricas muito utilizadas para avaliar o desempenho do modelo são:

- **MSE (Mean Squared Error - Erro Quadrático Médio):**
  Mede o valor médio dos quadrados das diferenças entre os valores reais e previstos. Penaliza mais fortemente grandes erros.
  
  Fórmula: $\text{MSE} = \frac{1}{N} \sum_{t=1}^{N} (y_t - \hat{y}_t)^2$
  
  Onde:
  - $N$ = número de amostras
  - $y_t$ = valor real na amostra $t$
  - $\hat{y}_t$ = valor previsto pelo modelo na amostra $t$

- **MAE (Mean Absolute Error - Erro Absoluto Médio):**
  Mede o valor médio das diferenças absolutas entre os valores reais e previstos. Penaliza todos os erros de forma linear.
  
  Fórmula: $\text{MAE} = \frac{1}{N} \sum_{t=1}^{N} |y_t - \hat{y}_t|$

Essas métricas ajudam a comparar modelos de regressão e entender o comportamento dos erros.

# Regressão - RMSE e NDEI

Além do MSE e MAE, outras métricas importantes para avaliar modelos de regressão são:

- **RMSE (Root Mean Squared Error - Raiz do Erro Quadrático Médio):**
  Mede a raiz quadrada da média dos erros ao quadrado, dando mais peso a grandes diferenças entre valores reais e previstos.
  
  Fórmula: $\text{RMSE} = \sqrt{\frac{1}{N} \sum_{t=1}^{N} (y_t - \hat{y}_t)^2}$

- **NDEI (Normalized Root Mean Square Error):**
  Avalia o erro de forma independente da escala dos dados, sendo obtido pela divisão do RMSE pelo desvio padrão ($\sigma$) da saída desejada.
  
  Fórmula: $\text{NDEI} = \frac{\text{RMSE}}{\sigma(y_t)}$

Onde:
- $N$ = número de amostras
- $y_t$ = valor real na amostra $t$
- $\hat{y}_t$ = valor previsto pelo modelo na amostra $t$
- $\sigma(y_t)$ = desvio padrão dos valores reais

Essas métricas são úteis para comparar modelos e avaliar o desempenho considerando a escala dos dados.

# Regressão - Soma dos Quadrados Residuais (SQR)

A Soma dos Quadrados Residuais (SQR) é uma métrica que avalia o ajuste de um modelo de regressão. Ela é calculada somando o quadrado dos resíduos (diferença entre valor real e previsto) para todas as amostras:

Fórmula: $\text{SQR} = \sum_{t=1}^{N} (y_t - \hat{y}_t)^2$

Quanto menor o valor da SQR, melhor o ajuste do modelo aos dados.

# Curva ROC, TVP, TFP, Especificidade e AUC

A curva ROC (Receiver Operating Characteristic) compara os benefícios de uma classificação correta (Taxa de Verdadeiros Positivos - TVP ou Sensibilidade) com o custo de uma classificação incorreta (Taxa de Falsos Positivos - TFP, também conhecida como 1 - Especificidade).

- **TVP (Sensibilidade):** Proporção de positivos corretamente classificados.
- **TFP:** Proporção de negativos classificados incorretamente como positivos.
- **Especificidade:** Proporção de negativos corretamente classificados (Especificidade = VN / (VN + FP)).
- **AUC (Area Under the ROC Curve):** Mede a área sob a curva ROC. Quanto mais próximo de 1, melhor o desempenho do classificador.

A curva ROC é útil para comparar classificadores e avaliar o desempenho em diferentes limiares de decisão.

```python
# Exemplo prático: Soma dos Quadrados Residuais (SQR)
import numpy as np
import matplotlib.pyplot as plt

# Gerando dados simulados
np.random.seed(42)
N = 30
y_true = np.linspace(1, 10, N)
y_pred = y_true + np.random.normal(0, 1, N)  # modelo com ruído

# Calculando os resíduos e a SQR
residuos = y_true - y_pred
sqr = np.sum(residuos**2)
print(f'Soma dos Quadrados Residuais (SQR): {sqr:.2f}')

# Plotando valores reais, previstos e resíduos
plt.figure(figsize=(8, 5))
plt.plot(y_true, label='Valor Real', marker='o')
plt.plot(y_pred, label='Valor Previsto', marker='x')
plt.bar(range(N), residuos, alpha=0.4, label='Resíduo')
plt.xlabel('Índice da Amostra')
plt.ylabel('Valor')
plt.title('Valores Reais, Previsto e Resíduos')
plt.legend()
plt.tight_layout()
plt.show()
```

# Regressão - Teste MGN (Morgan-Granger-Newbold)

O teste MGN (Morgan-Granger-Newbold) é utilizado para comparar o desempenho de dois modelos de previsão, assumindo que os erros (resíduos) dos modelos não são correlacionados. Ele testa a hipótese nula de que as variâncias dos erros de previsão dos dois modelos são iguais.

A estatística do teste é dada por:

$$
MGN = \hat{\rho}_e \sqrt{\frac{N-1}{1 - \hat{\rho}_e^2}}
$$

onde:
- $\hat{\rho}_e$ é a correlação amostral entre $l_1 = e_1 + e_2$ e $l_2 = e_1 - e_2$, sendo $e_1$ e $e_2$ os resíduos dos dois modelos.
- $N$ é o número de amostras.

A estatística MGN segue uma distribuição t com $N-1$ graus de liberdade. A hipótese nula ($H_0$) é de igualdade das variâncias dos erros de previsão (ou seja, $\rho_e = 0$). Valores extremos (positivos ou negativos) e p-valores baixos indicam rejeição da hipótese nula, sugerindo diferença significativa entre os modelos.

```python
# Exemplo prático: Teste MGN (Morgan-Granger-Newbold) entre modelos de regressão
import numpy as np
import pandas as pd
from scipy.stats import t

# Simulando resíduos de 4 modelos (eFCE, eFLS, eFMI, eFTLS)
np.random.seed(42)
N = 100
residuos = {
    'eFCE': np.random.normal(0, 0.03, N),
    'eFLS': np.random.normal(0, 0.04, N),
    'eFMI': np.random.normal(0, 0.035, N),
    'eFTLS': np.random.normal(0, 0.5, N)
}

def mgn_test(e1, e2):
    l1 = e1 + e2
    l2 = e1 - e2
    rho_e = np.corrcoef(l1, l2)[0, 1]
    mgn = rho_e * np.sqrt((N-1)/(1 - rho_e**2)) if abs(rho_e) < 1 else np.sign(rho_e) * np.inf
    p_value = 2 * (1 - t.cdf(abs(mgn), df=N-1))
    return mgn, p_value

# Pares de modelos
modelos = list(residuos.keys())
resultados = []
for i in range(len(modelos)):
    for j in range(i+1, len(modelos)):
        mgn, p = mgn_test(residuos[modelos[i]], residuos[modelos[j]])
        resultados.append({
            'Modelo 1': modelos[i],
            'Modelo 2': modelos[j],
            'MGN': mgn,
            'p-valor': p
        })

# Tabela de resultados
df_mgn = pd.DataFrame(resultados)
df_mgn['MGN'] = df_mgn['MGN'].round(2)
df_mgn['p-valor'] = df_mgn['p-valor'].round(4)
display(df_mgn)
```

# Regressão - R² Score (Coeficiente de Determinação)

O R² Score, ou coeficiente de determinação, é uma métrica que avalia o quão bem um modelo de regressão explica a variância dos dados de saída. O valor de R² varia entre 0 e 1, onde valores próximos de 1 indicam que o modelo explica grande parte da variância dos dados, enquanto valores próximos de 0 indicam baixo poder explicativo.

A fórmula do R² é:

$$
R^2 = 1 - \frac{\sum_{t=1}^N (y_t - \hat{y}_t)^2}{\sum_{t=1}^N (y_t - \bar{y})^2}
$$

onde:
- $y_t$ são os valores reais,
- $\hat{y}_t$ são os valores previstos pelo modelo,
- $\bar{y}$ é a média dos valores reais,
- $N$ é o número de amostras.

O R² representa a fração da variância total dos dados que é explicada pelo modelo. Quanto mais próximo de 1, melhor o ajuste do modelo aos dados.

# Regressão - Boxplot

O boxplot é uma ferramenta gráfica fundamental para analisar a distribuição dos resíduos ou dos valores previstos em problemas de regressão. Ele permite identificar:

- Assimetria dos resíduos
- Presença de outliers
- Dispersão dos erros
- Comparação entre diferentes modelos de regressão

Ao visualizar os resíduos de um modelo de regressão em um boxplot, espera-se que estejam distribuídos de forma aproximadamente simétrica em torno de zero, sem outliers extremos. Isso indica que o modelo não apresenta viés sistemático e que os erros são aleatórios.

```python
# Exemplo prático: Boxplot dos resíduos em regressão
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Gerando dados simulados
np.random.seed(42)
N = 50
y_true = np.linspace(10, 20, N)
y_pred_modelo1 = y_true + np.random.normal(0, 1, N)
y_pred_modelo2 = y_true + np.random.normal(0, 2, N)

# Calculando resíduos
residuo1 = y_true - y_pred_modelo1
residuo2 = y_true - y_pred_modelo2

# Criando DataFrame para boxplot
df_residuos = pd.DataFrame({
    'Modelo 1': residuo1,
    'Modelo 2': residuo2
})

# Plotando boxplot dos resíduos
plt.figure(figsize=(7, 5))
df_residuos.boxplot()
plt.title('Boxplot dos Resíduos dos Modelos de Regressão')
plt.ylabel('Resíduo (y_real - y_previsto)')
plt.grid(axis='y')
plt.show()
```

# Regressão - Histograma

O histograma é uma ferramenta gráfica importante para analisar a distribuição dos resíduos em problemas de regressão. Ele permite visualizar:

- A simetria ou assimetria dos resíduos
- A presença de viés (tendência dos erros para um lado)
- A dispersão dos erros
- Se os resíduos seguem uma distribuição aproximadamente normal (o ideal para muitos modelos)

Um histograma de resíduos centrado em zero e com formato aproximadamente simétrico indica que o modelo não apresenta viés sistemático e que os erros são aleatórios.

```python
# Exemplo prático: Histograma dos resíduos em regressão
import numpy as np
import matplotlib.pyplot as plt

# Usando os mesmos resíduos do exemplo anterior (Modelo 1 e Modelo 2)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(residuo1, bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma dos Resíduos - Modelo 1')
plt.xlabel('Resíduo (y_real - y_previsto)')
plt.ylabel('Frequência')
plt.grid(axis='y')

plt.subplot(1, 2, 2)
plt.hist(residuo2, bins=10, color='orange', edgecolor='black')
plt.title('Histograma dos Resíduos - Modelo 2')
plt.xlabel('Resíduo (y_real - y_previsto)')
plt.ylabel('Frequência')
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

# Outras Medidas Estatísticas em Avaliação de Modelos

Além das métricas tradicionais de erro e ajuste, outras medidas estatísticas são importantes para comparar modelos e validar hipóteses em experimentos:

## ANOVA (Análise de Variância)
A ANOVA é utilizada para comparar as médias de três ou mais grupos e verificar se pelo menos um grupo difere significativamente dos demais. Em avaliação de modelos, pode ser usada para comparar o desempenho de diferentes algoritmos em múltiplos conjuntos de dados.

- **Hipótese nula ($H_0$):** todas as médias são iguais.
- **Hipótese alternativa ($H_1$):** pelo menos uma média é diferente.
- O resultado é um valor F e um p-valor associado.

## Teste t Unilateral
O teste t unilateral compara as médias de dois grupos para verificar se uma média é significativamente maior (ou menor) que a outra. É útil para comparar o desempenho de dois modelos.

- **Hipótese nula ($H_0$):** as médias são iguais.
- **Hipótese alternativa ($H_1$):** uma média é maior (ou menor) que a outra.
- O resultado é um valor t e um p-valor associado.

## p-valor
O p-valor indica a probabilidade de obter um resultado tão extremo quanto o observado, assumindo que a hipótese nula é verdadeira. Valores de p-valor menores que 0.05 geralmente indicam diferença estatisticamente significativa.
