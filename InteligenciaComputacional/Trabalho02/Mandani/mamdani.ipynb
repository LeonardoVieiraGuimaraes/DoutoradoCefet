# Particionamento, Base de Regras, Operadores, Técnicas de Defuzzificação e Análise dos Resultados

## Particionamento dos Universos de Discurso

Cada variável de entrada foi particionada em três conjuntos fuzzy com nomes representativos:

- **Temperatura interna (°C):**
  - Baixa: 15 a 22
  - Média: 18 a 30
  - Alta: 25 a 35

- **Umidade relativa (%):**
  - Baixa: 20 a 45
  - Média: 35 a 85
  - Alta: 70 a 100

- **Número de pessoas:**
  - Poucas: 0 a 7
  - Moderado: 3 a 17
  - Muitas: 12 a 20

A variável de saída (intensidade da ventilação) foi definida qualitativamente:
- Fraca: 0 a 4
- Moderada: 2 a 8
- Forte: 6 a 10

Todas as funções de pertinência são triangulares.

## Base de Regras Fuzzy

A base de regras foi construída com base em conhecimento comum sobre conforto térmico e ventilação:

1. Se temperatura **alta** OU umidade **alta** OU pessoas **muitas** então ventilação **forte**
2. Se temperatura **média** E umidade **média** E pessoas **moderado** então ventilação **moderada**
3. Se temperatura **baixa** E umidade **baixa** E pessoas **poucas** então ventilação **fraca**
4. Se temperatura **alta** E umidade **baixa** então ventilação **moderada**
5. Se temperatura **média** E pessoas **muitas** então ventilação **forte**
6. Se umidade **alta** E pessoas **poucas** então ventilação **moderada**
7. Se temperatura **baixa** OU umidade **baixa** então ventilação **fraca**
8. Se pessoas **moderado** E umidade **média** então ventilação **moderada**
9. Se temperatura **média** E umidade **alta** então ventilação **forte**

## Operadores Lógicos e Agregação

- **AND:** operador mínimo (`min`)
- **OR:** operador máximo (`max`)
- **Agregação:** máximo entre as saídas das regras ativadas

## Técnicas de Defuzzificação

Foram implementadas três técnicas para comparação:

- **Centroide:** já implementada no código principal (média ponderada)
- **Média do Máximo:** média dos valores de saída onde a pertinência é máxima
- **Bissetriz:** ponto que divide a área agregada em duas áreas iguais

Exemplo de implementação para as três técnicas:

```python
def defuzz_centroide(x, mf):
    if np.sum(mf) == 0:
        return 0
    return np.sum(x * mf) / np.sum(mf)

def defuzz_media_maximo(x, mf):
    maxv = np.max(mf)
    if maxv == 0:
        return 0
    return np.mean(x[mf == maxv])

def defuzz_bissetriz(x, mf):
    area_total = np.sum(mf)
    if area_total == 0:
        return 0
    soma = 0
    for i in range(len(x)):
        soma += mf[i]
        if soma >= area_total/2:
            return x[i]
    return x[-1]
```

## Simulações e Análise dos Resultados

Foram simulados 5 cenários distintos de entrada. Os resultados variam conforme os operadores e técnicas de defuzzificação:

- **Centroide:** fornece um valor ponderado, geralmente intermediário.
- **Média do Máximo:** tende a privilegiar regiões onde a pertinência é máxima, podendo ser menos sensível a pequenas ativações.
- **Bissetriz:** busca o ponto de equilíbrio da área, podendo ser mais sensível a assimetrias na agregação.

### Exemplo Detalhado (Cenário 1)

- **Entradas:** temperatura=28, umidade=80, pessoas=15
- **Fuzzificação:** calcula-se o grau de pertinência de cada entrada em cada conjunto fuzzy.
- **Regras ativadas:** são aquelas cujo firing strength (nível de ativação) é maior que zero.
- **Agregação:** as saídas das regras ativadas são combinadas pelo máximo.
- **Defuzzificação:** calcula-se o valor final da ventilação pelas três técnicas.

#### Cálculos Matemáticos

- **Fuzzificação:**  
  Exemplo:  
  - temp_media(28) = (30-28)/(30-24) = 0.333...
  - temp_alta(28) = (28-25)/(35-25) = 0.3

- **Ativação das regras:**  
  Exemplo:  
  - R1: max(temp_alta, umid_alta, pess_muitas) = max(0.3, 0.33, 0.37) = 0.37

- **Agregação:**  
  Para cada ponto do universo de ventilação, pega-se o máximo entre as saídas das regras ativadas.

- **Defuzzificação:**  
  - Centroide: média ponderada dos valores de ventilação pela pertinência agregada.
  - Média do Máximo: média dos valores onde a pertinência agregada é máxima.
  - Bissetriz: ponto que divide a área agregada em duas áreas iguais.

#### Interpretação Qualitativa

No cenário 1, a saída foi próxima de "forte", pois tanto a temperatura quanto a umidade e o número de pessoas estavam elevados, ativando regras que recomendam ventilação intensa. O resultado é coerente com o objetivo de conforto térmico e eficiência energética.

#### Discussão

A escolha dos operadores e da técnica de defuzzificação pode alterar levemente o valor final, mas a tendência qualitativa se mantém. O método centroide é o mais utilizado por ser mais estável e sensível a todas as ativações.

````python
import numpy as np
import matplotlib.pyplot as plt
import os

# Função de pertinência triangular
def trimf(x, a, b, c):
    return np.maximum(np.minimum((x-a)/(b-a+1e-9), (c-x)/(c-b+1e-9)), 0)

# 1. Universos de discurso
temp_univ = np.arange(15, 36, 1)
umid_univ = np.arange(20, 101, 1)
pess_univ = np.arange(0, 21, 1)
vent_univ = np.arange(0, 11, 0.1)

# 2. Funções de pertinência (três conjuntos fuzzy para cada variável)
def temp_baixa(x): return trimf(x, 15, 15, 22)
def temp_media(x): return trimf(x, 18, 24, 30)
def temp_alta(x):  return trimf(x, 25, 35, 35)

def umid_baixa(x): return trimf(x, 20, 20, 45)
def umid_media(x): return trimf(x, 35, 60, 85)
def umid_alta(x):  return trimf(x, 70, 100, 100)

def pess_poucas(x):   return trimf(x, 0, 0, 7)
def pess_moderado(x): return trimf(x, 3, 10, 17)
def pess_muitas(x):   return trimf(x, 12, 20, 20)

def vent_fraca(x):    return trimf(x, 0, 0, 4)
def vent_moderada(x): return trimf(x, 2, 5, 8)
def vent_forte(x):    return trimf(x, 6, 10, 10)

# Função para salvar gráficos em pasta 'img'
def salvar_grafico(fig, nome_arquivo):
    base_dir = os.path.dirname(os.path.abspath(__file__)) if '__file__' in globals() else os.getcwd()
    img_dir = os.path.join(base_dir, "img")
    if not os.path.exists(img_dir):
        os.makedirs(img_dir)
    # Não apaga nem reescreve a pasta se já existir
    caminho = os.path.join(img_dir, nome_arquivo)
    fig.savefig(caminho, bbox_inches='tight')
    plt.close(fig)
    print(f"Gráfico salvo em: {caminho}")

# Visualização das funções de pertinência e salvar
fig, axs = plt.subplots(2, 2, figsize=(12,8))
axs[0,0].plot(temp_univ, temp_baixa(temp_univ), label='baixa')
axs[0,0].plot(temp_univ, temp_media(temp_univ), label='media')
axs[0,0].plot(temp_univ, temp_alta(temp_univ), label='alta')
axs[0,0].set_title('Temperatura')
axs[0,0].legend()

axs[0,1].plot(umid_univ, umid_baixa(umid_univ), label='baixa')
axs[0,1].plot(umid_univ, umid_media(umid_univ), label='media')
axs[0,1].plot(umid_univ, umid_alta(umid_univ), label='alta')
axs[0,1].set_title('Umidade')
axs[0,1].legend()

axs[1,0].plot(pess_univ, pess_poucas(pess_univ), label='poucas')
axs[1,0].plot(pess_univ, pess_moderado(pess_univ), label='moderado')
axs[1,0].plot(pess_univ, pess_muitas(pess_univ), label='muitas')
axs[1,0].set_title('Pessoas')
axs[1,0].legend()

axs[1,1].plot(vent_univ, vent_fraca(vent_univ), label='fraca')
axs[1,1].plot(vent_univ, vent_moderada(vent_univ), label='moderada')
axs[1,1].plot(vent_univ, vent_forte(vent_univ), label='forte')
axs[1,1].set_title('Ventilação')
axs[1,1].legend()
plt.tight_layout()
salvar_grafico(fig, "funcoes_pertinencia.png")

# 3. Base de regras fuzzy (coerente e baseada em conforto térmico)
def regras_fuzzy(mu_temp, mu_umid, mu_pess, operador='padrao'):
    regras = []
    min_op = min
    max_op = max
    if operador == 'and_produto':
        min_op = lambda a, b: a * b
    if operador == 'or_soma':
        max_op = lambda a, b, c: min(a + b + c, 1)
    regras.append(('forte', max_op(mu_temp['alta'], mu_umid['alta'], mu_pess['muitas']), 'R1'))
    regras.append(('moderada', min_op(mu_temp['media'], min_op(mu_umid['media'], mu_pess['moderado'])), 'R2'))
    regras.append(('fraca', min_op(mu_temp['baixa'], min_op(mu_umid['baixa'], mu_pess['poucas'])), 'R3'))
    regras.append(('moderada', min_op(mu_temp['alta'], mu_umid['baixa']), 'R4'))
    regras.append(('forte', min_op(mu_temp['media'], mu_pess['muitas']), 'R5'))
    regras.append(('moderada', min_op(mu_umid['alta'], mu_pess['poucas']), 'R6'))
    regras.append(('fraca', max_op(mu_temp['baixa'], mu_umid['baixa'], 0), 'R7'))
    regras.append(('moderada', min_op(mu_pess['moderado'], mu_umid['media']), 'R8'))
    regras.append(('forte', min_op(mu_temp['media'], mu_umid['alta']), 'R9'))
    return regras

# 4. Fuzzificação
def fuzzificar(entrada):
    mu_temp = {
        'baixa': temp_baixa(entrada['temperatura']),
        'media': temp_media(entrada['temperatura']),
        'alta':  temp_alta(entrada['temperatura'])
    }
    mu_umid = {
        'baixa': umid_baixa(entrada['umidade']),
        'media': umid_media(entrada['umidade']),
        'alta':  umid_alta(entrada['umidade'])
    }
    mu_pess = {
        'poucas':   pess_poucas(entrada['pessoas']),
        'moderado': pess_moderado(entrada['pessoas']),
        'muitas':   pess_muitas(entrada['pessoas'])
    }
    return mu_temp, mu_umid, mu_pess

# 5. Inferência Mamdani e Agregação
def mamdani(entrada, operador='padrao'):
    mu_temp, mu_umid, mu_pess = fuzzificar(entrada)
    regras = regras_fuzzy(mu_temp, mu_umid, mu_pess, operador)
    saida_vent = np.zeros_like(vent_univ)
    regras_ativadas = []
    for nome, firing, reg in regras:
        if firing > 0:
            regras_ativadas.append((reg, nome, firing))
            if nome == 'fraca':
                saida_vent = np.maximum(saida_vent, np.minimum(firing, vent_fraca(vent_univ)))
            elif nome == 'moderada':
                saida_vent = np.maximum(saida_vent, np.minimum(firing, vent_moderada(vent_univ)))
            elif nome == 'forte':
                saida_vent = np.maximum(saida_vent, np.minimum(firing, vent_forte(vent_univ)))
    return saida_vent, regras_ativadas, mu_temp, mu_umid, mu_pess

# 6. Técnicas de defuzzificação
def defuzz_centroide(x, mf):
    if np.sum(mf) == 0:
        return 0
    return np.sum(x * mf) / np.sum(mf)

def defuzz_media_maximo(x, mf):
    maxv = np.max(mf)
    if maxv == 0:
        return 0
    return np.mean(x[mf == maxv])

def defuzz_bissetriz(x, mf):
    area_total = np.sum(mf)
    if area_total == 0:
        return 0
    soma = 0
    for i in range(len(x)):
        soma += mf[i]
        if soma >= area_total/2:
            return x[i]
    return x[-1]

# 7. Simulações de cenários com diferentes operadores e defuzzificações
cenarios = [
    {'temperatura': 28, 'umidade': 80, 'pessoas': 15},  # Cenário 1
    {'temperatura': 22, 'umidade': 50, 'pessoas': 5},   # Cenário 2
    {'temperatura': 18, 'umidade': 30, 'pessoas': 2},   # Cenário 3
    {'temperatura': 32, 'umidade': 60, 'pessoas': 10},  # Cenário 4
    {'temperatura': 25, 'umidade': 90, 'pessoas': 8},   # Cenário 5
]

operadores = ['padrao', 'and_produto', 'or_soma']
defuzzificadores = [
    ('Centroide', defuzz_centroide),
    ('Média do Máximo', defuzz_media_maximo),
    ('Bissetriz', defuzz_bissetriz)
]

for op in operadores:
    print(f"\n--- Operador lógico: {op} ---")
    for i, c in enumerate(cenarios):
        saida_vent, regras_ativadas, _, _, _ = mamdani(c, operador=op)
        resultados = []
        for nome, func in defuzzificadores:
            val = func(vent_univ, saida_vent)
            resultados.append(f"{nome}: {val:.2f}")
        print(f"Cenário {i+1}: Entrada={c} | " + " | ".join(resultados))

# 8. Visualização da superfície fuzzy (temperatura x umidade, pessoas fixo)
x = np.arange(15, 36, 2)
y = np.arange(20, 101, 5)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X, dtype=float)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        entrada = {'temperatura': X[i, j], 'umidade': Y[i, j], 'pessoas': 10}
        saida_vent, _, _, _, _ = mamdani(entrada)
        Z[i, j] = defuzz_centroide(vent_univ, saida_vent)

def plot_superficie_fuzzy(X, Y, Z, nome_arquivo="superficie_fuzzy_pessoas10.png"):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_xlabel('Temperatura (°C)')
    ax.set_ylabel('Umidade (%)')
    ax.set_zlabel('Ventilação')
    plt.title('Superfície Fuzzy (Pessoas=10)')
    # Salva e mostra
    salvar_grafico(fig, nome_arquivo)
    plt.show()

# Chamada da função para plotar e salvar a superfície fuzzy
plot_superficie_fuzzy(X, Y, Z, nome_arquivo="superficie_fuzzy_pessoas10.png")

# 9. Processo detalhado para o cenário 1
entrada = cenarios[0]
print("\nProcesso detalhado para o cenário 1:")
saida_vent, regras_ativadas, mu_temp, mu_umid, mu_pess = mamdani(entrada)
print("Fuzzificação (graus de pertinência):")
print("  Temperatura:", mu_temp)
print("  Umidade:", mu_umid)
print("  Pessoas:", mu_pess)
print("\nRegras ativadas e firing strength:")
for reg, nome, firing in regras_ativadas:
    print(f"  {reg}: {nome} (nível={firing:.3f})")

centroide = defuzz_centroide(vent_univ, saida_vent)
media_max = defuzz_media_maximo(vent_univ, saida_vent)
bissetriz = defuzz_bissetriz(vent_univ, saida_vent)
print(f"\nDefuzzificação:")
print(f"  Centroide: {centroide:.2f}")
print(f"  Média do Máximo: {media_max:.2f}")
print(f"  Bissetriz: {bissetriz:.2f}")

# Função para plotar e salvar a saída fuzzy agregada
def plot_saida_fuzzy_salvar(vent_univ, vent_fraca, vent_moderada, vent_forte, saida_vent, centroide, media_max, bissetriz, nome_arquivo="saida_fuzzy_cenario1.png"):
    plt.figure(figsize=(7,4))
    plt.plot(vent_univ, vent_fraca(vent_univ), '--', label='fraca')
    plt.plot(vent_univ, vent_moderada(vent_univ), '--', label='moderada')
    plt.plot(vent_univ, vent_forte(vent_univ), '--', label='forte')
    plt.fill_between(vent_univ, 0, saida_vent, color='gray', alpha=0.5, label='Agregado')
    plt.axvline(centroide, color='red', linestyle=':', label='Centroide')
    plt.axvline(media_max, color='blue', linestyle=':', label='Média do Máximo')
    plt.axvline(bissetriz, color='green', linestyle=':', label='Bissetriz')
    plt.title('Saída Fuzzy Agregada (Cenário 1)')
    plt.xlabel('Ventilação')
    plt.ylabel('Pertinência')
    plt.legend()
    fig = plt.gcf()
    salvar_grafico(fig, nome_arquivo)
    plt.show()

# Salvar o gráfico da saída fuzzy agregada do cenário 1
plot_saida_fuzzy_salvar(
    vent_univ, vent_fraca, vent_moderada, vent_forte,
    saida_vent, centroide, media_max, bissetriz,
    nome_arquivo="saida_fuzzy_cenario1.png"
)

# Interpretação qualitativa
if centroide < 4:
    nivel = "fraca"
elif centroide < 7:
    nivel = "moderada"
else:
    nivel = "forte"
print(f"\nA intensidade de ventilação sugerida é '{nivel.upper()}' ({centroide:.2f}), adequada ao cenário de temperatura, umidade e ocupação elevadas.")
````
