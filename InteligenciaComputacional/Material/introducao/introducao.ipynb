# O que é Inteligência?

Inteligência é um conceito fundamental, porém de definição e mensuração complexas. É fácil reconhecer comportamentos inteligentes, mas difícil estabelecer uma definição precisa ou uma métrica universal para inteligência.

## Inteligência

- Fácil de reconhecer, difícil de definir ou medir.
- As leis naturais e o significado científico da inteligência ainda estão em aberto e são objeto de estudo contínuo.

## Atributos da Inteligência

Os principais atributos associados à inteligência incluem:

- **Raciocínio:** Capacidade de pensar logicamente e resolver problemas.
- **Senso comum:** Aplicação de conhecimentos práticos em situações cotidianas.
- **Planejamento:** Habilidade de traçar estratégias e antecipar consequências.
- **Percepção:** Capacidade de interpretar e compreender o ambiente ao redor.
- **Criatividade:** Geração de ideias novas e originais.
- **Memória:** Retenção e recuperação eficiente de conhecimento e experiências passadas.

## Definições Complementares de Inteligência

A inteligência pode ser entendida como a capacidade de alguém ou algo para lógica, abstração, memorização, compreensão, autoconhecimento, comunicação, aprendizado, controle emocional, planejamento e resolução de problemas.

Também pode ser definida como a faculdade de conhecer, aprender, compreender, resolver novos problemas, conflitos e adaptar-se a novas situações.

> "A inteligência é a capacidade de agir com eficácia ao ambiente, de aprender com a experiência, de engajar-se nas várias formas de raciocínio e de superar obstáculos mediante o pensamento. Embora tais diferenças individuais possam ser substanciais, nunca são completamente consistentes: o desempenho intelectual de uma pessoa pode variar em ocasiões distintas, em domínios distintos, a se julgar por critérios distintos. Os conceitos de 'inteligência' são tentativas de aclarar e organizar esse conjunto complexo de fenômenos."
>
> (Intelligence: Knowns and Unknowns, 1995)

> "Uma capacidade mental bastante geral que, entre outras coisas, envolve a habilidade de raciocinar, planejar, resolver problemas, pensar de forma abstrata, compreender ideias complexas, aprender rápido e aprender com a experiência. Não é uma mera aprendizagem literária, uma habilidade estritamente acadêmica ou um talento para sair-se bem em provas. Ao contrário disso, o conceito refere-se a uma capacidade mais ampla e mais profunda de compreensão do mundo à sua volta – 'pegar no ar', 'pegar' o sentido das coisas ou 'perceber' uma coisa."
>
> (Mainstream Science on Intelligence, 1994)

## Inteligência Artificial e o Estudo da Inteligência

O campo da Inteligência Artificial (IA) busca entender entidades inteligentes. Uma razão natural para estudá-lo é aprender sobre o comportamento humano e animal. Diferentemente da filosofia e da psicologia, que também estudam a inteligência, a IA se propõe a construir entidades inteligentes e compreender seu funcionamento.

Além disso, as entidades inteligentes construídas podem ser úteis na solução de problemas do cotidiano. Os produtos de IA, mesmo em estágios iniciais, sempre causaram grande impacto na sociedade.

## O Escopo da Inteligência Artificial

A Inteligência Artificial (IA) abrange uma ampla gama de atividades, desde áreas de propósito geral, como raciocínio lógico e percepção de máquina, até tarefas muito específicas, como jogar xadrez, provar teoremas matemáticos e diagnosticar doenças.

Frequentemente, pesquisadores de diferentes áreas migram para o campo da IA em busca de ferramentas para sistematizar e automatizar tarefas intelectuais. Por outro lado, cientistas de IA buscam aplicar suas pesquisas em diversas áreas do conhecimento humano, comprovando a aplicabilidade da IA.

## Universalidade e Polimorfismo da IA

Pode-se afirmar que a IA se tornou um campo verdadeiramente universal. Uma consequência dessa universalidade é o polimorfismo: IA significa coisas diferentes para pessoas diferentes.

A medida do sucesso em IA é tema de debate, pois as expectativas variam conforme a área — neurociência, ciência da computação, robótica, entre outras.

## Diferentes Perspectivas sobre IA

Diferentes pessoas abordam a IA com objetivos distintos. Segundo Russel e Norvig (2021), duas perguntas importantes são:

- Você está preocupado com o pensamento ou com o comportamento?
- Você deseja modelar humanos ou alcançar resultados ótimos?

## Definições Clássicas de Inteligência Artificial

- **Bellman:** "IA é a automação de atividades que a gente associa com o raciocínio humano, como tomada de decisão, solução de problemas, aprendizado."
- **Minsky:** "IA é uma ciência que possibilita que máquinas executem tarefas que requerem inteligência quando executadas por humanos."
- **Haugeland:** "É um esforço para se fazer computadores pensar: máquinas com mentes, em seu sentido amplo e literal."

## Abordagens para Inteligência Artificial

Segundo Russel e Norvig (2010), as abordagens para IA podem ser classificadas em quatro categorias:

- **Pensam humanamente**
- **Atuam humanamente**
- **Pensam racionalmente**
- **Atuam racionalmente**

## Sistemas que Pensam como Humanos

Esta abordagem busca desenvolver sistemas capazes de pensar de maneira semelhante aos humanos, reproduzindo processos mentais como raciocínio, tomada de decisão e aprendizado.

> “The exciting new effort to make computers think ... machines with minds, in the full and literal sense”  
> (Haugeland, 1985)

> “The automation of activities that we associate with human thinking, activities such as decision-making, problem solving, learning”  
> (Bellman, 1978)

## Como os Humanos Pensam?

Para entender como os humanos pensam, são utilizados diferentes métodos:

- **Introspecção:** tentar captar nossos próprios pensamentos.
- **Experimentos psicológicos:** observar uma pessoa em ação.
- **Imagens do cérebro:** observar o cérebro em ação.

Essas abordagens são fundamentais para a ciência da cognição e para a modelagem cognitiva.

## Ciência da Cognição e Modelagem Cognitiva

A ciência da cognição busca compreender os processos mentais humanos, muitas vezes utilizando modelos computacionais para simular funções cognitivas.

> “The art of creating machines that perform functions that require intelligence when performed by people”  
> (Kurzweil, 1990)

> “The study of how to make computers do things at which, at the moment, people are better”  
> (Rich and Knight, 1991)

## Abordagem do Teste de Turing

O Teste de Turing é um marco para avaliar se uma máquina pode exibir comportamento inteligente equivalente ao humano. Para passar no teste, um sistema deve ser capaz de:

- **Processamento de linguagem natural:** comunicar-se em linguagem humana.
- **Representação do conhecimento:** armazenar o que sabe e escuta.
- **Raciocínio automático:** responder perguntas e obter novas conclusões.
- **Aprendizagem de máquina:** adaptar-se, reconhecer e extrapolar padrões.

O Teste de Turing Total inclui ainda:

- **Visão computacional:** perceber objetos.
- **Robótica:** manipular e movimentar objetos.

## Modelagem Cognitiva

> “The study of mental faculties through the use of computational models.”  
> (Charniak and MacDermott, 1985)

> “The study of the computations that make it possible to perceive, reason, and act.”  
> (Winston, 1992)

## Leis do Pensamento: Lógica Matemática

A lógica matemática busca formalizar o raciocínio humano por meio de regras e leis do pensamento. Exemplo clássico:

- Sócrates é um homem.
- Todos os homens são mortais.
- Portanto, Sócrates é mortal.

## Definições de Inteligência Computacional e IA

> “Computational intelligence is the study of the design of intelligent agents.”  
> (Poole, 1998)

> “AI ... is concerned with intelligent behavior in artifacts.”  
> (Nilson, 1998)

## Agentes Racionais

A Inteligência Computacional e a IA buscam reproduzir, por meios computacionais, características normalmente atribuídas à inteligência humana, como:

- Capacidade de resolver problemas
- Indução e dedução lógica
- Raciocínio
- Aprendizagem
- Visão
- Compreensão de linguagem

## Paradigmas Clássicos da IA

O desenvolvimento da IA se apoiou em alguns paradigmas fundamentais:

- **Conexionismo:** criação de máquinas inteligentes a partir da reprodução e interconexão dos elementos de processamento do cérebro dos animais.
- **Evolucionismo:** criação de máquinas inteligentes baseada no processo de evolução e no comportamento de populações de seres vivos.
- **Simbolismo:** criação de máquinas inteligentes por meio do processamento de símbolos e expressões simbólicas, como ocorre no pensamento e raciocínio humano.

## Características de Sistemas Inteligentes

Sistemas de computação inteligentes devem ser capazes de tratar novas situações utilizando:

- Raciocínio
- Generalização
- Associação
- Abstração
- Descoberta de novas capacidades

Essas características envolvem autonomia, aprendizagem e raciocínio.

## Inteligência Computacional e Computação Flexível

Inteligência Computacional é frequentemente associada ao conceito de Computação Flexível (Soft Computing), que engloba técnicas capazes de lidar com incertezas, aproximações e adaptação.

## Influência da Filosofia na Inteligência Artificial

Desde a antiguidade, questões filosóficas influenciam o desenvolvimento da IA:

- Regras formais podem ser usadas para obter conclusões válidas?
- Como a mente surge do cérebro físico?
- De onde vem o conhecimento?
- Como o conhecimento produz a ação?

## Correntes Filosóficas Relevantes

- **Silogismo (Aristóteles, 384-322 A.C.):** modelo de raciocínio dedutivo composto por duas premissas que geram uma conclusão.
- **Dualismo (Descartes, 1596-1650):** separação entre mente e corpo, questionando como a mente emerge do físico.
- **Empirismo (Bacon, 1561-1626):** conhecimento adquirido por experiências sensoriais.
- **Indução (Hume, 1711-1776):** generalização de verdades a partir de casos particulares.
- **Positivismo lógico (Carnap, 1891-1970):** princípio da verificação, combinando racionalismo e empirismo.

## Silogismo

O silogismo, proposto por Aristóteles, é um modelo formal de dedução lógica. Exemplo:

- Premissa 1: Todos os homens são mortais.
- Premissa 2: Sócrates é um homem.
- Conclusão: Sócrates é mortal.

## GPS (Newell e Simon, 1961)

O General Problem Solver (GPS) foi o primeiro programa de computador útil no campo da IA. Seu objetivo era funcionar como uma máquina universal de resolução de problemas, aplicando regras formais para deduzir soluções.

## Fundamentos Matemáticos

A matemática fornece as bases formais para a IA e a Inteligência Computacional:

- Quais são as regras formais para se obter conclusões válidas?
- O que pode ser computado?
- Como raciocinar com informação incerta e imprecisa?

### Marcos Históricos

- **Algoritmo (al-Khowarazmi, século 9; Euclides):** O algoritmo de Euclides para máximo divisor comum é um dos primeiros algoritmos conhecidos.
- **Probabilidade (Cardano, 1501-1576):** Generalização da lógica para situações incertas.
- **Lógica proposicional (Boole, 1847):** Raciocínio com afirmações verdadeiras ou falsas.
- **Máquina de Turing (1950):** Modelo fundamental de computação.
- **Intratabilidade (Cobham, 1964; Edmonds, 1965):** Problemas cujo tempo de resolução cresce exponencialmente.
- **Fuzzy sets (Zadeh, 1965):** Conjunto fuzzy para lidar com incertezas.
- **Teorema da incompletude (Gödel, 1906-1978):** Existem verdades sem prova formal em sistemas matemáticos suficientemente fortes.
- **NP-completeness (Cook, 1971; Karp, 1972):** Classe de problemas para os quais não se conhece solução eficiente.

## Fundamentos Econômicos

A economia contribui com modelos de decisão e otimização:

- Como decidir para maximizar o benefício?
- Como decidir quando outros agentes também tomam decisões?
- Como decidir quando o benefício é de longo prazo?

### Conceitos Importantes

- **Utilidade (Walras, 1834-1910):** Medida de satisfação ou benefício.
- **Teoria de decisão:** Estrutura formal para decisões sob incerteza.
- **Teoria dos jogos (von Neumann e Morgenstern, 1944):** Estratégias racionais em ambientes competitivos, incluindo decisões aleatórias.
- **Programação dinâmica (Bellman, 1957):** Otimização por decomposição em subproblemas.
- **Decisões satisfatórias (Simon, 1916-2001):** Decisões “boas o suficiente” em vez de ótimas, refletindo o comportamento humano real.

## Fundamentos em Neurociência

A neurociência inspira modelos computacionais baseados no funcionamento do cérebro:

- Como o cérebro processa informação?

### Descobertas Relevantes

- **Regiões funcionais do cérebro (Broca, 1824-1880):** Identificação de áreas cerebrais responsáveis por funções específicas.
- **Neurônios (Golgi, 1843-1926; Cajal, 1852-1934):** Estrutura e funcionamento dos neurônios, base para redes neurais artificiais.

## Fundamentos em Psicologia

A psicologia busca entender como humanos e animais pensam e atuam.

- **Behaviorismo (Watson, 1878-1958):** estudo do comportamento observável com base em metodologia científica e experimental.
- **Psicologia cognitiva (William James, 1842-1910):** destaca a importância das cognições como reguladoras do comportamento.
- **Ciência cognitiva (MIT, 1956):** propõe que uma teoria cognitiva deve ser como um programa de computador (Anderson, 1980), descrevendo funções cognitivas em termos de processamento de informação.
- **Agente baseado em conhecimento (Craik, 1943):**
  - Tradução do estímulo em uma representação interna
  - Manipulação da representação por processos cognitivos
  - Tradução da representação em ação

## Engenharia de Computação

A engenharia de computação trata de como construir computadores eficientes.

- **ABC, ENIAC (Atanasoff, 1940/2; Mauchly e Eckert, 1942):** primeiros computadores eletrônicos.
- **Programação (Zuse, 1941):** desenvolvimento das primeiras linguagens de programação.
- **Linguagens, sistemas operacionais, estruturas de dados, paradigmas de programação:** fundamentos da computação moderna.

## Teoria de Controle e Cibernética

Como pode um artefato operar sob o próprio controle?

- **Watter clock (Ktesibios, c. 250 A.C.):** primeira máquina autocontrolada.
- **Regulador de máquina a vapor (Watt, 1736-1819)**
- **Termostato (Drebell, 1572-1633)**
- **Controle por realimentação (século 19)**
- **Teoria moderna de controle (1960)**

## Linguística

Como a linguagem se relaciona com o pensamento?

- **Comportamento verbal (Skinner, 1957)**
- **Estruturas sintáticas (Chomsky, 1957)**
- **Linguística computacional**
- **Processamento de linguagem natural**
- **Representação do conhecimento**

## Linha do Tempo Histórica da IA

- **300 A.C.:** Silogismo grego (Aristóteles)
- **1637:** Método de Descartes: "Penso, logo existo"
- **1726:** Máquina fictícia em "As Viagens de Gulliver" (Swift)
- **1818:** "Frankestein" (Mary Shelley)
- **1835:** Invenção do relé elétrico (Joseph Henry)
- **1847:** Lógica Booleana (George Boole)
- **1859:** "A Origem das Espécies" (Darwin)
- **1917:** Termo "robô" (Karel Capek)
- **1940-49:** Primeiros computadores (ABC, Robinson, Mark I, EDVAC, EDSAC)
- **1943:** Arquiteturas de McCulloch e Pitts
- **1950:** Teste de Turing (Alan Turing), programa de xadrez (Shannon)
- **1955-56:** IPL-11, Logic Theorist, termo "Inteligência Artificial" (Dartmouth)
- **1957:** General Problem Solver (Newell, Shaw, Simon)
- **1958:** Linguagem LISP (McCarthy), Perceptrons (Rosenblatt)
- **1959:** Laboratório de IA no MIT, programa de damas (Samuel)
- **1965:** Projeto DENDRAL, lógica fuzzy (Zadeh)
- **1970:** PROLOG (Colmerauer)
- **1974:** Sistema especialista MYCIN (Shortliffe)
- **1976-88:** Avanços em hardware, sistemas especialistas, redes neurais, aprendizado de máquina, raciocínio probabilístico, computação paralela
- **199x:** Inteligência Computacional
- **2001:** Big Data, redes profundas, sistemas evolutivos, computação granular
- **2011:** Watson vence no Jeopardy!
- **202x:** IA Generativa (ChatGPT, MidJourney, etc.)

## Sistemas Especialistas

Permitem o armazenamento e o processamento de regras do tipo:

- SE antecedente ENTÃO consequente1 SENÃO consequente2

Aplicações típicas: automação industrial, classificação bancária, detecção de fraudes, otimização de investimentos, tomada de decisão, classificação e otimização em diversas áreas.

## Evolução: Especialista vs Dados

- **Início:** Inserção do conhecimento do especialista em programas de computador ("expert in the box").
- **Atual:** Conhecimento extraído de dados coletados ou preexistentes sobre o domínio de aplicação.
