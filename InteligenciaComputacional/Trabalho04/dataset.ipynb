```python
import os
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Cria a pasta 'dataset' se não existir
os.makedirs('dataset', exist_ok=True)
```

### Energy Efficiency Dataset
- **Contexto:** Este conjunto de dados foi criado para avaliar o desempenho energético de edifícios residenciais. Cada linha representa um edifício simulado com diferentes características construtivas e ambientais.
- **Atributos de entrada:**
  - X1: Relative Compactness
  - X2: Surface Area
  - X3: Wall Area
  - X4: Roof Area
  - X5: Overall Height
  - X6: Orientation
  - X7: Glazing Area
  - X8: Glazing Area Distribution
- **Variáveis alvo:**
  - Y1: Heating Load (carga térmica de aquecimento)
  - Y2: Cooling Load (carga térmica de resfriamento)
- **Aplicação:** Utilizado para tarefas de regressão, visando prever o consumo energético de aquecimento e resfriamento de edifícios a partir das características físicas e ambientais.
- **Observações:** Não possui valores ausentes. Todos os atributos são numéricos.



```python
# Download e carregamento do Energy Efficiency Dataset
# Fonte: UCI Machine Learning Repository
energy_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00242/ENB2012_data.xlsx'
energy_path = 'dataset/ENB2012_data.xlsx'

# Faz o download do arquivo caso não exista localmente
if not os.path.exists(energy_path):
    urllib.request.urlretrieve(energy_url, energy_path)

# Carrega o dataset em um DataFrame
df_energy = pd.read_excel(energy_path)

# Exibe as primeiras linhas do dataset para inspeção inicial
print("Energy Efficiency Dataset:")
display(df_energy.head())

# Exibe informações detalhadas sobre o dataset
print("=== Energy Efficiency Dataset ===")
print("Shape:", df_energy.shape)  # Dimensão do dataset (linhas, colunas)
print("Colunas:", df_energy.columns.tolist())  # Lista de colunas
print("Tipos de dados:\n", df_energy.dtypes)  # Tipos de dados de cada coluna
print("Valores ausentes por coluna:\n", df_energy.isnull().sum())  # Checagem de valores ausentes
print("Estatísticas descritivas:\n", df_energy.describe())  # Estatísticas
```

```python
# Histograma para variáveis numéricas com barras azuis e curva normal vermelha
import scipy.stats as stats
import os
os.makedirs('graficos', exist_ok=True)
os.makedirs('datasets_salvos', exist_ok=True)

# Salvar Energy Efficiency Dataset
df_energy.to_csv('datasets_salvos/energy_efficiency.csv', index=False)

num_cols = df_energy.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    data = df_energy[col].dropna()
    counts, bins = np.histogram(data, bins=20)
    bin_centers = 0.5 * (bins[1:] + bins[:-1])
    plt.figure(figsize=(6, 3))
    plt.bar(bin_centers, counts, width=(bins[1] - bins[0]), color='skyblue', alpha=0.7, label='Frequência')
    mu, std = stats.norm.fit(data)
    x_fit = np.linspace(data.min(), data.max(), 200)
    p_fit = stats.norm.pdf(x_fit, mu, std)
    p_fit_scaled = p_fit * len(data) * (bins[1] - bins[0])
    plt.plot(x_fit, p_fit_scaled, color='red', lw=2, label='Distribuição Normal')
    plt.title(f'Distribuição de {col}')
    plt.xlabel(col)
    plt.ylabel('Frequência')
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'graficos/energy_{col}.png')
    plt.close()
```

### California Housing Dataset

- **Contexto:** Este conjunto de dados contém informações sobre distritos residenciais da Califórnia, coletados a partir do censo dos EUA de 1990. Cada linha representa uma região com diversas características socioeconômicas e estruturais.

- **Atributos de entrada:**
  - MedInc: renda média dos domicílios na região
  - HouseAge: idade média das casas
  - AveRooms: média de cômodos por casa
  - AveBedrms: média de quartos por casa
  - Population: população da região
  - AveOccup: média de ocupantes por domicílio
  - Latitude: latitude geográfica
  - Longitude: longitude geográfica

- **Variável alvo:**
  - MedHouseVal: valor médio das casas na região (em centenas de milhares de dólares)

- **Aplicação:** Utilizado para tarefas de regressão, visando prever o valor médio das casas em uma região da Califórnia a partir de suas características físicas e socioeconômicas.

- **Observações:** Todos os atributos são numéricos. Não possui valores ausentes. É um dos datasets mais utilizados para demonstração de algoritmos de regressão.

```python
# Carregamento do California Housing Dataset 
from sklearn.datasets import fetch_california_housing

# Carrega o dataset como DataFrame
california = fetch_california_housing(as_frame=True)
df_california = california.frame

# Exibe as primeiras linhas do dataset para inspeção inicial
print("California Housing Dataset:")
display(df_california.head())

# Exibe informações detalhadas sobre o dataset
print("=== California Housing Dataset ===")
print("Shape:", df_california.shape)
print("Colunas:", df_california.columns.tolist())
print("Tipos de dados:\n", df_california.dtypes)
print("Valores ausentes por coluna:\n", df_california.isnull().sum())
print("Estatísticas descritivas:\n", df_california.describe())
```

```python
# Visualização de distribuições - California Housing Dataset
import os
os.makedirs('graficos', exist_ok=True)
os.makedirs('datasets_salvos', exist_ok=True)

# Salvar California Housing Dataset
df_california.to_csv('datasets_salvos/california_housing.csv', index=False)

# Histograma para variáveis numéricas (um gráfico por variável)
num_cols = df_california.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    plt.figure(figsize=(6, 3))
    sns.histplot(df_california[col], kde=True)
    plt.title(f'Distribuição de {col}')
    plt.xlabel(col)
    plt.ylabel('Frequência')
    plt.tight_layout()
    plt.savefig(f'graficos/california_{col}.png')
    plt.close()
```

### Wine Quality Dataset (Red)

- **Contexto:** Este conjunto de dados contém informações físico-químicas de vinhos tintos portugueses, coletados para estudar a relação entre essas características e a qualidade sensorial do vinho.
- **Atributos de entrada:**
  - fixed acidity: acidez fixa
  - volatile acidity: acidez volátil
  - citric acid: ácido cítrico
  - residual sugar: açúcar residual
  - chlorides: cloretos
  - free sulfur dioxide: dióxido de enxofre livre
  - total sulfur dioxide: dióxido de enxofre total
  - density: densidade
  - pH: potencial hidrogeniônico
  - sulphates: sulfatos
  - alcohol: teor alcoólico
- **Variável alvo:**
  - quality: Nota de qualidade do vinho (escala de 0 a 10, atribuída por especialistas)
- **Aplicação:** Utilizado para tarefas de classificação multiclasse, buscando prever a nota de qualidade do vinho a partir das características físico-químicas.
- **Observações:** Não possui valores ausentes. Todos os atributos são numéricos. A distribuição das classes é desbalanceada (maioria dos vinhos tem nota entre 5 e 7).

```python
# Download e carregamento do Wine Quality Dataset (Red)
# Fonte: UCI Machine Learning Repository
wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
wine_path = 'dataset/winequality-red.csv'

# Faz o download do arquivo caso não exista localmente
if not os.path.exists(wine_path):
    print("Baixando o arquivo do Wine Quality (Red)...")
    urllib.request.urlretrieve(wine_url, wine_path)

# Carrega o dataset em um DataFrame
df_wine = pd.read_csv(wine_path, sep=';')

# Exibe as primeiras linhas do dataset para inspeção inicial
print("Wine Quality Dataset (Red):")
display(df_wine.head())

# Exibe informações detalhadas sobre o dataset
print("=== Wine Quality Dataset (Red) ===")
print("Shape:", df_wine.shape)  # Dimensão do dataset (linhas, colunas)
print("Colunas:", df_wine.columns.tolist())  # Lista de colunas
print("Tipos de dados:\n", df_wine.dtypes)  # Tipos de dados de cada coluna
print("Valores ausentes por coluna:\n", df_wine.isnull().sum())  # Checagem de valores ausentes
print("Estatísticas descritivas:\n", df_wine.describe())  # Estatísticas básicas das variáveis numéricas
```

```python
# Visualização de distribuições - Wine Quality Dataset (Red)
import scipy.stats as stats
import os
os.makedirs('graficos', exist_ok=True)
os.makedirs('datasets_salvos', exist_ok=True)

# Salvar Wine Quality Dataset (Red)
df_wine.to_csv('datasets_salvos/winequality_red.csv', index=False)

num_cols = [col for col in df_wine.select_dtypes(include=['float64', 'int64']).columns if col != 'quality']
for col in num_cols:
    data = df_wine[col].dropna()
    counts, bins = np.histogram(data, bins=20)
    bin_centers = 0.5 * (bins[1:] + bins[:-1])
    plt.figure(figsize=(6, 3))
    plt.bar(bin_centers, counts, width=(bins[1] - bins[0]), color='skyblue', alpha=0.7, label='Frequência')
    mu, std = stats.norm.fit(data)
    x_fit = np.linspace(data.min(), data.max(), 200)
    p_fit = stats.norm.pdf(x_fit, mu, std)
    p_fit_scaled = p_fit * len(data) * (bins[1] - bins[0])
    plt.plot(x_fit, p_fit_scaled, color='red', lw=2, label='Distribuição Normal')
    plt.title(f'Distribuição de {col}')
    plt.xlabel(col)
    plt.ylabel('Frequência')
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'graficos/wine_{col}.png')
    plt.close()

# Gráfico de barras para variável alvo com linha de distribuição normal suavizada
counts = df_wine['quality'].value_counts().sort_index()
x = counts.index.values
mu, std = stats.norm.fit(df_wine['quality'])

plt.figure(figsize=(6, 3))
plt.bar(x, counts.values, color='skyblue', alpha=0.7, label='Contagem')

x_fit = np.linspace(x.min(), x.max(), 200)
p_fit = stats.norm.pdf(x_fit, mu, std)
p_fit_scaled = p_fit * (counts.sum() * (x[1] - x[0]))
plt.plot(x_fit, p_fit_scaled, color='red', lw=2, label='Distribuição Normal Ajustada')

plt.title('Distribuição da variável alvo: quality')
plt.xlabel('quality')
plt.ylabel('Contagem')
plt.legend()
plt.tight_layout()
plt.savefig('graficos/wine_quality.png')
plt.close()
```

### Iris Dataset

- **Contexto:** O famoso conjunto de dados Iris contém amostras de três espécies de flores do gênero Iris (Iris setosa, Iris versicolor e Iris virginica). Cada amostra possui medidas de quatro características morfológicas.
- **Atributos de entrada:**
  - sepal length (cm): comprimento da sépala
  - sepal width (cm): largura da sépala
  - petal length (cm): comprimento da pétala
  - petal width (cm): largura da pétala
- **Variável alvo:**
  - species: espécie da flor (setosa, versicolor, virginica)
- **Aplicação:** Utilizado para tarefas de **classificação multiclasse**, buscando prever a espécie da flor a partir das medidas morfológicas.
- **Observações:** Não possui valores ausentes. Todos os atributos são numéricos. É um dos datasets mais utilizados para demonstração de algoritmos de classificação.

```python
# Download do arquivo Iris em CSV (opcional, pois o scikit-learn já fornece o dataset)
iris_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_path = 'dataset/iris.csv'

# Faz o download do arquivo caso não exista localmente
if not os.path.exists(iris_path):
    print("Baixando o arquivo do Iris Dataset...")
    urllib.request.urlretrieve(iris_url, iris_path)

# Carrega o dataset Iris do scikit-learn (mais recomendado)
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df_iris = iris.frame

# Exibe as primeiras linhas do dataset para inspeção inicial
print("Iris Dataset:")
display(df_iris.head())

# Exibe informações detalhadas sobre o dataset
print("=== Iris Dataset ===")
print("Shape:", df_iris.shape)  # Dimensão do dataset (linhas, colunas)
print("Colunas:", df_iris.columns.tolist())  # Lista de colunas
print("Tipos de dados:\n", df_iris.dtypes)  # Tipos de dados de cada coluna
print("Valores ausentes por coluna:\n", df_iris.isnull().sum())  # Checagem de valores ausentes
print("Estatísticas descritivas:\n", df_iris.describe())  # Estatísticas básicas das variáveis numéricas
```

```python
# Visualização de distribuições - Iris Dataset
import scipy.stats as stats
import os
os.makedirs('graficos', exist_ok=True)
os.makedirs('datasets_salvos', exist_ok=True)

# Salvar Iris Dataset
df_iris.to_csv('datasets_salvos/iris.csv', index=False)

# Histograma para variáveis numéricas com barras azuis e curva normal vermelha
num_cols = df_iris.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    data = df_iris[col].dropna()
    counts, bins = np.histogram(data, bins=20)
    bin_centers = 0.5 * (bins[1:] + bins[:-1])
    plt.figure(figsize=(6, 3))
    plt.bar(bin_centers, counts, width=(bins[1] - bins[0]), color='skyblue', alpha=0.7, label='Frequência')
    mu, std = stats.norm.fit(data)
    x_fit = np.linspace(data.min(), data.max(), 200)
    p_fit = stats.norm.pdf(x_fit, mu, std)
    p_fit_scaled = p_fit * len(data) * (bins[1] - bins[0])
    plt.plot(x_fit, p_fit_scaled, color='red', lw=2, label='Distribuição Normal')
    plt.title(f'Distribuição de {col}')
    plt.xlabel(col)
    plt.ylabel('Frequência')
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'graficos/iris_{col}.png')
    plt.close()

# Gráfico de barras para variável alvo (target) com nomes das espécies no eixo x
if 'target' in df_iris.columns:
    plt.figure(figsize=(6, 3))
    species_names = ['setosa', 'versicolor', 'virginica']
    counts = df_iris['target'].value_counts().sort_index()
    ax = counts.plot(kind='bar', color='skyblue', alpha=0.8)
    plt.title('Distribuição da variável alvo: target')
    plt.xlabel('Espécie')
    plt.ylabel('Contagem')
    ax.set_xticks(range(len(species_names)))
    ax.set_xticklabels(species_names, rotation=0)
    plt.tight_layout()
    plt.savefig('graficos/iris_target.png')
    plt.close()
```
