# Fundamentos e Conceitos de Sistemas Fuzzy

## Introdução

Sistemas fuzzy são ferramentas matemáticas e computacionais para tratar informações imprecisas, subjetivas e graduais, aproximando-se da forma como os humanos raciocinam e se comunicam. Eles permitem representar e manipular incertezas, subjetividades e graduações de pertinência, sendo fundamentais em aplicações de modelagem, previsão, classificação e controle.

**Exemplo:**  
Considere a afirmação "a temperatura está quente". Em lógica clássica, só é possível dizer se está quente (1) ou não (0). Na lógica fuzzy, podemos atribuir um grau de pertinência, como μ_quente(28°C) = 0.7, indicando que 28°C é "parcialmente quente".

**Exemplo em Python:**  
Abaixo, implementamos uma função de pertinência triangular para o conceito "quente" e avaliamos para diferentes temperaturas.

```python
import numpy as np
import matplotlib.pyplot as plt

def triangular(x, a, b, c):
    return np.maximum(np.minimum((x - a)/(b - a), (c - x)/(c - b)), 0)

x = np.linspace(15, 40, 200)
y = triangular(x, 20, 30, 40)

plt.plot(x, y)
plt.title('Função de Pertinência Triangular - "Quente"')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Grau de Pertinência')
plt.grid(True)
plt.show()

# Exemplo de avaliação
temp = 28
pertinencia = triangular(temp, 20, 30, 40)
print(f'μ_quente({temp}°C) = {pertinencia:.2f}')
```

## Conceitos-Chave

- **Conjuntos Fuzzy:**  
  Generalização dos conjuntos clássicos, permitindo graus de pertinência entre 0 e 1.  
  Exemplo: "Temperatura quente" pode ser parcialmente verdadeira para 28°C, ou seja, μ_quente(28) = 0.7.

  **Notação matemática:**  
  Um conjunto fuzzy A em X é definido por:  
  A = { (x, μA(x)) | x ∈ X }, onde μA(x): X → [0, 1]

- **Relações Fuzzy:**  
  Extensão das relações clássicas, permitindo associar elementos de dois conjuntos com graus de pertinência.  
  Exemplo: μR(x, y) = grau de relação entre x ∈ X e y ∈ Y.

- **Inferência Fuzzy:**  
  Processo de raciocínio baseado em regras do tipo SE-ENTÃO, usando operadores fuzzy (mínimo, máximo, etc.) para combinar informações.  
  Exemplo:  
  - Regra: Se temperatura é alta então ventilador é rápido.
  - Entrada: temperatura = 28°C, μ_alta(28) = 0.7
  - Saída: ventilador rápido com grau 0.7

- **Agrupamento (Clustering) Fuzzy:**  
  Técnica para agrupar dados permitindo que um elemento pertença a mais de um grupo com diferentes graus de pertinência.  
  Exemplo: Um ponto pode pertencer ao cluster A com grau 0.6 e ao cluster B com grau 0.4.

## Modelos Fuzzy em Aplicações

  Representação de sistemas complexos onde o conhecimento pode ser objetivo (modelos matemáticos) ou subjetivo (regras linguísticas).

  **Exemplo matemático:**

  - Modelo subjetivo: Se x é "grande" então y é "alto"

  Atribuição de categorias a elementos com base em graus de pertinência.

- **Controle:**

  **Exemplo:**  
  Controlar a velocidade de um motor com base em "erro" e "variação do erro", ambos representados por conjuntos fuzzy.

## Por que usar Fuzzy?

- Permite graduações de pertinência (granulação fuzzy), ou seja, valores intermediários entre 0 e 1.
- Resolve problemas em diferentes níveis de granularidade (detalhe/abstração).
- Lida com imperfeição, incerteza e subjetividade da informação.
- Aproxima o raciocínio computacional do raciocínio humano.
- Permite a integração de conhecimento objetivo (modelos matemáticos) e subjetivo (regras linguísticas).

**Exemplo matemático:**  
Para o conceito "idade jovem", podemos definir:  
μ_jovem(x) =  
- 1, se x ≤ 18  
- (30 - x)/(30 - 18), se 18 < x < 30  
- 0, se x ≥ 30

## Tipos de Informação

- **Perfeita:** "A aula começa às 13h." (informação exata)
- **Imprecisa:** "A aula começa entre 12h e 14h." (intervalo)
- **Incertas:** "Acho que a aula começa às 13h." (incerteza subjetiva)
- **Vaga:** "A aula começa lá pelas 13h." (vaguidade)
- **Probabilista:** "É provável que a aula comece às 13h." (probabilidade)
- **Possibilista:** "É possível que a aula comece às 13h." (possibilidade)
- **Inconsistente:** "José disse 13h, Maria disse 12h45." (informações conflitantes)
- **Ignorância:** "Não faço ideia do horário." (ausência de informação)

**Exemplo matemático:**

μ(x) = exp(−((x − 50)²)/(2·10²))

## Diferença entre Probabilidade e Fuzzy

- **Probabilidade:** Mede a chance de ocorrência de um evento.  
  Exemplo: P(chuva) = 0.7 significa 70% de chance de chover.

- **Fuzzy:** Mede o grau de pertencimento a um conceito.  
  Exemplo: μ_quente(28°C) = 0.7 significa que 28°C é "70% quente".

**Notação matemática:**  
- Probabilidade: P(A) ∈ [0, 1]
- Fuzzy: μA(x) ∈ [0, 1]

## Abordagens de Modelagem em Sistemas Fuzzy

### Baseada em Modelos

- **Informação objetiva:**  
  Utiliza modelos matemáticos tradicionais (equações diferenciais, funções analíticas) para descrever o comportamento do sistema.
  - Exemplo: y(t) = a·x(t) + b·dx/dt

- **Informação subjetiva:**  
  Utiliza afirmações linguísticas, regras fornecidas por especialistas ou extraídas de dados, para complementar ou substituir o modelo matemático.
  - Exemplo: Se temperatura é "alta" então potência é "baixa".

### Model Free (Livre de Modelo)

- **Informação linguística:**  
  Conhecimento prévio é coletado em forma de regras linguísticas (ex: "Se temperatura é alta, então potência é baixa").
- **Ajuste fino por dados:**  
  Parâmetros das funções de pertinência e regras podem ser ajustados automaticamente a partir de dados históricos, usando métodos como mínimos quadrados ou aprendizado por gradiente.
- **Alternativa:**  
  Em alguns casos, pode-se utilizar apenas dados para construir o sistema fuzzy, dispensando modelos matemáticos explícitos (ex: clustering fuzzy).

### Mapeamento Não Linear

Sistemas fuzzy realizam o mapeamento não linear de um vetor de entrada (características) para um vetor de saída, integrando conhecimento objetivo (matemático) e subjetivo (linguístico).

**Notação matemática:**  
Seja x ∈ ℝⁿ, y ∈ ℝᵐ, o sistema fuzzy implementa y = F(x), onde F é definido por regras fuzzy e funções de pertinência.

### Conjuntos Clássicos: Revisão e Propriedades

Conjuntos são usados para classificar elementos em conceitos gerais, como:
- Números pares
- Números ímpares
- Cidades que são capitais na América do Sul
- Times de futebol

Os elementos são divididos em dois grupos distintos:
- **MEMBROS:** pertencem ao conjunto
- **NÃO-MEMBROS:** não pertencem ao conjunto

---

#### Universo de Discurso

O universo de discurso é o espaço onde estão definidos os valores possíveis de análise.  
**Notação:** X

**Exemplo:**  
Defina o universo de discurso discreto X como todos os números inteiros entre -100 e 100:  
X = {x ∈ ℤ | -100 ≤ x ≤ 100}

Dentro desse universo, considere:
- Conjunto dos números positivos (A): A = {x ∈ X | x > 0}
- Conjunto dos números negativos (B): B = {x ∈ X | x < 0}

Exemplos de pertinência:
- 44 ∈ A, ou μA(44) = 1
- 101 ∉ B, ou μB(101) = 0
- -11 ∈ B, ou μB(-11) = 1

Onde μY(x) representa o grau de pertinência do elemento x em relação ao conjunto Y.

---

#### Formas de Definir um Conjunto

- **Lista dos membros:**  
  A = {a₁, a₂, ..., aₙ}

- **Método baseado em regra:**  
  A = {x | P(x)}, onde P(x) é uma propriedade lógica

- **Função característica:**  
  μA(x) = 1 se x ∈ A  
  μA(x) = 0 caso contrário

---

#### Propriedades da Função Característica

- μA(x) : X → {0, 1}
- μX(x) = 1 (todo elemento pertence ao conjunto universo)
- μ∅(x) = 0 (nenhum elemento pertence ao conjunto vazio)
- X é o conjunto universo ou universo de discurso
- ∅ é o conjunto vazio

**Exemplo:**  
A = {1, 2, 5}  
μA(1) = 1, μA(3) = 0

---

Esses conceitos são a base para a generalização fuzzy, onde a função característica é substituída por uma função de pertinência que pode assumir qualquer valor entre 0 e 1.

---

### Operações e Propriedades de Conjuntos

- **Cardinalidade:**  
  |A| = número de elementos de A.  
  Exemplo: A = {1, 2, 5} ⇒ |A| = 3

- **Conjunto das partes (power set):**  
  ℘(A) = conjunto de todos os subconjuntos de A  
  Exemplo:  
  A = {1, 2, 5}  
  ℘(A) = {∅, {1}, {2}, {5}, {1, 2}, {1, 5}, {2, 5}, {1, 2, 5}}  
  |℘(A)| = 2^|A| = 2³ = 8

- **Complemento:**  
  A^c = {x ∈ X | x ∉ A}

- **Diferença:**  
  B − A = {x | x ∈ B e x ∉ A}

- **União:**  
  A ∪ B = {x | x ∈ A ou x ∈ B}

- **Interseção:**  
  A ∩ B = {x | x ∈ A e x ∈ B}

- **Exemplo:**  
  A = {x | 0 ≤ x ≤ 4}, B = {x | 3 ≤ x ≤ 5}  
  A ∩ B = {3, 4}  
  A ∪ B = {0, 1, 2, 3, 4, 5}

---

### Operações em Conjuntos Fuzzy

- **União:**  
  μA∪B(x) = max[μA(x), μB(x)]

- **Interseção:**  
  μA∩B(x) = min[μA(x), μB(x)]

- **Complemento:**  
  μ¯A(x) = 1 − μA(x)

**Exemplo matemático:**  
Se μA(5) = 0.7 e μB(5) = 0.4:  
- União: μA∪B(5) = max(0.7, 0.4) = 0.7  
- Interseção: μA∩B(5) = min(0.7, 0.4) = 0.4  
- Complemento de A: μ¯A(5) = 1 - 0.7 = 0.3

---

### Propriedades dos Conjuntos

- **Involução:** ¯¯A = A
- **Comutatividade:** A ∪ B = B ∪ A, A ∩ B = B ∩ A
- **Associatividade:**  
  A ∪ (B ∪ C) = (A ∪ B) ∪ C  
  A ∩ (B ∩ C) = (A ∩ B) ∩ C
- **Distributividade:**  
  A ∪ (B ∩ C) = (A ∪ B) ∩ (A ∪ C)  
  A ∩ (B ∪ C) = (A ∩ B) ∪ (A ∩ C)
- **Idempotência:** A ∪ A = A, A ∩ A = A
- **Absorção:**  
  A ∪ (A ∩ B) = A  
  A ∩ (A ∪ B) = A
- **Absorção por X e ∅:**  
  A ∩ ∅ = ∅, A ∪ X = X
- **Identidade:**  
  A ∪ ∅ = A, A ∩ X = A
- **De Morgan:**  
  ¯(A ∪ B) = ¯A ∩ ¯B  
  ¯(A ∩ B) = ¯A ∪ ¯B
- **Exclusão do Meio:** A ∪ ¯A = X
- **Não Contradição:** A ∩ ¯A = ∅

---

### Relações Entre Conjuntos

- **Inclusão:**  
  Se todo elemento de A é também elemento de B, então A ⊆ B  
  Notação: ∀x ∈ X, μA(x) ≤ μB(x)

- **Igualdade:**  
  Se A ⊆ B e B ⊆ A, então A = B

- **Disjunção:**  
  A e B são disjuntos se A ∩ B = ∅

- **Partição:**  
  Família de subconjuntos de A: π(A) = {Aᵢ | i ∈ I, Aᵢ ⊆ A}

---

### Convexidade de Conjuntos

- Um conjunto definido por um único intervalo de números reais é convexo.
- Se for definido por mais de um intervalo sem pontos entre eles, não é convexo.

Exemplo:  
A = {x | x ∈ [0, 2] ∪ [4, 5]} não é convexo, pois, por exemplo, 1 ∈ A, 4 ∈ A, mas 2.8 ∉ A.

---

### Produto Cartesiano

- A × B = {(a, b) | a ∈ A, b ∈ B}
- Exemplo:  
  A = {1, 2}, B = {a, e, i, o, u}  
  A × B = {(1, a), (1, e), ..., (2, u)}

---

### Limitações dos Conjuntos Clássicos

- Um elemento x pertence ou não a um conjunto, mas há situações vagas:
  - Alta taxa de inflação
  - Pessoas altas
  - Pequeno erro de aproximação

- Dificuldade em definir limiares precisos usando conjuntos ordinários.

---

### Paradoxo do Monte

Quando um monte de areia deixa de ser um monte ao retirar grãos um a um?  
A teoria clássica de conjuntos só admite valores 0 (falso) ou 1 (verdadeiro).

---

### Conjuntos Fuzzy

A teoria de conjuntos fuzzy estende a clássica, permitindo valores-verdade no intervalo [0, 1], ou seja, cada elemento possui um grau de pertinência ao conjunto.

- **Exemplo:**  
  Conceitos imprecisos: poucos livros, uma longa história, um pouco de sal, pessoas altas.

---

#### Definição Formal

Um conjunto fuzzy A em X é caracterizado por uma função de pertinência (membership function) μA(x):

- μA : X → [0, 1]
- Alternativamente: A : X → [0, 1]

Um conjunto fuzzy A em X é definido por pares ordenados:  
A = {(x, μA(x)) | x ∈ X, μA(x) : X → [0, 1]}

O conjunto fuzzy é totalmente caracterizado por sua função de pertinência μA.

---

### Conjuntos Fuzzy Contínuos e Propriedades

#### Conjunto Fuzzy Contínuo

Um conjunto fuzzy contínuo é representado por:
A = ∫<sub>x</sub> μA(x) / x

**Exemplo:**  
Idade próxima de 50:
A = ∫<sub>ℝ⁺</sub> 1 / [1 + ((z - 50)/10)<sup>4</sup>] / x

- O universo de discurso (x) é o total de valores possíveis para os elementos do conjunto.

---

#### α-corte (α-cut)

- **Definição:**  
  Subconjunto de todos os elementos do domínio cuja função de pertinência é maior ou igual a um valor α.
  - A<sub>α</sub> = {x | μA(x) ≥ α} (α-corte)
  - A<sub>α+</sub> = {x | μA(x) > α} (α-corte forte)
  - α ∈ [0, 1]

- **Suporte:**  
  Subconjunto dos pontos com pertinência maior que 0:  
  Supp(A) = {x ∈ X | μA(x) > 0}

- **Suporte fechado:**  
  CSupp(A) = {x ∈ X | μA(x) ≥ 0}

- **Núcleo:**  
  Core(A) = {x ∈ X | μA(x) = 1}

**Exemplo matemático:**  
Se μA(x) = triangular(x, 20, 30, 40):  
- α-corte para α=0.5: valores de x onde μA(x) ≥ 0.5  
- Suporte: x ∈ (20, 40)  
- Núcleo: x = 30

---

#### Relações Entre Conjuntos Fuzzy

- **Inclusão:**  
  A ⊆ B ⇔ μA(x) ≤ μB(x), ∀x

**Exemplo:**  
Cidades grandes:  
A = {1/BH, 0.8/Uberlândia, 0.7/JF, 0.6/MC, 0.3/Div, 0/Tapiraí}  
- Altura de A: hgt(A) = 1 (A é normal)
- Suporte de A: Supp(A) = {BH, Uberlândia, JF, MC, Div}
- Núcleo de A: Core(A) = {BH}
- Subconjunto de A: B = {0.4/BH, 0.1/Uberlândia} ⇒ B ⊆ A

---

#### Cardinalidade de Conjuntos Fuzzy

- Para X discreto:  
  |A| = ∑<sub>x∈X</sub> μA(x)
- Para X contínuo:  
  |A| = ∫<sub>x</sub> μA(x) dx

**Exemplo:**  
A = {0.25/6.5, 0.5/7, 0.75/7.5, 1/8, 0.75/8.5, 0.5/9, 0.25/9.5}  
|A| = 0.25 + 0.5 + 0.75 + 1 + 0.75 + 0.5 + 0.25 = 4

---

#### Conjunto Fuzzy Degenerado

- Suporte é um único ponto em X, com valor de pertinência igual a 1.

---

#### Convexidade em Conjuntos Fuzzy

Um conjunto fuzzy A é convexo se e somente se:
μA(λx₁ + (1 − λ)x₂) ≥ min(μA(x₁), μA(x₂)), ∀λ ∈ [0, 1]

Se todos os conjuntos α-cut forem convexos, o conjunto fuzzy é convexo.

### Funções de Pertinência Comuns

- **Triangular:**  
  μA(x) = max{min[(x − a)/(b − a), (c − x)/(c − b)], 0}  
  a, b, c definem a forma do triângulo.

- **Trapezoidal:**  
  μA(x) = max{min[(x − a)/(b − a), 1, (d − x)/(d − c)], 0}  
  a, b, c, d definem a forma do trapézio.

- **Gaussiana:**  
  μA(x) = exp(−((x − c)²)/(2σ²))  
  c é o centro, σ é o desvio padrão.

- **Sigmoide:**  
  μA(x) = 1 / (1 + exp(−a(x − c)))  
  a controla a inclinação, c a posição central.

- **Generalizada:**  
  μA(x) = 1 / (1 + |(x − c)/a|<sup>2b</sup>)  
  a controla a largura, b a suavidade, c o ponto central.

- **Retangular:**  
  μA(x) = 1 se a ≤ x ≤ b, 0 caso contrário.

- **Singleton:**  
  μA(x) = 1 se x = a, 0 caso contrário.

---

#### Escolha da Função de Pertinência

A escolha da função de pertinência deve refletir:
- A natureza do problema
- A percepção do conceito a ser capturado
- O nível de detalhe desejado
- O contexto da aplicação
- A adequação para ajuste de parâmetros (otimização)

**Exemplo prático:**  
Para "idade jovem", uma função triangular pode ser adequada; para "temperatura confortável", uma gaussiana pode ser mais realista.

---

Esses conceitos são fundamentais para a modelagem fuzzy, permitindo representar e manipular informações imprecisas de forma flexível e adaptável.

### Operações Padrão e Generalizadas em Conjuntos Fuzzy

#### Operações Padrão

- **Complemento:**  
  ¯A = X − A ⇔ μ¯A(x) = 1 − μA(x)

- **União:**  
  A ∪ B ⇔ μA∪B(x) = max(μA(x), μB(x))

- **Interseção:**  
  A ∩ B ⇔ μA∩B(x) = min(μA(x), μB(x))

**Exemplo matemático:**  
Se μA(3) = 0.6, μB(3) = 0.8:  
- μA∪B(3) = max(0.6, 0.8) = 0.8  
- μA∩B(3) = min(0.6, 0.8) = 0.6  
- μ¯A(3) = 1 - 0.6 = 0.4

---

#### Operações Generalizadas

As operações fuzzy podem ser generalizadas por meio de normas triangulares (t-normas) e co-normas triangulares (s-normas):

- **T-normas (interseção generalizada):**  
  Função T : [0, 1] × [0, 1] → [0, 1]  
  μA∩B(x) = T(μA(x), μB(x))

  Exemplos:
  - Mínimo: T(a, b) = min(a, b)
  - Produto algébrico: T(a, b) = a·b
  - Diferença limitada: T(a, b) = max(0, a + b − 1)
  - Produto drástico: T(a, b) = a se b=1; b se a=1; 0 caso contrário

- **S-normas (união generalizada):**  
  Função S : [0, 1] × [0, 1] → [0, 1]  
  μA∪B(x) = S(μA(x), μB(x))

  Exemplos:
  - Máximo: S(a, b) = max(a, b)
  - Soma probabilística: S(a, b) = a + b − a·b
  - Soma limitada: S(a, b) = min(1, a + b)
  - Soma drástica: S(a, b) = a se b=0; b se a=0; 1 caso contrário

- **Complemento Generalizado:**  
  Função N : [0, 1] → [0, 1]  
  μ¯A(x) = N(μA(x))

  Exemplos:
  - Zadeh: N(a) = 1 − a
  - Sugeno: N(a) = (1−a)/(1+sa), s ∈ (−1,∞)
  - Yager: N(a) = (1 − a^w)^{1/w}, w ∈ (0,∞)

---

#### Propriedades e Dualidade

- **Leis de De Morgan:**  
  T(a, b) = N(S(N(a), N(b)))  
  S(a, b) = N(T(N(a), N(b)))

- **Idempotência:**  
  x T x = x, x S x = x (apenas min e max são idempotentes)

### Relações Fuzzy

- **Funções:** mapeiam vários para um
- **Relações:** podem mapear vários para vários
- Toda função é uma relação, mas nem toda relação é função

- **Produto Cartesiano:**  
  U × V = {(x, y) | x ∈ U, y ∈ V}

- **Relação Binária R(U, V):**  
  μR(x, y) = 1 ⇔ (x, y) ∈ R, 0 caso contrário  
  R ⊆ U × V

- **Relação Fuzzy:**  
  R(U, V) = {(x, y), μR(x, y) | (x, y) ∈ U × V}  
  μR(x, y) ∈ [0, 1]

**Exemplo:**  
Se U = {a, b}, V = {1, 2}, uma relação fuzzy pode ser:  
R = {((a, 1), 0.8), ((a, 2), 0.3), ((b, 1), 0.5), ((b, 2), 1.0)}

---

#### Operações em Relações Fuzzy

- **Interseção:**  
  μR1∩R2(x, y) = μR1(x, y) ∧ μR2(x, y) (t-norma)

- **União:**  
  μR1∪R2(x, y) = μR1(x, y) ∨ μR2(x, y) (s-norma)

- **Complemento:**  
  μ¯R1(x, y) = 1 − μR1(x, y)

---

### Composição de Relações Fuzzy

Dadas duas relações fuzzy R(U, V) e S(V, W), a composição é denotada por R ◦ S.

- **Composição Max-Min:**  
  μR◦S(u, w) = max<sub>v</sub> [min(μR(u, v), μS(v, w))]

- **Composição Max-Produto:**  
  μR◦S(u, w) = max<sub>v</sub> [μR(u, v) · μS(v, w)]

- **Composição Max-Estrela (⋆=t-norma):**  
  μR◦S(u, w) = max<sub>v</sub> [μR(u, v) ⋆ μS(v, w)]

- **Composição sup-t:**  
  R(x, y) = sup<sub>z∈Z</sub> [G(x, z) ∧ W(z, y)]

- **Composição inf-s:**  
  R(x, y) = inf<sub>z∈Z</sub> [G(x, z) ∨ W(z, y)]

**Exemplo matemático:**  
Seja R(a, 1) = 0.8, S(1, x) = 0.6, então na composição max-min:  
μR◦S(a, x) = max_v [min(μR(a, v), μS(v, x))]

---

Essas operações e propriedades são fundamentais para manipulação de conjuntos e relações fuzzy, permitindo modelar e resolver problemas complexos com incerteza e imprecisão.

### Conhecimento e Representação em Sistemas Fuzzy

O conhecimento em sistemas fuzzy pode ser obtido de duas formas principais:
- **Especialista:** Através de sessões de engenharia do conhecimento, como nos sistemas especialistas clássicos.
- **Dados:** Extraído de padrões ou categorias encontrados em conjuntos de dados coletados dos sistemas a serem representados.

#### Representação de Conhecimento por Regras

- Modela a experiência humana em situações específicas.
- Nos sistemas fuzzy, premissas e conclusões podem ser proposições fuzzy (descrição de ações sob condições específicas).

**Exemplo:**  
Se temperatura é "alta" então ventilador é "rápido".

---

### Variáveis Linguísticas

- **Variável numérica:** Assume valor numérico (ex: temperatura = 75°C).
- **Variável linguística:** Representa conceitos de forma imprecisa (ex: jovem, velho, alto, baixo, quente, frio), associados a valores numéricos via funções de pertinência.

Os termos primários de uma variável linguística formam sua estrutura de conhecimento (partição fuzzy).

**Partição fuzzy:**  
Uma família de conjuntos fuzzy {A₁, A₂, ..., Aₙ} constitui uma partição fuzzy A em um universo U se:
- Para cada elemento do universo, existe pelo menos um conjunto fuzzy com grau de pertinência não-nulo.
- Os elementos de A são conjuntos fuzzy unimodais e convexos.

**Exemplo matemático:**  
Para a variável "idade", termos primários podem ser "jovem", "adulto", "idoso", cada um com sua função de pertinência.

---

#### Caracterização Formal de Variável Linguística

Uma variável linguística é formalmente caracterizada por uma quíntupla (Nome, Termos, Universo, Funções de Pertinência, Regras Sintáticas/Semânticas).

---

### Regras Sintáticas e Semânticas

- **Regras sintáticas:** Definem a estrutura/formato das regras, como a informação será armazenada na base de conhecimento.
- **Regras semânticas:** Definem o significado dos termos usados, explicando o que significa, por exemplo, ser ALTO ou MÉDIO.

---

### Estrutura de uma Regra Fuzzy

- **Forma geral:** Se &lt;antecedente&gt; Então &lt;consequente&gt;
- **Exemplo:** Se x é A então y é B

Uma regra fuzzy é uma relação fuzzy a partir da qual se obtém um valor de pertinência (grau de ativação).

O conjunto das regras resulta numa relação fuzzy agregada por um operador de agregação (geralmente uma s-norma).

---

### Inferência Fuzzy: Modus Ponens Generalizado

- **Premissa 1 (fato):** x é A′
- **Premissa 2 (regra):** Se x é A então y é B
- **Conclusão:** y é B′

Se &lt;x é A&gt; é parcialmente verdade (&lt;x é A′&gt;), então &lt;y é B&gt; é parcialmente verdade (&lt;y é B′&gt;).

**Notação matemática:**  
Se μA′(x₀) = α, então o consequente é ativado com grau α.

---

### Raciocínio Fuzzy: Exemplo

- **Premissa 1:** O tomate está muito vermelho (μ_vermelho = 0.9)
- **Premissa 2:** Se o tomate está vermelho então o tomate está maduro
- **Conclusão:** O tomate está muito maduro (μ_maduro = 0.9)

---

### Raciocínio Fuzzy - 4 Etapas

1. **Graus de pertinência:**  
   Compare os fatos conhecidos com os antecedentes das regras fuzzy para encontrar os graus de pertinência em relação a cada função do antecedente.

2. **Força de ativação:**  
   Combine os graus de pertinência dos antecedentes de cada regra usando operadores AND (T-norma) ou OR (S-norma) para formar a força de ativação, indicando o grau em que a parte antecedente da regra é satisfeita.

3. **Implicação do consequente:**  
   Aplique a força de ativação ao consequente da regra para gerar o grau de ativação da regra (usualmente T-norma).

4. **Saída (Agregação):**  
   Agregue todos os consequentes ativados (usualmente S-norma) para obter uma saída geral.

- **Operador de Defuzzificação:**  
  Transforma a MF de saída em valor crisp ou escalar.

**Exemplo matemático:**  
Se temos duas regras ativadas com graus 0.6 e 0.8, a saída pode ser agregada por max(0.6, 0.8) = 0.8.

## Sistemas Fuzzy do Tipo Takagi-Sugeno-Kang (TSK)

Até agora, discutimos sistemas baseados em regras fuzzy em que tanto os antecedentes quanto os consequentes são conjuntos fuzzy (ex: Mamdani).

Agora, veremos sistemas em que os antecedentes são conjuntos fuzzy, mas os consequentes são funções das variáveis de entrada (modelos funcionais).

### Estrutura das Regras TSK

- **Forma geral:**  
  Se X é Ai e Y é Bi então z = fi(x, y)
  - Ai, Bi: conjuntos fuzzy dos antecedentes
  - fi(x, y): função (geralmente linear ou constante) das variáveis de entrada

- **Ordem do modelo:**  
  - Ordem 1: fi(x, y) é polinômio de 1ª ordem (ex: z = ax + by + c)
  - Ordem 0: fi(x, y) é constante (ex: z = c)

---

### Saída do Sistema TSK

A saída é a média ponderada das saídas individuais das regras:
z = (∑<sub>k=1</sub><sup>N</sup> m<sub>k</sub> · z<sub>k</sub>) / (∑<sub>k=1</sub><sup>N</sup> m<sub>k</sub>)

- z<sub>k</sub>: saída da regra k (função do consequente)
- m<sub>k</sub>: grau de disparo da regra k (T-norma dos antecedentes)

**Exemplo matemático:**  
Se temos duas regras:  
- Regra 1: Se x é A1 então z = 2x + 1  
- Regra 2: Se x é A2 então z = −x + 5  
Com graus de disparo m1 = 0.7, m2 = 0.3, para x = 2:  
z1 = 2·2 + 1 = 5  
z2 = −2 + 5 = 3  
Saída: z = (0.7·5 + 0.3·3)/(0.7 + 0.3) = (3.5 + 0.9)/1 = 4.4

---

### Exemplo de Sistema TSK

**Base de regras:**
- Se x é pequeno, então z = 0.1x + 6.4
- Se x é médio, então z = −0.5x + 4
- Se x é grande, então z = x − 2

**Com duas entradas:**
- Se x é pequeno E y é pequeno, então z = −x + y + 1
- Se x é pequeno E y é grande, então z = −y + 3
- Se x é grande E y é pequeno, então z = −x + 3
- Se x é grande E y é grande, então z = x + y + 2

---

### Exemplo Prático: Máquina de Lavar (TSK Ordem Zero)

- Entradas: peso das roupas (p) e nível de sujeira (s)
- Saída: quantidade de detergente

**Passos:**
1. Calcule a ativação de cada regra:  
   w<sub>i</sub> = μA1<sub>i</sub>(p) ∧ μA2<sub>i</sub>(s)
2. Calcule a saída:  
   y = (∑ w<sub>i</sub> Q<sub>i</sub>) / (∑ w<sub>i</sub>)

---

### Ajuste de Parâmetros em Modelos TSK

- Os parâmetros das funções consequentes podem ser ajustados a partir de dados, minimizando o erro quadrático entre a saída do modelo e os dados reais.
- Técnicas comuns: Mínimos Quadrados Recursivo (MQR), Gradiente Descendente, Algoritmos Genéticos.

---

### Comparação: Mamdani vs Takagi-Sugeno

- **FIS Mamdani:**
  - Mais intuitivo
  - Aceita entradas e saídas linguísticas
  - Apropriado para interação humana

- **FIS Takagi-Sugeno:**
  - Computacionalmente eficiente
  - Fácil integração com controle PID e métodos de otimização
  - Garante continuidade da superfície de saída
  - Consequentes podem ser ajustados por métodos de aprendizado

---

Esses conceitos permitem a modelagem de sistemas fuzzy tanto para controle quanto para previsão e classificação, com flexibilidade para diferentes tipos de conhecimento e aplicações.
