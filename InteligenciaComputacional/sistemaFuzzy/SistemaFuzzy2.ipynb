# Raciocínio Fuzzy - 4 Etapas

Este notebook descreve as quatro etapas principais do raciocínio fuzzy.

## 1. Graus de pertinência

Compare os fatos conhecidos com os antecedentes das regras fuzzy para encontrar os graus de pertinência em relação a cada função do antecedente.

## 2. Força de ativação

Combine os graus de pertinência dos antecedentes de cada regra usando operadores AND ou OR para formar uma força de ativação que indica o grau em que a parte antecedente da regra é satisfeita.

## 3. Implicação do consequente

Aplique a força de ativação ao consequente da regra para gerar o grau de ativação da regra.

## 4. Saída

Agregue todos os consequentes ativados para obter uma saída geral do sistema fuzzy.

### Operadores típicos em sistemas fuzzy

- **AND**: usualmente utiliza-se uma T-norma (por exemplo, mínimo) para calcular a força de ativação de uma regra com AND nos antecedentes.
- **OR**: usualmente utiliza-se uma S-norma (por exemplo, máximo) para calcular a força de ativação de uma regra com OR nos antecedentes.
- **Operador de Implicação**: normalmente uma T-norma é usada para calcular o consequente com base na força de disparo.
- **Operador de Agregação**: geralmente uma S-norma é utilizada para agregar os consequentes e gerar a saída geral.
- **Operador de Defuzzificação**: converte a função de pertinência de saída em um valor crisp ou escalar.

## Métodos de Inferência

Até agora, discutimos sistemas baseados em regras fuzzy em que tanto os antecedentes quanto os consequentes são conjuntos fuzzy.

---

Hoje veremos um novo tipo de sistema baseado em regras fuzzy, no qual os antecedentes são conjuntos fuzzy, mas os consequentes são funções das variáveis independentes (por exemplo, funções lineares). Esse tipo de sistema é conhecido como sistema fuzzy do tipo Takagi-Sugeno-Kang (TSK).

## Modelo de TSK (Takagi-Sugeno-Kang)

O modelo TSK é semelhante ao método de Mamdani em vários aspectos: utiliza fuzzificação das entradas, aplicação de operadores fuzzy e uma base de regras condicionais para inferência.

No entanto, a principal diferença está nos consequentes das regras: em vez de serem expressos por conjuntos fuzzy linguísticos, os consequentes do modelo TSK são funções paramétricas (por exemplo, funções lineares) que relacionam as entradas e as saídas do processo. Ou seja, os consequentes são funcionais e não linguísticos.

Essa diferença implica em uma sintaxe distinta para o banco de conhecimento em relação ao modelo Mamdani, o que também afeta as propriedades semânticas do sistema, como será detalhado posteriormente.

# Resumo e Fórmulas Principais

## 1. Grau de Pertinência (Membership)
Para um valor $x$ e conjunto fuzzy $A$:
$$
\mu_A(x) \in [0, 1]
$$

## 2. Força de Ativação (Activation Strength)
- **AND (T-norma, mínimo):**
$$
w = \min(\mu_{A_1}(x_1), \mu_{A_2}(x_2), \ldots)
$$
- **OR (S-norma, máximo):**
$$
w = \max(\mu_{A_1}(x_1), \mu_{A_2}(x_2), \ldots)
$$

## 3. Implicação do Consequente
- **Mamdani:** 
$$
\mu_{B}'(y) = \min(w, \mu_B(y))
$$
- **TSK:** Consequente é uma função, por exemplo:
$$
y = f(x_1, x_2) = a_0 + a_1 x_1 + a_2 x_2
$$

## 4. Agregação dos Consequentes
- **Mamdani (S-norma, máximo):**
$$
\mu_{B_{agg}}(y) = \max(\mu_{B_1}'(y), \mu_{B_2}'(y), \ldots)
$$
- **TSK (média ponderada):**
$$
y_{saida} = \frac{\sum_{i=1}^N w_i f_i(x_1, x_2, \ldots)}{\sum_{i=1}^N w_i}
$$

## 5. Defuzzificação (Mamdani)
- **Centroide:**
$$
y^* = \frac{\int y \cdot \mu_{B_{agg}}(y) \, dy}{\int \mu_{B_{agg}}(y) \, dy}
$$

---
**Resumo:**  
- Mamdani: consequentes são conjuntos fuzzy, saída obtida por agregação e defuzzificação.  
- TSK: consequentes são funções, saída é média ponderada das funções consequentes.

# Resumo dos Conceitos e Fórmulas de Sistemas Fuzzy

## 1. Variável Linguística e Funções de Pertinência

- Uma variável linguística $x$ é definida por uma quíntupla $(x, t(x), U, G, M)$.
- Cada termo linguístico $A$ é representado por uma função de pertinência $\mu_A(x)$.

Exemplo de função triangular:
$$
\mu_A(x) = \max(0, 1 - \frac{|x - c|}{w})
$$
onde $c$ é o centro e $w$ a largura.

## 2. Regra Fuzzy

- Estrutura: Se $x$ é $A$ então $y$ é $B$.

## 3. Inferência Fuzzy (Modus Ponens Generalizado)

- Dado um fato $x$ é $A'$, a conclusão é $y$ é $B'$:
$$
\mu_{B'}(y) = \max_x \min[\mu_{A'}(x), \mu_R(x, y)]
$$
onde $\mu_R(x, y) = \min[\mu_A(x), \mu_B(y)]$ (implicação de Mamdani).

## 4. Inferência com Duas Premissas

- Regra: Se $x$ é $A$ e $y$ é $B$ então $z$ é $C$.
- Fato: $x$ é $A'$, $y$ é $B'$.
- Grau de ativação:
$$
w_1 = \max_x [\mu_{A'}(x) \wedge \mu_A(x)]
$$
$$
w_2 = \max_y [\mu_{B'}(y) \wedge \mu_B(y)]
$$
- Saída fuzzy:
$$
\mu_{C'}(z) = (w_1 \wedge w_2) \wedge \mu_C(z)
$$

## 5. Métodos de Inferência

### Mamdani (Max-Min)
- Implicação: mínimo
- Agregação: máximo
$$
B'_i(y) = w_i \wedge B_i(y)
$$
$$
B'(y) = \max_i B'_i(y)
$$

### Larsen (Max-Produto)
- Implicação: produto
- Agregação: máximo
$$
B'_i(y) = w_i \cdot B_i(y)
$$
$$
B'(y) = \max_i B'_i(y)
$$

### Kosko/Mizumoto (MIN + Soma)
- Implicação: mínimo
- Agregação: soma ponto a ponto
$$
C(z) = \sum_{i=1}^n \min(w_i, \mu_{C_i}(z))
$$

## 6. Defuzzificação

### Centro da Área (CoA)
$$
\text{CoA}(Z) = \frac{\sum_{i=1}^n z_i \cdot C(z_i)}{\sum_{i=1}^n C(z_i)}
$$

### Centro das Somas (CoS)
$$
\text{CoS}(Z) = \frac{\sum_{i=1}^n z_i \cdot \sum_{k=1}^{N'} C'_k(z_i)}{\sum_{i=1}^n \sum_{k=1}^{N'} C'_k(z_i)}
$$

### Máximo/Mínimo/Média dos Máximos
- Mínimo dos máximos: menor $z$ com $C(z)$ máximo.
- Máximo dos máximos: maior $z$ com $C(z)$ máximo.
- Média dos máximos: média dos $z$ com $C(z)$ máximo.
