## 1. Introdução: O que são Sistemas Fuzzy?

Sistemas fuzzy são ferramentas matemáticas e computacionais para tratar informações imprecisas, subjetivas e graduais, aproximando-se do raciocínio humano. Eles permitem representar e manipular incertezas e graduações de pertinência, sendo fundamentais em aplicações de modelagem, previsão, classificação e controle.

**Exemplo prático:**  
No controle de um ar-condicionado, ao invés de "ligar/desligar" (lógica clássica), pode-se ajustar a potência conforme o quão "quente" está o ambiente, usando graus de pertinência.

---

## 2. Conceitos-Chave e Estrutura Matemática

### 2.1 Conjuntos Clássicos

- **Definição:** Um conjunto clássico A em um universo X é caracterizado por uma função característica:
  - μA(x) = 1 se x ∈ A
  - μA(x) = 0 se x ∉ A

**Exemplo:**  
A = {1, 2, 5} em X = {1, 2, 3, 4, 5}  
μA(3) = 0, μA(2) = 1

---

### 2.2 Conjuntos Fuzzy

- **Definição formal:**  
  Um conjunto fuzzy A em X é caracterizado por uma função de pertinência μA(x):  
  μA : X → [0, 1]

  A = { (x, μA(x)) | x ∈ X }

- **Exemplo:**  
  Conceito "Temperatura quente" em X = [0, 40]°C  
  μ_quente(20) = 0  
  μ_quente(30) = 1  
  μ_quente(25) = 0.5

---

### 2.3 Função de Pertinência

A função de pertinência define o grau com que cada elemento pertence ao conjunto fuzzy.

**Principais tipos:**
- Triangular
- Trapezoidal
- Gaussiana
- Sigmoide

**Exemplo matemático (triangular):**
- μA(x) = max{min[(x − a)/(b − a), (c − x)/(c − b)], 0}
- Parâmetros: a (início), b (pico), c (fim)

---

# Exemplo: Função de pertinência triangular para "Temperatura Quente"
import numpy as np
import matplotlib.pyplot as plt

def triangular(x, a, b, c):
    return np.maximum(np.minimum((x - a)/(b - a), (c - x)/(c - b)), 0)

x = np.linspace(0, 40, 200)
y = triangular(x, 20, 30, 40)

plt.plot(x, y)
plt.title('Função de Pertinência Triangular - "Temperatura Quente"')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Grau de Pertinência')
plt.grid(True)
plt.show()

### 2.4 Operações com Conjuntos Fuzzy

- **União:**  
  μA∪B(x) = max[μA(x), μB(x)]
- **Interseção:**  
  μA∩B(x) = min[μA(x), μB(x)]
- **Complemento:**  
  μ¯A(x) = 1 − μA(x)

**Exemplo:**  
Se μA(5) = 0.7 e μB(5) = 0.4:  
- União: max(0.7, 0.4) = 0.7  
- Interseção: min(0.7, 0.4) = 0.4  
- Complemento de A: 1 - 0.7 = 0.3

---

## 3. Propriedades e Elementos Matemáticos

### 3.1 α-corte, Suporte e Núcleo

- **α-corte:**  
  A<sub>α</sub> = {x | μA(x) ≥ α}
- **Suporte:**  
  Supp(A) = {x | μA(x) > 0}
- **Núcleo:**  
  Core(A) = {x | μA(x) = 1}

**Exemplo:**  
Para μA(x) = triangular(x, 20, 30, 40):  
- α-corte para α=0.5: valores de x onde μA(x) ≥ 0.5  
- Suporte: x ∈ (20, 40)  
- Núcleo: x = 30

---

### 3.2 Cardinalidade

- **Discreto:** |A| = ∑ μA(x)
- **Contínuo:** |A| = ∫ μA(x) dx

**Exemplo:**  
A = {0.2/1, 0.5/2, 1/3}  
|A| = 0.2 + 0.5 + 1 = 1.7

---

### 3.3 Convexidade

Um conjunto fuzzy A é convexo se:  
μA(λx₁ + (1 − λ)x₂) ≥ min(μA(x₁), μA(x₂)), ∀λ ∈ [0, 1]

---

## 4. Relações Fuzzy e Composição

- **Relação fuzzy:**  
  R(U, V) = {((x, y), μR(x, y)) | x ∈ U, y ∈ V}

- **Composição Max-Min:**  
  μR◦S(u, w) = max<sub>v</sub> [min(μR(u, v), μS(v, w))]

- **Composição Max-Produto:**  
  μR◦S(u, w) = max<sub>v</sub> [μR(u, v) · μS(v, w)]

**Exemplo prático:**  
Seja R(x, y) e S(y, z) relações fuzzy, a composição permite inferir relações indiretas entre x e z.

---

## 5. Variáveis Linguísticas e Regras Fuzzy

- **Variável linguística:**  
  Variável cujo valor é um termo qualitativo (ex: "temperatura" = "quente", "fria").

- **Estrutura formal:**  
  Quíntupla (nome, termos, universo, função de pertinência, regras sintáticas/semânticas).

- **Regras fuzzy (IF-THEN):**  
  Exemplo:  
  SE temperatura é "alta" ENTÃO ventilador é "rápido"

**Exemplo prático:**  
SE "idade" é "jovem" ENTÃO "risco" é "baixo"

---

## 6. Inferência Fuzzy: Passo a Passo Matemático

1. **Fuzzificação:**  
   Converte entradas nítidas em graus de pertinência.

2. **Avaliação das regras:**  
   Calcula o grau de ativação de cada regra (usando t-normas para AND, s-normas para OR).

3. **Implicação:**  
   Aplica o grau de ativação ao consequente da regra.

4. **Agregação:**  
   Combina as saídas fuzzy de todas as regras (usualmente s-norma).

5. **Defuzzificação:**  
   Converte o resultado fuzzy em um valor nítido (ex: centroide).

**Exemplo matemático:**  
Se entrada x = 25°C, calcule μ_frio(25), μ_médio(25), μ_quente(25), aplique as regras e obtenha a saída.

---

```python
# Exemplo: Fuzzificação e Defuzzificação simples
import numpy as np

# Funções de pertinência para "frio", "médio", "quente"
def frio(x): return np.clip((20 - x)/10, 0, 1)
def medio(x): return np.clip((x - 15)/10, 0, 1) * np.clip((25 - x)/10, 0, 1)
def quente(x): return np.clip((x - 20)/10, 0, 1)

x = 22
print("μ_frio(22):", frio(x))
print("μ_medio(22):", medio(x))
print("μ_quente(22):", quente(x))

# Suponha regra: SE quente ENTÃO ventilador=100%
# Defuzzificação (simples): saída = μ_quente(x) * 100
saida = quente(x) * 100
print("Saída ventilador:", saida, "%")
```

## 7. Sistemas Fuzzy Mamdani e Takagi-Sugeno (TSK)

### 7.1 Mamdani

- Regras com antecedentes e consequentes fuzzy.
- Saída obtida por defuzzificação do conjunto fuzzy agregado.

**Exemplo:**  
SE temperatura é "alta" ENTÃO ventilador é "rápido"

---

### 7.2 Takagi-Sugeno (TSK)

- Antecedentes fuzzy, consequentes como funções das entradas (geralmente lineares ou constantes).
- Saída: média ponderada das saídas das regras.

**Exemplo:**  
SE temperatura é "alta" ENTÃO saída = 0.8 * temperatura + 2

## Exercício Resolvido: Fuzzificação com Função Triangular

Considere o conceito fuzzy "Nota Boa" para notas de uma prova, definido por uma função triangular com início em 5, pico em 7 e fim em 9.

Pergunta: Qual o grau de pertinência de uma nota 6.5? Represente graficamente a função e destaque o valor encontrado.

```python
import numpy as np
import matplotlib.pyplot as plt

def triangular(x, a, b, c):
    return np.maximum(np.minimum((x - a)/(b - a), (c - x)/(c - b)), 0)

# Parâmetros da função triangular
a, b, c = 5, 7, 9
x = np.linspace(4, 10, 200)
y = triangular(x, a, b, c)

nota = 6.5
pertinencia = triangular(nota, a, b, c)

plt.plot(x, y, label='Nota Boa (triangular)')
plt.scatter([nota], [pertinencia], color='red', zorder=5)
plt.text(nota, pertinencia+0.05, f'({nota}, {pertinencia:.2f})', color='red')
plt.title('Função de Pertinência Triangular - Nota Boa')
plt.xlabel('Nota')
plt.ylabel('Grau de Pertinência')
plt.grid(True)
plt.legend()
plt.show()

print(f"Grau de pertinência da nota {nota}: {pertinencia:.2f}")
```

**Resposta:**  
O grau de pertinência da nota 6.5 é 0.75. Isso indica que 6.5 é "bastante boa", mas não totalmente, segundo a função triangular definida.

## Exercício Resolvido: Fuzzificação com Função Trapezoidal

Considere o conceito fuzzy "Nota Excelente" para notas de uma prova, definido por uma função trapezoidal com início em 7, platô entre 8 e 9, e fim em 10.

Pergunta: Qual o grau de pertinência de uma nota 8.5? Represente graficamente a função e destaque o valor encontrado.

```python
import numpy as np
import matplotlib.pyplot as plt

def trapezoidal(x, a, b, c, d):
    return np.maximum(np.minimum(np.minimum((x - a)/(b - a), 1), (d - x)/(d - c)), 0)

# Parâmetros da função trapezoidal
a, b, c, d = 7, 8, 9, 10
x = np.linspace(6, 11, 200)
y = trapezoidal(x, a, b, c, d)

nota = 8.5
pertinencia = trapezoidal(nota, a, b, c, d)

plt.plot(x, y, label='Nota Excelente (trapezoidal)')
plt.scatter([nota], [pertinencia], color='red', zorder=5)
plt.text(nota, pertinencia+0.05, f'({nota}, {pertinencia:.2f})', color='red')
plt.title('Função de Pertinência Trapezoidal - Nota Excelente')
plt.xlabel('Nota')
plt.ylabel('Grau de Pertinência')
plt.grid(True)
plt.legend()
plt.show()

print(f"Grau de pertinência da nota {nota}: {pertinencia:.2f}")
```

**Resposta:**  
O grau de pertinência da nota 8.5 é 1.00. Isso indica que 8.5 é totalmente considerada "excelente" segundo a função trapezoidal definida.

## Exercício Resolvido: Fuzzificação com Função Gaussiana

Considere o conceito fuzzy "Nota Muito Alta" para notas de uma prova, definido por uma função gaussiana com média em 9 e desvio padrão 0.5.

Pergunta: Qual o grau de pertinência de uma nota 8.5? Represente graficamente a função e destaque o valor encontrado.

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussiana(x, media, sigma):
    return np.exp(-0.5 * ((x - media)/sigma)**2)

media = 9
sigma = 0.5
x = np.linspace(7, 11, 200)
y = gaussiana(x, media, sigma)

nota = 8.5
pertinencia = gaussiana(nota, media, sigma)

plt.plot(x, y, label='Nota Muito Alta (gaussiana)')
plt.scatter([nota], [pertinencia], color='red', zorder=5)
plt.text(nota, pertinencia+0.05, f'({nota}, {pertinencia:.2f})', color='red')
plt.title('Função de Pertinência Gaussiana - Nota Muito Alta')
plt.xlabel('Nota')
plt.ylabel('Grau de Pertinência')
plt.grid(True)
plt.legend()
plt.show()

print(f"Grau de pertinência da nota {nota}: {pertinencia:.2f}")
```

**Resposta:**  
O grau de pertinência da nota 8.5 é aproximadamente 0.61. Isso indica que 8.5 é moderadamente considerada "muito alta" segundo a função gaussiana definida.

## Exercício Resolvido: Fuzzificação com Função Sigmoidal

Considere o conceito fuzzy "Nota Acima da Média" para notas de uma prova, definido por uma função sigmoidal com centro em 7 e inclinação 3.

Pergunta: Qual o grau de pertinência de uma nota 8.5? Represente graficamente a função e destaque o valor encontrado.

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoidal(x, a, c):
    return 1 / (1 + np.exp(-a * (x - c)))

a = 3      # inclinação
c = 7      # centro
x = np.linspace(5, 10, 200)
y = sigmoidal(x, a, c)

nota = 8.5
pertinencia = sigmoidal(nota, a, c)

plt.plot(x, y, label='Nota Acima da Média (sigmoidal)')
plt.scatter([nota], [pertinencia], color='red', zorder=5)
plt.text(nota, pertinencia+0.05, f'({nota}, {pertinencia:.2f})', color='red')
plt.title('Função de Pertinência Sigmoidal - Nota Acima da Média')
plt.xlabel('Nota')
plt.ylabel('Grau de Pertinência')
plt.grid(True)
plt.legend()
plt.show()

print(f"Grau de pertinência da nota {nota}: {pertinencia:.2f}")
```

**Resposta:**  
O grau de pertinência da nota 8.5 é aproximadamente 0.98. Isso indica que 8.5 é considerada "acima da média" com alto grau segundo a função sigmoidal definida.

## Exercício Resolvido: Fuzzificação com Função Sino (bell-shaped)

Considere o conceito fuzzy "Nota Muito Boa" para notas de uma prova, definido por uma função sino generalizada (bell-shaped) com parâmetros a=1, b=4, c=8.

Pergunta: Qual o grau de pertinência de uma nota 8.5? Represente graficamente a função e destaque o valor encontrado.

```python
import numpy as np
import matplotlib.pyplot as plt

def sino(x, a, b, c):
    return 1 / (1 + np.abs((x - c) / a) ** (2 * b))

a = 1    # largura
b = 4    # inclinação
c = 8    # centro
x = np.linspace(5, 11, 200)
y = sino(x, a, b, c)

nota = 8.5
pertinencia = sino(nota, a, b, c)

plt.plot(x, y, label='Nota Muito Boa (sino)')
plt.scatter([nota], [pertinencia], color='red', zorder=5)
plt.text(nota, pertinencia+0.05, f'({nota}, {pertinencia:.2f})', color='red')
plt.title('Função de Pertinência Sino - Nota Muito Boa')
plt.xlabel('Nota')
plt.ylabel('Grau de Pertinência')
plt.grid(True)
plt.legend()
plt.show()

print(f"Grau de pertinência da nota {nota}: {pertinencia:.2f}")
```

**Resposta:**  
O grau de pertinência da nota 8.5 é aproximadamente 0.68. Isso indica que 8.5 é considerada "muito boa" com grau significativo segundo a função sino definida.

## Exercício Resolvido: Fuzzificação com Função Z (Z-shaped)

Considere o conceito fuzzy "Nota Baixa" para notas de uma prova, definido por uma função Z com início em 6 e fim em 8.

Pergunta: Qual o grau de pertinência de uma nota 7? Represente graficamente a função e destaque o valor encontrado.

```python
import numpy as np
import matplotlib.pyplot as plt

def z_shape(x, a, b):
    x = np.asarray(x)
    z = np.zeros_like(x)
    idx1 = x <= a
    idx2 = (x > a) & (x < (a + b) / 2)
    idx3 = (x >= (a + b) / 2) & (x < b)
    idx4 = x >= b
    z[idx1] = 1
    z[idx2] = 1 - 2 * ((x[idx2] - a) / (b - a))**2
    z[idx3] = 2 * ((b - x[idx3]) / (b - a))**2
    z[idx4] = 0
    return z

a = 6  # início da transição
b = 8  # fim da transição
x = np.linspace(5, 10, 200)
y = z_shape(x, a, b)

nota = 7
pertinencia = z_shape(np.array([nota]), a, b)[0]

plt.plot(x, y, label='Nota Baixa (Z-shaped)')
plt.scatter([nota], [pertinencia], color='red', zorder=5)
plt.text(nota, pertinencia+0.05, f'({nota}, {pertinencia:.2f})', color='red')
plt.title('Função de Pertinência Z - Nota Baixa')
plt.xlabel('Nota')
plt.ylabel('Grau de Pertinência')
plt.grid(True)
plt.legend()
plt.show()

print(f"Grau de pertinência da nota {nota}: {pertinencia:.2f}")
```

**Resposta:**  
O grau de pertinência da nota 7 é 0.50. Isso indica que 7 é parcialmente considerada "baixa" segundo a função Z definida.

## Exercício Resolvido: Fuzzificação com Função S (S-shaped)

Considere o conceito fuzzy "Nota Alta" para notas de uma prova, definido por uma função S com início em 6 e fim em 8.

Pergunta: Qual o grau de pertinência de uma nota 7? Represente graficamente a função e destaque o valor encontrado.

```python
import numpy as np
import matplotlib.pyplot as plt

def s_shape(x, a, b):
    x = np.asarray(x)
    s = np.zeros_like(x)
    idx1 = x <= a
    idx2 = (x > a) & (x < (a + b) / 2)
    idx3 = (x >= (a + b) / 2) & (x < b)
    idx4 = x >= b
    s[idx1] = 0
    s[idx2] = 2 * ((x[idx2] - a) / (b - a))**2
    s[idx3] = 1 - 2 * ((b - x[idx3]) / (b - a))**2
    s[idx4] = 1
    return s

a = 6  # início da transição
b = 8  # fim da transição
x = np.linspace(5, 10, 200)
y = s_shape(x, a, b)

nota = 7
pertinencia = s_shape(np.array([nota]), a, b)[0]

plt.plot(x, y, label='Nota Alta (S-shaped)')
plt.scatter([nota], [pertinencia], color='red', zorder=5)
plt.text(nota, pertinencia+0.05, f'({nota}, {pertinencia:.2f})', color='red')
plt.title('Função de Pertinência S - Nota Alta')
plt.xlabel('Nota')
plt.ylabel('Grau de Pertinência')
plt.grid(True)
plt.legend()
plt.show()

print(f"Grau de pertinência da nota {nota}: {pertinencia:.2f}")
```

**Resposta:**  
O grau de pertinência da nota 7 é 0.50. Isso indica que 7 é parcialmente considerada "alta" segundo a função S definida.

Calcule e represente graficamente:

**Resposta:**  
Para x = 6:
- μ<sub>A</sub>(6) = 0.67
- μ<sub>B</sub>(6) = 0.67
- μ<sub>A∪B</sub>(6) = 0.67
- μ<sub>A∩B</sub>(6) = 0.67
- μ<sub>¬A</sub>(6) = 0.33

## Operações Fuzzy Generalizadas: T-normas e S-normas

Além das operações padrão (min, max, complemento), podemos utilizar operadores generalizados para interseção e união fuzzy:

- **T-normas (interseção):**
  - Mínimo: T(a, b) = min(a, b)  *(Zadeh padrão)*
  - Produto: T(a, b) = a · b
  - Lukasiewicz: T(a, b) = max(a + b - 1, 0)

- **S-normas (união):**
  - Máximo: S(a, b) = max(a, b)  *(Zadeh padrão)*
  - Soma Probabilística: S(a, b) = a + b - ab
  - Soma Limitada: S(a, b) = min(1, a + b)
  - Soma Drástica (1-4): 
  $$
  S(a, b) =
  \begin{cases}
    a, & \text{se } b = 0 \\
    b, & \text{se } a = 0 \\
    1, & \text{caso contrário}
  \end{cases}
  $$

A seguir, exemplos práticos dessas operações.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo as T-normas e S-normas
def T_min(a, b):
    return np.minimum(a, b)

def T_prod(a, b):
    return a * b

def T_lukasiewicz(a, b):
    return np.maximum(a + b - 1, 0)

def S_max(a, b):
    return np.maximum(a, b)

def S_soma_probabilistica(a, b):
    return a + b - a * b

def S_soma_limitada(a, b):
    return np.minimum(a + b, 1)

def S_lukasiewicz(a, b):
    return np.minimum(a + b, 1)

# Valores de exemplo
a = np.array([0.2, 0.5, 0.8])
b = np.array([0.3, 0.6, 0.9])

# Calculando as operações
resultado_T_min = T_min(a, b)
resultado_T_prod = T_prod(a, b)
resultado_T_lukasiewicz = T_lukasiewicz(a, b)
resultado_S_max = S_max(a, b)
resultado_S_soma_probabilistica = S_soma_probabilistica(a, b)
resultado_S_soma_limitada = S_soma_limitada(a, b)
resultado_S_lukasiewicz = S_lukasiewicz(a, b)

# Exibindo resultados
print("Resultados das T-normas:")
print("Mínimo:", resultado_T_min)
print("Produto:", resultado_T_prod)
print("Lukasiewicz:", resultado_T_lukasiewicz)

print("\nResultados das S-normas:")
print("Máximo:", resultado_S_max)
print("Soma probabilística:", resultado_S_soma_probabilistica)
print("Soma limitada:", resultado_S_soma_limitada)
print("Lukasiewicz:", resultado_S_lukasiewicz)
```

**Exemplos Práticos:**

Suponha que temos dois conjuntos fuzzy A e B, com funções de pertinência em um determinado universo.

- Para um dado x, temos:
  - μA(x) = 0.7 e μB(x) = 0.4

**Cenário 1: Interseção (AND) usando T-norma Mínimo:**
- Resultado: T_min(μA(x), μB(x)) = min(0.7, 0.4) = 0.4

**Cenário 2: Interseção (AND) usando T-norma Produto:**
- Resultado: T_prod(μA(x), μB(x)) = 0.7 · 0.4 = 0.28

**Cenário 3: Interseção (AND) usando T-norma Lukasiewicz:**
- Resultado: T_lukasiewicz(μA(x), μB(x)) = max(0.7 + 0.4 - 1, 0) = 0.1

**Cenário 4: União (OR) usando S-norma Máximo:**
- Resultado: S_max(μA(x), μB(x)) = max(0.7, 0.4) = 0.7

**Cenário 5: União (OR) usando S-norma Soma limitada:**
- Resultado: S_soma_limitada(μA(x), μB(x)) = min(0.7 + 0.4, 1) = 1

**Cenário 6: União (OR) usando S-norma Lukasiewicz:**
- Resultado: S_lukasiewicz(μA(x), μB(x)) = min(0.7 + 0.4, 1) = 1

Esses exemplos mostram como diferentes operadores podem ser aplicados em conjuntos fuzzy para obter interseções e uniões com base em T-normas e S-normas.

## Dualidade e Leis de De Morgan em Lógica Fuzzy

Assim como na lógica clássica, as operações fuzzy também obedecem à dualidade e às Leis de De Morgan, mas usando os operadores fuzzy (min, max ou t-normas/s-normas):

- **Dualidade:**  
  O complemento da união é igual à interseção dos complementos, e vice-versa.

- **Leis de De Morgan (fuzzy):
  - Complemento da união:  
    ¬(A ∪ B) = ¬A ∩ ¬B
  - Complemento da interseção:  
    ¬(A ∩ B) = ¬A ∪ ¬B

No caso padrão de Zadeh:
- ¬A(x) = 1 − μA(x)
- União: max
- Interseção: min

A seguir, um exemplo prático e gráfico dessas leis.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo uma função genérica de complemento
def complemento(f, x):
    return 1 - f(x)

# Funções de pertinência exemplo
def mu_A(x): return np.maximum(0, 1 - np.abs(x - 3)/2)
def mu_B(x): return np.maximum(0, 1 - np.abs(x - 5)/2)

x = np.linspace(0, 8, 200)
muA = mu_A(x)
muB = mu_B(x)

# Complementos
mu_not_A = complemento(mu_A, x)
mu_not_B = complemento(mu_B, x)

# Plotando
plt.plot(x, muA, label='μA(x)')
plt.plot(x, muB, label='μB(x)')
plt.plot(x, mu_not_A, label='¬A(x)', linestyle='--')
plt.plot(x, mu_not_B, label='¬B(x)', linestyle='--')
plt.title('Leis de De Morgan em Lógica Fuzzy')
plt.xlabel('x')
plt.ylabel('Grau de Pertinência')
plt.legend()
plt.grid(True)
plt.show()
```

## Função T-norma para Interseção Fuzzy

A função T-norma é uma função $T : [0, 1] \times [0, 1] \rightarrow [0, 1]$ que agrega dois graus de pertinência para modelar a interseção fuzzy.

### Axiomas das T-normas

1. **Condições limite:**  
   $T(0, 0) = 0$  
   $T(a, 1) = T(1, a) = a$

2. **Monotonicidade:**  
   Se $b \leq d$, então $T(a, b) \leq T(a, d)$

3. **Comutatividade:**  
   $T(a, b) = T(b, a)$

4. **Associatividade:**  
   $T(a, T(b, c)) = T(T(a, b), c)$

5. **Continuidade:**  
   $T(a, b)$ é contínua (opcional)

6. **Idempotência:**  
   $T(a, a) = a$ (opcional)

### Exemplos de T-normas

- **Mínimo (1-6):** $T(a, b) = \min(a, b)$
- **Produto Algébrico (1-5):** $T(a, b) = ab$
- **Diferença Limitada (1-5):** $T(a, b) = \max(0, a + b - 1)$
- **Produto Drástico (1-4):**
  $$
  T(a, b) =
  \begin{cases}
    a, & \text{se } b = 1 \\
    b, & \text{se } a = 1 \\
    0, & \text{caso contrário}
  \end{cases}
  $$

A T-norma mais comum é o mínimo:

$\mu_{A \cap B}(x) = \min(\mu_A(x), \mu_B(x)) = \mu_A(x) \wedge \mu_B(x)$

A seguir, um exemplo prático em Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Funções de pertinência exemplo
def mu_A(x): return np.maximum(0, 1 - np.abs(x - 3)/2)
def mu_B(x): return np.maximum(0, 1 - np.abs(x - 5)/2)

x = np.linspace(0, 8, 200)
muA = mu_A(x)
muB = mu_B(x)

# Interseção usando T-norma mínimo
mu_intersec = np.minimum(muA, muB)

plt.plot(x, muA, label='μA(x)')
plt.plot(x, muB, label='μB(x)')
plt.plot(x, mu_intersec, label='μA∩B(x) = min(μA, μB)', linestyle='--')
plt.title('Interseção Fuzzy via T-norma (mínimo)')
plt.xlabel('x')
plt.ylabel('Grau de Pertinência')
plt.legend()
plt.grid(True)
plt.show()
```

## União Fuzzy

A união fuzzy é definida por uma função S-norma $S : [0, 1] \times [0, 1] \rightarrow [0, 1]$ que agrega duas funções de pertinência:

$\mu_{A \cup B}(x) = S(\mu_A(x), \mu_B(x))$

### Axiomas das S-normas

1. **Condições limite:**  
   $S(0, 0) = 0$  
   $S(a, 0) = S(0, a) = a$

2. **Monotonicidade:**  
   Se $b \leq d$, então $S(a, b) \leq S(a, d)$

3. **Comutatividade:**  
   $S(a, b) = S(b, a)$

4. **Associatividade:**  
   $S(a, S(b, c)) = S(S(a, b), c)$

5. **Continuidade:**  
   $S(a, b)$ é contínua (opcional)

6. **Idempotência:**  
   $S(a, a) = a$ (opcional)

### Exemplos de S-normas

- **Máximo (1-6):** $S(a, b) = \max(a, b)$
- **Soma Probabilística (1-5):** $S(a, b) = a + b - ab$
- **Soma Limitada (1-5):** $S(a, b) = \min(1, a + b)$
- **Soma Drástica (1-4):**
  $$
  S(a, b) =
  \begin{cases}
    a, & \text{se } b = 0 \\
    b, & \text{se } a = 0 \\
    1, & \text{caso contrário}
  \end{cases}
  $$

A S-norma mais comum é o máximo:

$\mu_{A \cup B}(x) = \max(\mu_A(x), \mu_B(x)) = \mu_A(x) \vee \mu_B(x), \quad \forall x \in X$

### Observações sobre as Operações Padrão

- **Interseção padrão (operador min):** produz o maior conjunto fuzzy entre todos os produzidos por todas as possíveis interseções (t-normas).
- **União padrão (operador max):** produz o menor conjunto fuzzy entre todos os produzidos por todas as possíveis uniões (s-normas).
- **Idempotência:** são operações idempotentes ($x \, T \, x = x$ e $x \, S \, x = x$); são as únicas operações idempotentes entre as t-normas e s-normas.

## 10. T-normas e S-normas Duais: Definição, Tabela e Exemplos

Duas normas (T-norma e S-norma) são **duais** em relação ao complemento fuzzy $N$ se e somente se:
- $T(a, b) = N(S(N(a), N(b)))$
- $S(a, b) = N(T(N(a), N(b)))$
onde $N(a) = 1 - a$ (complemento padrão de Zadeh).

### Tabela de T-normas e S-normas mais comuns

| Nome                | T-norma $T(a, b)$                       | S-norma $S(a, b)$                       |
|---------------------|-----------------------------------------|-----------------------------------------|
| **Min Zadeh**       | $\min(a, b)$                            | $\max(a, b)$                            |
| **Produto Algébrico** | $a \cdot b$                            | $a + b - a \cdot b$ (Soma Probabilística)|
| **Lukasiewicz**     | $\max(0, a + b - 1)$                    | $\min(1, a + b)$                        |
| **Hamacher**        | $\frac{a b}{\gamma + (1-\gamma)(a + b - a b)}$ | $\frac{a + b - (1-\gamma)a b}{1 - (1-\gamma)a b}$ |
| **Diferença Limitada** | $\max(a + b - 1, 0)$                  | $\min(a + b, 1)$                        |
| **Weber Drástico**  | $a$ se $b=1$; $b$ se $a=1$; $0$ caso contrário | $a$ se $b=0$; $b$ se $a=0$; $1$ caso contrário |

### Exemplo Numérico de Dualidade

Seja $a = 0.6$, $b = 0.8$:

- **T-norma Min:** $T(a, b) = \min(0.6, 0.8) = 0.6$
- **S-norma Max:** $S(a, b) = \max(0.6, 0.8) = 0.8$
- **Dualidade:** $T(a, b) = 1 - S(1-0.6, 1-0.8) = 1 - S(0.4, 0.2) = 1 - 0.4 = 0.6$

### Gráfico das principais T-normas e S-normas

A seguir, um gráfico comparativo das principais T-normas e S-normas para $a$ variando de 0 a 1 e $b$ fixo.

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(0, 1, 200)
b = 0.7

# T-normas
t_min = np.minimum(a, b)
t_prod = a * b
t_lukas = np.maximum(0, a + b - 1)

# S-normas
s_max = np.maximum(a, b)
s_prob = a + b - a * b
s_lukas = np.minimum(1, a + b)

plt.figure(figsize=(10, 5))
plt.plot(a, t_min, label='T-norma Min (Zadeh)')
plt.plot(a, t_prod, label='T-norma Produto')
plt.plot(a, t_lukas, label='T-norma Lukasiewicz')
plt.plot(a, s_max, '--', label='S-norma Max (Zadeh)')
plt.plot(a, s_prob, '--', label='S-norma Probabilística')
plt.plot(a, s_lukas, '--', label='S-norma Lukasiewicz')
plt.title('Principais T-normas e S-normas (b=0.7)')
plt.xlabel('a')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()
```

## 11. Relações Binárias e Relações Fuzzy

### Relação Binária

- Uma relação binária $R \subseteq U \times V$ associa pares $(x, y)$.
- Exemplo: $U = \{10, 15, 20\}$, $V = \{2, 3, 5\}$, $R(x, y): x$ é divisível por $y$.
  - $R = \{(10,2), (10,5), (15,3), (15,5), (20,2), (20,5)\}$

### Relação Fuzzy

- Uma relação fuzzy $R$ é um conjunto fuzzy definido em $U \times V$:
  $$
  R(U, V) = \{((x, y), \mu_R(x, y)) \mid x \in U, y \in V\}
  $$
- $\mu_R(x, y) \in [0, 1]$ representa o grau de associação.

#### Exemplo Numérico

Considere $U = V = \{10, 40, 80, 100, 300\}$ e $R(x, y): x$ é muito maior que $y$:

| x/y   | 10  | 40  | 80  | 100 | 300 |
|-------|-----|-----|-----|-----|-----|
| **10**  | 0   | 0   | 0   | 0   | 0   |
| **40**  | 0.4 | 0   | 0   | 0   | 0   |
| **80**  | 0.8 | 0.2 | 0   | 0   | 0   |
| **100** | 1.0 | 0.6 | 0.2 | 0   | 0   |
| **300** | 1.0 | 0.8 | 0.4 | 0.2 | 0   |

#### Visualização de uma relação fuzzy

Vamos plotar a matriz de pertinência como um mapa de calor.

```python
import numpy as np
import matplotlib.pyplot as plt

U = V = np.array([10, 40, 80, 100, 300])
mu_R = np.array([
    [0,   0,   0,   0,   0],
    [0.4, 0,   0,   0,   0],
    [0.8, 0.2, 0,   0,   0],
    [1.0, 0.6, 0.2, 0,   0],
    [1.0, 0.8, 0.4, 0.2, 0]
])

plt.figure(figsize=(6,5))
plt.imshow(mu_R, cmap='Blues', interpolation='nearest', origin='upper')
plt.colorbar(label='Grau de pertinência')
plt.xticks(range(len(V)), V)
plt.yticks(range(len(U)), U)
plt.xlabel('y')
plt.ylabel('x')
plt.title('Relação fuzzy: x é muito maior que y')
for i in range(len(U)):
    for j in range(len(V)):
        plt.text(j, i, f"{mu_R[i, j]:.1f}", ha='center', va='center', color='black')
plt.show()
```

## 12. Operações com Relações Fuzzy

As operações de união, interseção e complemento também se aplicam a relações fuzzy:

- **Interseção:** $\mu_{R_1 \cap R_2}(x, y) = \mu_{R_1}(x, y) \wedge \mu_{R_2}(x, y)$
- **União:** $\mu_{R_1 \cup R_2}(x, y) = \mu_{R_1}(x, y) \vee \mu_{R_2}(x, y)$
- **Complemento:** $\mu_{\overline{R}}(x, y) = 1 - \mu_{R}(x, y)$

### Exemplo Prático

Considere $U = \{2, 12\}$, $V = \{1, 7, 13\}$:

- $\mu_p(u, v)$: $u$ é próximo de $v$
- $\mu_m(u, v)$: $u$ é muito menor que $v$

$\mu_p(u, v) = \begin{bmatrix} 0.9 & 0.4 & 0.1 \\ 0.1 & 0.4 & 0.9 \end{bmatrix}$

$\mu_m(u, v) = \begin{bmatrix} 0 & 0.6 & 1 \\ 0 & 0 & 0.3 \end{bmatrix}$

- **Interseção:** $\mu_{p \cap m}(u, v) = \min(\mu_p(u, v), \mu_m(u, v))$
- **União:** $\mu_{p \cup m}(u, v) = \max(\mu_p(u, v), \mu_m(u, v))$

Vamos calcular e visualizar essas operações.

```python
import numpy as np
import matplotlib.pyplot as plt

U = np.array([2, 12])
V = np.array([1, 7, 13])
mu_p = np.array([[0.9, 0.4, 0.1],
                 [0.1, 0.4, 0.9]])
mu_m = np.array([[0, 0.6, 1],
                 [0, 0, 0.3]])

mu_inter = np.minimum(mu_p, mu_m)
mu_union = np.maximum(mu_p, mu_m)

fig, axs = plt.subplots(1, 2, figsize=(10,4))
im1 = axs[0].imshow(mu_inter, cmap='Greens', interpolation='nearest', origin='upper')
axs[0].set_title('Interseção (min)')
axs[0].set_xticks(range(len(V)))
axs[0].set_xticklabels(V)
axs[0].set_yticks(range(len(U)))
axs[0].set_yticklabels(U)
for i in range(len(U)):
    for j in range(len(V)):
        axs[0].text(j, i, f"{mu_inter[i, j]:.1f}", ha='center', va='center', color='black')
fig.colorbar(im1, ax=axs[0])

im2 = axs[1].imshow(mu_union, cmap='Oranges', interpolation='nearest', origin='upper')
axs[1].set_title('União (max)')
axs[1].set_xticks(range(len(V)))
axs[1].set_xticklabels(V)
axs[1].set_yticks(range(len(U)))
axs[1].set_yticklabels(U)
for i in range(len(U)):
    for j in range(len(V)):
        axs[1].text(j, i, f"{mu_union[i, j]:.1f}", ha='center', va='center', color='black')
fig.colorbar(im2, ax=axs[1])

plt.tight_layout()
plt.show()
```

## 13. Composição de Relações Fuzzy

A composição de relações fuzzy permite combinar informações de diferentes domínios, semelhante à multiplicação de matrizes, mas usando mínimo (t-norma) e máximo (s-norma).

### Definição

Sejam $R(U, V)$ e $S(V, W)$ relações fuzzy, a composição $R \circ S$ é dada por:

- **Composição Max-Min:**  
  $\mu_{R \circ S}(u, w) = \max_{v} [\min(\mu_R(u, v), \mu_S(v, w))]$

- **Composição Max-Produto:**  
  $\mu_{R \circ S}(u, w) = \max_{v} [\mu_R(u, v) \cdot \mu_S(v, w)]$

### Exemplo Numérico

Considere $U = \{2, 12\}$, $V = \{1, 7, 13\}$, $W = \{4, 8\}$:

$\mu_p(u, v) = \begin{bmatrix} 0.9 & 0.4 & 0.1 \\ 0.1 & 0.4 & 0.9 \end{bmatrix}$

$\mu_m(v, w) = \begin{bmatrix} 0 & 0 \\ 0.6 & 0 \\ 1 & 0.7 \end{bmatrix}$

Vamos calcular a composição max-min $\mu_{p \circ m}(u, w)$.

```python
import numpy as np

U = np.array([2, 12])
V = np.array([1, 7, 13])
W = np.array([4, 8])

mu_p = np.array([[0.9, 0.4, 0.1],
                 [0.1, 0.4, 0.9]])
mu_m = np.array([[0, 0],
                 [0.6, 0],
                 [1, 0.7]])

# Composição max-min
mu_comp = np.zeros((len(U), len(W)))
for i in range(len(U)):
    for j in range(len(W)):
        min_vals = [min(mu_p[i, k], mu_m[k, j]) for k in range(len(V))]
        mu_comp[i, j] = max(min_vals)

import matplotlib.pyplot as plt
plt.figure(figsize=(4,3))
plt.imshow(mu_comp, cmap='Purples', interpolation='nearest', origin='upper')
plt.colorbar(label='Grau de pertinência')
plt.xticks(range(len(W)), W)
plt.yticks(range(len(U)), U)
plt.title('Composição Max-Min: μp∘m(u,w)')
for i in range(len(U)):
    for j in range(len(W)):
        plt.text(j, i, f"{mu_comp[i, j]:.1f}", ha='center', va='center', color='black')
plt.xlabel('w')
plt.ylabel('u')
plt.tight_layout()
plt.show()
```

**Resultado da composição:**

| u\w | 4   | 8   |
|-----|-----|-----|
| 2   | 0.4 | 0.1 |
| 12  | 0.9 | 0.7 |

A composição fuzzy permite inferir relações indiretas entre domínios, sendo fundamental em sistemas fuzzy para encadear regras e inferências.

---

### Outras Composições

- **Composição Max-Produto:**  
  $\mu_{R \circ S}(u, w) = \max_{v} [\mu_R(u, v) \cdot \mu_S(v, w)]$

- **Composição Max-Estrela:**  
  $\mu_{R \circ S}(u, w) = \max_{v} [\mu_R(u, v) \star \mu_S(v, w)]$  
  onde $\star$ é uma t-norma qualquer.

## 14. Propriedades das Composições de Relações Fuzzy

- **Inversiva:** $(R \circ S)^{-1} = S^{-1} \circ R^{-1}$
- **Associativa:** $(R \circ S) \circ T = R \circ (S \circ T)$
- **Comutatividade:** *Não é satisfeita em geral.*
