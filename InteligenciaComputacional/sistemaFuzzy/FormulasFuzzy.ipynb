# Fórmulas Fundamentais de Sistemas Fuzzy

Este notebook reúne e organiza as principais fórmulas matemáticas e conceitos de sistemas fuzzy, incluindo Mamdani, Takagi-Sugeno-Kang (TSK), operadores, propriedades, medidas e métodos de defuzzificação.

## 1. Funções de Pertinência

- **Triangular:**  
  $\displaystyle \mu_A(x) = \max\left(\min\left(\frac{x-a}{b-a}, \frac{c-x}{c-b}\right), 0\right)$

- **Trapezoidal:**  
  $\displaystyle \mu_A(x) = \max\left(\min\left(\frac{x-a}{b-a}, 1, \frac{d-x}{d-c}\right), 0\right)$

- **Gaussiana:**  
  $\displaystyle \mu_A(x) = \exp\left(-\frac{1}{2}\left(\frac{x-c}{\sigma}\right)^2\right)$

- **Sigmoidal:**  
  $\displaystyle \mu_A(x) = \frac{1}{1 + e^{-a(x-c)}}$

- **Sino (bell-shaped):**  
  $\displaystyle \mu_A(x) = \frac{1}{1 + \left|\frac{x-c}{a}\right|^{2b}}$

- **S-shaped:**  
  $$
  S(x; a, b) =
  \begin{cases}
    0, & x \leq a \\
    2\left(\frac{x-a}{b-a}\right)^2, & a < x < \frac{a+b}{2} \\
    1 - 2\left(\frac{b-x}{b-a}\right)^2, & \frac{a+b}{2} \leq x < b \\
    1, & x \geq b
  \end{cases}
  $$

- **Z-shaped:**  
  $$
  Z(x; a, b) =
  \begin{cases}
    1, & x \leq a \\
    1 - 2\left(\frac{x-a}{b-a}\right)^2, & a < x < \frac{a+b}{2} \\
    2\left(\frac{b-x}{b-a}\right)^2, & \frac{a+b}{2} \leq x < b \\
    0, & x \geq b
  \end{cases}
  $$

## 2. Operações Básicas com Conjuntos Fuzzy

- **União:**  
  $\mu_{A \cup B}(x) = \max(\mu_A(x), \mu_B(x))$

- **Interseção:**  
  $\mu_{A \cap B}(x) = \min(\mu_A(x), \mu_B(x))$

- **Complemento:**  
  $\mu_{\overline{A}}(x) = 1 - \mu_A(x)$

- **$\alpha$-corte:**  
  $A_\alpha = \{x \mid \mu_A(x) \geq \alpha\}$

- **Suporte:**  
  $\text{Supp}(A) = \{x \mid \mu_A(x) > 0\}$

- **Núcleo:**  
  $\text{Core}(A) = \{x \mid \mu_A(x) = 1\}$

- **Cardinalidade:**  
  Discreto: $|A| = \sum_x \mu_A(x)$  
  Contínuo: $|A| = \int \mu_A(x) dx$

## 3. T-normas e S-normas (Operadores Generalizados)

### T-normas (Interseção Generalizada)
- **Mínimo:** $T(a, b) = \min(a, b)$
- **Produto:** $T(a, b) = a \cdot b$
- **Lukasiewicz:** $T(a, b) = \max(0, a + b - 1)$
- **Drástica:** $T(a, b) = \begin{cases} a, & b = 1 \\ b, & a = 1 \\ 0, & \text{caso contrário} \end{cases}$

### S-normas (União Generalizada)
- **Máximo:** $S(a, b) = \max(a, b)$
- **Soma Probabilística:** $S(a, b) = a + b - ab$
- **Lukasiewicz:** $S(a, b) = \min(1, a + b)$
- **Drástica:** $S(a, b) = \begin{cases} a, & b = 0 \\ b, & a = 0 \\ 1, & \text{caso contrário} \end{cases}$

### Propriedades
- **Dualidade:** $T(a, b) = 1 - S(1-a, 1-b)$
- **Idempotência:** $T(a, a) = a$, $S(a, a) = a$
- **Leis de De Morgan:**  
  $\overline{A \cup B} = \overline{A} \cap \overline{B}$  
  $\overline{A \cap B} = \overline{A} \cup \overline{B}$

## 4. Relações Fuzzy e Composição

- **Relação fuzzy:**  
  $R(U, V) = \{((x, y), \mu_R(x, y)) \mid x \in U, y \in V\}$

- **Composição Max-Min:**  
  $\mu_{R \circ S}(u, w) = \max_v [\min(\mu_R(u, v), \mu_S(v, w))]$

- **Composição Max-Produto:**  
  $\mu_{R \circ S}(u, w) = \max_v [\mu_R(u, v) \cdot \mu_S(v, w)]$

## 5. Inferência Fuzzy

### Modus Ponens Generalizado (Mamdani)
- Regra: Se $x$ é $A$ então $y$ é $B$
- Fato: $x$ é $A'$
- Conclusão: $y$ é $B'$
- Fórmula:
  $$
  \mu_{B'}(y) = \max_x \min[\mu_{A'}(x), \mu_R(x, y)]
  $$
  onde $\mu_R(x, y) = \min[\mu_A(x), \mu_B(y)]$

### Inferência com Duas Premissas
- Regra: Se $x$ é $A$ e $y$ é $B$ então $z$ é $C$
- Fato: $x$ é $A'$, $y$ é $B'$
- Grau de ativação:
  $$
  w_1 = \max_x [\mu_{A'}(x) \wedge \mu_A(x)]
  $$
  $$
  w_2 = \max_y [\mu_{B'}(y) \wedge \mu_B(y)]
  $$
- Saída fuzzy:
  $$
  \mu_{C'}(z) = (w_1 \wedge w_2) \wedge \mu_C(z)
  $$

## 6. Métodos de Inferência

### Mamdani (Max-Min)
- Implicação: mínimo
- Agregação: máximo
  $$
  B'_i(y) = w_i \wedge B_i(y)
  $$
  $$
  B'(y) = \max_i B'_i(y)
  $$

### Larsen (Max-Produto)
- Implicação: produto
- Agregação: máximo
  $$
  B'_i(y) = w_i \cdot B_i(y)
  $$
  $$
  B'(y) = \max_i B'_i(y)
  $$

### Kosko/Mizumoto (MIN + Soma)
- Implicação: mínimo
- Agregação: soma ponto a ponto
  $$
  C(z) = \sum_{i=1}^n \min(w_i, \mu_{C_i}(z))
  $$

## 7. Defuzzificação

- **Centro da Área (CoA) / Centroide:**  
  $$
  x^* = \frac{\int x \mu_A(x) dx}{\int \mu_A(x) dx}
  $$
  (discreto: $x^* = \frac{\sum x_i \mu_A(x_i)}{\sum \mu_A(x_i)}$)

- **Centro das Somas (CoS):**  
  $$
  \text{CoS}(Z) = \frac{\sum_{i=1}^n z_i \cdot \sum_{k=1}^{N'} C'_k(z_i)}{\sum_{i=1}^n \sum_{k=1}^{N'} C'_k(z_i)}
  $$

- **Mínimo/Máximo/Média dos Máximos:**  
  - Mínimo dos máximos: menor $z$ com $C(z)$ máximo.
  - Máximo dos máximos: maior $z$ com $C(z)$ máximo.
  - Média dos máximos: média dos $z$ com $C(z)$ máximo.

- **Método das Alturas:**  
  $$
  y^* = \frac{\sum_{i=1}^n w_i z_i}{\sum_{i=1}^n w_i}
  $$
  onde $z_i$ é o centro do conjunto consequente da regra $i$.

## 8. Sistemas Fuzzy Takagi-Sugeno-Kang (TSK)

- **Forma geral da regra TSK:**  
  Se $x_1$ é $A_1^i$ e ... e $x_n$ é $A_n^i$ então $y = f_i(x_1, ..., x_n)$

- **Função consequente (tipicamente linear):**  
  $$
  f_i(x_1, ..., x_n) = a_{i0} + a_{i1} x_1 + ... + a_{in} x_n
  $$

- **Grau de ativação da regra $i$:**  
* Produto (mais comum em TSK): $$ w_i = \prod_{j=1}^n \mu_{A_j^i}(x_j) $$

* Mínimo (alternativo): $$ w_i = \min_j \mu_{A_j^i}(x_j) $$
- **Saída global do sistema TSK (média ponderada):**  
  $$
  y_{saida} = \frac{\sum_{i=1}^M w_i f_i(x_1, ..., x_n)}{\sum_{i=1}^M w_i}
  $$

- **TSK de ordem zero:** $f_i(x_1, ..., x_n) = c_i$ (constante)  
- **TSK de ordem um:** $f_i(x_1, ..., x_n)$ é linear

## 9. Medidas de Similaridade, Distância e Entropia

- **Índice de Jaccard (similaridade):**  
  $$
  S(A, B) = \frac{\sum_x \min(\mu_A(x), \mu_B(x))}{\sum_x \max(\mu_A(x), \mu_B(x))}
  $$

- **Distância Minkowski/Euclidiana:**  
  $$
  d(A, B) = \left( \sum_x |\mu_A(x) - \mu_B(x)|^p \right)^{1/p}
  $$

- **Entropia Fuzzy:**  
  $$
  H(A) = -\sum_x \left[ \mu_A(x) \log \mu_A(x) + (1 - \mu_A(x)) \log (1 - \mu_A(x)) \right]
  $$

## 10. Fuzzy C-means (Clustering Fuzzy)

- **Grau de pertinência de $x_k$ ao cluster $i$:**
  $$
  u_{ik} = \frac{1}{\sum_{j=1}^c \left( \frac{\|x_k - v_i\|}{\|x_k - v_j\|} \right)^{2/(m-1)}}
  $$
  onde $v_i$ é o centro do cluster $i$, $m > 1$ é o expoente de fuzzificação.

## 11. Propriedades Importantes

- **Normalização:**  
  Um conjunto fuzzy $A$ é normalizado se $\exists x$ tal que $\mu_A(x) = 1$.

- **Convexidade:**  
  $A$ é convexo se $\mu_A(\lambda x_1 + (1-\lambda)x_2) \geq \min(\mu_A(x_1), \mu_A(x_2)), \forall \lambda \in [0,1]$.

- **Partição fuzzy:**  
  Para cada $x$ do universo, $\sum_{i=1}^n \mu_{A_i}(x) > 0$

## 12. Notação e Observações

- $\mu_A(x)$: grau de pertinência de $x$ no conjunto fuzzy $A$
- $A'$, $B'$: conjuntos fuzzy resultantes de inferência
- $T$, $S$: operadores t-norma e s-norma
- $w_i$: grau de ativação da regra $i$
- $C(z)$: conjunto fuzzy de saída
- $x^*$: valor defuzzificado (crisp)
- $f_i$: função consequente da regra TSK
