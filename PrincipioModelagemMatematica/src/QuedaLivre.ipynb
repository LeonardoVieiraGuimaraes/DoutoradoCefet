## 2. Definindo Funções Principais

Aqui serão implementadas as funções para calcular posição, velocidade e aceleração da partícula em queda livre, considerando altura inicial (H) e velocidade inicial (v0).

```python
# Implementação do modelo discreto de queda livre usando o método de Euler
# O laço calcula posição e velocidade a cada passo de tempo Δt até a pedra atingir o solo

# Parâmetros iniciais
H = 100.0      # altura inicial (m)
v0 = 0.0       # velocidade inicial (m/s)
g = 9.81       # gravidade (m/s²)
dt = 0.1       # passo de tempo (s)

# Inicialização das listas para armazenar os resultados
t = [0.0]      # tempo
x = [H]        # posição
v = [v0]       # velocidade

# Loop do método de Euler
enquanto x[-1] > 0:
    t.append(t[-1] + dt)
    # Atualiza a posição e velocidade usando as fórmulas do método de Euler
    x_novo = x[-1] + v[-1] * dt
    v_novo = v[-1] - g * dt
    x.append(x_novo)
    v.append(v_novo)

# Ajusta o último valor para não passar do solo
if x[-1] < 0:
    x[-1] = 0

# Calcula a aceleração (constante)
a = [-g for _ in t]

# Exibe os primeiros valores como exemplo
print('t (s):', t[:5])
print('x (m):', x[:5])
print('v (m/s):', v[:5])
```

```python
# Salvando dados de entrada e saída em arquivos txt

# Salva os dados de entrada
with open('entrada_queda_livre.txt', 'w') as f:
    f.write(f'H (altura inicial): {H}\n')
    f.write(f'v0 (velocidade inicial): {v0}\n')
    f.write(f'g (gravidade): {g}\n')
    f.write(f'dt (passo de tempo): {dt}\n')

# Salva os dados de saída (tempo, posição, velocidade)
with open('saida_queda_livre.txt', 'w') as f:
    f.write('tempo(s)\tposicao(m)\tvelocidade(m/s)\n')
    for ti, xi, vi in zip(t, x, v):
        f.write(f'{ti:.4f}\t{xi:.4f}\t{vi:.4f}\n')
```

## 4. Gráficos do Modelo Discreto

Abaixo, apresentamos os gráficos de posição, velocidade e aceleração em função do tempo, conforme solicitado no roteiro. Estes gráficos são gerados a partir dos dados do modelo discreto (método de Euler).


```python
# Modelo discreto com resistência do ar proporcional à velocidade
# O termo c*v é adicionado na equação da velocidade

H = 100.0
v0 = 0.0
g = 9.81
dt = 0.1
c = 0.1  # coeficiente de arrasto (ajuste para testar)

t_ar = [0.0]
x_ar = [H]
v_ar = [v0]

while x_ar[-1] > 0:
    t_ar.append(t_ar[-1] + dt)
    # Resistência do ar proporcional à velocidade
    a_ar = -g + c * v_ar[-1]
    v_novo = v_ar[-1] + a_ar * dt
    x_novo = x_ar[-1] + v_ar[-1] * dt
    v_ar.append(v_novo)
    x_ar.append(x_novo)

if x_ar[-1] < 0:
    x_ar[-1] = 0

a_ar_list = [-g + c * v for v in v_ar]

# Gráfico da velocidade mostrando a velocidade terminal
plt.figure(figsize=(8,4))
plt.plot(t_ar, v_ar, label='Com resistência do ar')
plt.plot(t, v, '--', label='Sem resistência do ar')
plt.title('Velocidade x Tempo (com e sem resistência do ar)')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (m/s)')
plt.legend()
plt.grid(True)
plt.show()
```

## 6. Bônus: Resistência do Ar Proporcional à Velocidade

Agora, vamos refinar o modelo incluindo a resistência do ar proporcional à velocidade. O coeficiente de arrasto (c) pode ser ajustado para testar diferentes cenários. Observe como a velocidade tende a se estabilizar (velocidade terminal).


```python
# Estudo do erro em função do passo de tempo (dt)
import numpy as np

H = 100.0
v0 = 0.0
g = 9.81

# Tempo exato pela fórmula analítica
t_analitico = np.sqrt(2*H/g)
print(f"Tempo exato de queda: {t_analitico:.4f} s")

for dt in [1.0, 0.1, 0.01]:
    t = [0.0]
    x = [H]
    v = [v0]
    while x[-1] > 0:
        t.append(t[-1] + dt)
        x_novo = x[-1] + v[-1] * dt
        v_novo = v[-1] - g * dt
        x.append(x_novo)
        v.append(v_novo)
    # Ajusta o último valor para não passar do solo
    if x[-1] < 0:
        x[-1] = 0
    print(f"dt = {dt:.2f} s -> tempo de queda (modelo): {t[-1]:.4f} s | erro: {abs(t[-1]-t_analitico):.4f} s")
```

## 5. Discussão: Partícula vs. Corpo Rígido

No modelo básico, consideramos a pedra como uma partícula (sem tamanho, sem rotação, apenas massa). Se fosse um corpo rígido, teríamos que considerar:
- Rotação durante a queda (energia cinética rotacional).
- Área de contato com o ar (aerodinâmica).
- Possível resistência do ar, que mudaria a aceleração.

Essas diferenças tornam o modelo mais complexo e realista, levando à necessidade de refinar as equações do movimento.

```python
# Funções para queda livre sem resistência do ar

def posicao_queda_livre(t, H, v0, g=9.81):
    """Calcula a posição em função do tempo."""
    return H + v0 * t - 0.5 * g * t**2

def velocidade_queda_livre(t, v0, g=9.81):
    """Calcula a velocidade em função do tempo."""
    return v0 - g * t

def aceleracao_queda_livre(g=9.81):
    """A aceleração é constante (gravidade)."""
    return -g
```

## 3. Implementando Algoritmo Principal

Nesta etapa, será implementado o algoritmo para simular o movimento de queda livre, utilizando as funções definidas anteriormente.

```python
# Parâmetros iniciais
H = 100.0  # altura inicial (m)
v0 = 0.0   # velocidade inicial (m/s)
g = 9.81   # gravidade (m/s^2)

t_max = np.sqrt(2*H/g)  # tempo até atingir o solo

t = np.linspace(0, t_max, 200)
pos = posicao_queda_livre(t, H, v0, g)
vel = velocidade_queda_livre(t, v0, g)
acc = aceleracao_queda_livre(g) * np.ones_like(t)
```

## 4. Executando e Testando o Código

Vamos executar o algoritmo e verificar os resultados para diferentes cenários de altura e velocidade inicial.

```python
# Testando diferentes cenários
cenarios = [
    {'H': 100, 'v0': 0},
    {'H': 50, 'v0': 10},
    {'H': 200, 'v0': -5}
]

for c in cenarios:
    t_max = np.sqrt(2*c['H']/g)
    t = np.linspace(0, t_max, 200)
    pos = posicao_queda_livre(t, c['H'], c['v0'], g)
    vel = velocidade_queda_livre(t, c['v0'], g)
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(t, pos)
    plt.title(f"Posição - H={c['H']}m, v0={c['v0']}m/s")
    plt.xlabel('Tempo (s)')
    plt.ylabel('Posição (m)')
    plt.subplot(1,2,2)
    plt.plot(t, vel)
    plt.title(f"Velocidade - H={c['H']}m, v0={c['v0']}m/s")
    plt.xlabel('Tempo (s)')
    plt.ylabel('Velocidade (m/s)')
    plt.tight_layout()
    plt.show()
```
