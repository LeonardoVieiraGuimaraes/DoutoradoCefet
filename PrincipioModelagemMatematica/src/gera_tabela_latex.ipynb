# Geração de Tabela LaTeX a partir de arquivos .txt

Este notebook lê todos os arquivos `.txt` da pasta `input`, processa os dados e gera uma tabela em formato LaTeX para ser utilizada no relatório.

## 1. Importar bibliotecas necessárias
Vamos importar as bibliotecas pandas e os para manipulação de arquivos e dados.

```python
import pandas as pd
import os
from pathlib import Path
```

## 2. Listar arquivos .txt na pasta input
Vamos listar todos os arquivos `.txt` presentes na pasta `input`.

```python
from pathlib import Path
import os

# Caminho correto para a pasta de saída dos dados (relativo ao workspace)
input_dir = Path(os.getcwd()) / 'src' / 'saida'
if not input_dir.exists():
    # Tenta caminho absoluto direto, caso esteja executando de outro diretório
    input_dir = Path('d:/GitHub/DoutoradoCefet/PrincipioModelagemMatematica/src/saida')
print('Diretório de trabalho atual:', os.getcwd())
print('Caminho usado para a pasta de saída:', input_dir)

# Listar arquivos .txt (minúsculo e maiúsculo)
arquivos_txt = list(input_dir.glob('*.txt')) + list(input_dir.glob('*.TXT'))
print('Arquivos encontrados:', arquivos_txt)

# Listar todo o conteúdo da pasta para depuração
print('Conteúdo da pasta src/saida:')
for f in input_dir.iterdir():
    print(f, '->', 'arquivo' if f.is_file() else 'pasta')
```

## 3. Ler e processar arquivos .txt
Vamos ler o conteúdo de cada arquivo `.txt` e armazenar os dados em uma lista.

```python
dados = []
nomes_arquivos = []
cabecalho = None

for arquivo in arquivos_txt:
    print(f"Lendo arquivo: {arquivo.name}")
    with open(arquivo, 'r', encoding='utf-8') as f:
        linhas = [linha.strip() for linha in f if linha.strip()]
        print("Primeiras linhas:", linhas[:3])
        if not linhas:
            print("Arquivo vazio ou só com linhas em branco.")
            continue
        # Detecta cabeçalho na primeira linha de cada arquivo
        if cabecalho is None:
            cabecalho = linhas[0].split('\t') if '\t' in linhas[0] else linhas[0].split()
        for linha in linhas[1:]:
            valores = linha.split('\t') if '\t' in linha else linha.split()
            if valores and any(valores):
                dados.append(valores)
                nomes_arquivos.append(arquivo.name)

print(f'Total de linhas lidas: {len(dados)}')
if cabecalho:
    print('Cabeçalho detectado:', cabecalho)
else:
    print('Nenhum cabeçalho detectado!')
```

## 4. Criar DataFrame com os dados
Agora vamos montar um DataFrame com os dados extraídos dos arquivos.

```python
# Adiciona coluna com o nome do arquivo de origem
if dados:
    # Ajusta o número de colunas para o cabeçalho
    dados_filtrados = [linha for linha in dados if len(linha) == len(cabecalho)]
    if len(dados_filtrados) < len(dados):
        print(f"Linhas removidas por quantidade de colunas diferente do cabeçalho: {len(dados) - len(dados_filtrados)}")
    nomes_arquivos_filtrados = [nomes_arquivos[i] for i, linha in enumerate(dados) if len(linha) == len(cabecalho)]
    df = pd.DataFrame(dados_filtrados, columns=cabecalho)
    df['arquivo_origem'] = nomes_arquivos_filtrados
    display(df)
else:
    print('Nenhum dado encontrado nos arquivos .txt.')
```

## 5. Gerar tabela em formato LaTeX
Vamos gerar a tabela em formato LaTeX usando o método `to_latex()` do pandas.

```python
# Para cada arquivo .txt, gerar um .tex correspondente na pasta relatorio/input (fora da pasta src)
from pathlib import Path

if dados:
    for arquivo_nome in set(nomes_arquivos):
        # Seleciona apenas as linhas deste arquivo
        linhas_arquivo = [linha for linha, nome in zip(dados, nomes_arquivos) if nome == arquivo_nome]
        if not linhas_arquivo:
            continue
        # Garante que só linhas com o número correto de colunas sejam usadas
        linhas_arquivo = [linha for linha in linhas_arquivo if len(linha) == len(cabecalho)]
        if not linhas_arquivo:
            continue
        df_arq = pd.DataFrame(linhas_arquivo, columns=cabecalho)
        tabela_latex = df_arq.to_latex(index=False, caption=f'Tabela gerada a partir do arquivo {arquivo_nome}', label=f'tab:{arquivo_nome.replace(".txt", "")}', escape=False)
        saida_dir = Path('../relatorio/input').resolve()
        saida_dir.mkdir(parents=True, exist_ok=True)
        caminho_saida = saida_dir / f'{arquivo_nome.replace(".txt", "")}.tex'
        with open(caminho_saida, 'w', encoding='utf-8') as f:
            f.write(tabela_latex)
        print(f'Tabela LaTeX salva em: {caminho_saida}')
else:
    print('Tabela LaTeX não foi gerada pois não há dados.')
```
