# R15 - Velocidade do Som em um Gás

## Objetivo
Determinar, por análise dimensional, como a velocidade do som $c$ em um gás se relaciona com a pressão $p$ e a densidade mássica $\rho$, e verificar o resultado numericamente para o ar atmosférico.

## Fundamentação Teórica

Assume-se que $c \propto p^a \rho^b$. As dimensões no sistema $MLT$ são:

$$[c]=LT^{-1}, \quad [p]=ML^{-1}T^{-2}, \quad [\rho]=ML^{-3}$$

Igualando expoentes:

- $M$: $a + b = 0$
- $L$: $-a - 3b = 1$
- $T$: $-2a = -1 \Rightarrow a = 1/2,\ b = -1/2$

Resultado: $c = k\sqrt{p/\rho}$

Para gás ideal adiabático: $k = \sqrt{\gamma}$, com $\gamma \approx 1{,}4$ para o ar.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

gamma = 1.4
k = np.sqrt(gamma)
p0   = 101325.0
rho0 = 1.225

c_calc = k * np.sqrt(p0 / rho0)
c_ref  = 343.0

print(f'k = sqrt(gamma) = {k:.4f}')
print(f'c calculado  = {c_calc:.2f} m/s')
print(f'c referencia = {c_ref:.2f} m/s')
print(f'Erro relativo = {abs(c_calc - c_ref)/c_ref*100:.2f}%')
```

```python
p_vec = np.linspace(50000, 200000, 300)
rhos  = [0.8, 1.225, 1.8]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for rho in rhos:
    c_vec = k * np.sqrt(p_vec / rho)
    axes[0].plot(p_vec/1000, c_vec, lw=2, label=f'rho={rho} kg/m3')
    axes[1].loglog(p_vec, c_vec, lw=2, label=f'rho={rho} kg/m3')

axes[0].axvline(p0/1000, color='gray', ls='--', alpha=0.6, label='p atm')
axes[0].set_xlabel('Pressao p (kPa)'); axes[0].set_ylabel('c (m/s)')
axes[0].set_title('c vs pressao (linear)'); axes[0].legend(); axes[0].grid(alpha=0.4)

p_ref = np.array([50000, 200000])
axes[1].loglog(p_ref, k*np.sqrt(p_ref/rho0), 'k--', alpha=0.4, label='inclinacao=0.5')
axes[1].set_xlabel('Pressao p (Pa)'); axes[1].set_ylabel('c (m/s)')
axes[1].set_title('log-log: inclinacao = 1/2'); axes[1].legend()
axes[1].grid(True, which='both', alpha=0.4)

plt.tight_layout()
plt.savefig('r15_velocidade_som.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
dados = {'Altitude (m)': [0,1000,2000,5000,10000],
         'p (kPa)'     : [101.3,89.9,79.5,54.0,26.5],
         'rho (kg/m3)' : [1.225,1.112,1.007,0.736,0.414]}
df = pd.DataFrame(dados)
df['c (m/s)'] = (k * np.sqrt(df['p (kPa)']*1e3 / df['rho (kg/m3)'])).round(1)
print(df.to_string(index=False))
```

## Conclusao

A analise dimensional determina $c = k\sqrt{p/\rho}$ com expoentes $a=1/2$ e $b=-1/2$. O grafico log-log confirma a inclinacao 1/2. Para o ar padrao, $c \approx 343$ m/s, concordando com o valor experimental. A constante $k=\sqrt{\gamma}$ so pode ser obtida por experimento ou termodinamica.
