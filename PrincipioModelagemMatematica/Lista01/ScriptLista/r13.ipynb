# R13 - Rocha em movimento circular

## Objetivo
Demonstrar que a combinação adimensional para a tensão em movimento circular é $TL/(mv^2)=\text{constante}$.

## Fundamentação Teórica
Para massa $m$ girando com velocidade $v$ em raio/corda de comprimento $L$, a tensão centrípeta é:

$T = m\frac{v^2}{L}$

Rearranjando, obtém-se o grupo adimensional:

$\frac{TL}{mv^2} = 1$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1.5
L = 1.2
v = np.linspace(1.0, 12.0, 40)
T = m * v**2 / L
pi = T * L / (m * v**2)

print('Valor medio de TL/(mv^2):', pi.mean())
print('Desvio maximo absoluto de 1:', np.max(np.abs(pi - 1)))
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4.5))
axs[0].plot(v, T, color='tab:red')
axs[0].set_xlabel('v (m/s)')
axs[0].set_ylabel('T (N)')
axs[0].set_title('Tensao cresce com v^2')
axs[0].grid(True, alpha=0.3)

axs[1].plot(v, pi, color='tab:blue')
axs[1].set_xlabel('v (m/s)')
axs[1].set_ylabel('TL/(mv^2)')
axs[1].set_title('Grupo adimensional constante')
axs[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Conclusão
A simulação confirma que $TL/(mv^2)$ é constante (igual a 1 no modelo ideal), validando a forma adimensional pedida.
