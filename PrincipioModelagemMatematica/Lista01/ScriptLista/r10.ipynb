# R10 - Confirmação da formulação revisada do pêndulo (equação 2.32)

## Objetivo
Verificar numericamente que a formulação revisada preserva os mesmos grupos adimensionais do problema do pêndulo.

## Fundamentação Teórica
Uma reformulação correta de uma lei física deve manter invariantes os grupos adimensionais essenciais.

No pêndulo, $T_0\sqrt{g/l}$ deve permanecer constante (para pequenas amplitudes), evidenciando consistência da reescrita.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 9.81
l = np.array([0.4, 0.7, 1.0, 1.3, 1.8, 2.2])

# Formula fisica de referencia para pequenas amplitudes
T0_ref = 2*np.pi*np.sqrt(l/g)
pi_ref = T0_ref*np.sqrt(g/l)

# Formula revisada equivalente: C*sqrt(l/g), com C = 2*pi
C = 2*np.pi
T0_rev = C*np.sqrt(l/g)
pi_rev = T0_rev*np.sqrt(g/l)

print('pi_ref:', np.round(pi_ref, 6))
print('pi_rev:', np.round(pi_rev, 6))
print('erro maximo:', np.max(np.abs(pi_ref - pi_rev)))
```

```python
plt.figure(figsize=(7.5, 4.5))
plt.plot(l, pi_ref, 'o-', label='pi da forma de referencia')
plt.plot(l, pi_rev, 's--', label='pi da forma revisada')
plt.xlabel('l (m)')
plt.ylabel('pi = T0*sqrt(g/l)')
plt.title('R10: consistencia da formulacao revisada')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## Conclusão
A formulação revisada reproduz exatamente os mesmos grupos adimensionais da forma de referência, confirmando que a reformulação é fisicamente consistente.
