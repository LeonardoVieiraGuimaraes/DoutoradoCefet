# R9 - Buckingham Pi no pêndulo (equação 2.30)

## Objetivo
Confirmar os grupos adimensionais clássicos do pêndulo com período, ângulo e tensão.

## Fundamentação Teórica
Uma escolha comum para os grupos adimensionais é:

$\pi_1 = T_0\sqrt{g/l}$, $\pi_2 = \theta$, $\pi_3 = T/(mg)$.

Esses grupos devem permanecer invariantes sob mudança de escala dimensional.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 9.81
l = np.linspace(0.3, 2.0, 25)
T0 = 2*np.pi*np.sqrt(l/g)
theta = 0.15
m = 1.2
Tensao = m*g*np.ones_like(l)

pi1 = T0*np.sqrt(g/l)
pi2 = theta*np.ones_like(l)
pi3 = Tensao/(m*g)

print('pi1 medio (esperado 2*pi):', pi1.mean())
print('pi2 unico valor:', np.unique(np.round(pi2, 6)))
print('pi3 unico valor:', np.unique(np.round(pi3, 6)))
```

```python
plt.figure(figsize=(7.5, 4.5))
plt.plot(l, pi1, label='pi1 = T0*sqrt(g/l)')
plt.plot(l, pi2, '--', label='pi2 = theta')
plt.plot(l, pi3, ':', label='pi3 = T/(mg)')
plt.xlabel('l (m)')
plt.ylabel('Valor dos grupos Pi')
plt.title('R9: invariancia dos grupos adimensionais')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## Conclusão
Os três grupos adimensionais permanecem constantes nas variações testadas, confirmando a estrutura da relação adimensional do pêndulo.
