# R18 - Vazao em Tubo: Q = k * d^4 * (Dp/l) / mu

## Objetivo
Determinar, pelo metodo basico de analise dimensional, como a vazao volumetrica $Q$ num tubo depende do gradiente de pressao $\Delta p/l$, do diametro $d$ e da viscosidade $\mu$.

## Fundamentacao Teorica

Assume-se $Q = k (\Delta p / l)^\alpha d^\beta \mu^\gamma$. As dimensoes no sistema $MLT$:

| Variavel | Dimensoes |
|---|---|
| $Q$ | $L^3 T^{-1}$ |
| $\Delta p/l$ | $M L^{-2} T^{-2}$ |
| $d$ | $L$ |
| $\mu$ | $M L^{-1} T^{-1}$ |

Substituindo e igualando expoentes:

- $M$: $\alpha + \gamma = 0$
- $T$: $-2\alpha - \gamma = -1$
- $L$: $-2\alpha + \beta - \gamma = 3$

Resolvendo: $\alpha = 1$, $\gamma = -1$, $\beta = 4$.

$$\boxed{Q = k \cdot \frac{d^4}{\mu} \cdot \frac{\Delta p}{l}}$$

Pela Lei de Hagen-Poiseuille (escoamento laminar): $k = \pi/128$.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Constante da Lei de Hagen-Poiseuille
k = np.pi / 128

# Parametros de referencia (agua a 20C)
mu   = 1.002e-3   # Pa.s
dpdl = 100.0      # Pa/m (gradiente de pressao)

# Variando o diametro
d_vec = np.linspace(1e-3, 20e-3, 300)  # m (1 a 20 mm)
Q_vec = k * d_vec**4 * dpdl / mu       # m^3/s

print('=== Lei de Hagen-Poiseuille ===')
print(f'k = pi/128 = {k:.6f}')
print(f'mu = {mu:.4e} Pa.s')
print(f'Dp/l = {dpdl:.1f} Pa/m')
print()
for d in [5e-3, 10e-3, 15e-3, 20e-3]:
    Q = k * d**4 * dpdl / mu
    print(f'd = {d*1000:.0f} mm -> Q = {Q*1e6:.4f} cm3/s = {Q*60000:.4f} L/min')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

dpdl_vals = [50, 100, 200]  # Pa/m

for dp in dpdl_vals:
    Qv = k * d_vec**4 * dp / mu
    axes[0].plot(d_vec*1000, Qv*1e6, lw=2, label=f'Dp/l = {dp} Pa/m')
    axes[1].loglog(d_vec*1000, Qv*1e6, lw=2, label=f'Dp/l = {dp} Pa/m')

axes[0].set_xlabel('Diametro d (mm)', fontsize=12)
axes[0].set_ylabel('Vazao Q (cm3/s)', fontsize=12)
axes[0].set_title('Q vs d (linear)', fontsize=13)
axes[0].legend(); axes[0].grid(alpha=0.4)

axes[1].set_xlabel('Diametro d (mm)', fontsize=12)
axes[1].set_ylabel('Vazao Q (cm3/s)', fontsize=12)
axes[1].set_title('log-log: inclinacao = 4', fontsize=13)
axes[1].legend(); axes[1].grid(True, which='both', alpha=0.4)

# Anotacao da inclinacao
x_ann = np.array([5, 15])
y_ann = k * (x_ann*1e-3)**4 * 100 / mu * 1e6
axes[1].loglog(x_ann, y_ann, 'k--', alpha=0.4, label='inclinacao=4')

plt.tight_layout()
plt.savefig('r18_vazao_tubo.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Tabela de verificacao para diferentes fluidos
fluidos = {
    'Fluido'    : ['Agua (20C)', 'Oleo motor (40C)', 'Glicerina (20C)'],
    'mu (Pa.s)' : [1.002e-3, 0.056, 1.490],
    'rho (kg/m3)': [998.2, 875, 1261]
}
df = pd.DataFrame(fluidos)
d_t = 10e-3   # 10 mm
dp_t = 100    # Pa/m
df['Q (cm3/s) d=10mm'] = (k * d_t**4 * dp_t / df['mu (Pa.s)'] * 1e6).round(4)
print(df.to_string(index=False))
```

## Conclusao

A analise dimensional mostra que $Q = k\,d^4(\Delta p/l)/\mu$:

- A dependencia em $d^4$ implica que duplicar o diametro aumenta a vazao por um fator **16**
- A dependencia em $\mu^{-1}$ confirma que fluidos mais viscosos tem menor vazao
- A constante $k = \pi/128$ (Lei de Hagen-Poiseuille) so pode ser determinada por experimento ou analise exata
