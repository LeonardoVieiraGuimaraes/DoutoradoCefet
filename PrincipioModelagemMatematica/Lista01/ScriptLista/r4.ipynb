# R4 - Método básico de análise dimensional no pêndulo

## Objetivo
Aplicar a análise dimensional para confirmar a forma funcional do período do pêndulo simples e verificar numericamente o resultado.

## Fundamentação Teórica
Assumimos que o período depende de comprimento e gravidade:

$T_0 = C l^a g^b$

Dimensões:
- $[T_0] = T$
- $[l] = L$
- $[g] = LT^{-2}$

Logo:
$T = L^{a+b}T^{-2b}$

Igualando expoentes:
- em $L$: $a+b=0$
- em $T$: $-2b=1 ightarrow b=-1/2$

Então $a=1/2$ e:
$T_0 = C\sqrt{l/g}$

Para pequenas amplitudes, a teoria dinâmica completa fornece $C=2\pi$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 9.81  # m/s^2
l = np.linspace(0.2, 2.0, 25)

# Forma dimensional (constante C = 1)
T_dim = np.sqrt(l / g)

# Forma física para pequenas amplitudes (C = 2*pi)
T_fis = 2 * np.pi * np.sqrt(l / g)

print('Razao T_fis / T_dim (deve ser 2*pi):')
print(np.unique(np.round(T_fis / T_dim, 6)))
```

```python
plt.figure(figsize=(7.5, 4.8))
plt.plot(l, T_dim, 'o--', label='T0 ~ sqrt(l/g) (dimensional)')
plt.plot(l, T_fis, 's-', label='T0 = 2*pi*sqrt(l/g) (física)')
plt.xlabel('Comprimento l (m)')
plt.ylabel('Periodo T0 (s)')
plt.title('R4: Periodo do pendulo simples')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Verificacao de linearidade: T0^2 proporcional a l
coef = np.polyfit(l, T_fis**2, 1)
incl, inter = coef
print(f'Inclinação de T0^2 vs l: {incl:.4f} s^2/m')
print(f'Intercepto: {inter:.4e} s^2 (proximo de zero)')
```

## Conclusão
A análise dimensional confirma que o período do pêndulo segue

$T_0 = C\sqrt{l/g}$

e a forma física para pequenas oscilações é

$T_0 = 2\pi\sqrt{l/g}$.

O gráfico e a regressão de $T_0^2$ por $l$ validam o comportamento esperado.
