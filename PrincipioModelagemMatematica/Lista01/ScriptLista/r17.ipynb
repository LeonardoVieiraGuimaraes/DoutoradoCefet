# R17 - Pendulo em Fluido Viscoso: 6 Grupos Adimensionais

## Objetivo
Determinar, pelo Teorema Pi de Buckingham, o numero de grupos adimensionais necessarios para descrever um pendulo oscilando em fluido viscoso, e derivar os 6 grupos.

## Fundamentacao Teorica

### Variaveis do problema

Alem das 6 variaveis usuais do pendulo ($l, g, m, T_0, \theta, T$), temos 3 variaveis do fluido:

| Variavel | Simbolo | Dimensoes |
|---|---|---|
| Comprimento do pendulo | $l$ | $L$ |
| Acel. gravitacional | $g$ | $LT^{-2}$ |
| Massa | $m$ | $M$ |
| Periodo | $T_0$ | $T$ |
| Angulo | $\theta$ | $1$ (adimensional) |
| Tensao | $T$ | $MLT^{-2}$ |
| Viscosidade dinamica | $\mu$ | $ML^{-1}T^{-1}$ |
| Densidade do fluido | $\rho$ | $ML^{-3}$ |
| Diametro do pendulo | $d$ | $L$ |

### Aplicacao do Teorema Pi

- $n = 9$ variaveis
- $m = 3$ dimensoes primarias ($M, L, T$)
- $p = n - m = 6$ grupos adimensionais

Variaveis de repeticao: $l, g, m$. Os 6 grupos sao:

$$\Pi_1 = \frac{T_0}{\sqrt{l/g}}, \quad \Pi_2 = \theta, \quad \Pi_3 = \frac{T}{mg}, \quad \Pi_4 = \frac{\mu}{m}\sqrt{\frac{l^3}{g}}, \quad \Pi_5 = \frac{\rho l^3}{m}, \quad \Pi_6 = \frac{d}{l}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Resumo do sistema
variaveis = [
    ('l   (comprimento)', 'L',        'variavel de repeticao'),
    ('g   (aceleracao)',  'L T^-2',   'variavel de repeticao'),
    ('m   (massa)',       'M',        'variavel de repeticao'),
    ('T0  (periodo)',     'T',        '-> Pi_1'),
    ('th  (angulo)',      '1',        '-> Pi_2'),
    ('T   (tensao)',      'M L T^-2', '-> Pi_3'),
    ('mu  (viscosidade)', 'M L^-1 T^-1', '-> Pi_4'),
    ('rho (densidade)',   'M L^-3',   '-> Pi_5'),
    ('d   (diametro)',    'L',        '-> Pi_6'),
]
df = pd.DataFrame(variaveis, columns=['Variavel', 'Dimensoes', 'Papel'])
print('=== Sistema de variaveis ===')
print(df.to_string(index=False))
print(f'\nn = {len(variaveis)}, m = 3, p = n - m = {len(variaveis) - 3}')
```

```python
# Verificacao numerica dos grupos para um pendulo de referencia
g   = 9.81    # m/s^2
l   = 1.0     # m
m   = 0.5     # kg
T0  = 2*np.pi*np.sqrt(l/g)  # periodo analitico
theta = 0.1   # rad (angulo pequeno)
T_tensao = m*g  # tensao no equilibrio
mu  = 1e-3    # Pa.s (agua)
rho = 1000.0  # kg/m^3 (agua)
d   = 0.02    # m (diametro da esfera)

Pi1 = T0 / np.sqrt(l/g)
Pi2 = theta
Pi3 = T_tensao / (m*g)
Pi4 = (mu/m) * np.sqrt(l**3/g)
Pi5 = rho * l**3 / m
Pi6 = d / l

grupos = {
    'Grupo': ['Pi_1 = T0/sqrt(l/g)', 'Pi_2 = theta',
              'Pi_3 = T/(mg)', 'Pi_4 = (mu/m)*sqrt(l^3/g)',
              'Pi_5 = rho*l^3/m', 'Pi_6 = d/l'],
    'Formula': [f'T0/sqrt(l/g)', 'theta',
                'T/(mg)', '(mu/m)*sqrt(l^3/g)',
                'rho*l^3/m', 'd/l'],
    'Valor': [Pi1, Pi2, Pi3, Pi4, Pi5, Pi6]
}
df2 = pd.DataFrame(grupos)
df2['Valor'] = df2['Valor'].round(6)
print('=== Grupos adimensionais (verificacao numerica) ===')
print(df2.to_string(index=False))
print(f'\nPi_1 = {Pi1:.6f}  (esperado: 2pi = {2*np.pi:.6f})')
```

```python
# Grafico: Pi_1 (periodo adimensional) vs comprimento l
l_vec = np.linspace(0.1, 3.0, 300)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
gs_title = ['$\\Pi_1$ = periodo adim.', '$\\Pi_4$ = viscosidade adim.', '$\\Pi_5$ = densidade adim.']

# Pi1 = T0/sqrt(l/g) = 2pi (constante para pendulo simples)
Pi1_v = 2*np.pi * np.ones_like(l_vec)
axes[0].plot(l_vec, Pi1_v, lw=2.5, color='steelblue')
axes[0].set_xlabel('Comprimento l (m)'); axes[0].set_ylabel('$\\Pi_1$')
axes[0].set_title('$\\Pi_1 = T_0/\\sqrt{l/g}$', fontsize=12)
axes[0].set_ylim([0, 8]); axes[0].grid(alpha=0.4)
axes[0].axhline(2*np.pi, color='red', ls='--', alpha=0.6, label=f'= 2pi = {2*np.pi:.3f}')
axes[0].legend()

# Pi4 = (mu/m)*sqrt(l^3/g) -- varia com l^(3/2)
mu_vals = [1e-3, 1e-2, 1e-1]
for mu_v in mu_vals:
    Pi4_v = (mu_v/m) * np.sqrt(l_vec**3/g)
    axes[1].plot(l_vec, Pi4_v, lw=2, label=f'mu={mu_v:.3f} Pa.s')
axes[1].set_xlabel('Comprimento l (m)'); axes[1].set_ylabel('$\\Pi_4$')
axes[1].set_title('$\\Pi_4 = (\\mu/m)\\sqrt{l^3/g}$', fontsize=12)
axes[1].legend(); axes[1].grid(alpha=0.4)

# Pi5 = rho*l^3/m -- varia com l^3
rhos = [800, 1000, 1200]
for rho_v in rhos:
    Pi5_v = rho_v * l_vec**3 / m
    axes[2].plot(l_vec, Pi5_v, lw=2, label=f'rho={rho_v} kg/m3')
axes[2].set_xlabel('Comprimento l (m)'); axes[2].set_ylabel('$\\Pi_5$')
axes[2].set_title('$\\Pi_5 = \\rho l^3/m$', fontsize=12)
axes[2].legend(); axes[2].grid(alpha=0.4)

plt.suptitle('Pendulo em Fluido Viscoso - Grupos Adimensionais', fontsize=14)
plt.tight_layout()
plt.savefig('r17_pendulo_viscoso.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Conclusao

Para o pendulo em fluido viscoso, com $n=9$ variaveis e $m=3$ dimensoes primarias, o Teorema Pi de Buckingham estabelece $p = n - m = 6$ grupos adimensionais:

| Grupo | Expressao | Interpretacao fisica |
|---|---|---|
| $\Pi_1$ | $T_0/\sqrt{l/g}$ | Periodo adimensional (= $2\pi$ no vacuo) |
| $\Pi_2$ | $\theta$ | Angulo de oscilacao |
| $\Pi_3$ | $T/(mg)$ | Tensao adimensional |
| $\Pi_4$ | $(\mu/m)\sqrt{l^3/g}$ | Numero de viscosidade |
| $\Pi_5$ | $\rho l^3/m$ | Razao de densidades |
| $\Pi_6$ | $d/l$ | Razao geometrica |

A viscosidade e a densidade do fluido introduzem dois novos grupos ($\Pi_4$ e $\Pi_5$) alem do diametro geometrico ($\Pi_6$), totalizando 6.
