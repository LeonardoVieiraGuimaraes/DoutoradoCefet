# Justificativa: Por que são necessários 9 gráficos diferentes, cada um contendo 3 curvas?

Neste notebook, vamos justificar, por meio de simulação e visualização, a afirmação de que são necessários nove (9) gráficos diferentes, cada um contendo três (3) curvas, para relacionar força de arrasto e velocidade no exemplo do misturador de manteiga de amendoim. Seguindo o enunciado, consideraremos as variáveis largura da lâmina ($d$), densidade do fluido ($\rho$) e viscosidade do fluido ($\mu$), além da velocidade ($V$).

## 1. Importação das Bibliotecas Necessárias

Vamos importar as bibliotecas essenciais para manipulação de dados e geração dos gráficos: numpy e matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
```

## 2. Definição dos Parâmetros do Problema

Vamos definir três valores distintos para cada parâmetro extra: largura da lâmina ($d$), densidade do fluido ($\rho$) e viscosidade do fluido ($\mu$).

```python
# Três valores para cada parâmetro
larguras_d = [0.05, 0.10, 0.15]  # metros
rhos = [900, 1000, 1100]         # kg/m³
mus = [0.5, 1.0, 1.5]           # Pa.s
```

## 3. Geração das Combinações de Parâmetros

Vamos gerar todas as combinações possíveis de ($\rho$, $\mu$) para os 9 gráficos, e associar as 3 larguras de lâmina ($d$) para as curvas de cada gráfico.

```python
# Todas as combinações de (rho, mu) para os gráficos
combinacoes_fluido = list(product(rhos, mus))
print(f'Total de gráficos (combinações de fluido): {len(combinacoes_fluido)}')
print('Exemplo de combinações (rho, mu):', combinacoes_fluido)
```

## 4. Cálculo da Força de Arrasto para Cada Combinação

Vamos definir uma função exemplo para $F_D(V; d, \rho, \mu)$ e calcular os valores de força para uma faixa de velocidades.

```python
# Função exemplo para força de arrasto
# F_D = C * d * rho * mu * V (C arbitrário para fins didáticos)
def forca_arrasto(V, d, rho, mu, C=0.02):
    return C * d * rho * mu * V

# Faixa de velocidades
velocidades = np.linspace(0, 2, 50)  # de 0 a 2 m/s
```

## 5. Plotagem dos 9 Gráficos com 3 Curvas Cada

Para cada combinação de ($\rho$, $\mu$), vamos plotar um gráfico $F_D$ vs $V$ contendo 3 curvas (uma para cada $d$), totalizando 9 gráficos.

```python
# Plotando os 9 gráficos (3x3), cada um com 3 curvas (uma para cada d)
fig, axes = plt.subplots(3, 3, figsize=(18, 12), sharex=True, sharey=True)
fig.suptitle('Força de Arrasto $F_D$ vs Velocidade $V$\nPara cada combinação de ($\rho$, $\mu$), 3 curvas para $d$', fontsize=16)

for idx, (rho, mu) in enumerate(combinacoes_fluido):
    ax = axes[idx // 3, idx % 3]
    for d in larguras_d:
        F_D = forca_arrasto(velocidades, d, rho, mu)
        ax.plot(velocidades, F_D, label=f'd = {d:.2f} m')
    ax.set_title(f'$\rho$ = {rho} kg/m³, $\mu$ = {mu} Pa.s')
    ax.set_xlabel('Velocidade $V$ (m/s)')
    ax.set_ylabel('Força de Arrasto $F_D$ (N)')
    ax.legend()
    ax.grid(True)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## 6. Demonstração da Combinatória dos Gráficos e Curvas

Como cada gráfico corresponde a uma combinação única de ($\rho$, $\mu$) e, dentro de cada gráfico, temos 3 curvas para os diferentes valores de $d$, o total de gráficos é $3 \times 3 = 9$ e cada gráfico possui 3 curvas. Isso justifica a afirmação do enunciado.

```python
# Demonstração combinatória
print(f"Total de gráficos: {len(combinacoes_fluido)} (cada um para uma combinação de (rho, mu))")
print(f"Cada gráfico possui {len(larguras_d)} curvas (uma para cada d)")
print(f"Total de curvas: {len(combinacoes_fluido) * len(larguras_d)}")
```
