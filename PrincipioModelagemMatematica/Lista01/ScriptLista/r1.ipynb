# R1 - Justificativa de 9 gráficos com 3 curvas cada

## Objetivo
Justificar, com combinatória e visualização, por que o problema do misturador exige **9 gráficos diferentes**, cada um contendo **3 curvas** para relacionar força de arrasto e velocidade.

## Fundamentação Teórica
A força de arrasto depende de quatro variáveis independentes:

$F_D = F_D(V; d, \rho, \mu)$

- Para um gráfico 2D de $F_D$ por $V$, precisamos fixar as outras variáveis.
- Se $\rho$ tem 3 valores e $\mu$ também 3, então o número de condições de fluido é $3 \times 3 = 9$.
- Em cada condição, variamos $d$ em 3 valores, gerando 3 curvas no mesmo gráfico.

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
```

```python
# Parametros com 3 valores cada
larguras_d = [0.05, 0.10, 0.15]    # m
rhos = [900, 1000, 1100]          # kg/m^3
mus = [0.5, 1.0, 1.5]             # Pa.s

# Combinacoes de fluido (rho, mu) para os graficos
combinacoes_fluido = list(product(rhos, mus))

print(f'Total de graficos: {len(combinacoes_fluido)}')
print(f'Curvas por grafico: {len(larguras_d)}')
print(f'Total de curvas no experimento: {len(combinacoes_fluido) * len(larguras_d)}')
```

```python
# Modelo didatico de forca de arrasto
def forca_arrasto(V, d, rho, mu, C=0.02):
    return C * d * rho * mu * V

velocidades = np.linspace(0.0, 2.0, 50)
```

```python
fig, axes = plt.subplots(3, 3, figsize=(16, 11), sharex=True, sharey=True)
fig.suptitle('R1: 9 graficos (3x3), cada um com 3 curvas (d1, d2, d3)', fontsize=14)

for idx, (rho, mu) in enumerate(combinacoes_fluido):
    ax = axes[idx // 3, idx % 3]

    for d in larguras_d:
        FD = forca_arrasto(velocidades, d, rho, mu)
        ax.plot(velocidades, FD, label=f'd = {d:.2f} m')

    ax.set_title(f'rho={rho} kg/m^3, mu={mu} Pa.s')
    ax.set_xlabel('Velocidade V (m/s)')
    ax.set_ylabel('Forca de arrasto FD (N)')
    ax.grid(True, alpha=0.3)

axes[0, 0].legend(loc='upper left', fontsize=8)
plt.tight_layout(rect=[0, 0.02, 1, 0.95])
plt.show()
```

## Conclusão
A justificativa fica confirmada:

- **9 gráficos**: vêm das $9$ combinações de $(\rho, \mu) = 3 \times 3$.
- **3 curvas por gráfico**: vêm dos $3$ valores de largura $d$.

Portanto, para descrever completamente a relação entre força de arrasto e velocidade nesse arranjo experimental discreto, são necessários exatamente **9 gráficos com 3 curvas cada**.
