# R19 - Confirmacao por Buckingham Pi: Vazao em Tubo

## Objetivo
Confirmar o resultado do R18 aplicando o Teorema Pi de Buckingham ao sistema de vazao em tubo, mostrando que $p = n - m = 1$ grupo e que $\Pi_1 = Q\mu / [d^4(\Delta p/l)]$ e constante.

## Fundamentacao Teorica

**Variaveis** ($n=4$): $Q,\ d,\ \mu,\ \Delta p/l$

**Dimensoes** ($m=3$): $M, L, T$

**Numero de grupos Pi**: $p = n - m = 4 - 3 = 1$

Busca-se $\Pi_1 = Q \cdot d^\alpha \cdot \mu^\beta \cdot (\Delta p/l)^\gamma$ adimensional:

$$M^0 L^0 T^0 = (L^3 T^{-1}) \cdot L^\alpha \cdot (ML^{-1}T^{-1})^\beta \cdot (ML^{-2}T^{-2})^\gamma$$

Sistema:
- $M$: $\beta + \gamma = 0$
- $T$: $-1 - \beta - 2\gamma = 0$
- $L$: $3 + \alpha - \beta - 2\gamma = 0$

Solucao: $\gamma = -1$, $\beta = 1$, $\alpha = -4$.

$$\boxed{\Pi_1 = \frac{Q\mu}{d^4 (\Delta p/l)} = \text{constante}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Constante (k = pi/128 da Lei de Hagen-Poiseuille)
k = np.pi / 128
mu   = 1.002e-3   # Pa.s (agua a 20C)

# Variacoes de d e Dp/l
print('Verificacao: Pi_1 = Q*mu / (d^4 * Dp/l) deve ser constante')
print()
resultados = []
for d in [5e-3, 10e-3, 15e-3, 20e-3]:
    for dpdl in [50, 100, 200]:
        Q = k * d**4 * dpdl / mu
        Pi1 = Q * mu / (d**4 * dpdl)
        resultados.append({'d (mm)': d*1000, 'Dp/l (Pa/m)': dpdl,
                           'Q (m3/s)': f'{Q:.3e}', 'Pi_1': f'{Pi1:.6f}'})

df = pd.DataFrame(resultados)
print(df.to_string(index=False))
print(f'\nValor esperado: Pi_1 = k = pi/128 = {k:.6f}')
```

```python
# Grafico: Pi_1 constante vs d para varios Dp/l
d_vec   = np.linspace(1e-3, 30e-3, 300)
dpdl_vals = [50, 100, 200]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for dpdl in dpdl_vals:
    Q_v  = k * d_vec**4 * dpdl / mu
    Pi1_v = Q_v * mu / (d_vec**4 * dpdl)
    axes[0].plot(d_vec*1000, Pi1_v, lw=2.5, label=f'Dp/l={dpdl} Pa/m')
    axes[1].plot(dpdl_vals.index(dpdl), np.mean(Pi1_v), 'o', ms=10)

axes[0].axhline(k, color='red', ls='--', lw=2, label=f'k = pi/128 = {k:.5f}')
axes[0].set_xlabel('Diametro d (mm)', fontsize=12)
axes[0].set_ylabel('$\\Pi_1 = Q\\mu / [d^4 (\\Delta p/l)]$', fontsize=11)
axes[0].set_title('Grupo Pi invariante vs diametro', fontsize=13)
axes[0].legend(); axes[0].grid(alpha=0.4)
axes[0].set_ylim([0, 0.05])

# Segundo grafico: Pi_1 vs Dp/l
dpdl_range = np.linspace(10, 500, 200)
d_fixed = 10e-3
Q_fixed = k * d_fixed**4 * dpdl_range / mu
Pi1_fixed = Q_fixed * mu / (d_fixed**4 * dpdl_range)

axes[1].plot(dpdl_range, Pi1_fixed, lw=2.5, color='steelblue')
axes[1].axhline(k, color='red', ls='--', lw=2, label=f'k = {k:.5f}')
axes[1].set_xlabel('Gradiente Dp/l (Pa/m)', fontsize=12)
axes[1].set_ylabel('$\\Pi_1$', fontsize=11)
axes[1].set_title('Grupo Pi invariante vs gradiente de pressao', fontsize=13)
axes[1].legend(); axes[1].grid(alpha=0.4)
axes[1].set_ylim([0, 0.05])

plt.tight_layout()
plt.savefig('r19_pi_tubo.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Conclusao

Com $n=4$ variaveis e $m=3$ dimensoes primarias, o Teorema Pi determina $p = 1$ unico grupo adimensional:

$$\Pi_1 = \frac{Q\mu}{d^4(\Delta p/l)} = \text{constante}$$

Rearranjando: $Q = \Pi_1 \cdot d^4 (\Delta p/l)/\mu$, confirmando o resultado de R18. O valor $\Pi_1 = \pi/128 \approx 0{,}02454$ (Hagen-Poiseuille) e invariante com $d$ e $\Delta p/l$ para escoamento laminar.
