# R7 - Teorema de Buckingham Pi em etapas

## Objetivo
Organizar o Teorema de Buckingham Pi como sequência operacional análoga ao método básico de análise dimensional.

## Fundamentação Teórica
Passos típicos:
1. Listar variáveis relevantes.
2. Escrever dimensões fundamentais de cada variável.
3. Contar $n$ (variáveis) e $m$ (dimensões).
4. Calcular $p=n-m$ grupos adimensionais.
5. Escolher variáveis repetidas dimensionalmente independentes.
6. Formar os grupos $\Pi$.
7. Resolver os expoentes pela condição de adimensionalidade.
8. Escrever a relação final $\Phi(\Pi_1,\Pi_2,...) = 0$.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def buckingham_count(n, m):
    if n <= m:
        raise ValueError('E necessario n > m para haver grupos independentes.')
    return n - m

casos = [
    ('Misturador', 5, 3),
    ('Pendulo com tensao', 5, 3),
    ('Escoamento em tubo', 4, 3),
    ('Pendulo viscoso', 9, 3)
]

df = pd.DataFrame(casos, columns=['Sistema', 'n', 'm'])
df['p = n-m'] = df.apply(lambda r: buckingham_count(int(r['n']), int(r['m'])), axis=1)
df
```

```python
plt.figure(figsize=(7.2, 4.2))
plt.bar(df['Sistema'], df['p = n-m'], color='#0284c7')
plt.ylabel('Numero de grupos adimensionais (p)')
plt.title('R7: contagem de grupos pelo Teorema Pi')
plt.grid(axis='y', alpha=0.3)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

## Conclusão
A formula $p=n-m$ foi aplicada em diferentes problemas, confirmando quantos grupos adimensionais independentes devem ser construídos em cada caso.
