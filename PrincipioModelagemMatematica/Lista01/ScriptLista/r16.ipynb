# R16 - Numero de Weber

## Objetivo
Derivar por analise dimensional o numero adimensional de Weber $We$, que relaciona tensao superficial, velocidade, densidade e comprimento caracteristico em mecanica dos fluidos.

## Fundamentacao Teorica

As variaveis envolvidas e suas dimensoes no sistema $MLT$:

| Variavel | Simbolo | Dimensoes |
|---|---|---|
| Tensao superficial | $\sigma$ | $M T^{-2}$ |
| Velocidade | $v$ | $L T^{-1}$ |
| Densidade | $\rho$ | $M L^{-3}$ |
| Comprimento | $l$ | $L$ |

Buscando $We = \sigma^\alpha v^\beta \rho^\gamma l^\delta$ adimensional:

$$M^0 L^0 T^0 = (MT^{-2})^\alpha (LT^{-1})^\beta (ML^{-3})^\gamma L^\delta$$

Sistema:
- $M$: $\alpha + \gamma = 0 \Rightarrow \gamma = -\alpha$
- $T$: $-2\alpha - \beta = 0 \Rightarrow \beta = -2\alpha$
- $L$: $\beta - 3\gamma + \delta = 0 \Rightarrow \delta = -\alpha$

Para $\alpha = -1$ (forma convencional):

$$We = \frac{\rho v^2 l}{\sigma}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Propriedades da agua a 20 C
rho_agua  = 998.2    # kg/m^3
sigma_agua = 0.0728  # N/m (tensao superficial)
l_ref      = 1e-3    # m (comprimento caracteristico - gota de 1mm)

v_vec = np.linspace(0.01, 5.0, 300)  # m/s
We    = rho_agua * v_vec**2 * l_ref / sigma_agua

print('Agua a 20 C:')
print(f'  rho   = {rho_agua} kg/m3')
print(f'  sigma = {sigma_agua} N/m')
print(f'  l     = {l_ref*1000:.1f} mm')
print()
for v in [0.1, 0.5, 1.0, 2.0, 5.0]:
    we = rho_agua * v**2 * l_ref / sigma_agua
    regime = 'tensao sup. domina' if we < 1 else 'inercia domina'
    print(f'  v={v:.1f} m/s -> We={we:.3f}  ({regime})')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

ls_vals = [0.5e-3, 1e-3, 3e-3]  # comprimentos caracteristicos

for l in ls_vals:
    We_l = rho_agua * v_vec**2 * l / sigma_agua
    axes[0].plot(v_vec, We_l, lw=2, label=f'l = {l*1000:.1f} mm')
    axes[1].loglog(v_vec, We_l, lw=2, label=f'l = {l*1000:.1f} mm')

axes[0].axhline(1, color='red', ls='--', lw=1.5, label='We = 1 (transicao)')
axes[0].set_xlabel('Velocidade v (m/s)', fontsize=12)
axes[0].set_ylabel('Numero de Weber We', fontsize=12)
axes[0].set_title('We vs velocidade (linear)', fontsize=13)
axes[0].legend(); axes[0].grid(alpha=0.4)

axes[1].axhline(1, color='red', ls='--', lw=1.5, label='We = 1')
axes[1].set_xlabel('Velocidade v (m/s)', fontsize=12)
axes[1].set_ylabel('Numero de Weber We', fontsize=12)
axes[1].set_title('log-log: inclinacao = 2', fontsize=13)
axes[1].legend(); axes[1].grid(True, which='both', alpha=0.4)

plt.tight_layout()
plt.savefig('r16_weber.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Comparacao entre fluidos
fluidos = {
    'Fluido'   : ['Agua (20C)', 'Mercurio (20C)', 'Etanol (20C)'],
    'rho (kg/m3)': [998.2, 13546, 789],
    'sigma (N/m)': [0.0728, 0.4866, 0.0223]
}
df = pd.DataFrame(fluidos)
v_test = 1.0   # m/s
l_test = 1e-3  # m
df['We (v=1m/s, l=1mm)'] = (df['rho (kg/m3)'] * v_test**2 * l_test / df['sigma (N/m)']).round(2)
print(df.to_string(index=False))
```

## Conclusao

A analise dimensional determina o numero de Weber:

$$We = \frac{\rho v^2 l}{\sigma}$$

Ele representa a razao entre forcas inerciais ($\rho v^2 l$) e tensao superficial ($\sigma$). Quando $We < 1$, a tensao superficial domina (gotas esfericas). Quando $We > 1$, a inercia fragmenta o fluido. O grafico log-log confirma a inclinacao 2 em relacao a $v$ (pois $We \propto v^2$).
