# R3 - Constante na equação da velocidade de queda livre

## Objetivo
Determinar e justificar a constante da relação da velocidade de queda livre a partir da altura $h$.

## Fundamentação Teórica
Pela conservação de energia mecânica (sem perdas):

$mgh = \frac{1}{2}mv^2$

Cancelando $m$ e isolando $v$:

$v = \sqrt{2gh}$

Logo, a constante adimensional é **$\sqrt{2}$**. A análise dimensional sozinha fornece a forma $v \propto \sqrt{gh}$, mas não determina esse fator numérico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 9.81  # m/s^2
h = np.array([0.5, 1, 2, 3, 5, 10], dtype=float)

v_dimensional = np.sqrt(g * h)        # forma sem constante
v_fisica = np.sqrt(2 * g * h)         # forma correta pela energia

print('Constante multiplicativa = v_fisica / v_dimensional:')
print(v_fisica / v_dimensional)
print('Valor esperado: sqrt(2) =', np.sqrt(2))
```

```python
plt.figure(figsize=(7, 4.5))
plt.plot(h, v_dimensional, 'o--', label='v ~ sqrt(g h) (sem constante)')
plt.plot(h, v_fisica, 's-', label='v = sqrt(2 g h) (física)')
plt.xlabel('Altura h (m)')
plt.ylabel('Velocidade v (m/s)')
plt.title('R3: efeito da constante sqrt(2)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## Conclusão
A constante da equação é **$\sqrt{2}$**, obtida da conservação de energia.

- Análise dimensional: define a estrutura $v \propto \sqrt{gh}$.
- Lei física adicional (energia): define a constante exata, resultando em $v = \sqrt{2gh}$.
