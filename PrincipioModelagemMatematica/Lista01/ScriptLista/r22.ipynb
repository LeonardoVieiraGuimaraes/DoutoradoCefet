# R22 - Flexibilidade de uma Viga: CEd = F(l/d)

## Objetivo
Mostrar, pelo metodo basico de analise dimensional, que a complacencia $C$ de uma viga de secao quadrada $d \times d$ satisfaz a relacao $CEd = F_{CEd}(l/d)$.

## Fundamentacao Teorica

A complacencia $C$ e a deflexao por unidade de carga. Trabalhando com dimensoes de comprimento $L$ e forca $F$:

| Variavel | Significado | Dimensoes |
|---|---|---|
| $C$ | complacencia = deflexao/carga | $L/F$ |
| $E$ | modulo de elasticidade | $F/L^2$ |
| $l$ | comprimento da viga | $L$ |
| $d$ | lado da secao quadrada | $L$ |

Multiplicando $C$ por $E$:

$$[CE] = \left(\frac{L}{F}\right) \left(\frac{F}{L^2}\right) = L^{-1}$$

Para eliminar a dimensao remanescente, multiplica-se por $d$:

$$[CEd] = L^{-1} \cdot L = 1$$

Como $l$ e $d$ sao os comprimentos geometricos restantes, o unico outro grupo adimensional e:

$$\frac{l}{d}$$

Portanto, pela homogeneidade dimensional:

$$\boxed{CEd = F_{CEd}\left(\frac{l}{d}\right)}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

print('=== Verificacao dimensional ===')
print('[C]   = L/F')
print('[E]   = F/L^2')
print('[CE]  = 1/L')
print('[CEd] = 1/L * L = 1  -> adimensional')
print('[l/d] = L/L = 1      -> adimensional')
```

```python
# Exemplo numerico usando a formula conhecida da flexao de vigas
# Para uma viga simplesmente apoiada com carga central: C = l^3/(48 E I), I = d^4/12
E = 200e9      # Pa (aco)
d = 0.02       # m
l_vec = np.linspace(0.1, 2.0, 200)
I = d**4 / 12
C_vec = l_vec**3 / (48 * E * I)
CEd = C_vec * E * d
ld = l_vec / d

# Pela teoria elastica: CEd = (1/4)*(l/d)^3
CEd_teor = 0.25 * ld**3

erro_max = np.max(np.abs(CEd - CEd_teor))
print(f'Erro maximo entre CEd numerico e (1/4)*(l/d)^3: {erro_max:.3e}')
```

```python
# Graficos
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].plot(ld, CEd, lw=2.5, label='CEd numerico')
axes[0].plot(ld, CEd_teor, '--', lw=2, label='(1/4)*(l/d)^3')
axes[0].set_xlabel('Razao geometrica l/d', fontsize=12)
axes[0].set_ylabel('CEd', fontsize=12)
axes[0].set_title('CEd como funcao de l/d', fontsize=13)
axes[0].legend(); axes[0].grid(alpha=0.4)

axes[1].loglog(ld, CEd, lw=2.5, label='CEd numerico')
axes[1].loglog(ld, CEd_teor, '--', lw=2, label='(1/4)*(l/d)^3')
axes[1].set_xlabel('Razao geometrica l/d', fontsize=12)
axes[1].set_ylabel('CEd', fontsize=12)
axes[1].set_title('Log-log: inclinacao = 3', fontsize=13)
axes[1].legend(); axes[1].grid(True, which='both', alpha=0.4)

plt.tight_layout()
plt.savefig('r22_flexibilidade_viga.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Conclusao

A analise dimensional mostra que os dois grupos adimensionais relevantes sao $CEd$ e $l/d$, portanto:

$$CEd = F_{CEd}(l/d)$$

Isso significa que a flexibilidade depende apenas da combinacao material-geometria encapsulada em $CEd$ e da esbeltez $l/d$, independentemente da escala absoluta. A teoria elastica particular fornece depois a forma especifica $CEd = (1/4)(l/d)^3$.
