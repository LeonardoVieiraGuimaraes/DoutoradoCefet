# R8 - Buckingham Pi no problema do misturador

## Objetivo
Confirmar a forma adimensional do problema do misturador a partir de $F_D=f(V,d,\rho,\mu)$.

## Fundamentação Teórica
Com 5 variáveis e 3 dimensões fundamentais, há $p=2$ grupos adimensionais. Uma escolha comum é:

$\Pi_1 = \frac{\mu}{\rho V d}$
$\Pi_2 = \frac{F_D}{\rho V^2 d^2}$

onde $\Pi_1$ é o inverso do número de Reynolds e $\Pi_2$ é um coeficiente de arrasto adimensional.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parametros base
rho = 1000.0
d = 0.10
mu = 0.80
V = np.linspace(0.2, 2.5, 40)

# Modelo didatico: F_D = C * d * rho * mu * V
C = 0.02
FD = C * d * rho * mu * V

Pi1 = mu / (rho * V * d)
Pi2 = FD / (rho * V**2 * d**2)

print('p = 5-3 = 2 grupos')
print('Pi1 (primeiros 5):', np.round(Pi1[:5], 6))
print('Pi2 (primeiros 5):', np.round(Pi2[:5], 6))
```

```python
plt.figure(figsize=(7.5, 4.8))
plt.plot(Pi1, Pi2, 'o-', color='#7c3aed')
plt.xlabel('Pi1 = mu/(rho V d)')
plt.ylabel('Pi2 = FD/(rho V^2 d^2)')
plt.title('R8: relacao entre os dois grupos adimensionais')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Conclusão
A formulação adimensional com dois grupos está confirmada para o misturador: um grupo ligado à viscosidade relativa ao transporte inercial e outro ligado ao arrasto adimensional.
