# R5 - Massa não influencia a velocidade de queda livre

## Objetivo
Demonstrar, por análise física e verificação numérica, que a velocidade de queda livre não depende da massa do corpo.

## Fundamentação Teórica
Na ausência de resistência do ar:

$mgh = \frac{1}{2}mv^2$

Cancelando $m$:

$v = \sqrt{2gh}$

Logo, para uma mesma altura $h$, corpos com massas diferentes atingem o solo com a mesma velocidade.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 9.81
h = 3.0
massas = np.array([0.05, 0.1, 0.5, 1.0, 5.0, 10.0, 50.0])
velocidades = np.sqrt(2 * g * h) * np.ones_like(massas)

print(f'Altura fixa h = {h} m')
for m, v in zip(massas, velocidades):
    print(f'm = {m:>6.2f} kg -> v = {v:>6.3f} m/s')
```

```python
plt.figure(figsize=(7.5, 4.5))
plt.plot(massas, velocidades, 'o-', color='tab:blue')
plt.xlabel('Massa m (kg)')
plt.ylabel('Velocidade de queda v (m/s)')
plt.title('R5: v independe de m para h fixa')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Verificacao numerica da variacao (deve ser ~0)
variacao = velocidades.max() - velocidades.min()
print('Variacao max-min de v:', variacao)
```

## Conclusão
A teoria e os resultados numéricos confirmam que, para queda livre ideal, a massa não altera a velocidade final.

Portanto, a relação correta é $v = \sqrt{2gh}$, sem dependência de $m$.
