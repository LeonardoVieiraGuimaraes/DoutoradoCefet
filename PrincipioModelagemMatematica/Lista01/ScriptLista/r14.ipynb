# R14 - Buckingham Pi confirmando o Problema 13

## Objetivo
Confirmar, via Teorema de Buckingham Pi, o resultado do problema da rocha em movimento circular.

## Fundamentação Teórica
Variáveis: $T, L, m, v$.

- Número de variáveis: $n=4$
- Dimensões fundamentais envolvidas: $m=3$ ($M, L, T$)
- Número de grupos: $p=n-m=1$

O único grupo independente pode ser tomado como $\Pi = TL/(mv^2)$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n, m_dim = 4, 3
p = n - m_dim
print(f'n={n}, m={m_dim}, p=n-m={p}')

m = 2.0
L = 0.9
v = np.linspace(1.0, 10.0, 40)
T = m*v**2/L
Pi = T*L/(m*v**2)
print('Pi medio:', Pi.mean())
```

```python
plt.figure(figsize=(7.5, 4.5))
plt.plot(v, Pi, 'o-', color='#16a34a')
plt.xlabel('v (m/s)')
plt.ylabel('Pi = TL/(mv^2)')
plt.title('R14: unico grupo adimensional do sistema')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Conclusão
A contagem $p=1$ e a constância de $\Pi=TL/(mv^2)$ confirmam por Buckingham Pi a análise do Problema 13.
