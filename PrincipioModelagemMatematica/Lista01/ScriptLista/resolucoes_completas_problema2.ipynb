# Problema 2 - Resolucoes Completas com Implementacao Grafica
Este notebook consolida as resolucoes dos itens 1 a 23, com calculos, verificacoes numericas e graficos quando aplicavel.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import product
plt.rcParams['figure.figsize'] = (8, 4.5)
```

## 1) Nove graficos com tres curvas

```python
rhos = [900, 1000, 1100]
mus = [0.5, 1.0, 1.5]
ds = [0.05, 0.10, 0.15]
comb = list(product(rhos, mus))
print('Graficos =', len(comb), '| Curvas por grafico =', len(ds), '| Curvas totais =', len(comb)*len(ds))
V = np.linspace(0, 2, 50)
def FD(V, d, rho, mu, C=0.02):
    return C*d*rho*mu*V
fig, axes = plt.subplots(3, 3, figsize=(14, 10), sharex=True, sharey=True)
for i, (rho, mu) in enumerate(comb):
    ax = axes[i//3, i%3]
    for d in ds:
        ax.plot(V, FD(V, d, rho, mu), label=f'd={d:.2f} m')
    ax.set_title(f'rho={rho}, mu={mu}')
    ax.grid(True)
ax.legend(loc='upper left', fontsize=8)
fig.suptitle('Problema 1: 9 graficos, 3 curvas cada', fontsize=13)
plt.tight_layout()
plt.show()
```

## 2) Densidade e viscosidade: agua, mel e manteiga de amendoim

```python
df = pd.DataFrame({
    'Material': ['Agua', 'Mel', 'Manteiga de Amendoim'],
    'Densidade (kg/m^3)': [1000, 1400, 1200],
    'Viscosidade (Pa.s)': [1e-3, 5, 250]
})
display(df)
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].bar(df['Material'], df['Densidade (kg/m^3)'], color=['steelblue','goldenrod','sienna'])
axs[0].set_title('Densidade')
axs[0].grid(axis='y')
axs[1].bar(df['Material'], df['Viscosidade (Pa.s)'], color=['steelblue','goldenrod','sienna'])
axs[1].set_yscale('log')
axs[1].set_title('Viscosidade (escala log)')
axs[1].grid(axis='y', which='both')
plt.tight_layout()
plt.show()
```

## 3) Queda livre: constante sqrt(2)

```python
g = 9.81
h = np.array([1, 2, 5, 10.0])
v = np.sqrt(2*g*h)
print('v = sqrt(2gh) -> constante adimensional = sqrt(2)')
for hi, vi in zip(h, v):
    print(f'h={hi:>4.1f} m => v={vi:>6.3f} m/s')
```

## 4) Pendulo: T0 proporcional a sqrt(l/g)

```python
l = np.linspace(0.2, 2.0, 20)
T0 = 2*np.pi*np.sqrt(l/g)
plt.plot(l, T0, 'o-')
plt.xlabel('l (m)')
plt.ylabel('T0 (s)')
plt.title('T0 = 2pi*sqrt(l/g)')
plt.grid(True)
plt.show()
```

## 5) Massa nao afeta a velocidade de queda

```python
h0 = 3.0
for m in [0.1, 1.0, 10.0, 100.0]:
    print(f'm={m:>6.2f} kg -> v={np.sqrt(2*g*h0):.4f} m/s')
```

## 6) Escolha de massa de referencia (equivalencia)

```python
m1, m2 = 2.0, 5.0
a = g*(m2-m1)/(m1+m2)
print('a =', a, 'm/s^2')
print('Razoes m1/m2 e m2/m1:', m1/m2, m2/m1)
```

## 7) Buckingham Pi: contagem de grupos

```python
casos = [('Misturador', 5, 3), ('Tubo', 4, 3), ('Pendulo viscoso', 9, 3)]
for nome, n, m in casos:
    print(f'{nome}: p = n-m = {n-m}')
```

## 8) Grupos adimensionais do misturador

```python
rho, V0, d0, mu0, Fd0 = 1000.0, 1.2, 0.1, 0.8, 15.0
Pi1 = mu0/(rho*V0*d0)
Pi2 = Fd0/(rho*V0**2*d0**2)
print('Pi1 =', Pi1)
print('Pi2 =', Pi2)
```

## 9) Pendulo com grupos pi1, pi2, pi3

```python
T0_ex, l_ex, theta_ex, T_ex, m_ex = 2.01, 1.0, 0.1, 9.81, 1.0
pi1 = T0_ex*np.sqrt(g/l_ex)
pi2 = theta_ex
pi3 = T_ex/(m_ex*g)
print('pi1, pi2, pi3 =', pi1, pi2, pi3)
```

## 10) Verificacao da formulacao revisada do pendulo

```python
l_vals = np.array([0.5, 1.0, 1.5, 2.0])
T_vals = 2*np.pi*np.sqrt(l_vals/g)
pi1_vals = T_vals*np.sqrt(g/l_vals)
print('pi1 medio:', pi1_vals.mean(), '(esperado ~ 2pi =', 2*np.pi, ')')
```

## 11) Revolucao de dois corpos com G (Kepler)

```python
G = 6.674e-11
M = 1.989e30
r = np.array([5.79e10, 1.496e11, 2.279e11])
T = 2*np.pi*np.sqrt(r**3/(G*M))
k = T**2/r**3
print('T^2/r^3:', k)
plt.plot(r, T/86400, 'o-')
plt.xlabel('raio orbital (m)')
plt.ylabel('periodo (dias)')
plt.title('Problema 11: periodo orbital')
plt.grid(True)
plt.show()
```

## 12) Caso sem G (indeterminacao dimensional)

```python
print('Sem G, falta uma variavel para conectar L e T de forma unica.')
print('Sobra razao de massas como grupo adimensional principal.')
```

## 13 e 14) Rocha em circulo e confirmacao por Pi

```python
m, v, L = 1.5, 8.0, 1.2
Tensao = m*v**2/L
pi = Tensao*L/(m*v**2)
print('TL/(mv^2)=', pi, '=> grupo unico (p=1)')
```

## 15) Velocidade do som em gas: c proporcional a sqrt(p/rho)

```python
p = np.array([8e4, 1.01e5, 1.2e5])
rho_air = 1.225
c = np.sqrt(p/rho_air)
plt.plot(p, c, 'o-')
plt.xlabel('pressao p (Pa)')
plt.ylabel('c (m/s), com k=1')
plt.title('Problema 15')
plt.grid(True)
plt.show()
```

## 16) Numero de Weber

```python
rho_w, l_w, sigma = 1000.0, 0.05, 0.072
v_range = np.linspace(0.2, 4.0, 25)
We = rho_w*v_range**2*l_w/sigma
plt.plot(v_range, We)
plt.xlabel('v (m/s)')
plt.ylabel('We')
plt.title('Problema 16: crescimento de We com v^2')
plt.grid(True)
plt.show()
```

## 17) Pendulo em fluido viscoso: numero de grupos

```python
n, m = 9, 3
print('p =', n-m, 'grupos adimensionais')
print('Exemplos: T0*sqrt(g/l), theta, T/(mg), mu/(rho*l*sqrt(g*l)), d/l, rhof/rhos')
```

## 18 e 19) Fluxo em tubo e confirmacao por Pi

```python
dp_l = 1000.0
d = np.array([0.005, 0.01, 0.02])
mu = 1e-3
k = 1/128
Q = k*(d**4)*dp_l/mu
Pi = Q*mu/(d**4*dp_l)
print('Pi (deve ser constante=k):', Pi)
plt.plot(d, Q, 'o-')
plt.xlabel('diametro d (m)')
plt.ylabel('vazao Q (m^3/s)')
plt.title('Problema 18: Q proporcional a d^4')
plt.grid(True)
plt.show()
```

## 20) Tubo com rugosidade

```python
rho, v, d, mu = 1000.0, 1.5, 0.05, 1e-3
eps, l = 1e-4, 10.0
Re = rho*v*d/mu
print('Re =', Re)
print('eps/d =', eps/d)
print('l/d =', l/d)
```

## 21) Velocidade do som no aco

```python
E, rho_aco = 200e9, 7850.0
v_aco = np.sqrt(E/rho_aco)
print('v =', v_aco, 'm/s')
```

## 22) Flexibilidade de viga: CEd = F(l/d)

```python
x = np.linspace(5, 40, 50)
CEd = x**3/48
plt.plot(x, CEd)
plt.xlabel('l/d')
plt.ylabel('CEd')
plt.title('Problema 22: forma tipica F(l/d)= (l/d)^3/48')
plt.grid(True)
plt.show()
```

## 23) Ajuste log-log e deflexao final

```python
x = np.array([8, 12, 16, 24, 32], dtype=float)
log_CEd = 3*np.log10(x) - 0.6
CEd = 10**log_CEd
coef = np.polyfit(np.log10(x), np.log10(CEd), 1)
print('inclinacao ajustada ~', coef[0], '| intercepto ~', coef[1])

P, l, E, I = 500.0, 2.0, 200e9, 8e-6
delta = P*l**3/(48*E*I)
print('Deflexao exemplo delta =', delta, 'm')

plt.plot(np.log10(x), np.log10(CEd), 'o-', label='dados')
plt.xlabel('log10(l/d)')
plt.ylabel('log10(CEd)')
plt.title('Problema 23: relacao linear em escala log-log')
plt.grid(True)
plt.legend()
plt.show()
```
