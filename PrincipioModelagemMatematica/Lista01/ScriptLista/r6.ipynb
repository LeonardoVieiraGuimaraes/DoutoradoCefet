# R6 - Escolha da massa de referência

## Objetivo
Mostrar que usar a primeira ou a segunda massa como referência na etapa final da adimensionalização gera formas equivalentes, sem alterar a física do problema.

## Fundamentação Teórica
No sistema de duas massas (por exemplo, máquina de Atwood ideal), a aceleração pode ser escrita como:

$a = g\,\frac{m_2-m_1}{m_1+m_2}$

Escolhendo razões $m_1/m_2$ ou $m_2/m_1$, apenas a forma algébrica muda. O valor físico de $a$ permanece o mesmo.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 9.81
m1 = np.linspace(0.5, 9.5, 40)
m2 = 10.0

a_ref_m2 = g * (m2 - m1) / (m1 + m2)

r = m1 / m2
a_ref_ratio = g * (1 - r) / (1 + r)

erro = np.max(np.abs(a_ref_m2 - a_ref_ratio))
print('Erro maximo entre as duas formas =', erro)
```

```python
plt.figure(figsize=(7.5, 4.5))
plt.plot(m1, a_ref_m2, label='a usando m2 como referencia')
plt.plot(m1, a_ref_ratio, '--', label='a usando razao r = m1/m2')
plt.xlabel('m1 (kg), com m2 fixo = 10 kg')
plt.ylabel('a (m/s^2)')
plt.title('R6: equivalencia entre formas adimensionais')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## Conclusão
A troca da massa de referência altera apenas a parametrização adimensional, mas não altera o resultado físico. As duas curvas coincidem numericamente.
