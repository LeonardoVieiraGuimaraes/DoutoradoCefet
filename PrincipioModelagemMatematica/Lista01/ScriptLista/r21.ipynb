# R21 - Velocidade do Som no Aco

## Objetivo
Determinar por analise dimensional como a velocidade do som no aco depende do modulo de elasticidade $E$ e da densidade $\rho$, e calcular numericamente para o aco.

## Fundamentacao Teorica

Assume-se $v = k E^\alpha \rho^\beta$. As dimensoes no sistema $MLT$:

| Variavel | Dimensoes |
|---|---|
| $v$ (velocidade) | $L T^{-1}$ |
| $E$ (modulo elast.) | $M L^{-1} T^{-2}$ |
| $\rho$ (densidade) | $M L^{-3}$ |

Expandindo: $L T^{-1} = M^{\alpha+\beta} L^{-\alpha-3\beta} T^{-2\alpha}$

Sistema:
- $M$: $\alpha + \beta = 0$
- $T$: $-2\alpha = -1 \Rightarrow \alpha = 1/2$
- $L$: $-\alpha - 3\beta = 1 \Rightarrow \beta = -1/2$ (consistente)

Resultado:
$$\boxed{v = k\sqrt{\frac{E}{\rho}}}$$

Nota: $k=1$ para ondas longitudinais em barras (modulo de Young). Para aco: $E \approx 200$ GPa, $\rho \approx 7850$ kg/m$^3$.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

k = 1.0

# Propriedades do aco
E_aco   = 200e9   # Pa (200 GPa)
rho_aco = 7850    # kg/m^3

v_aco = k * np.sqrt(E_aco / rho_aco)
print('=== Velocidade do som no aco ===')
print(f'E   = {E_aco:.0e} Pa = {E_aco/1e9:.0f} GPa')
print(f'rho = {rho_aco} kg/m^3')
print(f'v   = sqrt(E/rho) = {v_aco:.1f} m/s')
print(f'v   = {v_aco:.1f} m/s  (~{v_aco/340:.1f} x velocidade do som no ar)')

# Em unidades britanicas: E = 30e6 psi
# 1 psi = 6894.76 Pa, 1 lb/in^3 = 27679.9 kg/m^3
E_brit   = 30e6 * 6894.76   # Pa
rho_brit = 0.284 * 27679.9  # kg/m^3 (0.284 lb/in^3)
v_brit = np.sqrt(E_brit / rho_brit)
print(f'\nUnidades britanicas: E=30e6 psi -> v = {v_brit:.1f} m/s = {v_brit*39.37/12:.1f} ft/s')
```

```python
# Comparacao entre materiais
materiais = {
    'Material'   : ['Aco',   'Aluminio', 'Cobre', 'Vidro',   'Concreto'],
    'E (GPa)'    : [200,     70,         120,     70,        30],
    'rho (kg/m3)': [7850,    2700,       8900,    2500,      2400]
}
df = pd.DataFrame(materiais)
df['v = sqrt(E/rho) (m/s)'] = (np.sqrt(df['E (GPa)']*1e9 / df['rho (kg/m3)'])).round(0).astype(int)
print(df.to_string(index=False))

# Grafico
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Barras: velocity per material
axes[0].barh(df['Material'], df['v = sqrt(E/rho) (m/s)'], color='steelblue', edgecolor='k')
axes[0].axvline(343, color='red', ls='--', lw=1.5, label='v_som_ar = 343 m/s')
axes[0].set_xlabel('Velocidade do som (m/s)', fontsize=12)
axes[0].set_title('v = sqrt(E/rho) por material', fontsize=13)
axes[0].legend(); axes[0].grid(axis='x', alpha=0.4)

# Parametrico: v vs E para densidades fixas
E_vec = np.linspace(10e9, 400e9, 300)
rhos  = [2700, 7850, 11300]  # Al, aco, chumbo
for rho in rhos:
    v_v = np.sqrt(E_vec / rho)
    axes[1].plot(E_vec/1e9, v_v, lw=2, label=f'rho={rho} kg/m3')
axes[1].set_xlabel('Modulo E (GPa)', fontsize=12)
axes[1].set_ylabel('Velocidade v (m/s)', fontsize=12)
axes[1].set_title('v = sqrt(E/rho) por densidade', fontsize=13)
axes[1].legend(); axes[1].grid(alpha=0.4)

plt.tight_layout()
plt.savefig('r21_som_aco.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Conclusao

A analise dimensional determina $v = k\sqrt{E/\rho}$ com $k=1$ para ondas em solidos. Para o aco ($E = 200$ GPa, $\rho = 7850$ kg/m$^3$), obtem-se $v \approx 5048$ m/s, cerca de 15 vezes maior que no ar. O grafico comparativo mostra que materiais rigidos (alto $E$) e leves (baixo $\rho$) conduzem o som com maior velocidade.
