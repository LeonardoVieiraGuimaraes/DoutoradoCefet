# R11 - Revolução de dois corpos com constante gravitacional

## Objetivo
Aplicar a relação dimensional orbital com $G$ e verificar a lei tipo Kepler: $T^2 \propto r^3$.

## Fundamentação Teórica
Para órbita circular em torno de massa dominante $M$:

$T = 2\pi\sqrt{\frac{r^3}{GM}}$

Portanto, o quociente $T^2/r^3$ deve ser aproximadamente constante para todos os corpos que orbitam o mesmo centro de massa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = 6.674e-11
M = 1.989e30  # Sol

# Raios orbitais aproximados: Mercurio, Terra, Marte
r = np.array([5.79e10, 1.496e11, 2.279e11])
nomes = ['Mercurio', 'Terra', 'Marte']

T = 2*np.pi*np.sqrt(r**3/(G*M))
k = T**2 / r**3

for nome, Ti, ki in zip(nomes, T, k):
    print(f'{nome:9s} -> T = {Ti/86400:8.2f} dias,  T^2/r^3 = {ki:.3e}')

print('Media de T^2/r^3 =', k.mean())
```

```python
plt.figure(figsize=(7.5, 4.8))
plt.plot(r, T**2, 'o-', label='dados')
coef = np.polyfit(np.log10(r), np.log10(T**2), 1)
incl = coef[0]
plt.xscale('log')
plt.yscale('log')
plt.xlabel('r (m) [log]')
plt.ylabel('T^2 (s^2) [log]')
plt.title(f'R11: Lei tipo Kepler (inclinação ~ {incl:.3f})')
plt.grid(True, which='both', alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## Conclusão
Os resultados numéricos confirmam que $T^2/r^3$ é praticamente constante e que a inclinação em escala log-log é próxima de 3 para $T^2$ em função de $r$, em acordo com a estrutura da terceira lei de Kepler.
