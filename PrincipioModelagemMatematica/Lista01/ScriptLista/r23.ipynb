# R23 - Da Lei Experimental ao Modelo de Deflexao da Viga

## Objetivo
Usar a reta experimental $\log_{10}(CEd)$ vs $\log_{10}(l/d)$, com inclinacao 3 e intercepto -0.60, para determinar a forma explicita da complacencia $C$ e mostrar que a deflexao sob carga $P$ e $\delta = Pl^3/(48EI)$.

## Fundamentacao Teorica

Do problema anterior:

$$CEd = F_{CEd}\left(\frac{l}{d}\right)$$

Os experimentos mostraram que:

$$\log_{10}(CEd) = 3\log_{10}\left(\frac{l}{d}\right) - 0.60$$

Exponenciando:

$$CEd = 10^{-0.60} \left(\frac{l}{d}\right)^3 \approx 0.2512 \left(\frac{l}{d}\right)^3 \approx \frac{1}{4} \left(\frac{l}{d}\right)^3$$

Isolando $C$:

$$C = \frac{1}{4} \cdot \frac{l^3}{E d^4}$$

Usando o segundo momento de area da secao quadrada $I = d^4/12$:

$$C = \frac{1}{4} \cdot \frac{l^3}{E(12I)} = \frac{l^3}{48EI}$$

Logo, a deflexao sob carga $P$ e:

$$\boxed{\delta = P C = \frac{Pl^3}{48EI}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Lei experimental
incl = 3.0
b = -0.60
A = 10**b

print('=== Lei experimental ===')
print(f'log10(CEd) = {incl:.1f} log10(l/d) + ({b:.2f})')
print(f'Constante A = 10^({b:.2f}) = {A:.6f}')
print(f'Comparacao com 1/4 = {1/4:.6f}')
print(f'Erro relativo de A em relacao a 1/4 = {abs(A - 1/4)/(1/4)*100:.2f}%')

# Forma explicita da complacencia
print('\nCEd = A * (l/d)^3')
print('C = A * l^3 / (E d^4)')
print('Com A ~ 1/4 e I = d^4/12 -> C = l^3/(48EI)')
```

```python
# Verificacao numerica da deflexao
E = 200e9    # Pa
d = 0.02     # m
P = 1000.0   # N
I = d**4 / 12
l_vec = np.linspace(0.2, 2.0, 200)
ld = l_vec / d

CEd_exp = A * ld**3
C_exp = CEd_exp / (E*d)
delta_exp = P * C_exp

delta_teor = P * l_vec**3 / (48 * E * I)

erro = np.max(np.abs(delta_exp - delta_teor))
print(f'Erro maximo entre modelo experimental e teorico: {erro:.3e} m')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Grafico log-log dos dados experimentais
x = np.linspace(0.5, 2.2, 100)
y = incl * x + b
axes[0].plot(x, y, lw=2.5, color='steelblue')
axes[0].set_xlabel('log10(l/d)', fontsize=12)
axes[0].set_ylabel('log10(CEd)', fontsize=12)
axes[0].set_title('Reta experimental: inclinacao 3, intercepto -0.60', fontsize=12)
axes[0].grid(alpha=0.4)
axes[0].text(0.6, incl*0.6+b+0.2, 'y = 3x - 0.60', fontsize=11, bbox=dict(facecolor='white', alpha=0.8))

# Grafico da deflexao
axes[1].plot(l_vec, delta_exp*1000, lw=2.5, label='Modelo experimental')
axes[1].plot(l_vec, delta_teor*1000, '--', lw=2, label='Formula teorica: Pl^3/(48EI)')
axes[1].set_xlabel('Comprimento l (m)', fontsize=12)
axes[1].set_ylabel('Deflexao delta (mm)', fontsize=12)
axes[1].set_title('Deflexao da viga sob carga central', fontsize=13)
axes[1].legend(); axes[1].grid(alpha=0.4)

plt.tight_layout()
plt.savefig('r23_deflexao_viga.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Tabela para alguns comprimentos
l_vals = np.array([0.5, 1.0, 1.5, 2.0])
I_vals = d**4 / 12 * np.ones_like(l_vals)
delta_vals = P * l_vals**3 / (48 * E * I_vals)
df = pd.DataFrame({
    'l (m)': l_vals,
    'l/d': (l_vals/d).round(1),
    'I (m^4)': I_vals,
    'delta (mm)': (delta_vals*1000).round(4)
})
print(df.to_string(index=False))
```

## Conclusao

A reta experimental $\log_{10}(CEd)$ vs $\log_{10}(l/d)$ fornece:

$$CEd = 10^{-0.60} \left(\frac{l}{d}\right)^3 \approx \frac{1}{4} \left(\frac{l}{d}\right)^3$$

Substituindo $I=d^4/12$, obtem-se a forma teorica classica da flexao de vigas:

$$\boxed{\delta = \frac{Pl^3}{48EI}}$$

O ajuste experimental e a teoria elastica coincidem, mostrando que a deflexao cresce com a terceira potencia do comprimento e diminui com a rigidez flexional $EI$.
