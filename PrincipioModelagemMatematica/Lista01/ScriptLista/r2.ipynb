# R2 - Comparação de densidade e viscosidade

## Objetivo
Comparar a densidade de massa e a viscosidade da água, do mel e da manteiga de amendoim, destacando a diferença de ordem de grandeza entre as viscosidades.

## Fundamentação Teórica
- A densidade de massa ($\rho$) desses materiais é da mesma ordem de grandeza ($10^3$ kg/m$^3$).
- A viscosidade dinâmica ($\mu$) varia drasticamente: água << mel << manteiga de amendoim.

Essa diferença explica por que fluidos muito viscosos exigem maior força de mistura, mesmo quando as densidades são relativamente próximas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Valores típicos aproximados (20 °C, quando aplicável)
dados = {
    'Material': ['Água', 'Mel', 'Manteiga de Amendoim'],
    'Densidade (kg/m^3)': [1000, 1400, 1200],
    'Viscosidade (Pa.s)': [1e-3, 5, 250]
}

df = pd.DataFrame(dados)
df
```

```python
# Razões para evidenciar contrastes
mu_agua = df.loc[df['Material'] == 'Água', 'Viscosidade (Pa.s)'].iloc[0]
mu_mel = df.loc[df['Material'] == 'Mel', 'Viscosidade (Pa.s)'].iloc[0]
mu_manteiga = df.loc[df['Material'] == 'Manteiga de Amendoim', 'Viscosidade (Pa.s)'].iloc[0]

rho_agua = df.loc[df['Material'] == 'Água', 'Densidade (kg/m^3)'].iloc[0]
rho_mel = df.loc[df['Material'] == 'Mel', 'Densidade (kg/m^3)'].iloc[0]

print(f'Viscosidade do mel / água: {mu_mel / mu_agua:.0f}x')
print(f'Viscosidade da manteiga de amendoim / água: {mu_manteiga / mu_agua:.0f}x')
print(f'Densidade do mel / água: {rho_mel / rho_agua:.2f}x')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4.5))
cores = ['#3b82f6', '#f59e0b', '#b45309']

# Densidade
axs[0].bar(df['Material'], df['Densidade (kg/m^3)'], color=cores)
axs[0].set_title('Comparação de Densidade')
axs[0].set_ylabel('kg/m^3')
axs[0].grid(axis='y', alpha=0.3)

# Viscosidade em escala log para visualizar ordens de grandeza
axs[1].bar(df['Material'], df['Viscosidade (Pa.s)'], color=cores)
axs[1].set_yscale('log')
axs[1].set_title('Comparação de Viscosidade (escala log)')
axs[1].set_ylabel('Pa.s (log)')
axs[1].grid(axis='y', which='both', alpha=0.3)

plt.tight_layout()
plt.show()
```

## Conclusão
A comparação confirma a ideia central do problema:

- As **densidades** são relativamente próximas.
- As **viscosidades** diferem por várias ordens de grandeza.

Logo, para modelar força de arrasto e esforço de mistura, a viscosidade é uma variável crítica.
