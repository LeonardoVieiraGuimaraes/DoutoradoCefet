# 🌉 O Problema das Sete Pontes de Königsberg

**O início da Teoria dos Grafos**

Este é um dos problemas matemáticos mais famosos da história, resolvido por Leonhard Euler em 1736, que deu origem à Teoria dos Grafos.

---

```python
# Importando bibliotecas
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

plt.style.use('default')
%matplotlib inline
```

## 1. Contexto Histórico

### 📜 A Cidade de Königsberg (século XVIII)

A cidade de Königsberg (hoje Kaliningrado, Rússia) foi construída em uma região onde havia:
- **Dois braços do Rio Pregel**
- **Uma ilha no meio do rio**
- **Sete pontes** ligando diferentes partes da cidade

### ❓ O Problema Original

> *É possível que uma pessoa faça um percurso na cidade de tal forma que inicie e volte à mesma posição passando por todas as pontes **somente uma única vez**?*

Os habitantes da cidade tentaram resolver este problema por décadas sem sucesso.

---

## 2. A Modelagem de Euler

### 💡 Insight Genial de Euler

Euler percebeu que:
1. **Todos os "pontos" de uma dada área de terra** podem ser representados por **um único vértice**
   - Não importa onde exatamente você está em uma região
   - O que importa é **qual região** você está

2. **Um vértice é conectado a outro** se houver **uma ponte** entre as regiões

3. O problema se torna: **encontrar um caminho que use cada aresta exatamente uma vez**

```python
# Criando o grafo de Königsberg
K = nx.MultiGraph()  # MultiGraph porque temos arestas paralelas

# Vértices representam as 4 áreas de terra:
# A = margem norte
# B = ilha central
# C = margem sul
# D = região entre os dois braços do rio

K.add_nodes_from(['A', 'B', 'C', 'D'])

# Adicionando as 7 pontes (arestas)
pontes = [
    ('A', 'B'), ('A', 'B'),  # 2 pontes entre A e B
    ('A', 'D'),              # 1 ponte entre A e D
    ('B', 'C'), ('B', 'C'),  # 2 pontes entre B e C
    ('B', 'D'),              # 1 ponte entre B e D
    ('C', 'D')               # 1 ponte entre C e D
]

for ponte in pontes:
    K.add_edge(*ponte)

print("🌉 GRAFO DE KÖNIGSBERG")
print("=" * 50)
print(f"Vértices (regiões): {list(K.nodes())}")
print(f"Arestas (pontes): {K.number_of_edges()}")
print("\nGrau de cada vértice:")
for v in K.nodes():
    print(f"  grau({v}) = {K.degree(v)}")
```

```python
# Visualização do grafo de Königsberg
plt.figure(figsize=(12, 8))

# Layout manual para melhor visualização
pos = {
    'A': (0, 1),    # Norte
    'B': (0, 0),    # Ilha (centro)
    'C': (0, -1),   # Sul
    'D': (1.5, 0)   # Leste
}

# Desenhando o grafo
nx.draw_networkx_nodes(K, pos,
                       node_color='lightblue',
                       node_size=3000,
                       edgecolors='black',
                       linewidths=3)

nx.draw_networkx_labels(K, pos,
                        font_size=20,
                        font_weight='bold')

nx.draw_networkx_edges(K, pos,
                       edge_color='brown',
                       width=4,
                       connectionstyle='arc3,rad=0.1')

# Anotações
plt.text(-0.3, 1, 'Margem Norte', fontsize=10, style='italic')
plt.text(-0.3, 0, 'Ilha Central', fontsize=10, style='italic')
plt.text(-0.3, -1, 'Margem Sul', fontsize=10, style='italic')
plt.text(1.5, -0.3, 'Região Leste', fontsize=10, style='italic')

plt.title('As 7 Pontes de Königsberg\n(Modelagem de Euler)', 
         fontsize=16, fontweight='bold', pad=20)
plt.axis('off')
plt.xlim(-0.8, 2.5)
plt.ylim(-1.5, 1.5)
plt.tight_layout()
plt.show()
```

## 3. O Problema Reformulado

### 🎯 Versão em Termos de Grafos:

> *É possível achar um **caminho** que comece e termine num vértice qualquer (A, B, C, ou D) e passe por cada aresta **exatamente uma vez**?*

Ou ainda:

> *É possível **desenhar este grafo** começando e terminando na mesma posição **sem levantar o lápis do papel**?*

Este tipo de caminho é chamado de **Caminho Euleriano** ou **Ciclo Euleriano**.

---

## 4. A Solução de Euler

### 🎓 Raciocínio de Euler:

Se for possível achar uma rota que:
- Usa todas as arestas
- Começa e termina no **mesmo vértice**

Então:
- **Cada vez que chegamos** a um vértice (exceto o inicial/final), precisamos **sair dele**
- Isso significa que as arestas devem vir em **pares** (entrada + saída)
- Portanto, **o grau de cada vértice deve ser PAR**

### 📊 Análise do Grafo de Königsberg:

| Vértice | Grau | Par/Ímpar |
|---------|------|----------|
| A | 3 | Ímpar ❌ |
| B | 5 | Ímpar ❌ |
| C | 3 | Ímpar ❌ |
| D | 3 | Ímpar ❌ |

### 💡 Conclusão:

**Existem 4 vértices de grau ímpar**, portanto **NÃO é possível** resolver o problema das 7 pontes! ❌

```python
# Verificando se existe caminho Euleriano
print("🔍 ANÁLISE DO PROBLEMA DE KÖNIGSBERG")
print("=" * 60)

# Contando vértices de grau ímpar
graus = dict(K.degree())
vertices_impares = [v for v, g in graus.items() if g % 2 == 1]

print("\nGraus dos vértices:")
for v, g in graus.items():
    status = "PAR ✓" if g % 2 == 0 else "ÍMPAR ❌"
    print(f"  grau({v}) = {g} ({status})")

print(f"\nVértices de grau ímpar: {vertices_impares}")
print(f"Quantidade: {len(vertices_impares)}")

print("\n" + "=" * 60)
print("TEOREMA DE EULER:")
print("• Um grafo tem CICLO EULERIANO se e somente se")
print("  TODOS os vértices têm grau PAR.")
print("\n• Um grafo tem CAMINHO EULERIANO (não fechado) se e somente se")
print("  tem EXATAMENTE 0 ou 2 vértices de grau ÍMPAR.")
print("=" * 60)

# Verificação usando NetworkX
tem_ciclo = nx.is_eulerian(K)
tem_caminho = nx.has_eulerian_path(K)

print(f"\nVerificação NetworkX:")
print(f"  Tem ciclo Euleriano (fecha)? {tem_ciclo}")
print(f"  Tem caminho Euleriano (não fecha)? {tem_caminho}")

print("\n" + "=" * 60)
print("CONCLUSÃO:")
print("❌ NÃO é possível cruzar as 7 pontes passando por cada uma")
print("   exatamente uma vez e voltar ao ponto de partida!")
print("❌ Também não é possível fazer isso começando em um ponto")
print("   e terminando em outro!")
print("=" * 60)
```

## 5. Exemplos de Grafos Eulerianos

Vamos ver exemplos de grafos que **têm** solução Euleriana:

```python
# Exemplos de grafos Eulerianos
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.flatten()

# 1. Grafo Ciclo (todos os graus = 2, par)
C5 = nx.cycle_graph(5)
pos1 = nx.circular_layout(C5)
nx.draw(C5, pos1, ax=axes[0],
        with_labels=True, node_color='lightgreen', node_size=1200,
        font_weight='bold', edge_color='green', width=3)
graus_C5 = [C5.degree(v) for v in C5.nodes()]
axes[0].set_title(f'C₅: Graus = {graus_C5}\nTodos PAR → Euleriano ✓', 
                 fontweight='bold', color='green')
axes[0].axis('off')

# 2. Grafo Completo K4 (todos os graus = 3, ímpar)
K4 = nx.complete_graph(4)
pos2 = nx.circular_layout(K4)
nx.draw(K4, pos2, ax=axes[1],
        with_labels=True, node_color='lightcoral', node_size=1200,
        font_weight='bold', edge_color='red', width=2)
graus_K4 = [K4.degree(v) for v in K4.nodes()]
axes[1].set_title(f'K₄: Graus = {graus_K4}\nTodos ÍMPAR → Não Euleriano ❌', 
                 fontweight='bold', color='red')
axes[1].axis('off')

# 3. Cubo (todos os graus = 3, mas 8 vértices ímpares)
Cubo = nx.cubical_graph()
pos3 = nx.spring_layout(Cubo, seed=42)
nx.draw(Cubo, pos3, ax=axes[2],
        with_labels=True, node_color='lightyellow', node_size=1200,
        font_weight='bold', edge_color='orange', width=2)
graus_cubo = list(dict(Cubo.degree()).values())
axes[2].set_title(f'Cubo: Todos grau = 3 (ímpar)\n8 ímpares → Não Euleriano ❌', 
                 fontweight='bold', color='red')
axes[2].axis('off')

# 4. Grafo com exatamente 2 vértices de grau ímpar
G_caminho = nx.Graph()
G_caminho.add_edges_from([(1,2), (2,3), (3,4), (4,2), (2,5)])
pos4 = nx.spring_layout(G_caminho, seed=42)
nx.draw(G_caminho, pos4, ax=axes[3],
        with_labels=True, node_color='lavender', node_size=1200,
        font_weight='bold', edge_color='purple', width=2)
graus_G = dict(G_caminho.degree())
impares_G = [v for v, g in graus_G.items() if g % 2 == 1]
axes[3].set_title(f'Grafo: Graus = {list(graus_G.values())}\n2 ímpares {impares_G} → Tem caminho (não ciclo) ⚠️', 
                 fontweight='bold', color='orange')
axes[3].axis('off')

plt.suptitle('Exemplos: Grafos Eulerianos e Não Eulerianos', 
             fontsize=14, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()

print("Resumo:")
print("✓ C₅: Ciclo Euleriano (todos os graus pares)")
print("✓ Grafo 4: Caminho Euleriano (exatamente 2 graus ímpares)")
print("❌ K₄: Não Euleriano (4 graus ímpares)")
print("❌ Cubo: Não Euleriano (8 graus ímpares)")
```

## 6. Encontrando Caminhos Eulerianos

```python
# Exemplo: Encontrando um ciclo Euleriano no C5
C5 = nx.cycle_graph(5)

if nx.is_eulerian(C5):
    ciclo = list(nx.eulerian_circuit(C5, source=0))
    print("🎯 CICLO EULERIANO ENCONTRADO no C₅:")
    print("\nSequência de arestas:")
    for i, (u, v) in enumerate(ciclo, 1):
        print(f"  {i}. {u} → {v}")
    
    vertices_ordem = [ciclo[0][0]] + [v for u, v in ciclo]
    print(f"\nCaminho completo: {' → '.join(map(str, vertices_ordem))}")
    print("\n✓ Todas as arestas foram percorridas exatamente uma vez!")
    print("✓ O caminho começa e termina no mesmo vértice!")
else:
    print("❌ Este grafo não tem ciclo Euleriano")
```

---

## 7. E se Modificarmos Königsberg?

```python
# E se removêssemos uma ponte?
K_modificado = K.copy()
# Removendo uma ponte entre A e B
K_modificado.remove_edge('A', 'B')

print("🔧 KÖNIGSBERG MODIFICADO (removendo 1 ponte)")
print("=" * 60)
print("\nNovos graus:")
graus_mod = dict(K_modificado.degree())
for v, g in graus_mod.items():
    status = "PAR ✓" if g % 2 == 0 else "ÍMPAR ❌"
    print(f"  grau({v}) = {g} ({status})")

vertices_impares_mod = [v for v, g in graus_mod.items() if g % 2 == 1]
print(f"\nVértices de grau ímpar: {vertices_impares_mod}")
print(f"Quantidade: {len(vertices_impares_mod)}")

tem_caminho_mod = nx.has_eulerian_path(K_modificado)
tem_ciclo_mod = nx.is_eulerian(K_modificado)

print("\n" + "=" * 60)
if tem_ciclo_mod:
    print("✓ Agora TEM ciclo Euleriano (fecha)!")
elif tem_caminho_mod:
    print("⚠️ Agora TEM caminho Euleriano (não fecha)!")
    print(f"   Deve começar em {vertices_impares_mod[0]} e terminar em {vertices_impares_mod[1]}")
    print("   (ou vice-versa)")
else:
    print("❌ Ainda não tem solução Euleriana")
print("=" * 60)
```

## 8. Exercício Final

```python
# Exercício: Analisar diferentes configurações
print("🎯 EXERCÍCIO: Analise estes grafos e determine se são Eulerianos\n")

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Grafo 1
G1 = nx.Graph([(1,2), (2,3), (3,4), (4,1), (1,3)])
nx.draw(G1, nx.circular_layout(G1), ax=axes[0],
        with_labels=True, node_color='lightblue', node_size=1000,
        font_weight='bold', edge_color='blue', width=2)
axes[0].set_title('Grafo 1', fontweight='bold')
axes[0].axis('off')

# Grafo 2
G2 = nx.Graph([(1,2), (2,3), (3,1), (1,4), (4,5), (5,1)])
nx.draw(G2, nx.spring_layout(G2, seed=42), ax=axes[1],
        with_labels=True, node_color='lightgreen', node_size=1000,
        font_weight='bold', edge_color='green', width=2)
axes[1].set_title('Grafo 2', fontweight='bold')
axes[1].axis('off')

# Grafo 3
G3 = nx.Graph([(1,2), (2,3), (3,4), (4,5), (5,1)])
nx.draw(G3, nx.circular_layout(G3), ax=axes[2],
        with_labels=True, node_color='lightyellow', node_size=1000,
        font_weight='bold', edge_color='orange', width=2)
axes[2].set_title('Grafo 3', fontweight='bold')
axes[2].axis('off')

plt.tight_layout()
plt.show()

# Respostas
print("\n" + "="*60)
print("RESPOSTAS:")
print("="*60)

for i, G in enumerate([G1, G2, G3], 1):
    graus = dict(G.degree())
    impares = [v for v, g in graus.items() if g % 2 == 1]
    tem_ciclo = nx.is_eulerian(G)
    tem_caminho = nx.has_eulerian_path(G)
    
    print(f"\nGrafo {i}:")
    print(f"  Graus: {list(graus.values())}")
    print(f"  Vértices de grau ímpar: {len(impares)}")
    if tem_ciclo:
        print(f"  ✓ Tem CICLO Euleriano (todos os graus pares)")
    elif tem_caminho:
        print(f"  ⚠️ Tem CAMINHO Euleriano (2 graus ímpares: {impares})")
    else:
        print(f"  ❌ NÃO é Euleriano ({len(impares)} graus ímpares)")
```

## 📌 Resumo

### O Problema de Königsberg nos ensinou:

1. ✅ **Modelagem matemática**: Problemas do mundo real podem ser representados por grafos
2. ✅ **Ciclo Euleriano**: Existe SE E SOMENTE SE todos os vértices têm grau PAR
3. ✅ **Caminho Euleriano**: Existe SE E SOMENTE SE tem 0 ou 2 vértices de grau ÍMPAR
4. ✅ **Início da Teoria dos Grafos**: Este problema deu origem a toda uma área da matemática!

### Teorema de Euler (1736):

| Condição | Resultado |
|----------|----------|
| Todos os graus PARES | ✓ Tem ciclo Euleriano |
| Exatamente 2 graus ÍMPARES | ⚠️ Tem caminho Euleriano (não fecha) |
| Mais de 2 graus ÍMPARES | ❌ Não é Euleriano |

### 🎓 Importância Histórica:

O trabalho de Euler sobre este problema é considerado o **nascimento da Teoria dos Grafos** e da **Topologia**!
