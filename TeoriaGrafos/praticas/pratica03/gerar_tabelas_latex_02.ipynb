````python
import pandas as pd
import os
from pathlib import Path

# Configurar diretórios
DIR_RESULTADOS = './resultados_2'
DIR_TABELAS = './relatorio_02/tabelas'

# Criar diretório de tabelas se não existir
Path(DIR_TABELAS).mkdir(parents=True, exist_ok=True)

print("✓ Ambiente configurado")
print(f"  Lendo dados de: {DIR_RESULTADOS}")
print(f"  Salvando tabelas em: {DIR_TABELAS}")
````

## 1. Tabela: Parâmetros dos Grafos (Parte 1)

Informações estruturais dos grafos aleatórios gerados.

````python
# Verificar e carregar dados
csv_path = f'{DIR_RESULTADOS}/parte1/csv/parametros_grafos.csv'
print(f"Verificando: {csv_path}")

if os.path.exists(csv_path):
    df_params = pd.read_csv(csv_path)
    print(f"✓ Arquivo encontrado: {len(df_params)} linhas")
    
    # Selecionar colunas principais e renomear para nomes legíveis
    df_tabela = df_params[['id_instancia', 'num_vertices', 'num_arestas', 'densidade', 'grau_medio']].head(15)
    df_tabela.columns = ['ID', 'Vértices', 'Arestas', 'Densidade', 'Grau Médio']
    
    # Gerar LaTeX
    latex = df_tabela.to_latex(
        index=False,
        float_format="%.4f",
        caption='Parâmetros Estruturais dos Grafos (Primeiras 15 instâncias)',
        label='tab:parametros_grafos_2',
        column_format='ccccc'
    )
    
    # Adicionar \centering para centralizar tabela
    latex = latex.replace(r'\begin{table}', r'\begin{table}' + '\n\\centering')
    
    # Salvar
    with open(f'{DIR_TABELAS}/tabela_parametros_grafos_02.tex', 'w', encoding='utf-8') as f:
        f.write(latex)
    
    print("✓ Tabela salva: tabela_parametros_grafos_02.tex")
    print(df_tabela)
else:
    print(f"⚠ Arquivo não encontrado: {csv_path}")
````

## 2. Tabela: Resultados Força Bruta (Parte 1)

Número cromático exato e tempo de execução.

````python
# Verificar e carregar dados
csv_path = f'{DIR_RESULTADOS}/parte1/csv/resultados_forca_bruta.csv'
print(f"Verificando: {csv_path}")

if os.path.exists(csv_path):
    df_fb = pd.read_csv(csv_path)
    print(f"✓ Arquivo encontrado: {len(df_fb)} linhas")
    
    # Agrupar por tamanho (média)
    df_agrupado = df_fb.groupby('tamanho').agg({
        'num_cores': 'mean',
        'tempo_execucao': ['mean', 'min', 'max']
    }).reset_index()
    
    df_agrupado.columns = ['Vértices', 'χ(G) Médio', 'Tempo Médio (s)', 'Tempo Mín (s)', 'Tempo Máx (s)']
    
    # Gerar LaTeX
    latex = df_agrupado.to_latex(
        index=False,
        float_format="%.4f",
        caption='Resultados da Força Bruta por Tamanho de Grafo',
        label='tab:resultados_forca_bruta_2',
        column_format='ccccc'
    )
    
    # Substituir caractere χ por comando LaTeX \chi
    latex = latex.replace('χ', r'$\chi$')
    
    # Adicionar \centering para centralizar tabela
    latex = latex.replace(r'\begin{table}', r'\begin{table}' + '\n\\centering')
    
    # Salvar
    with open(f'{DIR_TABELAS}/tabela_resultados_forca_bruta_02.tex', 'w', encoding='utf-8') as f:
        f.write(latex)
    
    print("✓ Tabela salva: tabela_resultados_forca_bruta_02.tex")
    print(df_agrupado)
else:
    print(f"⚠ Arquivo não encontrado: {csv_path}")
````

## 3. Tabela: Informações Instâncias DIMACS (Parte 2)

Características das instâncias DIMACS processadas.

````python
# Verificar e carregar dados
csv_path = f'{DIR_RESULTADOS}/parte2/csv/informacoes_instancias.csv'
print(f"Verificando: {csv_path}")

if os.path.exists(csv_path):
    df_info = pd.read_csv(csv_path)
    print(f"✓ Arquivo encontrado: {len(df_info)} linhas")
    
    # Selecionar e renomear colunas
    df_tabela = df_info[['instancia', 'num_vertices', 'num_arestas', 'densidade', 'grau_medio']]
    df_tabela.columns = ['Instância', 'Vértices', 'Arestas', 'Densidade', 'Grau Médio']
    
    # Gerar LaTeX
    latex = df_tabela.to_latex(
        index=False,
        float_format="%.4f",
        caption='Informações das Instâncias DIMACS',
        label='tab:informacoes_dimacs_2',
        column_format='ccccc'
    )
    
    # Adicionar \centering para centralizar tabela
    latex = latex.replace(r'\begin{table}', r'\begin{table}' + '\n\\centering')
    
    # Salvar
    with open(f'{DIR_TABELAS}/tabela_informacoes_dimacs_02.tex', 'w', encoding='utf-8') as f:
        f.write(latex)
    
    print("✓ Tabela salva: tabela_informacoes_dimacs_02.tex")
    print(df_tabela)
else:
    print(f"⚠ Arquivo não encontrado: {csv_path}")
````

## 4. Tabela: Resultados Welsh-Powell (Parte 2)

Cores encontradas pela heurística Welsh-Powell.

````python
# Verificar e carregar dados
csv_path = f'{DIR_RESULTADOS}/parte2/csv/resultados_welsh_powell.csv'
print(f"Verificando: {csv_path}")

if os.path.exists(csv_path):
    df_wp = pd.read_csv(csv_path)
    print(f"✓ Arquivo encontrado: {len(df_wp)} linhas")
    
    # Selecionar colunas
    df_tabela = df_wp[['instancia_id', 'num_vertices', 'cores_heuristica', 'tempo_segundos']]
    df_tabela.columns = ['Instância', 'Vértices', 'Cores (WP)', 'Tempo (s)']
    
    # Gerar LaTeX
    latex = df_tabela.to_latex(
        index=False,
        float_format="%.4f",
        caption='Resultados da Heurística Welsh-Powell',
        label='tab:resultados_welsh_powell_2',
        column_format='cccc'
    )
    
    # Adicionar \centering para centralizar tabela
    latex = latex.replace(r'\begin{table}', r'\begin{table}' + '\n\\centering')
    
    # Salvar
    with open(f'{DIR_TABELAS}/tabela_resultados_welsh_powell_02.tex', 'w', encoding='utf-8') as f:
        f.write(latex)
    
    print("✓ Tabela salva: tabela_resultados_welsh_powell_02.tex")
    print(df_tabela)
else:
    print(f"⚠ Arquivo não encontrado: {csv_path}")
````

## 5. Tabela: Comparação dos 2 Algoritmos (Final)

Comparação consolidada: Força Bruta vs Welsh-Powell.

````python
# Verificar e carregar dados
csv_path = f'{DIR_RESULTADOS}/comparacao_algoritmos.csv'
print(f"Verificando: {csv_path}")

if os.path.exists(csv_path):
    df_comp = pd.read_csv(csv_path)
    print(f"✓ Arquivo encontrado: {len(df_comp)} linhas")
    
    # Renomear colunas para nomes mais curtos
    df_comp_rename = df_comp.copy()
    df_comp_rename.columns = ['Algoritmo', 'Inst.', 'Cores Méd.', 'Cores Mín.', 'Cores Máx.', 'Tempo (s)', 'Qualidade', 'Complexidade']
    
    # Gerar LaTeX com 8 colunas (l + 7 c's para as 8 colunas de dados)
    latex = df_comp_rename.to_latex(
        index=False,
        caption='Comparação Entre os 2 Algoritmos',
        label='tab:comparacao_algoritmos_2',
        column_format='lccccccc'
    )
    
    # Substituir caractere χ por comando LaTeX \chi
    latex = latex.replace('χ', r'$\chi$')
    
    # Adicionar [H] para forçar posição exata da tabela
    latex = latex.replace(r'\begin{table}', r'\begin{table}[H]' + '\n\\centering')
    
    # Envolver tabular em resizebox para ajustar largura
    latex = latex.replace(r'\begin{tabular}', r'\resizebox{\textwidth}{!}{\begin{tabular}')
    latex = latex.replace(r'\end{tabular}', r'\end{tabular}}')
    
    # Salvar
    with open(f'{DIR_TABELAS}/tabela_comparacao_algoritmos_02.tex', 'w', encoding='utf-8') as f:
        f.write(latex)
    
    print("✓ Tabela salva: tabela_comparacao_algoritmos_02.tex")
    print(df_comp_rename)
else:
    print(f"⚠ Arquivo não encontrado: {csv_path}")
````

## 6. Copiar Imagens para Pasta img

Copia todos os gráficos e visualizações de grafos para `relatorio_02/img/`.

````python
import shutil
from glob import glob

# Definir diretório de destino para imagens
DIR_IMG = './relatorio_02/img'
Path(DIR_IMG).mkdir(parents=True, exist_ok=True)

# Padrões de imagens a copiar
padroes_imagens = [
    f'{DIR_RESULTADOS}/parte1/graficos/*.png',
    f'{DIR_RESULTADOS}/parte1/grafos/*.png',
    f'{DIR_RESULTADOS}/parte2/graficos/*.png',
    f'{DIR_RESULTADOS}/parte2/grafos/*.png',
]

total_copiadas = 0
print("Copiando imagens para relatorio_02/img/...\n")

for padrao in padroes_imagens:
    arquivos = glob(padrao)
    if arquivos:
        # Extrair nome da categoria (parte1/graficos, parte1/grafos, etc)
        categoria = padrao.split('/')[1] + '_' + padrao.split('/')[2]
        print(f"📁 {categoria}: {len(arquivos)} arquivos")
        
        for arquivo in arquivos:
            nome_arquivo = os.path.basename(arquivo)
            # Adicionar prefixo da categoria ao nome do arquivo
            nome_destino = f"{categoria.replace('_', '-')}_{nome_arquivo}"
            destino = os.path.join(DIR_IMG, nome_destino)
            
            shutil.copy2(arquivo, destino)
            total_copiadas += 1
            print(f"  ✓ {nome_destino}")
    else:
        print(f"⚠ Nenhum arquivo encontrado em: {padrao}")

print(f"\n{'='*70}")
print(f"✅ Total de imagens copiadas: {total_copiadas}")
print(f"📁 Destino: {DIR_IMG}")
print(f"{'='*70}")
````
