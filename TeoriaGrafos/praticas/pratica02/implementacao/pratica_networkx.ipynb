## Parte I: Teste com Instância Apresentada em Aula

Nesta seção, será implementado e testado o algoritmo para o problema do Caixeiro Chinês utilizando a biblioteca NetworkX, com uma instância simples para validar a corretude da solução.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Exemplo de grafo apresentado em aula
G = nx.Graph()

# Adiciona nós e arestas com pesos (distâncias)
G.add_edge('A', 'B', weight=2)
G.add_edge('A', 'C', weight=3)
G.add_edge('B', 'C', weight=1)
G.add_edge('B', 'D', weight=4)
G.add_edge('C', 'D', weight=5)

# Visualiza o grafo
pos = nx.spring_layout(G)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw(G, pos, with_labels=True, node_color='lightblue')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.show()

# Algoritmo do Caixeiro Chinês (Chinese Postman Problem)
# Para grafos não direcionados, pode-se usar nx.eulerize para tornar o grafo euleriano
G_euler = nx.eulerize(G)

# Encontrar circuito euleriano
circuito = list(nx.eulerian_circuit(G_euler))
print('Percurso encontrado:')
for u, v in circuito:
    print(f'{u} -> {v}')
```

## Parte II: Teste com Instâncias do AVA

Nesta seção, o algoritmo será testado em 4 instâncias adicionais, conforme disponibilizadas no AVA. Cada instância representa um conjunto de cidades e as distâncias entre elas. O objetivo é validar a solução do Caixeiro Chinês em diferentes cenários.

> **Observação:** Substitua os dados das instâncias abaixo pelos dados reais fornecidos no AVA.

```python
# Exemplo de instância do AVA (substitua pelos dados reais)
G_ava = nx.Graph()

# Adiciona nós e arestas com pesos (exemplo fictício)
G_ava.add_edge('A', 'B', weight=7)
G_ava.add_edge('A', 'C', weight=9)
G_ava.add_edge('B', 'C', weight=10)
G_ava.add_edge('B', 'D', weight=15)
G_ava.add_edge('C', 'D', weight=11)
G_ava.add_edge('C', 'E', weight=2)
G_ava.add_edge('D', 'E', weight=6)

# Visualiza o grafo
pos = nx.spring_layout(G_ava)
labels = nx.get_edge_attributes(G_ava, 'weight')
nx.draw(G_ava, pos, with_labels=True, node_color='lightgreen')
nx.draw_networkx_edge_labels(G_ava, pos, edge_labels=labels)
plt.show()

# Euleriza e encontra o circuito euleriano
G_ava_euler = nx.eulerize(G_ava)
circuito_ava = list(nx.eulerian_circuit(G_ava_euler))
print('Percurso encontrado na instância do AVA:')
for u, v in circuito_ava:
    print(f'{u} -> {v}')
```
