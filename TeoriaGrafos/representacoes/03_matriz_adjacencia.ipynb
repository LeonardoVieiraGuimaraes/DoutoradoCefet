## 1. Estrutura da Matriz de Adjacência

### 📊 Conceito

Para um grafo G = (V, E) com |V| = n:
- Mantemos uma **matriz** `mat[n][n]`
- `mat[i][j] = peso` se existe aresta (i, j)
- `mat[i][j] = 0` (ou ∞) se não existe aresta
- Para grafo não ponderado: `mat[i][j] = 1` ou `0`

### 🔧 Componentes Necessários:

1. **Classe `Aresta`**: Representa uma aresta com peso
2. **Matriz `mat[][]`**: Armazena os pesos das arestas
3. **Array `pos[]`**: Para percorrer adjacências de um vértice
4. **Classe `GrafoMatriz`**: Grafo usando matriz de adjacência

**Complexidade de espaço**: O(|V|²)

```python
# Importando bibliotecas
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

plt.style.use('default')
%matplotlib inline
```

## 2. Classe Aresta (Reutilizada)

Mesma classe usada na lista de adjacência.

```python
class Aresta:
    """Representa uma aresta do grafo com peso."""
    
    def __init__(self, v1, v2, peso=1):
        self.v1 = v1
        self.v2 = v2
        self.peso = peso
    
    def __str__(self):
        return f"({self.v1} -[{self.peso}]→ {self.v2})"
    
    def __repr__(self):
        return self.__str__()

print("🔗 CLASSE ARESTA")
print("=" * 50)
print("Reutilizando a mesma classe da Lista de Adjacência")
```

---

## 3. Classe GrafoMatriz (Matriz de Adjacência)

Implementação completa do grafo usando matriz de adjacência.

```python
class GrafoMatriz:
    """Grafo implementado com matriz de adjacência."""
    
    def __init__(self, num_vertices):
        """
        Inicializa um grafo com n vértices.
        Complexidade: O(V²)
        """
        self.num_vertices = num_vertices
        # Matriz n×n inicializada com zeros
        self.mat = [[0 for _ in range(num_vertices)] for _ in range(num_vertices)]
        # Array para percorrer adjacências
        self.pos = [0] * num_vertices
    
    def insere_aresta(self, v1, v2, peso=1):
        """
        Insere uma aresta (v1, v2) no grafo.
        Para grafo não direcionado, mat[v1][v2] = mat[v2][v1] = peso
        Complexidade: O(1)
        """
        if v1 < 0 or v1 >= self.num_vertices or v2 < 0 or v2 >= self.num_vertices:
            raise ValueError(f"Vértices devem estar entre 0 e {self.num_vertices-1}")
        
        # Inserir peso na matriz (grafo não direcionado)
        self.mat[v1][v2] = peso
        self.mat[v2][v1] = peso
    
    def existe_aresta(self, v1, v2):
        """
        Verifica se existe aresta entre v1 e v2.
        Complexidade: O(1) - acesso direto à matriz!
        """
        if v1 < 0 or v1 >= self.num_vertices or v2 < 0 or v2 >= self.num_vertices:
            return False
        return self.mat[v1][v2] != 0
    
    def retira_aresta(self, v1, v2):
        """
        Remove a aresta (v1, v2) do grafo.
        Complexidade: O(1) - acesso direto!
        """
        if v1 < 0 or v1 >= self.num_vertices or v2 < 0 or v2 >= self.num_vertices:
            return
        
        self.mat[v1][v2] = 0
        self.mat[v2][v1] = 0
    
    def primeiro_lista_adj(self, v):
        """
        Retorna a primeira aresta adjacente a v.
        Inicializa o percurso: pos[v] = -1
        Complexidade: O(V) - pior caso percorre toda linha
        """
        # Inicializar percurso
        self.pos[v] = -1
        return self.prox_adj(v)
    
    def prox_adj(self, v):
        """
        Retorna a próxima aresta adjacente a v.
        Usa pos[v] para manter o estado do percurso.
        Complexidade: O(V) - pior caso
        """
        # Avançar posição
        self.pos[v] += 1
        
        # Buscar próxima adjacência válida
        while self.pos[v] < self.num_vertices:
            if self.mat[v][self.pos[v]] != 0:
                # Encontrou adjacência
                return Aresta(v, self.pos[v], self.mat[v][self.pos[v]])
            self.pos[v] += 1
        
        # Não há mais adjacências
        return None
    
    def lista_adj_vazia(self, v):
        """
        Verifica se v não tem adjacentes.
        Complexidade: O(V) - verifica linha inteira
        """
        for j in range(self.num_vertices):
            if self.mat[v][j] != 0:
                return False
        return True
    
    def grau(self, v):
        """
        Retorna o grau do vértice v.
        Complexidade: O(V) - conta adjacências na linha
        """
        return sum(1 for j in range(self.num_vertices) if self.mat[v][j] != 0)
    
    def imprime(self):
        """
        Imprime a matriz de adjacência.
        Complexidade: O(V²)
        """
        print("\n🔢 MATRIZ DE ADJACÊNCIA")
        print("=" * 60)
        
        # Cabeçalho
        print("    ", end="")
        for j in range(self.num_vertices):
            print(f"{j:4}", end="")
        print()
        
        # Linhas
        for i in range(self.num_vertices):
            print(f"{i} [ ", end="")
            for j in range(self.num_vertices):
                if self.mat[i][j] == 0:
                    print("  -", end=" ")
                else:
                    print(f"{int(self.mat[i][j]):3}", end=" ")
            print("]")
        print("=" * 60)
    
    def imprime_lista_adj(self):
        """
        Imprime no formato de lista de adjacência.
        Útil para comparação.
        """
        print("\n📋 REPRESENTAÇÃO COMO LISTA")
        print("=" * 60)
        for v in range(self.num_vertices):
            vizinhos = []
            for j in range(self.num_vertices):
                if self.mat[v][j] != 0:
                    vizinhos.append(f"{j}({int(self.mat[v][j])})")
            
            print(f"{v} (grau={self.grau(v)}): ", end="")
            if vizinhos:
                print(" → ".join(vizinhos))
            else:
                print("[]")
        print("=" * 60)

print("\n🌐 CLASSE GRAFOMATRIZ")
print("=" * 50)
print("Implementação completa da matriz de adjacência!")
```

## 4. Testando a Implementação

```python
# Criar um grafo com 5 vértices
print("\n🔨 CRIANDO GRAFO COM MATRIZ")
print("=" * 50)
g = GrafoMatriz(5)
print(f"Grafo criado com {g.num_vertices} vértices")

# Inserir arestas
print("\nInserindo arestas...")
arestas = [(0, 1, 5), (0, 2, 3), (1, 2, 2), (1, 3, 6), (2, 3, 7), (3, 4, 4)]

for v1, v2, peso in arestas:
    g.insere_aresta(v1, v2, peso)
    print(f"  ✓ Inserida aresta ({v1}, {v2}) com peso {peso}")

# Imprimir matriz
g.imprime()

# Imprimir como lista (para comparação)
g.imprime_lista_adj()

# Visualizar com NetworkX
G_nx = nx.Graph()
for v1, v2, peso in arestas:
    G_nx.add_edge(v1, v2, weight=peso)

fig, ax = plt.subplots(figsize=(10, 7), constrained_layout=True, constrained_layout=True)
pos = nx.spring_layout(G_nx, seed=42)
nx.draw(G_nx, pos,
        with_labels=True,
        node_color='lightcoral',
        node_size=1500,
        font_size=16,
        font_weight='bold',
        edge_color='gray',
        width=3, ax=ax)

edge_labels = nx.get_edge_attributes(G_nx, 'weight')
nx.draw_networkx_edge_labels(G_nx, pos, edge_labels, font_size=12, ax=ax)
ax.set_title('Grafo Implementado com Matriz de Adjacência', fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
```

## 5. Testando Operações

```python
print("\n🔍 TESTANDO OPERAÇÕES")
print("=" * 60)

# 1. Verificar adjacência (O(1) - muito rápido!)
print("\n1. Verificando adjacências (O(1)):")
testes = [(0, 1), (0, 3), (1, 4), (3, 4)]
for v1, v2 in testes:
    existe = g.existe_aresta(v1, v2)
    print(f"   Existe aresta ({v1}, {v2})? {existe} {'✓' if existe else '✗'}")

# 2. Percorrer adjacências
print("\n2. Percorrendo adjacências do vértice 1:")
aresta = g.primeiro_lista_adj(1)
vizinhos = []
while aresta is not None:
    vizinhos.append(f"{aresta.v2}(peso={aresta.peso})")
    aresta = g.prox_adj(1)
print(f"   Vizinhos: {', '.join(vizinhos)}")

# 3. Grau dos vértices
print("\n3. Grau de cada vértice:")
for v in range(g.num_vertices):
    print(f"   grau({v}) = {g.grau(v)}")

# 4. Verificar Teorema do Aperto de Mãos
soma_graus = sum(g.grau(v) for v in range(g.num_vertices))
num_arestas = len(arestas)
print(f"\n4. Teorema do Aperto de Mãos:")
print(f"   ∑ grau(v) = {soma_graus}")
print(f"   2|E| = {2 * num_arestas}")
print(f"   Válido? {soma_graus == 2 * num_arestas} ✓")

# 5. Remover aresta
print("\n5. Removendo aresta (1, 3)...")
g.retira_aresta(1, 3)
print(f"   Existe (1, 3) agora? {g.existe_aresta(1, 3)} ✗")

g.imprime()
```

## 6. Comparação: Lista vs. Matriz

Vamos comparar as duas representações lado a lado.

```python
print("\n📊 COMPARAÇÃO: LISTA vs MATRIZ")
print("=" * 80)

# Tabela de complexidades
print("\n⏱️ COMPLEXIDADE DAS OPERAÇÕES:")
print("-" * 80)
print(f"{'Operação':<30} {'Lista de Adjacência':<25} {'Matriz de Adjacência'}")
print("-" * 80)
print(f"{'Espaço':<30} {'O(V + E)':<25} {'O(V²)'}")
print(f"{'Criar grafo':<30} {'O(V)':<25} {'O(V²)'}")
print(f"{'Inserir aresta':<30} {'O(1)':<25} {'O(1)'}")
print(f"{'Remover aresta':<30} {'O(grau(v))':<25} {'O(1) ✓'}")
print(f"{'Verificar adjacência':<30} {'O(grau(v))':<25} {'O(1) ✓✓'}")
print(f"{'Listar vizinhos de v':<30} {'O(grau(v)) ✓':<25} {'O(V)'}")
print(f"{'Grau do vértice':<30} {'O(1) ✓':<25} {'O(V)'}")
print(f"{'Percorrer todas arestas':<30} {'O(V + E) ✓':<25} {'O(V²)'}")
print("-" * 80)

# Exemplo numérico
print("\n💾 EXEMPLO NUMÉRICO (V=100, E=500):")
print("-" * 80)
V, E = 100, 500
lista_espaco = V + E
matriz_espaco = V * V

print(f"Lista de Adjacência:")
print(f"  • Espaço: O({V} + {E}) = {lista_espaco} unidades")
print(f"  • Consulta adjacência: O(grau) ≈ O({E//V}) médio")

print(f"\nMatriz de Adjacência:")
print(f"  • Espaço: O({V}²) = {matriz_espaco} unidades (!)")
print(f"  • Consulta adjacência: O(1) - instantâneo!")

print(f"\n📈 Diferença de espaço: {matriz_espaco / lista_espaco:.1f}x mais memória na matriz!")
print("-" * 80)
```

## 8. Quando Usar Matriz de Adjacência?

### ✅ Use Matriz quando:

1. **Grafo DENSO**: |E| ≈ |V|² (muitas arestas)
2. **Consultas frequentes**: Precisa verificar adjacência O(1)
3. **Grafo pequeno**: |V| ≤ 1000 (memória não é problema)
4. **Algoritmos matriciais**: Floyd-Warshall, potências de matriz

### ❌ NÃO use Matriz quando:

1. **Grafo ESPARSO**: |E| ≪ |V|² (poucas arestas)
2. **Grafo grande**: |V| > 10.000 (memória explode!)
3. **Percorrer vizinhos**: O(V) é lento
4. **Algoritmos de busca**: BFS, DFS preferem lista

```python
# Exemplo: Grafo denso vs esparso
print("\n🎯 EXEMPLO: ESCOLHENDO A REPRESENTAÇÃO")
print("=" * 60)

# Caso 1: Grafo esparso (rede social)
print("\nCASO 1: Rede Social (grafo esparso)")
v1, e1 = 1_000_000, 10_000_000  # 1M usuários, 10M amizades
lista1 = v1 + e1
matriz1 = v1 ** 2
print(f"  Vértices: {v1:,}")
print(f"  Arestas: {e1:,}")
print(f"  Lista: {lista1:,} unidades (~44 MB)")
print(f"  Matriz: {matriz1:,} unidades (~4 TB!) ❌")
print(f"  ✓ USAR LISTA!")

# Caso 2: Grafo denso (torneio completo)
print("\nCASO 2: Torneio Completo (grafo denso)")
v2 = 100
e2 = v2 * (v2 - 1) // 2  # Grafo completo
lista2 = v2 + 2 * e2
matriz2 = v2 ** 2
print(f"  Vértices: {v2}")
print(f"  Arestas: {e2}")
print(f"  Lista: {lista2} unidades")
print(f"  Matriz: {matriz2} unidades")
print(f"  ✓ USAR MATRIZ! (similar em memória, O(1) nas consultas)")

# Caso 3: Algoritmo de caminhos mínimos
print("\nCASO 3: Floyd-Warshall (todos caminhos mínimos)")
print(f"  Algoritmo precisa de matriz V×V")
print(f"  ✓ USAR MATRIZ! (necessário para o algoritmo)")
```

## 9. Propriedades da Matriz de Adjacência

### 🔍 Propriedades Matemáticas Interessantes

```python
# Propriedades da matriz
print("\n🔬 PROPRIEDADES MATEMÁTICAS DA MATRIZ")
print("=" * 60)

# Criar matriz NumPy para cálculos
M = np.array(g.mat)
print("\nMatriz M:")
print(M)

# Propriedade 1: Simetria (grafo não direcionado)
print("\n1. Simetria (grafo não direcionado):")
simetrica = np.allclose(M, M.T)
print(f"   M = Mᵀ? {simetrica} ✓")

# Propriedade 2: Diagonal principal (laços)
print("\n2. Diagonal principal:")
diagonal = np.diag(M)
tem_lacos = np.any(diagonal != 0)
print(f"   Diagonal: {diagonal}")
print(f"   Tem laços? {tem_lacos}")

# Propriedade 3: Soma da linha = grau do vértice
print("\n3. Soma da linha = grau do vértice:")
for i in range(len(M)):
    soma = np.sum(M[i] != 0)  # Contar não-zeros
    grau_real = g.grau(i)
    print(f"   Linha {i}: soma = {soma}, grau({i}) = {grau_real} ✓")

# Propriedade 4: M² conta caminhos de comprimento 2
print("\n4. M² conta caminhos de comprimento 2:")
M_bin = (M > 0).astype(int)  # Matriz binária (0 ou 1)
M2 = M_bin @ M_bin  # Multiplicação matricial
print("\nM² (caminhos de comprimento 2):")
print(M2)
print(f"\nM²[0][2] = {M2[0][2]} → Existem {M2[0][2]} caminhos de comprimento 2 entre 0 e 2")
```

## 10. Exercício: Floyd-Warshall Simplificado

```python
print("\n🎯 EXERCÍCIO: Caminho Mínimo com Matriz")
print("=" * 60)

def floyd_warshall_simples(grafo):
    """
    Algoritmo Floyd-Warshall simplificado.
    Encontra distâncias mínimas entre todos os pares de vértices.
    
    Usa matriz de adjacência!
    Complexidade: O(V³)
    """
    n = grafo.num_vertices
    
    # Inicializar matriz de distâncias
    dist = [[float('inf')] * n for _ in range(n)]
    
    # Distância de um vértice para ele mesmo é 0
    for i in range(n):
        dist[i][i] = 0
    
    # Copiar arestas existentes
    for i in range(n):
        for j in range(n):
            if grafo.mat[i][j] != 0:
                dist[i][j] = grafo.mat[i][j]
    
    # Floyd-Warshall: testar caminhos através de k
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
    
    return dist

# Criar grafo ponderado
g_peso = GrafoMatriz(5)
g_peso.insere_aresta(0, 1, 4)
g_peso.insere_aresta(0, 2, 1)
g_peso.insere_aresta(1, 3, 1)
g_peso.insere_aresta(2, 1, 2)
g_peso.insere_aresta(2, 3, 5)
g_peso.insere_aresta(3, 4, 3)

print("\nGrafo:")
g_peso.imprime()

# Executar Floyd-Warshall
dist = floyd_warshall_simples(g_peso)

print("\n🔍 DISTÂNCIAS MÍNIMAS (Floyd-Warshall):")
print("    ", end="")
for j in range(5):
    print(f"{j:4}", end="")
print()

for i in range(5):
    print(f"{i} [ ", end="")
    for j in range(5):
        if dist[i][j] == float('inf'):
            print(" ∞", end=" ")
        else:
            print(f"{int(dist[i][j]):3}", end=" ")
    print("]")

print("\n✓ Floyd-Warshall usa matriz de adjacência!")
print("  Complexidade: O(V³) com matriz")
```
