## 1. Estrutura da Lista de Adjacência

### 📊 Conceito

Para um grafo G = (V, E):
- Mantemos um **array** `adj[]` de tamanho |V|
- Cada `adj[v]` contém uma **lista encadeada** dos vértices adjacentes a v
- Para cada u ∈ V, `adj[u]` contém todos os vértices v tais que (u,v) ∈ E

### 🔧 Componentes Necessários:

1. **Classe `Aresta`**: Representa uma aresta com peso
2. **Classe `Celula`**: Nó da lista encadeada
3. **Classe `Lista`**: Lista encadeada genérica
4. **Classe `GrafoLista`**: Grafo usando lista de adjacência

**Complexidade de espaço**: O(|V| + |E|)

## 2. Classe Aresta

Representa as informações de uma aresta do grafo.

```python
class Aresta:
    """Representa uma aresta do grafo com peso."""
    
    def __init__(self, v1, v2, peso=1):
        """
        Inicializa uma aresta.
        
        Args:
            v1 (int): Vértice de origem
            v2 (int): Vértice de destino
            peso (float): Peso da aresta (padrão: 1)
        """
        self.v1 = v1
        self.v2 = v2
        self.peso = peso
    
    def v1(self):
        """Retorna o primeiro vértice."""
        return self.v1
    
    def v2(self):
        """Retorna o segundo vértice."""
        return self.v2
    
    def peso(self):
        """Retorna o peso da aresta."""
        return self.peso
    
    def __str__(self):
        """Representação em string."""
        return f"({self.v1} -[{self.peso}]→ {self.v2})"
    
    def __repr__(self):
        return self.__str__()
    
    def __eq__(self, other):
        """Verifica igualdade de arestas."""
        if not isinstance(other, Aresta):
            return False
        return self.v1 == other.v1 and self.v2 == other.v2

# Teste
print("🔗 CLASSE ARESTA")
print("=" * 50)
a1 = Aresta(0, 1, 5)
a2 = Aresta(1, 2, 3)
print(f"Aresta 1: {a1}")
print(f"Aresta 2: {a2}")
print(f"a1 == a2? {a1 == a2}")
```

---

## 3. Classe Celula

Representa um nó da lista encadeada que armazena uma aresta.

```python
class Celula:
    """Célula de uma lista encadeada que armazena uma Aresta."""
    
    def __init__(self, aresta):
        """
        Inicializa uma célula.
        
        Args:
            aresta (Aresta): Aresta a ser armazenada
        """
        self.item = aresta
        self.prox = None
    
    def __str__(self):
        return f"[{self.item}]"

# Teste
print("\n📦 CLASSE CELULA")
print("=" * 50)
c1 = Celula(Aresta(0, 1, 5))
c2 = Celula(Aresta(0, 2, 3))
c1.prox = c2  # Encadear

print(f"Célula 1: {c1}")
print(f"Célula 2 (próxima): {c1.prox}")
```

## 4. Classe Lista (Encadeada Genérica)

Lista encadeada para armazenar as arestas adjacentes a cada vértice.

```python
class Lista:
    """Lista encadeada genérica para armazenar arestas."""
    
    def __init__(self):
        """Cria uma lista vazia."""
        self.primeiro = None
        self.ultimo = None
        self.tamanho = 0
    
    def insere(self, aresta):
        """
        Insere uma aresta no final da lista.
        Complexidade: O(1)
        """
        nova = Celula(aresta)
        
        if self.primeiro is None:
            self.primeiro = nova
            self.ultimo = nova
        else:
            self.ultimo.prox = nova
            self.ultimo = nova
        
        self.tamanho += 1
    
    def retira(self, aresta):
        """
        Remove uma aresta da lista.
        Complexidade: O(n) onde n é o tamanho da lista
        """
        if self.primeiro is None:
            return None
        
        # Caso especial: primeiro elemento
        if self.primeiro.item == aresta:
            item = self.primeiro.item
            self.primeiro = self.primeiro.prox
            if self.primeiro is None:
                self.ultimo = None
            self.tamanho -= 1
            return item
        
        # Buscar na lista
        anterior = self.primeiro
        atual = self.primeiro.prox
        
        while atual is not None:
            if atual.item == aresta:
                item = atual.item
                anterior.prox = atual.prox
                if atual == self.ultimo:
                    self.ultimo = anterior
                self.tamanho -= 1
                return item
            anterior = atual
            atual = atual.prox
        
        return None
    
    def vazia(self):
        """Verifica se a lista está vazia. Complexidade: O(1)"""
        return self.primeiro is None
    
    def primeiro_item(self):
        """Retorna o primeiro item sem remover. Complexidade: O(1)"""
        if self.primeiro is None:
            return None
        return self.primeiro.item
    
    def imprime(self):
        """Imprime todos os itens da lista. Complexidade: O(n)"""
        if self.vazia():
            print("[]")
            return
        
        items = []
        atual = self.primeiro
        while atual is not None:
            items.append(str(atual.item))
            atual = atual.prox
        print(" → ".join(items))
    
    def __len__(self):
        return self.tamanho
    
    def __iter__(self):
        """Permite iterar sobre a lista."""
        atual = self.primeiro
        while atual is not None:
            yield atual.item
            atual = atual.prox

# Teste
print("\n📚 CLASSE LISTA")
print("=" * 50)
lista = Lista()
print(f"Lista vazia? {lista.vazia()}")

lista.insere(Aresta(0, 1, 5))
lista.insere(Aresta(0, 2, 3))
lista.insere(Aresta(0, 3, 7))

print(f"\nApós inserções (tamanho={len(lista)}):")
lista.imprime()

print(f"\nPrimeiro item: {lista.primeiro_item()}")

lista.retira(Aresta(0, 2, 3))
print(f"\nApós remover (0,2):")
lista.imprime()
```

## 5. Classe GrafoLista (Lista de Adjacência)

Implementação completa do grafo usando lista de adjacência.

```python
class GrafoLista:
    """Grafo implementado com lista de adjacência."""
    
    def __init__(self, num_vertices):
        """
        Inicializa um grafo com n vértices.
        Complexidade: O(V)
        """
        self.num_vertices = num_vertices
        self.adj = [Lista() for _ in range(num_vertices)]  # Array de listas
    
    def insere_aresta(self, v1, v2, peso=1):
        """
        Insere uma aresta (v1, v2) no grafo.
        Para grafo não direcionado, insere em ambas as direções.
        Complexidade: O(1)
        """
        if v1 < 0 or v1 >= self.num_vertices or v2 < 0 or v2 >= self.num_vertices:
            raise ValueError(f"Vértices devem estar entre 0 e {self.num_vertices-1}")
        
        # Inserir v2 na lista de adjacência de v1
        self.adj[v1].insere(Aresta(v1, v2, peso))
        
        # Para grafo não direcionado, inserir v1 na lista de v2
        if v1 != v2:  # Evitar duplicar laços
            self.adj[v2].insere(Aresta(v2, v1, peso))
    
    def existe_aresta(self, v1, v2):
        """
        Verifica se existe aresta entre v1 e v2.
        Complexidade: O(grau(v1))
        """
        if v1 < 0 or v1 >= self.num_vertices:
            return False
        
        # Percorrer lista de adjacência de v1
        for aresta in self.adj[v1]:
            if aresta.v2 == v2:
                return True
        return False
    
    def lista_adj_vazia(self, v):
        """
        Verifica se a lista de adjacência de v está vazia.
        Complexidade: O(1)
        """
        return self.adj[v].vazia()
    
    def primeiro_lista_adj(self, v):
        """
        Retorna a primeira aresta na lista de adjacência de v.
        Complexidade: O(1)
        """
        return self.adj[v].primeiro_item()
    
    def prox_adj(self, v, aresta_atual):
        """
        Retorna a próxima aresta na lista de adjacência de v.
        Complexidade: O(grau(v))
        """
        encontrou = False
        for aresta in self.adj[v]:
            if encontrou:
                return aresta
            if aresta == aresta_atual:
                encontrou = True
        return None
    
    def retira_aresta(self, v1, v2):
        """
        Remove a aresta (v1, v2) do grafo.
        Complexidade: O(grau(v1) + grau(v2))
        """
        # Remover v2 da lista de v1
        aresta_alvo = Aresta(v1, v2)
        self.adj[v1].retira(aresta_alvo)
        
        # Remover v1 da lista de v2 (grafo não direcionado)
        if v1 != v2:
            aresta_alvo_inv = Aresta(v2, v1)
            self.adj[v2].retira(aresta_alvo_inv)
    
    def grau(self, v):
        """
        Retorna o grau do vértice v.
        Complexidade: O(1)
        """
        return len(self.adj[v])
    
    def imprime(self):
        """
        Imprime a lista de adjacência do grafo.
        Complexidade: O(V + E)
        """
        print("\n📋 LISTA DE ADJACÊNCIA")
        print("=" * 60)
        for v in range(self.num_vertices):
            print(f"{v} (grau={self.grau(v)}): ", end="")
            if self.adj[v].vazia():
                print("[]")
            else:
                # Mostrar apenas os vértices destino
                vizinhos = [f"{aresta.v2}" for aresta in self.adj[v]]
                print(" → ".join(vizinhos))
        print("=" * 60)

print("\n🌐 CLASSE GRAFOLISTA")
print("=" * 50)
print("Implementação completa da lista de adjacência!")
```

## 6. Testando a Implementação

```python
# Criar um grafo com 5 vértices
print("\n🔨 CRIANDO GRAFO")
print("=" * 50)
g = GrafoLista(5)
print(f"Grafo criado com {g.num_vertices} vértices")

# Inserir arestas
print("\nInserindo arestas...")
arestas = [(0, 1, 5), (0, 2, 3), (1, 2, 2), (1, 3, 6), (2, 3, 7), (3, 4, 4)]

for v1, v2, peso in arestas:
    g.insere_aresta(v1, v2, peso)
    print(f"  ✓ Inserida aresta ({v1}, {v2}) com peso {peso}")

# Imprimir grafo
g.imprime()

# Visualizar com NetworkX
G_nx = nx.Graph()
for v1, v2, peso in arestas:
    G_nx.add_edge(v1, v2, weight=peso)

fig, ax = plt.subplots(figsize=(10, 7), constrained_layout=True, constrained_layout=True)
pos = nx.spring_layout(G_nx, seed=42)
nx.draw(G_nx, pos,
        with_labels=True,
        node_color='lightblue',
        node_size=1500,
        font_size=16,
        font_weight='bold',
        edge_color='gray',
        width=3, ax=ax)

edge_labels = nx.get_edge_attributes(G_nx, 'weight')
nx.draw_networkx_edge_labels(G_nx, pos, edge_labels, font_size=12, ax=ax)
ax.set_title('Grafo Implementado com Lista de Adjacência', fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
```

## 7. Testando Operações

```python
print("\n🔍 TESTANDO OPERAÇÕES")
print("=" * 60)

# 1. Verificar adjacência
print("\n1. Verificando adjacências:")
testes = [(0, 1), (0, 3), (1, 4), (3, 4)]
for v1, v2 in testes:
    existe = g.existe_aresta(v1, v2)
    print(f"   Existe aresta ({v1}, {v2})? {existe} {'✓' if existe else '✗'}")

# 2. Listar vizinhos
print("\n2. Listando vizinhos do vértice 1:")
if not g.lista_adj_vazia(1):
    aresta = g.primeiro_lista_adj(1)
    vizinhos = []
    while aresta is not None:
        vizinhos.append(aresta.v2)
        aresta = g.prox_adj(1, aresta)
    print(f"   Vizinhos de 1: {vizinhos}")

# 3. Grau dos vértices
print("\n3. Grau de cada vértice:")
for v in range(g.num_vertices):
    print(f"   grau({v}) = {g.grau(v)}")

# 4. Remover aresta
print("\n4. Removendo aresta (1, 3)...")
g.retira_aresta(1, 3)
print(f"   Existe (1, 3) agora? {g.existe_aresta(1, 3)} ✗")

g.imprime()
```

## 8. Análise de Complexidade

### ⏱️ Tabela de Complexidades

| Operação | Complexidade | Explicação |
|----------|-------------|------------|
| **Criar grafo** | O(V) | Criar array de V listas |
| **Inserir aresta** | O(1) | Inserir no fim da lista |
| **Remover aresta** | O(grau(v)) | Buscar na lista de v |
| **Verificar adjacência** | O(grau(v)) | Percorrer lista de v |
| **Listar vizinhos** | O(grau(v)) | Percorrer lista |
| **Grau do vértice** | O(1) | Tamanho da lista |
| **Imprimir grafo** | O(V + E) | Percorrer todas as listas |

### 💾 Complexidade de Espaço

- **Array `adj[]`**: O(V) para os |V| ponteiros
- **Listas**: O(E) para armazenar todas as arestas (cada aresta aparece 2× em grafo não direcionado)
- **Total**: **O(V + E)**

```python
# Demonstração: Comparando complexidades
import time
import random

def benchmark_operacoes(n_vertices):
    """Testa performance das operações."""
    g = GrafoLista(n_vertices)
    
    # Inserir arestas (grafo aleatório)
    num_arestas = n_vertices * 2  # Grafo esparso
    
    print(f"\n📊 Benchmark com {n_vertices} vértices, {num_arestas} arestas")
    print("=" * 60)
    
    # 1. Inserção
    inicio = time.time()
    for _ in range(num_arestas):
        v1 = random.randint(0, n_vertices-1)
        v2 = random.randint(0, n_vertices-1)
        if v1 != v2:
            g.insere_aresta(v1, v2)
    tempo_insercao = time.time() - inicio
    print(f"Inserção de {num_arestas} arestas: {tempo_insercao:.4f}s")
    
    # 2. Verificação de adjacência
    inicio = time.time()
    for _ in range(1000):
        v1 = random.randint(0, n_vertices-1)
        v2 = random.randint(0, n_vertices-1)
        g.existe_aresta(v1, v2)
    tempo_consulta = time.time() - inicio
    print(f"1000 consultas de adjacência: {tempo_consulta:.4f}s")
    
    # 3. Cálculo de graus
    inicio = time.time()
    graus = [g.grau(v) for v in range(n_vertices)]
    tempo_graus = time.time() - inicio
    print(f"Calcular grau de todos vértices: {tempo_graus:.4f}s (O(V))")
    print(f"Grau médio: {sum(graus)/len(graus):.2f}")

benchmark_operacoes(100)
benchmark_operacoes(500)
```

## 9. Vantagens e Desvantagens

### ✅ Vantagens da Lista de Adjacência:

1. **Eficiente em espaço** para grafos esparsos (|E| ≪ |V|²)
2. **Percorrer vizinhos** é proporcional ao grau do vértice
3. **Fácil adicionar/remover vértices** (redimensionar array)
4. **Grafo esparso comum** na prática (redes sociais, web)
5. **Memória linear**: O(V + E)

### ⚠️ Desvantagens:

1. **Verificar adjacência lenta**: O(grau(v)) no pior caso
2. **Não ideal para grafos densos** (muitas arestas)
3. **Remover aresta** requer busca na lista
4. **Mais complexa de implementar** que matriz

### 🎯 Quando Usar:

- ✅ Grafos **esparsos** (poucas arestas)
- ✅ Algoritmos de **percurso** (BFS, DFS)
- ✅ Grafos **grandes** (milhões de vértices)
- ✅ **90% dos casos práticos**

## 10. Exercício Final

```python
print("\n🎯 EXERCÍCIO: Implementar BFS usando Lista de Adjacência")
print("=" * 60)

def bfs_lista(grafo, origem):
    """
    Busca em Largura (BFS) usando lista de adjacência.
    
    Args:
        grafo (GrafoLista): Grafo representado por lista
        origem (int): Vértice de origem
    
    Returns:
        list: Ordem de visitação dos vértices
    """
    from collections import deque
    
    visitado = [False] * grafo.num_vertices
    fila = deque([origem])
    visitado[origem] = True
    ordem = []
    
    while fila:
        v = fila.popleft()
        ordem.append(v)
        
        # Percorrer vizinhos usando a lista de adjacência
        for aresta in grafo.adj[v]:
            vizinho = aresta.v2
            if not visitado[vizinho]:
                visitado[vizinho] = True
                fila.append(vizinho)
    
    return ordem

# Teste
g_teste = GrafoLista(7)
arestas_teste = [(0,1), (0,2), (1,3), (1,4), (2,5), (2,6)]
for v1, v2 in arestas_teste:
    g_teste.insere_aresta(v1, v2)

print("\nGrafo para BFS:")
g_teste.imprime()

ordem_bfs = bfs_lista(g_teste, 0)
print(f"\nOrdem BFS a partir de 0: {ordem_bfs}")
print("\n✓ BFS explora vizinhos nível por nível!")

# Visualizar
G_bfs = nx.Graph()
G_bfs.add_edges_from(arestas_teste)

fig, ax = plt.subplots(figsize=(10, 6), constrained_layout=True, constrained_layout=True)
pos = nx.spring_layout(G_bfs, seed=42)

# Colorir por ordem de visita
cores = ['red' if i == 0 else 'lightblue' for i in range(7)]
nx.draw(G_bfs, pos,
        with_labels=True,
        node_color=cores,
        node_size=1200,
        font_weight='bold',
        edge_color='gray',
        width=2, ax=ax)

ax.set_title(f'BFS a partir do vértice 0\nOrdem: {" → ".join(map(str, ordem_bfs))}',
         fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
```
