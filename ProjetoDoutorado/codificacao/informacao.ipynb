````python
import os
import pandas as pd

def info_csv(pasta=r'D:\OneDrive\Pessoais\Doutorado\Cefet\data'):
    """Mostra informações básicas dos arquivos CSV."""
    
    if not os.path.exists(pasta):
        print(f"❌ Pasta '{pasta}' não encontrada!")
        return
    
    arquivos = [f for f in os.listdir(pasta) if f.endswith('.csv')]
    
    if not arquivos:
        print("❌ Nenhum arquivo CSV encontrado!")
        return
    
    print("📊 INFORMAÇÕES DOS ARQUIVOS CSV")
    print("=" * 40)
    
    for arquivo in arquivos:
        caminho = os.path.join(pasta, arquivo)
        
        try:
            # Tamanho
            tamanho_mb = os.path.getsize(caminho) / (1024*1024)
            
            # Dados
            df = pd.read_csv(caminho)
            
            print(f"\n📄 {arquivo}")
            print(f"   Tamanho: {tamanho_mb:.2f} MB")
            print(f"   Linhas: {df.shape[0]}")
            print(f"   Colunas: {df.shape[1]}")
            print(f"   Variáveis: {list(df.columns)}")
            
        except Exception as e:
            print(f"\n❌ {arquivo}: Erro - {e}")

# Executar
info_csv()
````

# 📊 Informações CSV

## Função: `info_csv(pasta='data')`

**O que faz:** Mostra informações básicas dos arquivos CSV

**Informações mostradas:**
- Tamanho do arquivo (MB)
- Número de linhas
- Número de colunas  
- Lista das variáveis

**Como usar:**
```python
info_csv()              # pasta 'data'
info_csv('minha_pasta') # pasta específica
```
