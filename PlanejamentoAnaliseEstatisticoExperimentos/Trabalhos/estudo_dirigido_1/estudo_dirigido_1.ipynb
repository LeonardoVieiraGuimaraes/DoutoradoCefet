# Exercício de Simulação: TCL e IC em Distribuições Assimétricas


Resumo
Este exercício prático visa demonstrar numericamente a validade do Teorema Central do Limite (TCL) e a precisão dos Intervalos de Confiança (IC) baseados na distribuição t (ou z), mesmo quando a população original segue uma distribuição assimétrica, como a Exponencial.

## 1. Simulação do TCL para População Assimétrica (Exponencial)

### 1.1  Setup Inicial

Utilizaremos a Distribuição Exponencial com parâmetro de taxa λ = 0,2.
Média Populacional: μ = 1/λ = 1/0,2 = 5.

O objetivo é observar como a distribuição das médias amostrais se aproxima da Normalidade à medida que o tamanho da amostra (n) aumenta, apesar da forte assimetria da população.

- População Gerada: Distribuição Exponencial, N = 100.000, λ = 0,2 (μ = 5).
- Tamanhos Amostrais: n = 5 (pequena) e n = 50 (grande).
- Número de Simulações: k = 10.000.



```python
# ══════════════════════════════════════════════════════════════════════════════
# SETUP INICIAL: Imports e Configuração de Parâmetros
# ══════════════════════════════════════════════════════════════════════════════

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, t, probplot
from pathlib import Path

# 📋 PARÂMETROS GLOBAIS DO EXPERIMENTO
TRUE_MEAN = 5              # Média teórica da distribuição exponencial
RATE = 0.2                 # Taxa (λ) da distribuição exponencial
POP_SIZE = 100000          # Tamanho da população
NUM_SIMULATIONS = 10000    # Número de simulações (replicações)
SAMPLE_SIZE_SMALL = 5      # Tamanho pequeno de amostra
SAMPLE_SIZE_LARGE = 50     # Tamanho grande de amostra
CONFIDENCE_LEVEL = 0.95    # Nível de confiança (95%)

# 🎨 CONFIGURAÇÕES VISUAIS
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 10)

# 📁 DIRETÓRIO DE RESULTADOS
RESULTS_DIR = Path("relatorio/estudo_dirigido_1/resultados")
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

# 🌱 SEED PARA REPRODUTIBILIDADE
np.random.seed(42)

print(f"\n{'='*80}")
print("✅ SETUP INICIAL CONCLUÍDO")
print(f"{'='*80}")
print(f"Parâmetros:")
print(f"  ├─ Distribuição: Exponencial (λ={RATE}, μ={TRUE_MEAN})")
print(f"  ├─ População: N = {POP_SIZE:,}")
print(f"  ├─ Simulações: k = {NUM_SIMULATIONS:,}")
print(f"  ├─ Tamanhos amostrais: n = {SAMPLE_SIZE_SMALL} e n = {SAMPLE_SIZE_LARGE}")
print(f"  └─ Nível de confiança: {CONFIDENCE_LEVEL*100:.0f}%")
print(f"{'='*80}\n")
```

### 1.2  Tarefa 1: Implementação do TCL
Implemente a simulação nos dois ambientes (R e Python). Em ambos os casos, você deve gerar a população, coletar as 10.000 médias amostrais para n = 5 e n = 50, e gerar os histogramas para comparação.

### 1.2.1 Código em Python
Listing 2: Simulação do TCL em Python (Assimetria)

```python
# Listing 2: Simulação do TCL em Python (Assimetria)

# ══════════════════════════════════════════════════════════════════════════════
# TAREFA 1: SIMULAÇÃO DO TEOREMA CENTRAL DO LIMITE (TCL)
# ══════════════════════════════════════════════════════════════════════════════
# Usa parâmetros definidos no Setup Inicial

print(f"\n{'='*80}")
print("TAREFA 1: SIMULAÇÃO DO TEOREMA CENTRAL DO LIMITE (TCL)")
print(f"{'='*80}")

# 1. Gerar a População Exponencial
population = np.random.exponential(scale=1/RATE, size=POP_SIZE)
pop_mean = np.mean(population)
pop_std = np.std(population, ddof=1)
pop_skewness = skew(population)

print(f"\n✓ População Exponencial gerada (N={POP_SIZE:,})")
print(f"  ├─ Parâmetro λ: {RATE}")
print(f"  ├─ Média teórica: {TRUE_MEAN} | Observada: {pop_mean:.4f}")
print(f"  ├─ Desvio Padrão: {pop_std:.4f}")
print(f"  └─ Assimetria: {pop_skewness:.4f}")

# 2. Simular o TCL para n=5 e n=50
print(f"\n✓ Coletando {NUM_SIMULATIONS:,} médias amostrais...")
means_n5 = []
means_n50 = []

for _ in range(NUM_SIMULATIONS):
    sample_n5 = np.random.choice(population, SAMPLE_SIZE_SMALL, replace=False)
    sample_n50 = np.random.choice(population, SAMPLE_SIZE_LARGE, replace=False)
    means_n5.append(np.mean(sample_n5))
    means_n50.append(np.mean(sample_n50))

means_n5 = np.array(means_n5)
means_n50 = np.array(means_n50)

# 3. Análise Quantitativa
print(f"\n✓ Simulações TCL completadas")
print(f"  ├─ Assimetria das Médias (n={SAMPLE_SIZE_SMALL}):  {skew(means_n5):.4f}")
print(f"  ├─ Assimetria das Médias (n={SAMPLE_SIZE_LARGE}): {skew(means_n50):.4f}")
print(f"  └─ Redução de assimetria: {100*(1-abs(skew(means_n50))/abs(pop_skewness)):.1f}% (com n={SAMPLE_SIZE_LARGE})")
print(f"\n{'='*80}\n")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# TAREFA 1: SALVAR RESULTADOS E IMAGENS
# ══════════════════════════════════════════════════════════════════════════════

print(f"\n{'='*80}")
print("SALVANDO RESULTADOS DA TAREFA 1")
print(f"{'='*80}\n")

# Imagem 1: População Exponencial
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(population[:10000], bins=60, density=True, alpha=0.7, color='steelblue', edgecolor='black')
ax.axvline(pop_mean, color='red', linestyle='--', linewidth=2, label=f'Média = {pop_mean:.4f}')
ax.set_title('População Exponencial', fontweight='bold', fontsize=14)
ax.set_xlabel('Valor', fontsize=12)
ax.set_ylabel('Densidade', fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(RESULTS_DIR / "tarefa_1_populacao.png", dpi=300, bbox_inches='tight')
plt.close()
print("✓ Imagem salva: tarefa_1_populacao.png")

# Imagem 2: Médias (n=5)
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(means_n5, bins=50, density=True, alpha=0.7, color='coral', edgecolor='black')
ax.axvline(np.mean(means_n5), color='red', linestyle='--', linewidth=2, label=f'Média = {np.mean(means_n5):.4f}')
ax.set_title('Distribuição das Médias (n=5)', fontweight='bold', fontsize=14)
ax.set_xlabel('Média Amostral', fontsize=12)
ax.set_ylabel('Densidade', fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(RESULTS_DIR / "tarefa_1_medias_n5.png", dpi=300, bbox_inches='tight')
plt.close()
print("✓ Imagem salva: tarefa_1_medias_n5.png")

# Imagem 3: Médias (n=50)
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(means_n50, bins=50, density=True, alpha=0.7, color='lightgreen', edgecolor='black')
ax.axvline(np.mean(means_n50), color='red', linestyle='--', linewidth=2, label=f'Média = {np.mean(means_n50):.4f}')
ax.set_title('Distribuição das Médias (n=50)', fontweight='bold', fontsize=14)
ax.set_xlabel('Média Amostral', fontsize=12)
ax.set_ylabel('Densidade', fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(RESULTS_DIR / "tarefa_1_medias_n50.png", dpi=300, bbox_inches='tight')
plt.close()
print("✓ Imagem salva: tarefa_1_medias_n50.png")

# Imagem 4: Q-Q Plot
fig, ax = plt.subplots(figsize=(10, 6))
probplot(means_n50, dist="norm", plot=ax)
ax.set_title('Q-Q Plot: Médias (n=50)', fontweight='bold', fontsize=14)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(RESULTS_DIR / "tarefa_1_qqplot.png", dpi=300, bbox_inches='tight')
plt.close()
print("✓ Imagem salva: tarefa_1_qqplot.png")

# Criar CSV com resultados estatísticos
dados_t1 = {
    'Métrica': ['Média', 'Desvio Padrão', 'Assimetria'],
    'População': [f'{pop_mean:.4f}', f'{pop_std:.4f}', f'{pop_skewness:.4f}'],
    'Médias (n=5)': [f'{np.mean(means_n5):.4f}', f'{np.std(means_n5, ddof=1):.4f}', f'{skew(means_n5):.4f}'],
    'Médias (n=50)': [f'{np.mean(means_n50):.4f}', f'{np.std(means_n50, ddof=1):.4f}', f'{skew(means_n50):.4f}'],
}
pd.DataFrame(dados_t1).to_csv(RESULTS_DIR / "tarefa_1_resultados.csv", index=False)

print("✓ Dados salvos: tarefa_1_resultados.csv")
print(f"\n{'='*80}\n")
```

## 2  Simulação da Cobertura do Intervalo de Confiança (IC)

### 2.1  Tarefa 2: Cobertura do IC
Usando a mesma população assimétrica (μ = 5) e o tamanho de amostra grande (n = 50), simule a construção do IC para verificar se o Nível de Confiança Nominal (95%) é mantido, graças ao TCL.

- Média Populacional: μ = 5.
- Tamanho da Amostra: n = 50.
- Nível de Confiança: 95%.

#### 2.1.1  Código em Python
Listing 4: Cobertura do IC em Python para Assimetria

```python
# Listing 4: Cobertura do IC em Python para Assimetria

# ══════════════════════════════════════════════════════════════════════════════
# TAREFA 2: SIMULAÇÃO DA COBERTURA DO INTERVALO DE CONFIANÇA (IC)
# ══════════════════════════════════════════════════════════════════════════════
# Usa população e parâmetros da Tarefa 1 e Setup Inicial

print(f"\n{'='*80}")
print("TAREFA 2: SIMULAÇÃO DA COBERTURA DO INTERVALO DE CONFIANÇA (IC)")
print(f"{'='*80}")
print(f"\n✓ Simulando {NUM_SIMULATIONS:,} Intervalos de Confiança")
print(f"  ├─ Tamanho da amostra: n = {SAMPLE_SIZE_LARGE}")
print(f"  ├─ Nível de confiança: {CONFIDENCE_LEVEL*100:.0f}%")
print(f"  └─ Verdadeira média populacional: μ = {TRUE_MEAN}")

# Calcular cobertura do IC
ic_contains_mean = 0

for _ in range(NUM_SIMULATIONS):
    sample = np.random.choice(population, SAMPLE_SIZE_LARGE, replace=False)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    standard_error = sample_std / np.sqrt(SAMPLE_SIZE_LARGE)
    degrees_freedom = SAMPLE_SIZE_LARGE - 1
    
    # Calcular IC usando distribuição t
    ic_lower, ic_upper = t.interval(
        confidence=CONFIDENCE_LEVEL,
        df=degrees_freedom,
        loc=sample_mean,
        scale=standard_error
    )
    
    # Verificar se IC contém a verdadeira média
    if ic_lower <= TRUE_MEAN <= ic_upper:
        ic_contains_mean += 1

proportion_success = ic_contains_mean / NUM_SIMULATIONS

# Resultados
print(f"\n✓ Simulação de IC completada")
print(f"  ├─ ICs que capturaram μ = {TRUE_MEAN}: {ic_contains_mean:,} de {NUM_SIMULATIONS:,}")
print(f"  ├─ Taxa de Cobertura Observada: {proportion_success:.4f} ({proportion_success*100:.2f}%)")
print(f"  ├─ Taxa de Cobertura Esperada:  {CONFIDENCE_LEVEL:.4f} ({CONFIDENCE_LEVEL*100:.2f}%)")
print(f"  └─ Diferença: {abs(proportion_success - CONFIDENCE_LEVEL):.4f} ({abs(proportion_success - CONFIDENCE_LEVEL)/CONFIDENCE_LEVEL*100:.2f}%)")
print(f"\n{'='*80}\n")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# TAREFA 2: SALVAR RESULTADOS E IMAGENS
# ══════════════════════════════════════════════════════════════════════════════

print(f"\n{'='*80}")
print("SALVANDO RESULTADOS DA TAREFA 2")
print(f"{'='*80}\n")

# Imagem 1: Taxa de Cobertura
fig, ax = plt.subplots(figsize=(10, 6))
taxas = [proportion_success * 100, CONFIDENCE_LEVEL * 100]
labels = ['Observada', 'Esperada']
cores = ['#2ecc71', '#3498db']
ax.bar(labels, taxas, color=cores, edgecolor='black', linewidth=2, width=0.6)
ax.set_ylabel('Taxa de Cobertura (%)', fontweight='bold', fontsize=12)
ax.set_title('Taxa de Cobertura do IC', fontweight='bold', fontsize=14)
ax.set_ylim([85, 100])
ax.grid(True, alpha=0.3, axis='y')

# Adicionar valores nas barras
for i, v in enumerate(taxas):
    ax.text(i, v + 0.5, f'{v:.2f}%', ha='center', fontweight='bold', fontsize=11)

plt.tight_layout()
plt.savefig(RESULTS_DIR / "tarefa_2_cobertura.png", dpi=300, bbox_inches='tight')
plt.close()
print("✓ Imagem salva: tarefa_2_cobertura.png")

# Imagem 2: Sucesso vs Falha
fig, ax = plt.subplots(figsize=(10, 6))
sucesso_falha = [ic_contains_mean, NUM_SIMULATIONS - ic_contains_mean]
labels_sf = [f'Sucesso\n({ic_contains_mean:,})', f'Falha\n({NUM_SIMULATIONS - ic_contains_mean:,})']
cores_sf = ['#2ecc71', '#e74c3c']
ax.bar(labels_sf, sucesso_falha, color=cores_sf, edgecolor='black', linewidth=2, width=0.6)
ax.set_ylabel('Número de ICs', fontweight='bold', fontsize=12)
ax.set_title('Resultado dos ICs (10.000 simulações)', fontweight='bold', fontsize=14)
ax.grid(True, alpha=0.3, axis='y')

# Adicionar valores nas barras
for i, v in enumerate(sucesso_falha):
    ax.text(i, v + 100, f'{v:,}', ha='center', fontweight='bold', fontsize=11)

plt.tight_layout()
plt.savefig(RESULTS_DIR / "tarefa_2_sucessofalha.png", dpi=300, bbox_inches='tight')
plt.close()
print("✓ Imagem salva: tarefa_2_sucessofalha.png")

# Criar CSV com resultados do IC
dados_t2 = {
    'Métrica': [
        'Tamanho da Amostra',
        'Nível de Confiança',
        'Número de Simulações',
        'ICs com Sucesso',
        'ICs com Falha',
        'Taxa Observada (%)',
        'Taxa Esperada (%)',
        'Diferença (%)',
    ],
    'Valor': [
        f'{SAMPLE_SIZE_LARGE}',
        f'{CONFIDENCE_LEVEL*100:.0f}%',
        f'{NUM_SIMULATIONS:,}',
        f'{ic_contains_mean:,}',
        f'{NUM_SIMULATIONS - ic_contains_mean:,}',
        f'{proportion_success*100:.2f}%',
        f'{CONFIDENCE_LEVEL*100:.2f}%',
        f'{abs(proportion_success - CONFIDENCE_LEVEL)*100:.2f}%',
    ]
}
pd.DataFrame(dados_t2).to_csv(RESULTS_DIR / "tarefa_2_resultados.csv", index=False)

print("✓ Dados salvos: tarefa_2_resultados.csv")
print(f"\n{'='*80}\n")
```
