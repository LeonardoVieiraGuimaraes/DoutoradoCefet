# Exercício de Simulação: TCL e IC em Distribuições Assimétricas


Resumo
Este exercício prático visa demonstrar numericamente a validade do Teorema Central do Limite (TCL) e a precisão dos Intervalos de Confiança (IC) baseados na distribuição t (ou z), mesmo quando a população original segue uma distribuição assimétrica, como a Exponencial.

## 1. Simulação do TCL para População Assimétrica (Exponencial)

### 1.1  Setup Inicial

Utilizaremos a Distribuição Exponencial com parâmetro de taxa λ = 0,2.
Média Populacional: μ = 1/λ = 1/0,2 = 5.

O objetivo é observar como a distribuição das médias amostrais se aproxima da Normalidade à medida que o tamanho da amostra (n) aumenta, apesar da forte assimetria da população.

- População Gerada: Distribuição Exponencial, N = 100.000, λ = 0,2 (μ = 5).
- Tamanhos Amostrais: n = 5 (pequena) e n = 50 (grande).
- Número de Simulações: k = 10.000.



### 1.2  Tarefa 1: Implementação do TCL
Implemente a simulação nos dois ambientes (R e Python). Em ambos os casos, você deve gerar a população, coletar as 10.000 médias amostrais para n = 5 e n = 50, e gerar os histogramas para comparação.

### 1.2.1 Código em Python
Listing 2: Simulação do TCL em Python (Assimetria)

```python
# Listing 2: Simulação do TCL em Python (Assimetria)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, shapiro
import pandas as pd
import os
from pathlib import Path

# Configuração
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# Criar diretório para resultados
RESULTS_DIR = Path("relatorio/estudo_dirigido_1/resultados")
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

print("="*70)
print("TAREFA 1: SIMULAÇÃO DO TEOREMA CENTRAL DO LIMITE")
print("="*70)

# 1. Parâmetros e Geração da População Exponencial (mu=5)
TRUE_MEAN = 5
RATE = 0.2
POP_SIZE = 100000
NUM_SIMULATIONS = 10000

np.random.seed(42)
# np.random.exponential usa scale = 1/rate, então scale=5
population = np.random.exponential(scale=1/RATE, size=POP_SIZE)

# Estatísticas da população
pop_mean = np.mean(population)
pop_std = np.std(population, ddof=1)
pop_skewness = skew(population)

print(f"\n1. POPULAÇÃO (Distribuição Exponencial)")
print(f"   Parâmetro de taxa (λ): {RATE}")
print(f"   Tamanho da população: {POP_SIZE:,}")
print(f"   Média populacional teórica (μ): {TRUE_MEAN}")
print(f"   Média observada: {pop_mean:.4f}")
print(f"   Desvio padrão: {pop_std:.4f}")
print(f"   Assimetria (Skewness): {pop_skewness:.4f}")

# 2. Simular o TCL para n=5 e n=50
SAMPLE_SIZE_SMALL = 5
SAMPLE_SIZE_LARGE = 50
means_n5 = []
means_n50 = []

print(f"\n2. SIMULANDO DISTRIBUIÇÃO DAS MÉDIAS AMOSTRAIS")
print(f"   Número de simulações: {NUM_SIMULATIONS:,}")
print(f"   Tamanho amostra pequena (n): {SAMPLE_SIZE_SMALL}")
print(f"   Tamanho amostra grande (n): {SAMPLE_SIZE_LARGE}")

for _ in range(NUM_SIMULATIONS):
    sample_n5 = np.random.choice(population, SAMPLE_SIZE_SMALL, replace=False)
    sample_n50 = np.random.choice(population, SAMPLE_SIZE_LARGE, replace=False)
    means_n5.append(np.mean(sample_n5))
    means_n50.append(np.mean(sample_n50))

means_n5 = np.array(means_n5)
means_n50 = np.array(means_n50)

# 3. Análise dos resultados (Quantitativa)
print(f"\n3. ANÁLISE QUANTITATIVA")
print(f"\n   Para n = {SAMPLE_SIZE_SMALL}:")
print(f"   - Média das médias: {np.mean(means_n5):.4f}")
print(f"   - Desvio padrão: {np.std(means_n5, ddof=1):.4f}")
print(f"   - Assimetria: {skew(means_n5):.4f}")
print(f"   - Teste Shapiro-Wilk p-value: {shapiro(means_n5)[1]:.4f}")

print(f"\n   Para n = {SAMPLE_SIZE_LARGE}:")
print(f"   - Média das médias: {np.mean(means_n50):.4f}")
print(f"   - Desvio padrão: {np.std(means_n50, ddof=1):.4f}")
print(f"   - Assimetria: {skew(means_n50):.4f}")
print(f"   - Teste Shapiro-Wilk p-value: {shapiro(means_n50)[1]:.4f}")

# Erro padrão teórico
se_n5_teorico = pop_std / np.sqrt(SAMPLE_SIZE_SMALL)
se_n50_teorico = pop_std / np.sqrt(SAMPLE_SIZE_LARGE)
print(f"\n   Erro padrão teórico para n=5: {se_n5_teorico:.4f}")
print(f"   Erro padrão teórico para n=50: {se_n50_teorico:.4f}")

# Salvar resultados em CSV
results_tcl = {
    'Parâmetro': [
        'Média Populacional Teórica',
        'Média Populacional Observada',
        'Desvio Padrão Populacional',
        'Assimetria da População',
        'Média das Médias (n=5)',
        'Desvio Padrão das Médias (n=5)',
        'Assimetria das Médias (n=5)',
        'Erro Padrão Teórico (n=5)',
        'Média das Médias (n=50)',
        'Desvio Padrão das Médias (n=50)',
        'Assimetria das Médias (n=50)',
        'Erro Padrão Teórico (n=50)'
    ],
    'Valor': [
        f'{TRUE_MEAN:.4f}',
        f'{pop_mean:.4f}',
        f'{pop_std:.4f}',
        f'{pop_skewness:.4f}',
        f'{np.mean(means_n5):.4f}',
        f'{np.std(means_n5, ddof=1):.4f}',
        f'{skew(means_n5):.4f}',
        f'{se_n5_teorico:.4f}',
        f'{np.mean(means_n50):.4f}',
        f'{np.std(means_n50, ddof=1):.4f}',
        f'{skew(means_n50):.4f}',
        f'{se_n50_teorico:.4f}'
    ]
}

df_tcl = pd.DataFrame(results_tcl)
df_tcl.to_csv(RESULTS_DIR / "tarefa1_tcl_resultados.csv", index=False)
print(f"\n✓ Resultados TCL salvos em: {RESULTS_DIR}/tarefa1_tcl_resultados.csv")
```

## 2  Simulação da Cobertura do Intervalo de Confiança (IC)

### 2.1  Tarefa 2: Cobertura do IC
Usando a mesma população assimétrica (μ = 5) e o tamanho de amostra grande (n = 50), simule a construção do IC para verificar se o Nível de Confiança Nominal (95%) é mantido, graças ao TCL.

- Média Populacional: μ = 5.
- Tamanho da Amostra: n = 50.
- Nível de Confiança: 95%.

#### 2.1.1  Código em Python
Listing 4: Cobertura do IC em Python para Assimetria

```python
# Listing 4: Cobertura do IC em Python para Assimetria
import numpy as np
import pandas as pd
from scipy.stats import t
from pathlib import Path

print("\n" + "="*70)
print("TAREFA 2: VALIDAÇÃO DA COBERTURA DO INTERVALO DE CONFIANÇA")
print("="*70)

# População e parâmetros do Item 1
TRUE_MEAN = 5
RATE = 0.2
POP_SIZE = 100000
CONFIDENCE_LEVEL = 0.95
SAMPLE_SIZE = 50
NUM_SIMULATIONS = 10000
ALPHA = 1 - CONFIDENCE_LEVEL

# Diretório de resultados
RESULTS_DIR = Path("relatorio/estudo_dirigido_1/resultados")
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

np.random.seed(42)
population = np.random.exponential(scale=1/RATE, size=POP_SIZE)

print(f"\n1. PARÂMETROS DA SIMULAÇÃO")
print(f"   População: Distribuição Exponencial (λ={RATE})")
print(f"   Média populacional: μ = {TRUE_MEAN}")
print(f"   Tamanho da amostra: n = {SAMPLE_SIZE}")
print(f"   Nível de confiança: {CONFIDENCE_LEVEL*100:.0f}%")
print(f"   Número de simulações: {NUM_SIMULATIONS:,}")

# Simulação de IC
ic_contains_mean = 0
ic_limits = []
sample_means = []
sample_stds = []

print(f"\n2. EXECUTANDO SIMULAÇÕES...")

for i in range(NUM_SIMULATIONS):
    sample = np.random.choice(population, SAMPLE_SIZE, replace=False)
    
    # Estatísticas amostrais
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    standard_error = sample_std / np.sqrt(SAMPLE_SIZE)
    degrees_freedom = SAMPLE_SIZE - 1
    
    # Armazenar para análise
    sample_means.append(sample_mean)
    sample_stds.append(sample_std)
    
    # Cálculo do IC usando t.interval da scipy
    ic_lower, ic_upper = t.interval(
        confidence=CONFIDENCE_LEVEL,
        df=degrees_freedom,
        loc=sample_mean,
        scale=standard_error
    )
    
    ic_limits.append({'lower': ic_lower, 'upper': ic_upper})
    
    # Verifica se o IC contém a média populacional
    if (ic_lower <= TRUE_MEAN) and (TRUE_MEAN <= ic_upper):
        ic_contains_mean += 1

# Proporção de Sucesso (Taxa de Cobertura)
proportion_success = ic_contains_mean / NUM_SIMULATIONS
confidence_interval_95 = t.ppf((1 + CONFIDENCE_LEVEL) / 2, NUM_SIMULATIONS - 1) * \
                         np.sqrt(proportion_success * (1 - proportion_success) / NUM_SIMULATIONS)

print(f"\n3. RESULTADOS DA SIMULAÇÃO")
print(f"\n   Taxa de Cobertura do IC:")
print(f"   - Observada: {proportion_success:.4f} ({ic_contains_mean}/{NUM_SIMULATIONS})")
print(f"   - Esperada (teórica): {CONFIDENCE_LEVEL:.4f}")
print(f"   - Erro relativo: {abs(proportion_success - CONFIDENCE_LEVEL):.4f}")

print(f"\n   Estatísticas das Médias Amostrais:")
print(f"   - Média: {np.mean(sample_means):.4f}")
print(f"   - Desvio padrão: {np.std(sample_means, ddof=1):.4f}")
print(f"   - Mín: {np.min(sample_means):.4f}")
print(f"   - Máx: {np.max(sample_means):.4f}")

print(f"\n   Amplitude dos ICs:")
ic_limits_array = np.array([(ic['upper'] - ic['lower']) for ic in ic_limits])
print(f"   - Média: {np.mean(ic_limits_array):.4f}")
print(f"   - Mín: {np.min(ic_limits_array):.4f}")
print(f"   - Máx: {np.max(ic_limits_array):.4f}")

# Salvar resultados em CSV
results_ic = {
    'Parâmetro': [
        'Média Populacional',
        'Nível de Confiança',
        'Tamanho da Amostra',
        'Taxa de Cobertura Observada',
        'Taxa de Cobertura Esperada',
        'Número de ICs que cobrem μ',
        'Número total de simulações',
        'Erro Relativo',
        'Média das Médias Amostrais',
        'Desvio Padrão das Médias',
        'Amplitude Média dos ICs'
    ],
    'Valor': [
        f'{TRUE_MEAN:.4f}',
        f'{CONFIDENCE_LEVEL:.2%}',
        f'{SAMPLE_SIZE}',
        f'{proportion_success:.4f}',
        f'{CONFIDENCE_LEVEL:.4f}',
        f'{ic_contains_mean}',
        f'{NUM_SIMULATIONS}',
        f'{abs(proportion_success - CONFIDENCE_LEVEL):.4f}',
        f'{np.mean(sample_means):.4f}',
        f'{np.std(sample_means, ddof=1):.4f}',
        f'{np.mean(ic_limits_array):.4f}'
    ]
}

df_ic = pd.DataFrame(results_ic)
df_ic.to_csv(RESULTS_DIR / "tarefa2_ic_resultados.csv", index=False)

# Salvar primeiros 100 ICs em CSV detalhado
ic_details = []
for i in range(min(100, NUM_SIMULATIONS)):
    ic_details.append({
        'Simulação': i + 1,
        'Média Amostral': f'{sample_means[i]:.4f}',
        'IC Inferior': f'{ic_limits[i]["lower"]:.4f}',
        'IC Superior': f'{ic_limits[i]["upper"]:.4f}',
        'Contém μ?': 'Sim' if ic_limits[i]["lower"] <= TRUE_MEAN <= ic_limits[i]["upper"] else 'Não',
        'Amplitude': f'{ic_limits[i]["upper"] - ic_limits[i]["lower"]:.4f}'
    })

df_ic_details = pd.DataFrame(ic_details)
df_ic_details.to_csv(RESULTS_DIR / "tarefa2_ic_detalhado.csv", index=False)

print(f"\n✓ Resultados do IC salvos em: {RESULTS_DIR}/tarefa2_ic_resultados.csv")
print(f"✓ Detalhes dos primeiros 100 ICs salvos em: {RESULTS_DIR}/tarefa2_ic_detalhado.csv")
```
