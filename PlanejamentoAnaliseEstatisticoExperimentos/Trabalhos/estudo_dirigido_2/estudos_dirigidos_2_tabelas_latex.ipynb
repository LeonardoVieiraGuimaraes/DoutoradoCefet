````python
import pandas as pd
import os
from pathlib import Path
import numpy as np

# Diretórios
BASE_DIR = Path('relatorio/resultados')
LATEX_DIR = Path('relatorio/tabelas_latex')

# Criar diretórios para tabelas LaTeX organizadas por análise
LATEX_TCL_DIR = LATEX_DIR / 'parte_1_tcl'
LATEX_ONEWAY_DIR = LATEX_DIR / 'parte_2_oneway'
LATEX_RCBD_DIR = LATEX_DIR / 'parte_2_rcbd'
LATEX_FATORIAL_DIR = LATEX_DIR / 'parte_2_fatorial'

# Criar todos os diretórios se não existirem
for dir_path in [LATEX_DIR, LATEX_TCL_DIR, LATEX_ONEWAY_DIR, LATEX_RCBD_DIR, LATEX_FATORIAL_DIR]:
    dir_path.mkdir(parents=True, exist_ok=True)

print("✓ Diretórios configurados")
print(f"  - Dados: {BASE_DIR}")
print(f"  - Saída Central: {LATEX_DIR}")
print(f"  - TCL: {LATEX_TCL_DIR}")
print(f"  - One-Way: {LATEX_ONEWAY_DIR}")
print(f"  - RCBD: {LATEX_RCBD_DIR}")
print(f"  - Fatorial: {LATEX_FATORIAL_DIR}")
````

## Parte I - Teorema Central do Limite (TCL)

````python
# Carregar dados TCL
df_param_tcl = pd.read_csv(BASE_DIR / 'parte_1_tcl' / 'parametros_tcl.csv')
df_result_tcl = pd.read_csv(BASE_DIR / 'parte_1_tcl' / 'resultados_tcl.csv')

# Função para escapar caracteres especiais em LaTeX
def escape_latex_chars(text):
    """Escapa caracteres especiais para LaTeX"""
    if isinstance(text, str):
        text = text.replace('μ', r'$\mu$')
        text = text.replace('λ', r'$\lambda$')
        text = text.replace('α', r'$\alpha$')
        text = text.replace('σ', r'$\sigma$')
        text = text.replace('β', r'$\beta$')
        text = text.replace('τ', r'$\tau$')
        text = text.replace('γ', r'$\gamma$')
        text = text.replace('ε', r'$\varepsilon$')
        text = text.replace('∼', r'$\sim$')
        text = text.replace('≈', r'$\approx$')
    return text

# Função para gerar apenas o conteúdo da tabela (sem \begin{table}...\end{table})
def generate_table_content(df, caption='', label='', column_format='lp{8cm}', float_format=None):
    """Gera tabela completa com wrapper [H]"""
    latex_full = df.to_latex(
        index=False,
        caption=caption,
        label=label,
        column_format=column_format,
        escape=False,
        float_format=float_format
    )
    
    # Remover o wrapper gerado pelo pandas
    lines = latex_full.split('\n')
    content_lines = []
    skip_begin_end = False
    for line in lines:
        if '\\begin{table}' in line or '\\end{table}' in line:
            continue
        content_lines.append(line)
    
    inner_content = '\n'.join(content_lines).strip()
    
    # Adicionar wrapper [H] customizado
    wrapped = f'''\\begin{{table}}[H]
\\centering
{inner_content}
\\end{{table}}'''
    
    return wrapped

# Aplicar escape aos DataFrames
df_param_tcl = df_param_tcl.applymap(escape_latex_chars)
df_result_tcl = df_result_tcl.applymap(escape_latex_chars)

# 1. Tabela de Parâmetros TCL
latex_param_tcl = generate_table_content(
    df_param_tcl,
    caption='Parâmetros da Simulação - Teorema Central do Limite',
    label='tab:parametros_tcl',
    column_format='lp{8cm}'
)

# Salvar em uma localização
with open(LATEX_TCL_DIR / '01_parametros_tcl.tex', 'w', encoding='utf-8') as f:
    f.write(latex_param_tcl)

print("✓ Tabela de parâmetros TCL gerada")

# 2. Tabela de Resultados TCL
latex_result_tcl = generate_table_content(
    df_result_tcl,
    caption='Resultados da Simulação - Intervalo de Confiança',
    label='tab:resultados_tcl',
    column_format='lrrr',
    float_format=lambda x: f'{x:.6f}' if abs(x) < 1 else f'{x:.4f}'
)

# Salvar em uma localização
with open(LATEX_TCL_DIR / '02_resultados_tcl.tex', 'w', encoding='utf-8') as f:
    f.write(latex_result_tcl)

print("✓ Tabela de resultados TCL gerada")
````

## Parte II.1 - ANOVA One-Way

````python
# Carregar dados One-Way
df_param_oneway = pd.read_csv(BASE_DIR / 'parte_2_oneway' / 'parametros_oneway.csv')
df_result_oneway = pd.read_csv(BASE_DIR / 'parte_2_oneway' / 'resultados_oneway.csv')

# Aplicar escape aos DataFrames
df_param_oneway = df_param_oneway.applymap(escape_latex_chars)
df_result_oneway = df_result_oneway.applymap(escape_latex_chars)

# 1. Tabela de Parâmetros One-Way
latex_param_oneway = generate_table_content(
    df_param_oneway,
    caption='Parâmetros do Experimento - ANOVA One-Way',
    label='tab:parametros_oneway',
    column_format='lp{8cm}'
)

# Salvar em uma localização
with open(LATEX_ONEWAY_DIR / '03_parametros_oneway.tex', 'w', encoding='utf-8') as f:
    f.write(latex_param_oneway)

print("✓ Tabela de parâmetros One-Way gerada")

# 2. Tabela de Resultados ANOVA One-Way
latex_result_oneway = generate_table_content(
    df_result_oneway,
    caption='Tabela ANOVA - Análise de Variância One-Way',
    label='tab:anova_oneway',
    column_format='lrrrr',
    float_format=lambda x: f'{x:.6f}' if x < 0.001 else f'{x:.4f}' if x < 1 else f'{x:.2f}'
)

# Salvar em uma localização
with open(LATEX_ONEWAY_DIR / '04_anova_oneway.tex', 'w', encoding='utf-8') as f:
    f.write(latex_result_oneway)

print("✓ Tabela ANOVA One-Way gerada")

# 3. Tabela de Teste de Tukey (leitura do TXT)
txt_path = BASE_DIR / 'parte_2_oneway' / 'analise_oneway_resultados.txt'
if txt_path.exists():
    with open(txt_path, 'r', encoding='utf-8') as f:
        content = f.read()
    
    # Extrair informações do teste de Tukey
    lines = content.split('\n')
    tukey_start = None
    for i, line in enumerate(lines):
        if 'Teste de Tukey HSD' in line:
            tukey_start = i
            break
    
    if tukey_start:
        tukey_text = '\n'.join(lines[tukey_start:tukey_start+20])
        # Salvar em uma localização
        with open(LATEX_ONEWAY_DIR / '05_tukey_oneway.tex', 'w', encoding='utf-8') as f:
            f.write(f"% Teste de Tukey HSD - ANOVA One-Way\n")
            f.write(f"\\begin{{verbatim}}\n")
            f.write(tukey_text)
            f.write(f"\n\\end{{verbatim}}\n")
        print("✓ Tabela Teste de Tukey gerada")
````

## Parte II.2 - ANOVA em Blocos Completos Casualizados (RCBD)

````python
# Carregar dados RCBD
df_param_rcbd = pd.read_csv(BASE_DIR / 'parte_2_rcbd' / 'parametros_rcbd.csv')
df_result_rcbd = pd.read_csv(BASE_DIR / 'parte_2_rcbd' / 'resultados_rcbd.csv')

print(f"df_param_rcbd shape: {df_param_rcbd.shape}")
print(f"df_result_rcbd shape: {df_result_rcbd.shape}")

# Aplicar escape aos DataFrames
df_param_rcbd = df_param_rcbd.applymap(escape_latex_chars)
df_result_rcbd = df_result_rcbd.applymap(escape_latex_chars)

# 1. Tabela de Parâmetros RCBD
latex_param_rcbd = generate_table_content(
    df_param_rcbd,
    caption='Parâmetros do Experimento - ANOVA RCBD',
    label='tab:parametros_rcbd',
    column_format='lp{8cm}'
)

print(f"latex_param_rcbd length: {len(latex_param_rcbd)}")

# Salvar em uma localização
with open(LATEX_RCBD_DIR / '06_parametros_rcbd.tex', 'w', encoding='utf-8') as f:
    f.write(latex_param_rcbd)

print("✓ Tabela de parâmetros RCBD gerada")

# 2. Tabela de Resultados RCBD
latex_result_rcbd = generate_table_content(
    df_result_rcbd,
    caption='Tabela ANOVA RCBD - Blocos Casualizados',
    label='tab:anova_rcbd',
    column_format='lrrrr',
    float_format=lambda x: f'{x:.6f}' if x < 0.001 else f'{x:.4f}' if x < 1 else f'{x:.2f}'
)

print(f"latex_result_rcbd length: {len(latex_result_rcbd)}")

# Salvar em uma localização
with open(LATEX_RCBD_DIR / '07_anova_rcbd.tex', 'w', encoding='utf-8') as f:
    f.write(latex_result_rcbd)

print("✓ Tabela de resultados RCBD gerada")
````

