## Setup Inicial: Importações e Configurações

```python
import pandas as pd
import os
from pathlib import Path
import re

# Configuração de diretórios
RESULTADOS_BASE = Path("relatorio/resultados")
ANALISES_DIR = RESULTADOS_BASE / "analises"
ANALISES_DIR.mkdir(parents=True, exist_ok=True)

print("\n" + "="*80)
print("ESTUDO DIRIGIDO 2 - PARTE 2: GERAÇÃO DE TABELAS LATEX")
print("="*80 + "\n")
print(f"Diretório base de resultados: {RESULTADOS_BASE}")
print(f"Diretório para LaTeX: {ANALISES_DIR}\n")
```

# Parte I: Tabelas TCL (Teorema Central do Limite)

```python
print("\n" + "="*80)
print("PROCESSANDO RESULTADOS TCL")
print("="*80 + "\n")

# Ler CSVs TCL (parâmetros e resultados)
csv_param_tcl = RESULTADOS_BASE / "parte_1_tcl" / "parametros_tcl.csv"
csv_resultado_tcl = RESULTADOS_BASE / "parte_1_tcl" / "analise_tcl_resultados.csv"

df_param_tcl = None
df_resultado_tcl = None

if csv_param_tcl.exists():
    df_param_tcl = pd.read_csv(csv_param_tcl, encoding='utf-8')
    print(f"✓ Arquivo de Parâmetros TCL encontrado: {csv_param_tcl}\n")
    print("PARÂMETROS:")
    print(df_param_tcl.to_string(index=False))
    print()
else:
    print(f"✗ Arquivo não encontrado: {csv_param_tcl}\n")

if csv_resultado_tcl.exists():
    df_resultado_tcl = pd.read_csv(csv_resultado_tcl, encoding='utf-8')
    print(f"✓ Arquivo de Resultados TCL encontrado: {csv_resultado_tcl}\n")
    print("RESULTADOS:")
    print(df_resultado_tcl.to_string(index=False))
    print()
else:
    print(f"✗ Arquivo não encontrado: {csv_resultado_tcl}\n")
```

## Geração da Tabela LaTeX - TCL

```python
# Gerar tabelas LaTeX TCL (Parâmetros e Resultados)
output_dir_tcl = ANALISES_DIR / "tcl"
output_dir_tcl.mkdir(parents=True, exist_ok=True)

if df_param_tcl is not None:
    # Tabela 1: Parâmetros TCL
    latex_param_tcl = """\\begin{table}[H]
\\centering
\\caption{Parâmetros da Simulação do TCL}
\\label{tab:parametros_tcl}
\\begin{tabular}{|l|r|}
\\hline
\\textbf{Métrica} & \\textbf{Valor} \\\\
\\hline
"""
    for _, row in df_param_tcl.iterrows():
        metrica = row['Métrica'].replace('_', '\\_')
        valor = str(row['Valor'])
        latex_param_tcl += f"{metrica} & {valor} \\\\\n"

    latex_param_tcl += """\\hline
\\end{tabular}
\\end{table}
"""
    
    output_param_tcl = output_dir_tcl / "parametros_tcl.tex"
    with open(output_param_tcl, 'w', encoding='utf-8') as f:
        f.write(latex_param_tcl)
    print(f"✓ Tabela LaTeX Parâmetros TCL salva em: {output_param_tcl}\n")

if df_resultado_tcl is not None:
    # Tabela 2: Resultados TCL
    latex_resultado_tcl = """\\begin{table}[H]
\\centering
\\caption{Resultados da Simulação do TCL}
\\label{tab:resultados_tcl}
\\begin{tabular}{|l|r|}
\\hline
\\textbf{Métrica} & \\textbf{Valor} \\\\
\\hline
"""
    for _, row in df_resultado_tcl.iterrows():
        metrica = row['Métrica'].replace('_', '\\_')
        valor = str(row['Valor'])
        latex_resultado_tcl += f"{metrica} & {valor} \\\\\n"

    latex_resultado_tcl += """\\hline
\\end{tabular}
\\end{table}
"""
    
    output_resultado_tcl = output_dir_tcl / "resultados_tcl.tex"
    with open(output_resultado_tcl, 'w', encoding='utf-8') as f:
        f.write(latex_resultado_tcl)
    print(f"✓ Tabela LaTeX Resultados TCL salva em: {output_resultado_tcl}\n")
```

# Parte II, Exercício 1: Tabelas ANOVA One-Way

```python
print("\n" + "="*80)
print("PROCESSANDO RESULTADOS ONE-WAY ANOVA")
print("="*80 + "\n")

# Ler CSVs One-Way
csv_param_oneway = RESULTADOS_BASE / "parte_2_oneway" / "parametros_oneway.csv"
csv_resultado_oneway = RESULTADOS_BASE / "parte_2_oneway" / "analise_oneway_resultados.csv"

df_param_oneway = None
df_resultado_oneway = None

if csv_param_oneway.exists():
    df_param_oneway = pd.read_csv(csv_param_oneway, encoding='utf-8')
    print(f"✓ Arquivo de Parâmetros One-Way encontrado: {csv_param_oneway}\n")
    print("PARÂMETROS:")
    print(df_param_oneway.to_string(index=False))
    print()
else:
    print(f"✗ Arquivo não encontrado: {csv_param_oneway}\n")

if csv_resultado_oneway.exists():
    df_resultado_oneway = pd.read_csv(csv_resultado_oneway, encoding='utf-8')
    print(f"✓ Arquivo de Resultados One-Way encontrado: {csv_resultado_oneway}\n")
    print("RESULTADOS:")
    print(df_resultado_oneway.to_string(index=False))
    print()
else:
    print(f"✗ Arquivo não encontrado: {csv_resultado_oneway}\n")
```
