## 🔬 Metodologia Estatística

### 1️⃣ Teorema Central do Limite (TCL)

**Conceito**: O TCL afirma que, para amostras de tamanho suficientemente grande, a distribuição das médias amostrais aproxima-se de uma distribuição normal, **independentemente da forma da distribuição original**.

**Procedimento**:
- Extrair n=50 observações aleatoriamente da população
- Calcular a média amostral
- Repetir 10.000 vezes
- Analisar a distribuição das médias

**Testes Aplicados**:
- **Shapiro-Wilk**: Testar normalidade (H₀: dados são normais)
- **Assimetria**: Verificar simetria da distribuição
- **Comparação Empírica vs Teórica**: Desvio padrão observado vs esperado

---

### 2️⃣ Intervalos de Confiança (IC) a 95%

**Conceito**: Um intervalo onde, com 95% de confiança, a verdadeira média populacional se localiza.

**Fórmula**: 
$$IC = \bar{x} \pm t_{crítico} \times SE$$

Onde:
- $\bar{x}$ = média amostral
- $t_{crítico}$ = valor crítico da distribuição t (gl = n-1, α=0.05)
- $SE$ = erro padrão = $\frac{s}{\sqrt{n}}$

**Cálculos**:
- IC por **mês** (12 grupos)
- IC por **ano** (13 grupos)

---

### 3️⃣ ANOVA Fatorial 2-Way (Mês × Ano)

**Conceito**: Testa se há diferenças significativas entre grupos em dois fatores simultâneos.

**Modelo Linear**:
$$qtd_{ijk} = \mu + \alpha_i + \beta_j + (\alpha\beta)_{ij} + \varepsilon_{ijk}$$

Onde:
- $\mu$ = média geral
- $\alpha_i$ = efeito do mês i
- $\beta_j$ = efeito do ano j
- $(\alpha\beta)_{ij}$ = efeito de interação
- $\varepsilon_{ijk}$ = erro aleatório

**Hipóteses Testadas**:
- H₀: Não há efeito de mês (todos os α_i = 0)
- H₀: Não há efeito de ano (todos os β_j = 0)
- H₀: Não há interação (todos os (αβ)_ij = 0)

**Pressupostos Verificados**:
- Normalidade dos resíduos
- Homogeneidade de variâncias (Levene)
- Independência dos resíduos

---

### 4️⃣ Teste Post-Hoc (Tukey HSD)

**Conceito**: Quando ANOVA rejeita H₀, Tukey HSD identifica **quais pares de grupos** diferem significativamente.

**Vantagem**: Controla o erro tipo I em múltiplas comparações através da correção Honest Significant Difference.

**Comparações Realizadas**:
- Pairwise entre 12 meses (66 comparações)
- Pairwise entre 13 anos (78 comparações)

**Critério de Significância**: α = 0.05

### Teste de Normalidade (Shapiro-Wilk)

Aplica o teste de Shapiro-Wilk para validar se as médias amostrais seguem distribuição normal.

- **H0**: Os dados seguem distribuição normal
- **H1**: Os dados NÃO seguem distribuição normal
- **Critério**: Se p-value > 0.05, não rejeitamos H0 (dados são aproximadamente normais)

```python
# ========== TESTE DE NORMALIDADE ==========

# Teste Shapiro-Wilk nos dados originais
stat_original, p_original = shapiro(np.random.choice(populacao, size=5000))

# Teste Shapiro-Wilk nas médias amostrais
stat_medias, p_medias = shapiro(medias_amostrais)

print(f'\n🔍 TESTES DE NORMALIDADE (Shapiro-Wilk):')
print(f'\n  Dados Originais:')
print(f'    Estatística: {stat_original:.6f}')
print(f'    p-value:     {p_original:.4e}')
print(f'    Resultado:   {"❌ NÃO normal" if p_original < 0.05 else "✅ Normal"}')

print(f'\n  Médias Amostrais (TCL):')
print(f'    Estatística: {stat_medias:.6f}')
print(f'    p-value:     {p_medias:.4e}')
print(f'    Resultado:   {"❌ NÃO normal" if p_medias < 0.05 else "✅ Normal"}')

print(f'\n💡 CONCLUSÃO:')
if p_medias > 0.05:
    print(f'   ✅ As médias amostrais apresentam distribuição APROXIMADAMENTE NORMAL')
    print(f'   ✅ TCL validado para estes dados!')
else:
    print(f'   ⚠️ As médias amostrais NÃO apresentam distribuição perfeitamente normal')
    print(f'   Mas com n={n_simulacoes:,} simulações, a distribuição é muito próxima da normal')
```

## 3️⃣ Intervalos de Confiança 95% (IC)

### O que é um Intervalo de Confiança?

Um intervalo de confiança de 95% é um **intervalo estimado onde acreditamos, com 95% de confiança, que a verdadeira média populacional se encontra**.

**Fórmula**: IC = Média ± (t-crítico × Erro Padrão)

**Objetivo**: Calcular IC para cada mês e cada ano, permitindo identificar períodos com maior/menor quantidade de bovinos.

```python
# ========== INTERVALOS DE CONFIANÇA POR MÊS ==========

print('\n' + '='*80)
print('📊 INTERVALOS DE CONFIANÇA 95% (IC)')
print('='*80)

from scipy import stats as sp_stats

# Calcular IC por mês
ic_mes_list = []

print(f'\n🔍 Calculando IC para cada mês...')

for mes in sorted(df_analise['mes'].unique()):
    dados_mes = df_analise[df_analise['mes'] == mes]['qtd']
    n = len(dados_mes)
    media = np.mean(dados_mes)
    se = sp_stats.sem(dados_mes)  # Standard error
    
    # Calcular t-crítico para 95% (α=0.05, bicaudal)
    t_critico = sp_stats.t.ppf(0.975, df=n-1)
    
    # Margem de erro
    margem = t_critico * se
    ic_inf = media - margem
    ic_sup = media + margem
    
    ic_mes_list.append({
        'Mês': nomes_meses[mes],
        'N': n,
        'Média': media,
        'SE': se,
        'IC Inferior': ic_inf,
        'IC Superior': ic_sup,
        'Margem de Erro': margem
    })

df_ic_mes = pd.DataFrame(ic_mes_list)
print(f'✅ IC por mês calculado')
```

### Intervalos de Confiança por Ano

```python
# ========== INTERVALOS DE CONFIANÇA POR ANO ==========

ic_ano_list = []

print(f'\n🔍 Calculando IC para cada ano...')

for ano in sorted(df_analise['ano'].unique()):
    dados_ano = df_analise[df_analise['ano'] == ano]['qtd']
    n = len(dados_ano)
    media = np.mean(dados_ano)
    se = sp_stats.sem(dados_ano)
    
    # Calcular t-crítico
    t_critico = sp_stats.t.ppf(0.975, df=n-1)
    
    # Margem de erro
    margem = t_critico * se
    ic_inf = media - margem
    ic_sup = media + margem
    
    ic_ano_list.append({
        'Ano': int(ano),
        'N': n,
        'Média': media,
        'SE': se,
        'IC Inferior': ic_inf,
        'IC Superior': ic_sup,
        'Margem de Erro': margem
    })

df_ic_ano = pd.DataFrame(ic_ano_list)
print(f'✅ IC por ano calculado')
```

## 4️⃣ ANOVA Fatorial (Mês × Ano)

### O que é ANOVA Fatorial?

A ANOVA (Analysis of Variance) testa se há diferenças significativas nas médias entre grupos.

**Modelo**: Quantidade ~ Mês + Ano + Mês:Ano

**Hipóteses**:
- **H0**: Não há diferença significativa entre os grupos
- **H1**: Há diferença significativa entre os grupos

**Objetivo**: Identificar se mês, ano ou ambos influenciam significativamente a quantidade de bovinos.

```python
# ========== ANOVA FATORIAL ==========

print('\n' + '='*80)
print('📊 ANOVA FATORIAL (MÊS × ANO)')
print('='*80)

# Preparar dados para ANOVA
df_anova = df_analise.copy()
df_anova['mes'] = df_anova['mes'].astype('category')
df_anova['ano'] = df_anova['ano'].astype('category')

# Ajustar modelo linear
print(f'\n🔧 Ajustando modelo: qtd ~ mes + ano + mes:ano')
modelo = ols('qtd ~ C(mes) + C(ano) + C(mes):C(ano)', data=df_anova).fit()

# ANOVA
anova_table = anova_lm(modelo, typ=2)

print(f'✅ Modelo ajustado!')

# Resultado ANOVA
df_anova_resultado = pd.DataFrame({
    'Efeito': ['Mês', 'Ano', 'Mês:Ano'],
    'F-statistic': [
        anova_table.loc['C(mes)', 'F'],
        anova_table.loc['C(ano)', 'F'],
        anova_table.loc['C(mes):C(ano)', 'F']
    ],
    'p-value': [
        anova_table.loc['C(mes)', 'PR(>F)'],
        anova_table.loc['C(ano)', 'PR(>F)'],
        anova_table.loc['C(mes):C(ano)', 'PR(>F)']
    ]
})

# Adicionar significância
df_anova_resultado['Significante (α=0.05)'] = df_anova_resultado['p-value'].apply(
    lambda x: 'SIM ✅' if x < 0.05 else 'NÃO ❌'
)

print(f'\n📈 RESULTADOS DA ANOVA:\n')
print(df_anova_resultado.to_string(index=False))

# Salvar
salvar_dados(df_anova_resultado, '05_ANOVA_fatorial_efeitos')
```

### Qualidade do Modelo

```python
# ========== QUALIDADE DO MODELO ==========

print(f'\n📊 QUALIDADE DO MODELO:')
print(f'  R² (variância explicada):        {modelo.rsquared:.6f}')
print(f'  R² Ajustado:                     {modelo.rsquared_adj:.6f}')
print(f'  F-statistic geral:               {modelo.fvalue:.4f}')
print(f'  p-value geral:                   {modelo.f_pvalue:.4e}')
```

## 5️⃣ Teste Post-Hoc (Tukey HSD)

### O que é Tukey HSD?

Quando ANOVA indica diferenças significativas, o teste de Tukey HSD identifica **quais pares de grupos diferem significativamente**.

**Vantagem**: Controla o erro tipo I quando fazemos múltiplas comparações pairwise.

**Objetivo**: Comparar meses e anos entre si para identificar diferenças significativas.

```python
# ========== TESTE TUKEY HSD ==========

print('\n' + '='*80)
print('🔬 TESTE POST-HOC (TUKEY HSD)')
print('='*80)

# Teste Tukey para MÊS
if df_anova_resultado[df_anova_resultado['Efeito'] == 'Mês']['p-value'].values[0] < 0.05:
    print(f'\n📊 COMPARAÇÕES PAIRWISE ENTRE MESES:')
    
    tukey_mes = pairwise_tukeyhsd(endog=df_anova['qtd'], groups=df_anova['mes'], alpha=0.05)
    print(tukey_mes)
    
    # Converter para DataFrame
    tukey_mes_df = pd.DataFrame(data=tukey_mes.summary().data[1:], columns=tukey_mes.summary().data[0])
    salvar_dados(tukey_mes_df, '05_Tukey_HSD_meses')
else:
    print(f'\n⚠️ Efeito de MÊS não significante (p>0.05)')
    print(f'   Tukey HSD não é necessário para meses')

# Teste Tukey para ANO
if df_anova_resultado[df_anova_resultado['Efeito'] == 'Ano']['p-value'].values[0] < 0.05:
    print(f'\n📅 COMPARAÇÕES PAIRWISE ENTRE ANOS:')
    
    tukey_ano = pairwise_tukeyhsd(endog=df_anova['qtd'], groups=df_anova['ano'], alpha=0.05)
    print(tukey_ano)
    
    # Converter para DataFrame
    tukey_ano_df = pd.DataFrame(data=tukey_ano.summary().data[1:], columns=tukey_ano.summary().data[0])
    salvar_dados(tukey_ano_df, '05_Tukey_HSD_anos')
else:
    print(f'\n⚠️ Efeito de ANO não significante (p>0.05)')
    print(f'   Tukey HSD não é necessário para anos')

print(f'\n✅ Testes Tukey concluídos!')
```
