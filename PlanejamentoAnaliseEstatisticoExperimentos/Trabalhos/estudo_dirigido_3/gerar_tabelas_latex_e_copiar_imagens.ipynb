```python
# Função para sanitizar LaTeX - VERSÃO CORRIGIDA
def sanitizar_latex(texto):
    """Remove caracteres problemáticos e converte para LaTeX compatível em tabelas"""
    if not isinstance(texto, str):
        texto = str(texto)
    
    # Mapa de substituições - SEM colocar em math mode dentro de tabelas
    substituicoes = {
        'α': 'alfa',  # Não usar $\alpha$ em tabelas
        '❌': 'Nao',  # Remover emoji
        '✅': 'Sim',  # Remover emoji
        'ã': 'a',
        'õ': 'o',
        'ç': 'c',
        'é': 'e',
        'á': 'a',
        'ó': 'o',
        'ú': 'u',
        'í': 'i',
        '^': 'circunflexo',
    }
    
    for char, replace in substituicoes.items():
        if char in texto:
            texto = texto.replace(char, replace)
    
    # Escapar caracteres especiais LaTeX
    especiais = {
        '_': r'\_',
        '$': r'\$',
        '%': r'\%',
        '&': r'\&',
        '#': r'\#',
        '{': r'\{',
        '}': r'\}',
        '~': r'\textasciitilde{}',
    }
    
    for char, replace in especiais.items():
        if char in texto:
            texto = texto.replace(char, replace)
    
    return texto

print("✅ Função sanitizar_latex carregada (VERSÃO CORRIGIDA)")
```

## Geração de Tabelas LaTeX

Lê todos os CSVs e converte para arquivos .tex formatados.

```python
# ========== FUNÇÃO PARA ESCAPAR CARACTERES LaTeX ==========
def escape_latex(text):
    """
    Escapa caracteres especiais para LaTeX.
    Trata acentuação e caracteres especiais Unicode.
    """
    text = str(text)
    
    # PRIMEIRA ETAPA: Substituir caracteres acentuados e especiais
    acentos = {
        'ã': 'a',
        'õ': 'o', 
        'ç': 'c',
        'á': 'a',
        'é': 'e',
        'í': 'i',
        'ó': 'o',
        'ú': 'u',
        'â': 'a',
        'ê': 'e',
        'ô': 'o',
        'à': 'a',
        'è': 'e',
        'ì': 'i',
        'ò': 'o',
        'ù': 'u',
        'α': 'alfa',
    }
    
    for char, replace in acentos.items():
        if char in text:
            text = text.replace(char, replace)
    
    # SEGUNDA ETAPA: Escapar caracteres especiais LaTeX
    # Ordem é importante: backslash deve ser primeiro!
    text = text.replace('\\', r'\textbackslash{}')
    text = text.replace('&', r'\&')
    text = text.replace('%', r'\%')
    text = text.replace('$', r'\$')
    text = text.replace('#', r'\#')
    text = text.replace('_', r'\_')
    text = text.replace('{', r'\{')
    text = text.replace('}', r'\}')
    text = text.replace('~', r'\textasciitilde{}')
    text = text.replace('^', r'\textasciicircum{}')
    
    return text
```
