```python
# TRABALHO FINAL: Análise Estatística com Dados Reais de CSV
# ==============================================================
# Objetivo: Aplicar TCL, IC, ANOVA usando dados reais

import os
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Configurar estilos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 7)
plt.rcParams['font.size'] = 10

print("✅ Bibliotecas carregadas com sucesso!")
print("\n📊 ANÁLISE ESTATÍSTICA COM DADOS REAIS")
print("=" * 70)
```

```python
# 1. CARREGAR DADOS DO CSV
# =========================

pasta_dados = r'D:\OneDrive\Pessoais\Doutorado\Cefet\data'

# Verificar se pasta existe
if not os.path.exists(pasta_dados):
    print(f"⚠️  Pasta '{pasta_dados}' não encontrada!")
    print("\n💡 Alternativas:")
    print("   1. Criar arquivo CSV de teste")
    print("   2. Usar caminho alternativo")
    
    # Criar dados de exemplo se pasta não existir
    print("\n🔧 Gerando dados de exemplo...")
    np.random.seed(42)
    
    # Simular dados de um experimento
    num_observacoes = 500
    dados_dict = {
        'Grupo': np.repeat(['Controle', 'Tratamento A', 'Tratamento B'], 
                          num_observacoes // 3),
        'Variavel_1': np.concatenate([
            np.random.exponential(1.0, num_observacoes // 3),
            np.random.exponential(1.5, num_observacoes // 3),
            np.random.exponential(2.0, num_observacoes // 3)
        ]),
        'Variavel_2': np.concatenate([
            np.random.normal(10, 2, num_observacoes // 3),
            np.random.normal(12, 2, num_observacoes // 3),
            np.random.normal(14, 2, num_observacoes // 3)
        ])
    }
    
    df = pd.DataFrame(dados_dict)
    print(f"✅ Dados de exemplo criados: {df.shape[0]} observações, {df.shape[1]} variáveis")
else:
    # Carregar arquivos CSV da pasta
    arquivos_csv = [f for f in os.listdir(pasta_dados) if f.endswith('.csv')]
    
    if not arquivos_csv:
        print(f"❌ Nenhum arquivo CSV encontrado em '{pasta_dados}'")
        # Usar dados de exemplo
        print("\n🔧 Gerando dados de exemplo...")
        np.random.seed(42)
        num_observacoes = 500
        dados_dict = {
            'Grupo': np.repeat(['Controle', 'Tratamento A', 'Tratamento B'], 
                              num_observacoes // 3),
            'Variavel_1': np.concatenate([
                np.random.exponential(1.0, num_observacoes // 3),
                np.random.exponential(1.5, num_observacoes // 3),
                np.random.exponential(2.0, num_observacoes // 3)
            ]),
            'Variavel_2': np.concatenate([
                np.random.normal(10, 2, num_observacoes // 3),
                np.random.normal(12, 2, num_observacoes // 3),
                np.random.normal(14, 2, num_observacoes // 3)
            ])
        }
        df = pd.DataFrame(dados_dict)
        print(f"✅ Dados de exemplo criados: {df.shape[0]} observações, {df.shape[1]} variáveis")
    else:
        # Carregar primeiro arquivo CSV
        arquivo_principal = os.path.join(pasta_dados, arquivos_csv[0])
        df = pd.read_csv(arquivo_principal)
        print(f"✅ Dados carregados: {arquivo_principal}")
        print(f"   Dimensões: {df.shape[0]} linhas × {df.shape[1]} colunas")
        print(f"   Variáveis: {list(df.columns)}")

print(f"\n📊 Dados carregados com sucesso!")
print(f"   Forma: {df.shape}")
print(f"\nPrimeiras 5 linhas:")
print(df.head())
```

```python
# 2. ANÁLISE EXPLORATÓRIA DOS DADOS
# ===================================

print("\n📌 ANÁLISE EXPLORATÓRIA DOS DADOS")
print("-" * 70)

# Estatísticas descritivas
print("\nEstatísticas Descritivas:")
print(df.describe())

# Tipos de dados
print("\nTipos de Dados:")
print(df.dtypes)

# Valores ausentes
print(f"\nValores Ausentes:")
print(df.isnull().sum())

# Selecionar coluna numérica para análise principal
colunas_numericas = df.select_dtypes(include=[np.number]).columns.tolist()

if len(colunas_numericas) == 0:
    print("⚠️  Nenhuma coluna numérica encontrada!")
    coluna_principal = None
else:
    # Usar primeira coluna numérica
    coluna_principal = colunas_numericas[0]
    print(f"\n✅ Coluna selecionada para análise principal: '{coluna_principal}'")
```

```python
# 3. DISTRIBUIÇÃO DA POPULAÇÃO - VERIFICAR ASSIMETRIA
# ====================================================

if coluna_principal is not None:
    populacao = df[coluna_principal].dropna().values
    
    mu_pop = np.mean(populacao)
    sigma_pop = np.std(populacao, ddof=0)
    mediana_pop = np.median(populacao)
    assimetria_pop = stats.skew(populacao)
    curtose_pop = stats.kurtosis(populacao)
    
    print("\n📌 CARACTERÍSTICAS DA POPULAÇÃO")
    print("-" * 70)
    print(f"Coluna analisada: '{coluna_principal}'")
    print(f"Tamanho: {len(populacao):,} observações")
    print(f"Média (μ): {mu_pop:.4f}")
    print(f"Mediana: {mediana_pop:.4f}")
    print(f"Desvio padrão (σ): {sigma_pop:.4f}")
    print(f"Assimetria (skewness): {assimetria_pop:.4f}", end="")
    
    if abs(assimetria_pop) < 0.5:
        print(" → Distribuição aproximadamente simétrica")
    elif assimetria_pop > 0.5:
        print(" → Distribuição assimétrica à direita (cauda positiva)")
    else:
        print(" → Distribuição assimétrica à esquerda (cauda negativa)")
    
    print(f"Curtose (excesso): {curtose_pop:.4f}")
    print(f"Mínimo: {populacao.min():.4f}")
    print(f"Máximo: {populacao.max():.4f}")
    print(f"Amplitude: {populacao.max() - populacao.min():.4f}")
    
    # Visualizar distribuição
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Histograma
    axes[0].hist(populacao, bins=50, color='steelblue', alpha=0.7, edgecolor='black')
    axes[0].axvline(mu_pop, color='red', linestyle='--', linewidth=2, label=f'Média = {mu_pop:.2f}')
    axes[0].axvline(mediana_pop, color='green', linestyle='--', linewidth=2, label=f'Mediana = {mediana_pop:.2f}')
    axes[0].set_xlabel('Valores')
    axes[0].set_ylabel('Frequência')
    axes[0].set_title(f'Distribuição de {coluna_principal}')
    axes[0].legend()
    axes[0].grid(alpha=0.3)
    
    # Q-Q Plot
    stats.probplot(populacao, dist="norm", plot=axes[1])
    axes[1].set_title(f'Q-Q Plot: {coluna_principal} vs Distribuição Normal')
    axes[1].grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n✅ Distribuição visualizada!")
else:
    print("\n❌ Nenhuma coluna numérica disponível para análise!")
```

```python
# 4. TEOREMA CENTRAL DO LIMITE (TCL)
# ===================================

if coluna_principal is not None and len(populacao) >= 30:
    print("\n📌 TEOREMA CENTRAL DO LIMITE")
    print("-" * 70)
    
    # Diferentes tamanhos de amostra
    tamanhos_amostra = [5, 10, 30, 50] if len(populacao) >= 50 else [5, 10, 30]
    num_amostras = min(1000, len(populacao) // 5)  # Ajustar número de amostras
    
    print(f"Número de amostras para cada tamanho: {num_amostras:,}")
    
    resultados_tcl = {}
    
    fig, axes = plt.subplots(1, len(tamanhos_amostra), figsize=(15, 4))
    if len(tamanhos_amostra) == 1:
        axes = [axes]
    
    for idx, n in enumerate(tamanhos_amostra):
        # Gerar amostras e calcular suas médias
        medias = np.array([np.mean(np.random.choice(populacao, size=n, replace=True)) 
                          for _ in range(num_amostras)])
        resultados_tcl[n] = medias
        
        # Estatísticas
        media_das_medias = np.mean(medias)
        dp_das_medias = np.std(medias, ddof=1)
        teorico_dp_das_medias = sigma_pop / np.sqrt(n)
        assimetria_medias = stats.skew(medias)
        
        print(f"\nTamanho de amostra: n = {n}")
        print(f"  Média das médias: {media_das_medias:.4f} (esperado: {mu_pop:.4f})")
        print(f"  DP das médias: {dp_das_medias:.4f} (teórico: {teorico_dp_das_medias:.4f})")
        print(f"  Assimetria: {assimetria_medias:.4f}")
        
        # Visualização
        ax = axes[idx]
        ax.hist(medias, bins=30, color='skyblue', alpha=0.7, edgecolor='black', density=True)
        
        # Sobrepor distribuição normal teórica
        x = np.linspace(medias.min(), medias.max(), 100)
        y_normal = stats.norm.pdf(x, mu_pop, teorico_dp_das_medias)
        ax.plot(x, y_normal, 'r-', linewidth=2, label='Normal teórica')
        
        ax.set_xlabel('Média da Amostra')
        ax.set_ylabel('Densidade')
        ax.set_title(f'TCL: n = {n}')
        ax.legend(fontsize=8)
        ax.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n✅ TCL verificado!")
else:
    print("\n⚠️  Dados insuficientes para TCL!")
```

```python
# 5. INTERVALO DE CONFIANÇA (IC)
# ===============================

if coluna_principal is not None and len(populacao) >= 30:
    print("\n📌 INTERVALO DE CONFIANÇA")
    print("-" * 70)
    
    # Parâmetros do IC
    nivel_confianca = 0.95
    alpha = 1 - nivel_confianca
    tamanho_amostra_ic = min(50, len(populacao) // 5)
    num_amostras_ic = min(500, len(populacao) // 10)
    
    # Gerar amostras
    amostras_ic = [np.random.choice(populacao, size=tamanho_amostra_ic, replace=True) 
                   for _ in range(num_amostras_ic)]
    
    # Calcular ICs
    ics = []
    contem_media = 0
    
    for amostra in amostras_ic:
        x_bar = np.mean(amostra)
        s = np.std(amostra, ddof=1)
        se = s / np.sqrt(tamanho_amostra_ic)
        t_critico = stats.t.ppf(1 - alpha/2, df=tamanho_amostra_ic - 1)
        margem_erro = t_critico * se
        ic_inf = x_bar - margem_erro
        ic_sup = x_bar + margem_erro
        ics.append((ic_inf, ic_sup))
        
        if ic_inf <= mu_pop <= ic_sup:
            contem_media += 1
    
    print(f"Tamanho de amostra: n = {tamanho_amostra_ic}")
    print(f"Número de amostras: {num_amostras_ic}")
    print(f"Nível de confiança: {nivel_confianca*100:.0f}%")
    print(f"\nICs que contêm a média populacional: {contem_media}/{num_amostras_ic} ({contem_media/num_amostras_ic*100:.1f}%)")
    print(f"Esperado: ~{nivel_confianca*100:.0f}%")
    
    # Visualizar ICs
    num_vis = min(100, len(ics))
    fig, ax = plt.subplots(figsize=(12, 8))
    
    for i in range(num_vis):
        ic_inf, ic_sup = ics[i]
        cor = 'green' if ic_inf <= mu_pop <= ic_sup else 'red'
        ax.plot([ic_inf, ic_sup], [i, i], color=cor, alpha=0.6, linewidth=1)
        ax.plot(np.mean([ic_inf, ic_sup]), i, 'o', color=cor, markersize=3)
    
    ax.axvline(mu_pop, color='blue', linestyle='--', linewidth=2, label=f'Média = {mu_pop:.2f}')
    ax.set_xlabel('Valor')
    ax.set_ylabel(f'Amostra (primeiras {num_vis})')
    ax.set_title(f'Intervalos de Confiança {nivel_confianca*100:.0f}% (n={tamanho_amostra_ic})')
    ax.legend()
    ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    print("\n✅ ICs calculados!")
else:
    print("\n⚠️  Dados insuficientes para IC!")
```

```python
# 6. ANOVA - ANÁLISE DE VARIÂNCIA
# ================================

# Procurar coluna de grupo/categoria
colunas_categoricas = df.select_dtypes(include=['object']).columns.tolist()

if coluna_principal is not None and len(colunas_categoricas) > 0 and len(df) >= 30:
    print("\n📌 ANOVA - ANÁLISE DE VARIÂNCIA")
    print("-" * 70)
    
    # Usar primeira coluna categórica
    coluna_grupo = colunas_categoricas[0]
    
    # Criar grupos
    grupos = []
    labels_grupos = []
    
    for grupo_nome in df[coluna_grupo].unique():
        dados_grupo = df[df[coluna_grupo] == grupo_nome][coluna_principal].dropna().values
        if len(dados_grupo) > 0:
            grupos.append(dados_grupo)
            labels_grupos.append(str(grupo_nome))
    
    if len(grupos) >= 2:
        print(f"Grupos analisados: '{coluna_grupo}'")
        print(f"Variável: '{coluna_principal}'")
        print(f"\nEstatísticas dos grupos:")
        
        for label, grupo in zip(labels_grupos, grupos):
            print(f"\n{label}:")
            print(f"  n = {len(grupo)}")
            print(f"  Média: {np.mean(grupo):.4f}")
            print(f"  DP: {np.std(grupo, ddof=1):.4f}")
        
        # Realizar ANOVA
        f_stat, p_value = stats.f_oneway(*grupos)
        
        print(f"\n{'RESULTADO DA ANOVA':^70}")
        print(f"Estatística F: {f_stat:.4f}")
        print(f"p-value: {p_value:.6f}")
        
        if p_value < 0.05:
            print("✅ Resultado: SIGNIFICATIVO (rejeita-se H₀)")
            print("   Conclusão: Há diferenças estatísticas significativas entre os grupos")
        else:
            print("❌ Resultado: NÃO SIGNIFICATIVO (falha em rejeitar H₀)")
            print("   Conclusão: Não há diferenças estatísticas significativas entre os grupos")
        
        # Visualizações
        fig, axes = plt.subplots(1, 2, figsize=(14, 6))
        
        # Boxplot
        bp = axes[0].boxplot(grupos, labels=labels_grupos, patch_artist=True)
        cores = plt.cm.Set3(np.linspace(0, 1, len(grupos)))
        for patch, color in zip(bp['boxes'], cores):
            patch.set_facecolor(color)
        axes[0].set_ylabel('Valores')
        axes[0].set_title(f'Boxplot dos Grupos')
        axes[0].grid(alpha=0.3, axis='y')
        plt.setp(axes[0].xaxis.get_majorticklabels(), rotation=15, ha='right')
        
        # Histogramas sobrepostos
        for grupo, label, color in zip(grupos, labels_grupos, cores):
            axes[1].hist(grupo, bins=15, alpha=0.5, label=label, color=color, edgecolor='black')
        axes[1].set_xlabel('Valores')
        axes[1].set_ylabel('Frequência')
        axes[1].set_title('Distribuição dos Grupos')
        axes[1].legend()
        axes[1].grid(alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        print("\n✅ ANOVA realizada com sucesso!")
    else:
        print(f"\n⚠️  Número insuficiente de grupos para ANOVA ({len(grupos)} grupos)")
else:
    print("\n⚠️  Dados insuficientes para ANOVA!")
    print("   (Necessário: coluna numérica e coluna categórica)")
```

```python
# 7. RESUMO FINAL
# ===============

print("\n" + "="*70)
print("📊 RESUMO FINAL DA ANÁLISE")
print("="*70)

if coluna_principal is not None:
    print(f"\n✅ Análises Realizadas:")
    print(f"   1. Análise Exploratória dos Dados (n={len(populacao)})")
    print(f"   2. Verificação de Distribuição (Assimetria={assimetria_pop:.2f})")
    
    if len(populacao) >= 30:
        print(f"   3. Teorema Central do Limite (TCL)")
        print(f"   4. Intervalo de Confiança (95%)")
    
    if len(grupos) >= 2:
        print(f"   5. ANOVA (F={f_stat:.2f}, p={p_value:.4f})")
    
    print(f"\n📈 Características dos Dados:")
    print(f"   - Média: {mu_pop:.4f}")
    print(f"   - Desvio padrão: {sigma_pop:.4f}")
    print(f"   - Assimetria: {assimetria_pop:.4f}", end="")
    if abs(assimetria_pop) < 0.5:
        print(" (Distribuição aproximadamente simétrica)")
    elif assimetria_pop > 0:
        print(" (Assimétrica à direita)")
    else:
        print(" (Assimétrica à esquerda)")
    
    print(f"\n✅ Análise concluída com sucesso!")
else:
    print("\n❌ Nenhuma análise foi realizada - dados insuficientes!")

print("\n" + "="*70)
```
