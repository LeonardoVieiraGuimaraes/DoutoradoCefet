# 📊 Explorador de Arquivos CSV - Informações de Dados

## Objetivo
Criar uma ferramenta que **automaticamente explore e exiba informações estruturadas** sobre todos os arquivos CSV de um diretório específico.

---

## 🎯 O que Será Feito

Este notebook vai:
1. ✅ Verificar se a pasta de dados existe
2. ✅ Procurar todos os arquivos CSV na pasta
3. ✅ Para cada arquivo CSV, extrair:
   - 📏 Tamanho do arquivo (em MB)
   - 📊 Número de linhas (registros)
   - 📋 Número de colunas (variáveis)
   - 🏷️ Nomes de todas as variáveis
4. ✅ Exibir um relatório formatado e colorido

````python
# ============================================
# PASSO 1: IMPORTAR BIBLIOTECAS NECESSÁRIAS
# ============================================
import os
import pandas as pd
from pathlib import Path

print("✅ Bibliotecas importadas com sucesso!")
print("   - os: para manipulação de arquivos e diretórios")
print("   - pandas: para leitura e análise de dados CSV")
print("   - pathlib: para trabalhar com caminhos de forma mais segura")
print()
````

## Passo 2: Definir a Função Principal

A função `info_csv()` faz todo o trabalho de exploração. Ela:

1. **Recebe um parâmetro:** `pasta` (caminho para o diretório)
2. **Valida a existência:** Verifica se a pasta existe
3. **Procura CSVs:** Lista todos os arquivos .csv
4. **Extrai informações:** Para cada arquivo, coleta tamanho, linhas, colunas e variáveis
5. **Exibe relatório:** Mostra tudo de forma organizada

---

````python
def info_csv(pasta=r'D:\OneDrive\Pessoais\Doutorado\Cefet\data'):
    """
    Explora e exibe informações sobre todos os arquivos CSV em um diretório.
    
    Parâmetros
    ----------
    pasta : str
        Caminho para o diretório contendo os arquivos CSV
        Padrão: D:\OneDrive\Pessoais\Doutorado\Cefet\data
    
    Retorna
    -------
    None - apenas exibe as informações formatadas
    
    Exemplos
    --------
    >>> info_csv()  # Usa o caminho padrão
    >>> info_csv('C:/dados')  # Usa um caminho customizado
    """
    
    # ==========================================
    # ETAPA 1: VALIDAR SE A PASTA EXISTE
    # ==========================================
    print("\n" + "=" * 60)
    print("🔍 EXPLORADOR DE ARQUIVOS CSV")
    print("=" * 60)
    print(f"\n📂 Buscando CSV em: {pasta}\n")
    
    if not os.path.exists(pasta):
        print(f"❌ ERRO: Pasta '{pasta}' não encontrada!")
        print(f"   Verifique se o caminho está correto e tente novamente.")
        return
    
    print("✅ Pasta encontrada!")
    
    # ==========================================
    # ETAPA 2: PROCURAR TODOS OS ARQUIVOS CSV
    # ==========================================
    arquivos = [f for f in os.listdir(pasta) if f.endswith('.csv')]
    
    if not arquivos:
        print("❌ Nenhum arquivo CSV encontrado nesta pasta!")
        return
    
    print(f"✅ Encontrados {len(arquivos)} arquivo(s) CSV\n")
    
    # ==========================================
    # ETAPA 3: EXTRAIR INFORMAÇÕES DE CADA CSV
    # ==========================================
    print("=" * 60)
    print("📊 INFORMAÇÕES DOS ARQUIVOS CSV")
    print("=" * 60)
    
    for i, arquivo in enumerate(arquivos, 1):
        caminho_completo = os.path.join(pasta, arquivo)
        
        try:
            # --- CÁLCULO 1: Tamanho do arquivo ---
            tamanho_bytes = os.path.getsize(caminho_completo)
            tamanho_mb = tamanho_bytes / (1024 * 1024)
            
            # --- CÁLCULO 2: Leitura do CSV e coleta de metadados ---
            df = pd.read_csv(caminho_completo)
            
            num_linhas = df.shape[0]
            num_colunas = df.shape[1]
            nomes_variaveis = list(df.columns)
            
            # --- EXIBIÇÃO: Formatação e apresentação ---
            print(f"\n{i}. 📄 {arquivo}")
            print(f"   ├─ 💾 Tamanho: {tamanho_mb:.2f} MB ({tamanho_bytes:,} bytes)")
            print(f"   ├─ 📊 Linhas: {num_linhas:,} registros")
            print(f"   ├─ 📋 Colunas: {num_colunas} variáveis")
            print(f"   └─ 🏷️  Variáveis:")
            
            # Listar as variáveis de forma organizada (até 5 por linha)
            for j in range(0, len(nomes_variaveis), 5):
                variaveis_grupo = nomes_variaveis[j:j+5]
                indentacao = "      "
                print(f"      {', '.join(variaveis_grupo)}")
            
        except Exception as e:
            print(f"\n❌ {arquivo}: Erro ao processar")
            print(f"   Motivo: {e}")
    
    # ==========================================
    # ETAPA 4: RESUMO FINAL
    # ==========================================
    print("\n" + "=" * 60)
    print("✅ RESUMO")
    print("=" * 60)
    print(f"Total de arquivos CSV processados: {len(arquivos)}")
    print("=" * 60 + "\n")


# ============================================
# PASSO 3: EXECUTAR A FUNÇÃO
# ============================================
print("\n🚀 Iniciando exploração de dados...\n")
info_csv()
````

````python
# ============================================
# BÔNUS: Função Alternativa Mais Detalhada
# ============================================

def info_csv_detalhada(pasta=r'D:\OneDrive\Pessoais\Doutorado\Cefet\data'):
    """
    Versão DETALHADA com estatísticas adicionais.
    Mostra tipos de dados e primeiras linhas.
    """
    
    if not os.path.exists(pasta):
        print(f"❌ Pasta '{pasta}' não encontrada!")
        return
    
    arquivos = [f for f in os.listdir(pasta) if f.endswith('.csv')]
    
    if not arquivos:
        print("❌ Nenhum arquivo CSV encontrado!")
        return
    
    print("\n" + "=" * 70)
    print("🔍 EXPLORADOR DETALHADO DE ARQUIVOS CSV")
    print("=" * 70)
    
    for arquivo in arquivos:
        caminho = os.path.join(pasta, arquivo)
        
        try:
            df = pd.read_csv(caminho)
            tamanho_mb = os.path.getsize(caminho) / (1024 * 1024)
            
            print(f"\n📄 {arquivo}")
            print(f"   Tamanho: {tamanho_mb:.2f} MB")
            print(f"   Linhas: {df.shape[0]:,} | Colunas: {df.shape[1]}")
            
            # Tipos de dados
            print(f"\n   📋 Tipos de Dados:")
            for col, dtype in df.dtypes.items():
                print(f"      {col}: {dtype}")
            
            # Estatísticas básicas para colunas numéricas
            print(f"\n   📊 Estatísticas (primeiras colunas numéricas):")
            numeric_df = df.select_dtypes(include=['number'])
            if not numeric_df.empty:
                print(numeric_df.describe().to_string())
            
            # Valores faltantes
            print(f"\n   🔍 Valores Faltantes:")
            faltantes = df.isnull().sum()
            if faltantes.sum() == 0:
                print(f"      ✅ Nenhum valor faltante!")
            else:
                print(faltantes[faltantes > 0])
            
        except Exception as e:
            print(f"\n❌ Erro ao processar {arquivo}: {e}")
    
    print("\n" + "=" * 70)


# Descomente a linha abaixo para usar a versão detalhada
# info_csv_detalhada()
````
