# Estimadores Não-Viciados e Teorema Central do Limite (TCL)

Este notebook explora o conceito de **Estimadores Não-Viciados** através de simulações computacionais.

## Teoria

Um **Estimador Não-Viciado** é aquele cuja Esperança é igual ao parâmetro populacional.

### Conceitos Importantes:

1. **Média Amostral ($\bar{X}$)**: Estimador não-viciado para μ
   - $E(\bar{X}) = \mu$

2. **Desvio Padrão Viciado ($s_e$)**: Utiliza denominador n
   - $s_e = \sqrt{\frac{\sum(x_i - \bar{x})^2}{n}}$
   - Tende a **subestimar** σ

3. **Desvio Padrão Não-Viciado (s)**: Utiliza denominador n-1
   - $s = \sqrt{\frac{\sum(x_i - \bar{x})^2}{n-1}}$
   - Estimador não-viciado para σ

## Importação das Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configuração de estilo para os gráficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 6)

print("Bibliotecas importadas com sucesso!")
```

# Parte 1: Simulação da Média Amostral - $E(\bar{X})$

## Objetivo

Demonstrar que a **média das médias amostrais** converge para $\mu = 180$.

Isso comprova que $\bar{X}$ é um estimador não-viciado para $\mu$.

```python
# Parâmetros da Simulação
TRUE_MEAN = 180
TRUE_STD = 20
POP_SIZE = 100000
SAMPLE_SIZE = 100
NUM_SIMULATIONS = 10000

# Geração da População
np.random.seed(42)
population = np.random.normal(TRUE_MEAN, TRUE_STD, POP_SIZE)

print(f"População gerada:")
print(f"  Tamanho: {POP_SIZE:,}")
print(f"  Média verdadeira (μ): {TRUE_MEAN}")
print(f"  Desvio padrão (σ): {TRUE_STD}")
print(f"\nSimulação:")
print(f"  Tamanho de cada amostra: {SAMPLE_SIZE}")
print(f"  Número de amostras: {NUM_SIMULATIONS:,}")
```

```python
# Coleta de amostras e cálculo das médias amostrais
sample_means = []

for i in range(NUM_SIMULATIONS):
    sample = np.random.choice(population, SAMPLE_SIZE, replace=False)
    sample_means.append(np.mean(sample))

# Média das médias amostrais (E(X̄))
mean_of_sample_means = np.mean(sample_means)

print("="*70)
print("RESULTADOS - MÉDIA AMOSTRAL")
print("="*70)
print(f"Média das Médias Amostrais E(X̄): {mean_of_sample_means:.4f}")
print(f"Valor Teórico (μ): {TRUE_MEAN}")
print(f"Diferença: {abs(mean_of_sample_means - TRUE_MEAN):.4f}")
print("="*70)
print(f"\n✓ A média das médias amostrais está muito próxima de μ = {TRUE_MEAN}!")
print("✓ Isso confirma que X̄ é um estimador NÃO-VICIADO de μ.")
```

## Visualização: Distribuição das Médias Amostrais

Vamos visualizar a distribuição das médias amostrais e verificar sua aproximação com a média populacional.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 1: Histograma das Médias Amostrais
axes[0].hist(sample_means, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
axes[0].axvline(TRUE_MEAN, color='red', linewidth=2, linestyle='--', label=f'μ verdadeiro = {TRUE_MEAN}')
axes[0].axvline(mean_of_sample_means, color='green', linewidth=2, linestyle='-', 
                label=f'E(X̄) = {mean_of_sample_means:.2f}')
axes[0].set_xlabel('Valor da Média Amostral', fontsize=12)
axes[0].set_ylabel('Densidade', fontsize=12)
axes[0].set_title('Distribuição das Médias Amostrais', fontsize=14, fontweight='bold')
axes[0].legend(fontsize=11)
axes[0].grid(True, alpha=0.3)

# Gráfico 2: Convergência da Média das Médias
cumulative_means = np.cumsum(sample_means) / np.arange(1, NUM_SIMULATIONS + 1)
axes[1].plot(cumulative_means, color='darkblue', linewidth=1.5, label='Média Acumulada de X̄')
axes[1].axhline(TRUE_MEAN, color='red', linewidth=2, linestyle='--', label=f'μ = {TRUE_MEAN}')
axes[1].set_xlabel('Número de Simulações', fontsize=12)
axes[1].set_ylabel('Média das Médias Amostrais', fontsize=12)
axes[1].set_title('Convergência de E(X̄) para μ', fontsize=14, fontweight='bold')
axes[1].legend(fontsize=11)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n📊 Os gráficos demonstram:")
print("   1. A distribuição das médias amostrais segue uma distribuição normal (TCL)")
print("   2. A média converge para μ conforme aumentamos o número de simulações")
```

# Parte 2: Simulação do Viés do Estimador de Desvio Padrão

## Objetivo

Demonstrar que:
1. O estimador **viciado** ($s_e$) com denominador **n** **subestima** $\sigma = 20$
2. O estimador **não-viciado** (s) com denominador **n-1** estima corretamente $\sigma = 20$

```python
# Função para o estimador viciado (denominador n)
def se_biased(x):
    """Calcula o desvio padrão viciado usando denominador n"""
    return np.std(x, ddof=0)  # ddof=0 usa denominador n

print("Função se_biased() definida.")
print("\nExplicação dos parâmetros:")
print("  - ddof=0: Usa denominador n (VICIADO)")
print("  - ddof=1: Usa denominador n-1 (NÃO-VICIADO)")
```

```python
# Parâmetros (mesmos da simulação anterior)
print(f"Usando os mesmos parâmetros:")
print(f"  σ verdadeiro: {TRUE_STD}")
print(f"  Tamanho da amostra: {SAMPLE_SIZE}")
print(f"  Número de simulações: {NUM_SIMULATIONS:,}")
print(f"\nIniciando simulação...")
```

```python
# Coleta de amostras e cálculo dos desvios padrão
sample_stds_biased = []    # Estimador viciado (denominador n)
sample_stds_unbiased = []  # Estimador não-viciado (denominador n-1)

for i in range(NUM_SIMULATIONS):
    sample = np.random.choice(population, SAMPLE_SIZE, replace=False)
    sample_stds_biased.append(se_biased(sample))           # ddof=0
    sample_stds_unbiased.append(np.std(sample, ddof=1))    # ddof=1

# Médias dos estimadores
mean_of_biased_stds = np.mean(sample_stds_biased)
mean_of_unbiased_stds = np.mean(sample_stds_unbiased)

print("="*70)
print("RESULTADOS - DESVIO PADRÃO")
print("="*70)
print(f"Média do Estimador VICIADO (sₑ, ddof=0):     {mean_of_biased_stds:.4f}")
print(f"Média do Estimador NÃO-VICIADO (s, ddof=1):  {mean_of_unbiased_stds:.4f}")
print(f"Valor Teórico (σ):                           {TRUE_STD}")
print("="*70)
print(f"\nViés do estimador viciado:     {mean_of_biased_stds - TRUE_STD:.4f}")
print(f"Viés do estimador não-viciado:  {mean_of_unbiased_stds - TRUE_STD:.4f}")
print("="*70)
print(f"\n✓ O estimador VICIADO (ddof=0) subestima σ em aproximadamente {TRUE_STD - mean_of_biased_stds:.4f}")
print(f"✓ O estimador NÃO-VICIADO (ddof=1) está muito próximo do valor real de σ!")
```

## Visualização: Comparação dos Estimadores

Vamos visualizar as distribuições dos dois estimadores e compará-las com o valor verdadeiro de σ.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 1: Histogramas sobrepostos
axes[0].hist(sample_stds_biased, bins=50, density=True, alpha=0.6, color='red', 
             label=f'Viciado (ddof=0): {mean_of_biased_stds:.2f}', edgecolor='black')
axes[0].hist(sample_stds_unbiased, bins=50, density=True, alpha=0.6, color='green', 
             label=f'Não-Viciado (ddof=1): {mean_of_unbiased_stds:.2f}', edgecolor='black')
axes[0].axvline(TRUE_STD, color='blue', linewidth=2, linestyle='--', 
                label=f'σ verdadeiro = {TRUE_STD}')
axes[0].set_xlabel('Valor do Desvio Padrão Amostral', fontsize=12)
axes[0].set_ylabel('Densidade', fontsize=12)
axes[0].set_title('Distribuição dos Estimadores de Desvio Padrão', fontsize=14, fontweight='bold')
axes[0].legend(fontsize=11)
axes[0].grid(True, alpha=0.3)

# Gráfico 2: Boxplot comparativo
data_to_plot = [sample_stds_biased, sample_stds_unbiased]
bp = axes[1].boxplot(data_to_plot, labels=['Viciado\n(ddof=0)', 'Não-Viciado\n(ddof=1)'],
                     patch_artist=True, widths=0.6)
bp['boxes'][0].set_facecolor('red')
bp['boxes'][0].set_alpha(0.6)
bp['boxes'][1].set_facecolor('green')
bp['boxes'][1].set_alpha(0.6)
axes[1].axhline(TRUE_STD, color='blue', linewidth=2, linestyle='--', 
                label=f'σ = {TRUE_STD}')
axes[1].set_ylabel('Desvio Padrão Amostral', fontsize=12)
axes[1].set_title('Comparação: Estimador Viciado vs Não-Viciado', fontsize=14, fontweight='bold')
axes[1].legend(fontsize=11)
axes[1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\n📊 Os gráficos demonstram claramente:")
print("   1. O estimador VICIADO (vermelho) sistematicamente subestima σ")
print("   2. O estimador NÃO-VICIADO (verde) está centrado em σ = 20")
print("   3. A diferença entre os dois é consistente ao longo das simulações")
```

## Análise Estatística Detalhada

```python
# Cálculo de estatísticas descritivas
print("="*70)
print("ESTATÍSTICAS DESCRITIVAS")
print("="*70)

print("\n📊 Estimador VICIADO (ddof=0):")
print(f"   Média:           {np.mean(sample_stds_biased):.4f}")
print(f"   Mediana:         {np.median(sample_stds_biased):.4f}")
print(f"   Desvio Padrão:   {np.std(sample_stds_biased):.4f}")
print(f"   Mínimo:          {np.min(sample_stds_biased):.4f}")
print(f"   Máximo:          {np.max(sample_stds_biased):.4f}")
print(f"   Viés:            {np.mean(sample_stds_biased) - TRUE_STD:.4f}")

print("\n📊 Estimador NÃO-VICIADO (ddof=1):")
print(f"   Média:           {np.mean(sample_stds_unbiased):.4f}")
print(f"   Mediana:         {np.median(sample_stds_unbiased):.4f}")
print(f"   Desvio Padrão:   {np.std(sample_stds_unbiased):.4f}")
print(f"   Mínimo:          {np.min(sample_stds_unbiased):.4f}")
print(f"   Máximo:          {np.max(sample_stds_unbiased):.4f}")
print(f"   Viés:            {np.mean(sample_stds_unbiased) - TRUE_STD:.4f}")

print("\n📈 Diferença entre os estimadores:")
diff = np.mean(sample_stds_unbiased) - np.mean(sample_stds_biased)
print(f"   Diferença absoluta: {diff:.4f}")
print(f"   Diferença relativa: {(diff/TRUE_STD)*100:.2f}%")
print("="*70)
```

# Conclusões

## 1. Média Amostral ($\bar{X}$)

- ✅ **$\bar{X}$ é um estimador NÃO-VICIADO de μ**
- A média das médias amostrais converge para o valor verdadeiro μ = 180
- A distribuição das médias amostrais segue uma distribuição normal (Teorema Central do Limite)

## 2. Desvio Padrão Amostral

### Estimador Viciado ($s_e$, ddof=0):
- ❌ **Sistematicamente SUBESTIMA σ**
- Usa denominador n: $s_e = \sqrt{\frac{\sum(x_i - \bar{x})^2}{n}}$
- Não deve ser usado para inferência estatística

### Estimador Não-Viciado (s, ddof=1):
- ✅ **Estimador NÃO-VICIADO de σ**
- Usa denominador n-1: $s = \sqrt{\frac{\sum(x_i - \bar{x})^2}{n-1}}$
- Correção de Bessel: o uso de n-1 compensa o viés
- **Sempre use ddof=1 em Python/NumPy para análises estatísticas!**

## 3. Importância Prática

Ao realizar análises estatísticas:
- Use `np.std(x, ddof=1)` para calcular o desvio padrão amostral
- Use `np.var(x, ddof=1)` para calcular a variância amostral
- O ddof=1 garante estimadores não-viciados

Esta simulação demonstra **empiricamente** por que a correção de Bessel (n-1) é necessária!
