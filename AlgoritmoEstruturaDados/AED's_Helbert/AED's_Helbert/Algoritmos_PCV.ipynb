```python
#bloco de definição de variáveis e funções (Força Bruta e vizinho mais próximo)

import numpy as np
import time
import timeit
import random
from datetime import datetime

#bloco de definição de variáveis e funções


def fb(grafo, s):
    vertex = []
    vertex2 = []
    for i in range(v): #Gera vetor dos vértices
        if i != s:
            vertex.append(i)
        
    custo_min = 999999999
    iter = 0
    vertex2=vertex.copy()
    solucao = [0,vertex,0]
    while True:

        custo_atual = 0
        k = s
        for i in range(len(vertex)):
            custo_atual += grafo[k][vertex[i]] #Atualiza custo da rota da permutação analisada
            k = vertex[i]
        custo_atual += grafo[k][s]
        if custo_atual < custo_min:
            vertex2=vertex.copy()
            solucao=np.array([0,vertex2,0])
        custo_min = min(custo_min, custo_atual) #Caso o custo atual seja menor do que o da permutação de custo mínimo, trocar
        #print (iter, ' - Solucao testada: x=  ', [0, vertex, 0],"f(x)= ", custo_atual) #imprimir a solução atual na tela
        
        iter = iter + 1
        if not permutacao(vertex): #Gerar nova permutação para avaliar a solução
            break
    return [custo_min,solucao, iter]


def permutacao(l):
    n = len(l)
    i = n - 2

    while i >= 0 and l[i] > l[i + 1]:
        i -= 1

    if i == -1:
        return False

    j = i + 1
    while j < n and l[j] > l[i]:
        j += 1

    j -= 1

    #Permutando posições e gerando as novas soluções para analisar
    
    l[i], l[j] = l[j], l[i]
    left = i + 1
    right = n - 1

    while left < right:
        l[left], l[right] = l[right], l[left]
        left += 1
        right -= 1
    return True

def retornaMatrizSimestricaAleatoria(tamanho):
    #tamanho=100;

    m=np.zeros( (tamanho, tamanho) )
    for i in range(tamanho):
        for j in range(tamanho):
            if j!=i:
                m[i][j] = random.randint(1,10)
            
            if m[i][j]!=m[j][i]:
                m[j][i]=m[i][j]

    return m
    #print (m)

 #Funções para leitura das instâncias

def ler_dados(filename):
    if filename == 'si535.tsp' or filename == 'si1032.tsp':
        f = open(filename) #abre o arquivo
        line = f.readline()
        while line.find("DIMENSION") == -1: #procura a dimensão para salvar a qtde de cidades
            line = f.readline()
        n = int(line.split()[-1])
        c = np.zeros([n,n])
        i,j = 0,0

        while line.find("EDGE_WEIGHT_SECTION") == -1: #começa a ler os pesos após os textos iniciais
            line = f.readline()
        data = '0'
        while data != "EOF": #Critério para parar de ler os números das linhas 
            line=f.readline() #ler a linha
            for data in line.split(): #faz um split do conteúdo e lê cada número
                if data != 'EOF':
                    c[i,j]=int(data) #atribuições para formar matriz simétrica de pesos
                    c[j,i]=int(data)  #atribuições para formar matriz simétrica de pesos
                j+= 1 #incrementa próxima coluna da matriz de pesos
                if j>=n: #incrementa próxima linhada matriz de pesos
                    i+= 1
                    j=i
    elif filename == 'pa561.tsp':
        f = open(filename) #abre o arquivo
        line = f.readline()
        while line.find("DIMENSION")==-1: #procura a dimensão para salvar a qtde de cidades
            line = f.readline()
        n=int(line.split()[-1])
        c=np.zeros([n,n])
        i=0
        
        while line.find("EDGE_WEIGHT_SECTION")==-1: #começa a ler os pesos após os textos iniciais
            line=f.readline()
        data='0'
        while data!='DISPLAY_DATA_SECTION': #Critério para parar de ler os números das linhas 
            j=0
            line = f.readline() #ler a linha
            for data in line.split(): #faz um split do conteúdo da linha e lê cada número
                if data!='DISPLAY_DATA_SECTION':
                    c[i,j]=int(data)
                    c[j,i]=int(data)
                    j+=1
            i+=1
    else: return('Instância desconhecida')
    
    return c   


def proxvertice(grafo,s, visits): #funcao para buscar a aresta de menor caminho, a partir de um vertice selecionado s
    indice = 0
    menor = 99999999
    n=grafo.shape[0]
    
    for i in range(n):
        if i!= s and i not in visits: #exclui da verificação o vértice atual e os vértices já visitados
            if grafo[i,s]<menor:
                menor = grafo[i,s]
                indice = i
    return indice, menor
    

def vizproximo(grafo,s):
    print("-------------- Construção Gulosa pelo Vizinho mais Próximo ---------------------")
    print('Vértice inicial: ',s)
    ultimo = s
    n=grafo.shape[0]
    vertex=[]
    custo = 0
    for i in range(n): #Gera vetor dos vértices
        if i != s:
            vertex.append(i)
    
    visits = []
    visits.append(s) #marca o primeiro vértice como visitado
    
    for i in range(n-1):
        aux = proxvertice(grafo,s, visits) #indice e distancia do vértice mais próximo de s (ainda não visitado)
        s=aux[0]
        custo = custo+aux[1]
        visits.append(s) #marca o vértice escolhido como visitado
        print ('Próximo vértice: ', s, 'Custo atual: ', custo)
    
    custo = custo+grafo[s,ultimo] #soma o custo para voltar ao inicio
    visits.append(ultimo)

    print('Último vértice: ', ultimo,'Custo total: ', custo)
    
    return visits


    
```

```python
#Heurísticas Computacionais
import random
from datetime import datetime



def avaliarsolucao(grafo, solucao): #função para calcular o custo de uma rota para o PCV
    n=grafo.shape[0]
    f=0
    for i in range (n):
        f=f+grafo[solucao[i],solucao[i+1]]
    return f

def descidarand(grafo, s_ini, itermax): #Descida Randômica, o vizinho é definido pela troca de um par, selecionado aleatoriamente
    s=s_ini
    n=grafo.shape[0]
    iter = 0
    #print('Solução inicial: ', s,' Distância: ', avaliarsolucao(grafo, s))
    #print('Distância Inicial: ', avaliarsolucao(grafo, s))
    while iter < itermax:
        #print(iter)
        #random.seed(datetime.now())
        i=random.randint(1,n-1) #não considera o primeiro e o ultimo
        #random.seed(datetime.now())
        j=random.randint(1,n-1)
        while j==i: #não selecionar mesma posição
            #random.seed(datetime.now())
            j=random.randint(1,n-1)
        #print('i= ', i, 'j= ', j)
        slinha= s.copy()
        aux = slinha[i]
        slinha[i]=slinha[j]
        slinha[j]=aux
        dist1= avaliarsolucao(grafo,s)
        dist2= avaliarsolucao(grafo,slinha)
        #print(dist2)
        if dist2 < dist1: #se a troca melhor a função objetivo, trocar
            s=slinha.copy()
            iter = 0
            print('Aplicando Busca Local (Descida Randômica) - Solução atualizada (distância): ', dist2, 'Vértices trocados: ', s[i],' e ', s[j])
            dist1=dist2
            #print('Solução atualizada: - Nova Distância: ', dist2)
        else:
            slinha=s.copy()
            iter +=1
    return s
                
def descida(grafo, s_ini): #Descida Completa - vizinho: troca contígua
    #são trocadas as posições contíguas
    s=s_ini
    n=grafo.shape[0]
    #iter = 0
    #print('Solução inicial: ', s,' Distância: ', avaliarsolucao(grafo, s))
    #print('Distância Inicial: ', avaliarsolucao(grafo, s))
    i=1
    while i < n-2:
        j=i+1
        #print('i= ', i, 'j= ', j)
        slinha= s.copy()
        aux = slinha[i]
        slinha[i]=slinha[j]
        slinha[j]=aux
        dist1= avaliarsolucao(grafo,s)
        dist2= avaliarsolucao(grafo,slinha)
        #print(dist2)
        if dist2 < dist1: #se a troca melhorar a função objetivo, trocar
            s=slinha.copy()
            dist1=dist2
            print('Aplicando Busca Local (Descida) - Solução atualizada: - Nova Distância: ', dist1, 'Posições trocadas: ', i,' e ', j)
            i=1
            #print('Solução atualizada: ', s, 'Distância: ', dist2, 'Vértices trocados: ', s[i],' e ', s[j])
                
        else:
            slinha=s.copy()
            i+=1
    return s

def ils(grafo,s_ini, itermax, pert): #Metaherística Iterated Local Seach - itermax (Critério da Busca Local), pert (número máx. de perturbações)
       
        n=grafo.shape[0]
        s=0
        
        #Busca Local
        #x=descida(grafo, s_ini)
        x=descidarand(grafo,s_ini, itermax)
        dist1=avaliarsolucao(grafo,x)
        
        k=1
        while k <= pert:
        #for k in range(pert):
            #criar vetor de trocas
            print("Perturbação: ", k, "pares trocados")
            trocas=[] #vetor de trocas inicial nulo
            m=1
            while m <= 2*k: #montar vetor com os indices para efetuar as trocas de acordo com o nível da perturbação
                #random.seed(datetime.now())
                trocas.append(random.randint(1,n-1))
                #print('trocas: ', trocas, 'm: ', m)
                while m>1 and trocas[m-1] in trocas[0:m-2]: #retirar vértices já escolhidos
                    #random.seed(datetime.now())
                    trocas[m-1]=random.randint(1,n-1)
                m=m+1
            x2=x.copy()
            print('Trocas: ', trocas)
            for i in range(m-2): #realização das trocas (perturbação)
                #print('i: ', i, trocas[i])
                aux=x2[trocas[i]]
                x2[trocas[i]]=x2[trocas[i+1]]
                x2[trocas[i+1]]=aux
            #xlinha= descida(grafo, x2) #Fazer a busca local na solução perturbada
            xlinha=descidarand(grafo,s_ini, itermax)
            dist2=avaliarsolucao(grafo,xlinha)
            print('Busca Local na Solução Perturbada: ', dist2,'- Ótimo até o momento: ', dist1)
            if dist2<dist1:
                x=xlinha.copy()
                dist1=dist2
                print('Solução atualizada: - Nova Distância após perturbação', dist1)
                k=1
            else: 
                k+=1
        
        print('Solução Final pelo ILS: ', dist1)
        
        return dist1
    
```
