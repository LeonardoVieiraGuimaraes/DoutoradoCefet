# Heapsort - Princípio e Relação com Seleção

- Possui o mesmo princípio de funcionamento da ordenação por seleção.

**Algoritmo:**
1. Selecione o menor item do vetor.
2. Troque-o com o item da primeira posição do vetor.
3. Repita estas operações com os $n-1$ itens restantes, depois com os $n-2$ itens, e assim sucessivamente.

- O custo para encontrar o menor (ou o maior) item entre $n$ itens é $n-1$ comparações.
- Isso pode ser reduzido utilizando uma **fila de prioridades**.

# Heapsort - Filas de Prioridades

- É uma estrutura de dados onde a chave de cada item reflete sua habilidade relativa de abandonar o conjunto de itens rapidamente.

**Aplicações:**
- Sistemas operacionais usam filas de prioridades, nas quais as chaves representam o tempo em que eventos devem ocorrer.
- Métodos numéricos iterativos são baseados na seleção repetida de um item com maior (ou menor) valor.
- Sistemas de gerência de memória usam a técnica de substituir a página menos utilizada na memória principal por uma nova página.

# Heapsort - Filas de Prioridades: Representação e Custos

**Representação através de uma lista linear ordenada:**
- Construção: $O(\log n)$
- Inserção: $O(n)$
- Retirada: $O(1)$
- Ajunta (merge): $O(n)$

**Representação através de uma lista linear não ordenada:**
- Construção: custo linear $O(n)$
- Inserção: $O(1)$
- Retirada: $O(n)$
- Ajunta (merge): $O(1)$ para apontadores, $O(n)$ para arranjos

# Heapsort - Filas de Prioridades: Representação com Heap

- A melhor representação é através de uma estrutura de dados chamada **heap**.

**Custos das operações:**
- Construção (Constrói): $O(n)$
- Inserção: $O(\log n)$
- Retirada: $O(\log n)$
- Substituição e alteração: $O(\log n)$

# Heapsort - Filas de Prioridades e Algoritmos de Ordenação

- As operações das filas de prioridades podem ser utilizadas para implementar algoritmos de ordenação.
- Basta utilizar repetidamente a operação **Insere** para construir a fila de prioridades.
- Em seguida, utilizar repetidamente a operação **Retira** para receber os itens na ordem reversa.

# Heap - Definição e Estrutura

- Um **heap** é uma sequência de itens com chaves $c[1], c[2], \ldots, c[n]$, tal que:
  - $c[i] \geq c[2i]$
  - $c[i] \geq c[2i+1]$
  - para todo $i = 1, 2, \ldots, \left\lfloor n/2 \right\rfloor$

- Esta definição pode ser facilmente visualizada em uma **árvore binária completa**:
  - Os nós são numerados de $1$ a $n$.
  - O primeiro nó é chamado **raiz**.
  - O nó $\left\lfloor k/2 \right\rfloor$ é o pai do nó $k$, para $1 < k \leq n$.
  - Os nós $2k$ e $2k+1$ são os filhos à esquerda e à direita do nó $k$, para $1 \leq k \leq \left\lfloor n/2 \right\rfloor$.

- As chaves na árvore satisfazem a **condição do heap**:
  - A chave em cada nó é maior do que as chaves em seus filhos.
  - A chave no nó raiz é a maior chave do conjunto.

- Uma árvore binária completa pode ser representada por um **array**:
  - Os filhos de um nó $i$ estão nas posições $2i$ e $2i+1$.
  - O pai de um nó $i$ está na posição $i \div 2$.
  - Esta representação é extremamente compacta e permite caminhar facilmente pelos nós da árvore.

# Heap - Construção, Algoritmo de Floyd e Manutenção

- Na representação do heap em um arranjo, a maior chave está sempre na posição 1 do vetor.
- Os algoritmos para implementar as operações sobre o heap operam ao longo de um dos caminhos da árvore.
- Um algoritmo elegante para construir o heap foi proposto por Floyd em 1964.
  - O algoritmo não necessita de nenhuma memória auxiliar.

**Construção do heap:**
- Dado um vetor $A[1], A[2], \ldots, A[n]$,
- Os itens $A[\lfloor n/2 \rfloor + 1], A[\lfloor n/2 \rfloor + 2], \ldots, A[n]$ formam um heap (são folhas).
- Neste intervalo não existem dois índices $i$ e $j$ tais que $j = 2i$ ou $j = 2i + 1$.

**Exemplo de chaves iniciais:**

| 1 | 2 | 3 | 4 | 5 | 6 | 7 |
|---|---|---|---|---|---|---|
| S | R | D | S | D | O | O |

- Os itens de $A[4]$ a $A[7]$ formam um heap.
- O heap é estendido para a esquerda ($\text{Esq} = 3$), englobando o item $A[3]$, pai dos itens $A[6]$ e $A[7]$.
- Se a condição de heap for violada, o heap é refeito trocando os itens conforme necessário (por exemplo, trocando D e S).
- O item R é incluído no heap ($\text{Esq} = 2$), o que pode ou não violar a condição de heap.
- O item O é incluído no heap ($\text{Esq} = 1$).
- Se a condição de heap for violada, o heap é refeito trocando os itens O e S, encerrando o processo.

**Programa que implementa a operação que informa o item com maior chave:**  
Basta acessar o item na posição 1 do vetor.

# Heapsort - Métodos para Refazer, Construir e Retirar do Heap (Java)

```java
// Refaz o heap a partir da posição esq até dir
public void refaz(int esq, int dir) {
    int j = esq * 2;
    Item x = this.v[esq];
    while (j <= dir) {
        if ((j < dir) && (this.v[j].compara(this.v[j + 1]) < 0)) j++;
        if (x.compara(this.v[j]) >= 0) break;
        this.v[esq] = this.v[j];
        esq = j;
        j = esq * 2;
    }
    this.v[esq] = x;
}

// Constrói o heap usando o método refaz
public void constroi() {
    int esq = n / 2;
    while (esq >= 1) {
        this.refaz(esq, this.n);
        esq--;
    }
}

// Retira o item com maior chave (raiz do heap)
public Item retiraMax() throws Exception {
    if (this.n < 1) throw new Exception("Erro: heap vazio");
    Item max = this.v[1];
    this.v[1] = this.v[this.n--];
    this.refaz(1, this.n);
    return max;
}
```

# Heap - Operação de Aumentar Chave

```java
public void aumentaChave(int i, Object chaveNova) throws Exception {
    Item x = this.v[i];
    if (chaveNova == null)
        throw new Exception("Erro: chaveNova valor null");
    x.alteraChave(chaveNova);
    while ((i > 1) && (x.compara(this.v[i / 2]) >= 0)) {
        this.v[i] = this.v[i / 2];
        i = i / 2;
    }
    this.v[i] = x;
}
```
- Exemplo da operação de aumentar o valor da chave do item na posição $i$.
- O tempo de execução do procedimento `aumentaChave` em um item do heap é $O(\log n)$.

# Heapsort - Algoritmo

1. Construir o heap.
2. Troque o item na posição 1 do vetor (raiz do heap) com o item da posição $n$.
3. Use o procedimento Refaz para reconstituir o heap para os itens restantes.
4. Repita os passos 2 e 3 com os $n-1$ itens restantes, depois com os $n-2$, até que reste apenas um item.

# Heapsort - Exemplo de Aplicação

Considere as chaves iniciais:

| 1 | 2 | 3 | 4 | 5 | 6 | 7 |
|---|---|---|---|---|---|---|
| S | R | O | N | E | D | A |

Após construir o heap:

| 1 | 2 | 3 | 4 | 5 | 6 | 7 |
|---|---|---|---|---|---|---|
| S | R | O | N | E | D | A |

Troca S e A, refaz heap:

| 1 | 2 | 3 | 4 | 5 | 6 | 7 |
|---|---|---|---|---|---|---|
| R | N | O | A | E | D | S |

Troca R e D, refaz heap:

| 1 | 2 | 3 | 4 | 5 | 6 | 7 |
|---|---|---|---|---|---|---|
| O | N | D | A | E | R | S |

Troca O e E, refaz heap:

| 1 | 2 | 3 | 4 | 5 | 6 | 7 |
|---|---|---|---|---|---|---|
| N | E | D | A | O | R | S |

Troca N e A, refaz heap:

| 1 | 2 | 3 | 4 | 5 | 6 | 7 |
|---|---|---|---|---|---|---|
| E | A | D | N | O | R | S |

Troca E e D, refaz heap:

| 1 | 2 | 3 | 4 | 5 | 6 | 7 |
|---|---|---|---|---|---|---|
| D | A | E | N | O | R | S |

Troca D e A, final ordenado:

| 1 | 2 | 3 | 4 | 5 | 6 | 7 |
|---|---|---|---|---|---|---|
| A | D | E | N | O | R | S |

O caminho seguido pelo procedimento Refaz para reconstituir a condição do heap está em negrito nas trocas.
Por exemplo, após a troca dos itens S e D, o item D volta para a posição 5, após passar pelas posições 1 e 2.

# Heapsort - Algoritmo em Pseudocódigo

```pascal
procedure Heapsort(var A: Vetor; var n: Indice)
var Esq, Dir: Indice;
    x: Item;
begin
    Constroi(A, n); { constroi o heap }
    Esq := 1; Dir := n;
    while Dir > 1 do { ordena o vetor }
    begin
        x := A[1];
        A[1] := A[Dir];
        A[Dir] := x;
        Dir := Dir - 1;
        Refaz(Esq, Dir, A);
    end;
end;
```
- O procedimento Refaz gasta cerca de $\log n$ operações, no pior caso.
- Logo, Heapsort gasta um tempo de execução proporcional a $n \log n$, no pior caso.

# Heapsort - Vantagens e Desvantagens

**Vantagens:**
- O comportamento do Heapsort é sempre $O(n \log n)$, qualquer que seja a entrada.

**Desvantagens:**
- O anel interno do algoritmo é bastante complexo se comparado com o do Quicksort.
- O Heapsort não é estável.

**Recomendado:**
- Para aplicações que não podem tolerar eventualmente um caso desfavorável.
- Não é recomendado para arquivos com poucos registros, por causa do tempo necessário para construir o heap.

# Comparação entre os Métodos de Ordenação

| Método      | Complexidade |
|-------------|-------------|

## Observações sobre os Métodos

1. Shellsort, Quicksort e Heapsort têm a mesma ordem de grandeza.
2. O Quicksort é o mais rápido para todos os tamanhos aleatórios experimentados.
3. A relação Heapsort/Quicksort é constante para todos os tamanhos.
4. A relação Shellsort/Quicksort aumenta se o número de elementos aumenta.
5. Para arquivos pequenos (500 elementos), o Shellsort é mais rápido que o Heapsort.
6. Se a entrada aumenta, o Heapsort é mais rápido que o Shellsort.
7. O Inserção é o mais rápido se os elementos estão ordenados.
8. O Inserção é o mais lento para qualquer tamanho se os elementos estão em ordem descendente.
9. Entre os algoritmos de custo $O(n^2)$, o Inserção é melhor para todos os tamanhos aleatórios experimentados.

## Influência da Ordem Inicial dos Registros

- Shellsort é bastante sensível à ordenação ascendente ou descendente da entrada.
- Em arquivos do mesmo tamanho, o Shellsort executa mais rápido para arquivos ordenados.
- Quicksort é sensível à ordenação ascendente ou descendente da entrada.
- Em arquivos do mesmo tamanho, o Quicksort executa mais rápido para arquivos ordenados.
- O Quicksort é o mais rápido para qualquer tamanho para arquivos na ordem ascendente.
- O Heapsort praticamente não é sensível à ordenação da entrada.

## Considerações Finais

- Para registros muito grandes é desejável que o método de ordenação realize apenas $n$ movimentos dos registros.
- Com o uso de uma **ordenação indireta** é possível conseguir isso.
- Suponha que o arquivo $A$ contenha os seguintes registros: $A[1], A[2], \ldots, A[n]$.
- Seja $P$ um arranjo $P[1], P[2], \ldots, P[n]$ de apontadores.
- Os registros somente são acessados para fins de comparações e toda movimentação é realizada sobre os apontadores.
- Ao final, $P[1]$ contém o índice do menor elemento de $A$, $P[2]$ o índice do segundo menor e assim sucessivamente.
- Essa estratégia pode ser utilizada para qualquer dos métodos de ordenação interna.
