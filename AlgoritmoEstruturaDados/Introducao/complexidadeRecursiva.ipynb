# Procedimento Recursivo e Equações de Recorrência

- Para cada procedimento recursivo é associada uma função de complexidade $f(n)$ desconhecida, onde $n$ mede o tamanho dos argumentos para o procedimento.
- Obtemos uma **equação de recorrência** para $f(n)$.
- Equação de recorrência: maneira de definir uma função por uma expressão envolvendo a mesma função.

---

## Exemplo de Procedimento Recursivo

```pseudo
Pesquisa(n);
(1) // inspecione elemento e termine
(2) // ...
(3) para cada um dos n elementos 'inspecione elemento';
(4) Pesquisa(n/3);
```

- Seja $T(n)$ uma função de complexidade que represente o número de inspeções nos $n$ elementos do conjunto.
- O custo de execução das linhas (1) e (2) é $O(1)$ e o da linha (3) é exatamente $n$.
- Usa-se uma equação de recorrência para determinar o número de chamadas recursivas.
- O termo $T(n)$ é especificado em função dos termos anteriores $T(k)$.

---

## Exemplo de Equação de Recorrência

- $T(n) = n + T(n/3)$, $T(1) = 1$ (para $n = 1$ fazemos uma inspeção)

Exemplo de cálculo:
- $T(3) = T(3/3) + 3 = T(1) + 3 = 1 + 3 = 4$
- $T(9) = T(9/3) + 9 = T(3) + 9 = 4 + 9 = 13$

Para calcular o valor da função seguindo a definição são necessários $k-1$ passos para computar o valor de $T(3^k)$.

---

## Resolução da Equação de Recorrência

Substitui-se os termos $T(k)$, $k < n$, até que todos os termos $T(k)$, $k > 1$, tenham sido substituídos por fórmulas contendo apenas $T(1)$.

- $T(n) = n + T(n/3)$
- $T(n/3) = n/3 + T(n/9)$
- $T(n/9) = n/9 + T(n/27)$
- $\ldots$

Somando todos os termos:

$T(n) = n + n/3 + n/9 + n/27 + \ldots + T(1)$

Esta é uma soma de uma série geométrica de razão $1/3$, multiplicada por $n$, e adicionada de $T(1)$.

---

## Solução da Recorrência

A soma da série geométrica é:

$T(n) = n \left(1 + \frac{1}{3} + \frac{1}{9} + \ldots \right) + 1$

Como a soma infinita $1 + \frac{1}{3} + \frac{1}{9} + \ldots = \frac{1}{1 - 1/3} = \frac{3}{2}$,

$T(n) \approx n \cdot \frac{3}{2} + 1 = \frac{3n}{2} + 1$

Portanto, $T(n) = O(n)$.

---

## Exemplo Geral de Recorrência

Seja:

$$
T(n) =
\begin{cases}
0, & \text{se } n < 1 \\
1, & \text{se } n = 1 \\
2T(n/2) + 2, & \text{se } n > 1
\end{cases}
$$

- Para resolver, substitua recursivamente até chegar ao caso base.
- O método geral é expandir a recorrência até que todos os termos dependam apenas de constantes.
