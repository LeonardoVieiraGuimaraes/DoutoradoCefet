# Comportamento Assintótico de Funções

- O parâmetro $n$ fornece uma medida da dificuldade para se resolver o problema.
- Para valores suficientemente pequenos de $n$, qualquer algoritmo custa pouco para ser executado, mesmo os ineficientes.
- A escolha do algoritmo não é um problema crítico para problemas de tamanho pequeno.
- Logo, a análise de algoritmos é realizada para valores grandes de $n$.
- Estuda-se o comportamento assintótico das funções de custo (comportamento para valores grandes de $n$).

---

## Dominação Assintótica

- A análise de um algoritmo geralmente conta com apenas algumas operações elementares.
- A medida de custo ou medida de complexidade relata o crescimento assintótico da operação considerada.
- **Definição:** Uma função $f(n)$ domina assintoticamente outra função $g(n)$ se existem duas constantes positivas $c$ e $m$ tais que, para $n > m$, temos $g(n) < c \cdot f(n)$.

---

## Notação O (Ordem Superior)

- Escrevemos $g(n) = O(f(n))$ para expressar que $f(n)$ domina assintoticamente $g(n)$. Lê-se: "$g(n)$ é da ordem no máximo $f(n)$".
- Exemplo: quando dizemos que o tempo de execução $T(n)$ de um programa é $O(n^2)$, significa que existem constantes $c$ e $m$ tais que, para $n > m$, $T(n) < c n^2$.

### Exemplos de Notação O

- $g(n) = (n + 1)^2$. Logo, $g(n)$ é $O(n^2)$, pois $(n + 1)^2 < 4n^2$ para $n \geq 1$.
- $g(n) = n$ e $f(n) = n^2$. $g(n)$ é $O(n^2)$, pois para $n > 0$, $n < n^2$.
- $g(n) = 3n^3 + 2n^2 + n$ é $O(n^3)$, pois $3n^3 + 2n^2 + n < 6n^3$ para $n \geq 1$.
- $g(n) = \log_5 n$ é $O(\log n)$, pois logaritmos de diferentes bases diferem apenas por uma constante multiplicativa.

---

## Operações com a Notação O

- Se $f(n)$ e $g(n)$ são funções de complexidade, então $O(f(n) + g(n)) = O(\max\{f(n), g(n)\})$.
- Exemplo: se temos trechos de código com tempos $O(n)$, $O(n^2)$ e $O(\log n)$, o tempo total é $O(n^2)$.

---

## Notação Ω (Ômega)

- Especifica um limite inferior para $g(n)$.
- **Definição:** Uma função $g(n)$ é $\Omega(f(n))$ se existirem duas constantes positivas $c$ e $m$ tais que $g(n) > c f(n)$ para todo $n > m$.
- Exemplo: $g(n) = 3n^3 + 2n^2$ é $\Omega(n^3)$, pois para $n$ suficientemente grande, $g(n) > c n^3$.

---

## Notação Θ (Theta)

- **Definição:** Uma função $g(n)$ é $\Theta(f(n))$ se existirem constantes $c_1, c_2$ e $m$ tais que $c_1 f(n) \leq g(n) \leq c_2 f(n)$ para todo $n > m$.
- $g(n) = \Theta(f(n))$ significa que $g(n)$ está limitada superior e inferiormente por múltiplos de $f(n)$ para $n$ suficientemente grande.
- Exemplo: $g(n) = n^2/3 - 2n$ é $\Theta(n^2)$.

---

## Resumo das Notações

- $O(f(n))$: limite superior assintótico (no máximo $f(n)$)
- $\Omega(f(n))$: limite inferior assintótico (pelo menos $f(n)$)
- $\Theta(f(n))$: limite assintótico firme (exatamente da ordem de $f(n)$)
