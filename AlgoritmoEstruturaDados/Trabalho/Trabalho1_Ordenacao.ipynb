## 1) Implementação dos métodos de ordenação
- Quicksort (pivô: primeiro elemento)
- Heapsort

```python
# Implementação do Quicksort (pivô: primeiro elemento)
def quicksort(arr, contador):  # quicksort
    r"""
    Ordena o vetor 'arr' usando o algoritmo Quicksort, escolhendo sempre o primeiro elemento como pivô.
    O contador é uma lista de um elemento para contar o número de comparações realizadas.
    """
    def _quicksort(a, baixo, alto):  # _quicksort
        if baixo < alto:
            p = particiona(a, baixo, alto)
            _quicksort(a, baixo, p-1)
            _quicksort(a, p+1, alto)
    def particiona(a, baixo, alto):  # particiona
        # O pivô é sempre o primeiro elemento
        pivo = a[baixo]
        i = baixo + 1
        for j in range(baixo+1, alto+1):
            contador[0] += 1  # Conta a comparação
            if a[j] < pivo:
                a[i], a[j] = a[j], a[i]
                i += 1
        a[baixo], a[i-1] = a[i-1], a[baixo]
        return i-1
    _quicksort(arr, 0, len(arr)-1)
```

```python
# Implementação do Heapsort
def heapsort(arr, contador):  # heapsort
    r"""
    Ordena o vetor 'arr' usando o algoritmo Heapsort.
    O contador é uma lista de um elemento para contar o número de comparações realizadas.
    """
    def heapifica(a, n, i):  # heapifica
        maior = i
        esq = 2 * i + 1
        dir = 2 * i + 2
        if esq < n:
            contador[0] += 1  # Conta a comparação
            if a[esq] > a[maior]:
                maior = esq
        if dir < n:
            contador[0] += 1  # Conta a comparação
            if a[dir] > a[maior]:
                maior = dir
        if maior != i:
            a[i], a[maior] = a[maior], a[i]
            heapifica(a, n, maior)
    n = len(arr)
    # Constroi o heap máximo
    for i in range(n//2 - 1, -1, -1):
        heapifica(arr, n, i)
    # Extrai elementos do heap um a um
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapifica(arr, i, 0)
```

# 2) Para cada método implementado faça os seguintes testes:


a) gerar vetores de n elementos ORDENADOS EM ORDEM CRESCENTE, com n variando de 1000 até 10.000, com intervalo de 1000; ordenar esses vetores com ambos os métodos e computar o número de comparações realizadas;

```python
import sys
sys.setrecursionlimit(20000)

# Teste (a): Vetores ordenados em ordem crescente
ns = list(range(1000, 10001, 1000))
comparacoes_crescente = {'quicksort': [], 'heapsort': []}

for n in ns:
    vetor_crescente = list(range(n))
    # Quicksort
    arr_q = vetor_crescente.copy()
    contador_q = [0]
    quicksort(arr_q, contador_q)
    comparacoes_crescente['quicksort'].append(contador_q[0])
    # Heapsort
    arr_h = vetor_crescente.copy()
    contador_h = [0]
    heapsort(arr_h, contador_h)
    comparacoes_crescente['heapsort'].append(contador_h[0])

print("Comparações Quicksort (crescente):", comparacoes_crescente['quicksort'])
print("Comparações Heapsort (crescente):", comparacoes_crescente['heapsort'])
```

b) gerar vetores de n elementos ORDENADOS EM ORDEM DECRESCENTE, com n variando de 1.000 até 10.000, com intervalo de 1.000; Ordenar esses vetores com ambos os métodos e computar o número de comparações realizadas;

```python
# Teste (b): Vetores ordenados em ordem decrescente
ns = list(range(1000, 10001, 1000))
comparacoes_decrescente = {'quicksort': [], 'heapsort': []}

for n in ns:
    vetor_decrescente = list(range(n, 0, -1))
    # Quicksort
    arr_q = vetor_decrescente.copy()
    contador_q = [0]
    quicksort(arr_q, contador_q)
    comparacoes_decrescente['quicksort'].append(contador_q[0])
    # Heapsort
    arr_h = vetor_decrescente.copy()
    contador_h = [0]
    heapsort(arr_h, contador_h)
    comparacoes_decrescente['heapsort'].append(contador_h[0])

print("Comparações Quicksort (decrescente):", comparacoes_decrescente['quicksort'])
print("Comparações Heapsort (decrescente):", comparacoes_decrescente['heapsort'])
```

c) gerar vetores de n elementos ALEATÓRIOS  com n variando de 1.000 até 10.000, comintervalo de 1.000;  Ordenar esses vetores com ambos os métodos e computar o número de comparações realizadas;



```python
# Teste (c): Vetores aleatórios
import numpy as np

ns = list(range(1000, 10001, 1000))
comparacoes_aleatorio = {'quicksort': [], 'heapsort': []}

for n in ns:
    vetor_aleatorio = np.random.permutation(n).tolist()
    # Quicksort
    arr_q = vetor_aleatorio.copy()
    contador_q = [0]
    quicksort(arr_q, contador_q)
    comparacoes_aleatorio['quicksort'].append(contador_q[0])
    # Heapsort
    arr_h = vetor_aleatorio.copy()
    contador_h = [0]
    heapsort(arr_h, contador_h)
    comparacoes_aleatorio['heapsort'].append(contador_h[0])

print("Comparações Quicksort (aleatório):", comparacoes_aleatorio['quicksort'])
print("Comparações Heapsort (aleatório):", comparacoes_aleatorio['heapsort'])
```
