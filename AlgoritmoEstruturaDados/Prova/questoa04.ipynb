**Vetor inicial:**  
4, 1, 7, 2, 16, 9, 11, 14, 18, 71

## O que é um Heap?

Um **heap** é uma estrutura de dados baseada em árvore binária completa, onde cada nível da árvore está completamente preenchido, exceto possivelmente o último, que é preenchido da esquerda para a direita.

Existem dois tipos principais de heap:
- **Heap máximo (max-heap):** O valor de cada nó é maior ou igual ao valor de seus filhos. O maior elemento está sempre na raiz.
- **Heap mínimo (min-heap):** O valor de cada nó é menor ou igual ao valor de seus filhos. O menor elemento está sempre na raiz.

**Propriedades principais:**
- Estrutura de árvore binária completa.
- Propriedade de heap: relação de ordem entre pais e filhos (máximo ou mínimo).
- Permite acesso eficiente ao maior (ou menor) elemento.

**Aplicações:**
- Implementação de filas de prioridade.
- Algoritmo de ordenação Heapsort.
- Algoritmos de grafos (Dijkstra, Prim).

No contexto desta questão, usamos o heap máximo para garantir que o maior elemento esteja sempre na raiz da árvore.

### 2. Aplicando heapify de baixo para cima

#### Heapify no índice 3 (valor 2)

Novo vetor: 4, 1, 7, 18, 71, 9, 11, 14, 2, 16

#### Heapify no índice 2 (valor 7)
- Filhos: 5 (valor 9) e 6 (valor 11)
- 11 > 9 > 7, troca 7 com 11

Novo vetor: 4, 1, 11, 18, 71, 9, 7, 14, 2, 16

#### Heapify no índice 1 (valor 1)
- Filhos: 3 (valor 18) e 4 (valor 71)

Novo vetor: 4, 71, 11, 18, 1, 9, 7, 14, 2, 16

- 16 > 1, troca 1 com 16

#### Heapify no índice 0 (valor 4)
- Filhos: 1 (valor 71) e 2 (valor 11)

## Pseudocódigo para construção do heap máximo

```python
def print_heap(arr, passo):
    print(f"Passo {passo}: {arr}")

def heapify(arr, n, i, passo):
    maior = i
    esq = 2*i + 1
    dir = 2*i + 2
    if esq < n and arr[esq] > arr[maior]:
        maior = esq
    if dir < n and arr[dir] > arr[maior]:
        maior = dir
    if maior != i:
        arr[i], arr[maior] = arr[maior], arr[i]
        print_heap(arr, passo)
        heapify(arr, n, maior, passo + 1)

def construir_heap_maximo(arr):
    n = len(arr)
    passo = 1
    print_heap(arr, "Inicial")
    for i in range(n//2 - 1, -1, -1):
        heapify(arr, n, i, passo)
        passo += 1

# Vetor inicial
vetor = [4, 1, 7, 2, 16, 9, 11, 14, 18, 71]
construir_heap_maximo(vetor)
print("Heap máximo final:", vetor)
```

````python
def print_heap(arr, passo, acao, explicacao):
    print(f"Passo {passo}: {arr}  <- {acao}")
    if explicacao:
        print("   ", explicacao)

def heapify(arr, n, i, passo):
    """
    Garante a propriedade de heap máximo para o nó i e seus filhos.
    Se algum filho for maior que o pai, faz a troca e aplica heapify recursivamente.
    """
    maior = i
    esq = 2*i + 1  # filho esquerdo
    dir = 2*i + 2  # filho direito
    acao = ""
    explicacao = ""
    if esq < n and arr[esq] > arr[maior]:
        maior = esq
    if dir < n and arr[dir] > arr[maior]:
        maior = dir
    if maior != i:
        acao = f"Troca {arr[i]} com {arr[maior]}"
        explicacao = (
            f"O valor {arr[maior]} é maior que o pai {arr[i]}. "
            f"Para manter a propriedade do heap máximo, trocamos {arr[i]} com {arr[maior]}. "
            f"O maior valor sobe para a posição do pai."
        )
        arr[i], arr[maior] = arr[maior], arr[i]
        print_heap(arr, passo, acao, explicacao)
        heapify(arr, n, maior, passo + 1)
    else:
        if passo != "Inicial":
            acao = "Nenhuma troca necessária"
            explicacao = (
                "O valor do pai já é maior que seus filhos, "
                "então a propriedade do heap máximo está mantida."
            )
            print_heap(arr, passo, acao, explicacao)

def construir_heap_maximo(arr):
    """
    Constrói o heap máximo a partir de um vetor.
    Começa do último nó pai e aplica heapify para cada um até a raiz.
    """
    n = len(arr)
    passo = 1
    print_heap(arr, "Inicial", "Vetor original", "")
    print_heap(arr, passo, "Construindo heap máximo", "Iniciando a construção do heap máximo")
    passo += 1
    print(n//2 - 1)
    for i in range(n//2 - 1, -1, -1):
        acao = f"Aplicando heapify no nó {arr[i]} (índice {i})"
        print_heap(arr, passo, acao, "Aplicando heapify para garantir a propriedade do heap máximo")
        heapify(arr, n, i, passo)
        passo += 1

# Vetor inicial
vetor = [4, 1, 7, 2, 16, 9, 11, 14, 18, 71]
construir_heap_maximo(vetor)
print("Heap máximo final:", vetor)
````
