````python
# Exemplo: vetor com chaves idênticas
def selecao(v):
    n = len(v)
    trocas = 0
    for i in range(n-1):
        min_idx = i
        for j in range(i+1, n):
            if v[j] < v[min_idx]:
                min_idx = j
        if min_idx != i:
            v[i], v[min_idx] = v[min_idx], v[i]
            trocas += 1
    return trocas

def insercao(v):
    n = len(v)
    movimentos = 0
    for i in range(1, n):
        chave = v[i]
        j = i - 1
        while j >= 0 and v[j] > chave:
            v[j+1] = v[j]
            j -= 1
            movimentos += 1
        v[j+1] = chave
    return movimentos

vetor = [5] * 10

trocas_selecao = selecao(vetor.copy())
mov_insercao = insercao(vetor.copy())

print("Trocas (Seleção):", trocas_selecao)
print("Movimentações (Inserção):", mov_insercao)
````

### Como fica a soma $\sum_{j=i+1}^{n} 1$

A expressão $\sum_{j=i+1}^{n} 1$ representa a quantidade de inteiros de $j$ indo de $i+1$ até $n$.

Para cada valor de $i$, o somatório conta quantos valores $j$ existem nesse intervalo:

- O menor valor de $j$ é $i+1$
- O maior valor de $j$ é $n$
- Portanto, o número de termos é $n - (i+1) + 1 = n - i$

Assim:
$$
\sum_{j=i+1}^{n} 1 = n - i
$$

Se quisermos somar isso para todos os valores de $i$ de $0$ até $n-1$:
$$
\sum_{i=0}^{n-1} \sum_{j=i+1}^{n} 1 = \sum_{i=0}^{n-1} (n - i)
$$

Expandindo:
$$
= n + (n-1) + (n-2) + \ldots + 1
$$

Essa soma é igual à soma dos $n$ primeiros inteiros positivos:
$$
\sum_{k=1}^{n} k = \frac{n(n+1)}{2}
$$

Portanto:
$$
\sum_{i=0}^{n-1} \sum_{j=i+1}^{n} 1 = \frac{n(n+1)}{2}
$$

**Saída esperada:**

- Trocas (Seleção): 0 (ou próximo de 0, pois todos são iguais, mas sempre percorre o vetor)
- Movimentações (Inserção): 0 (nenhuma movimentação, pois nunca precisa mover elementos)

**Conclusão:**  
O método de inserção é mais rápido em vetores com chaves idênticas, pois evita movimentações desnecessárias.

### Análise de melhor, médio e pior caso para seleção e inserção

#### Seleção (Selection Sort)

- **Melhor caso:** Todos os elementos são iguais.
    - O algoritmo ainda percorre todo o vetor para buscar o menor elemento em cada iteração, mesmo que não haja trocas.
    - Número de comparações: sempre $\frac{n(n-1)}{2}$.
    - Número de trocas: $0$ (se considerar que não troca elementos iguais) ou $n$ (se sempre trocar).
    - **Complexidade:** $O(n^2)$ no melhor, médio e pior caso.

- **Médio caso:** Elementos em ordem aleatória.
    - Mesmo número de comparações, trocas dependem da ordem.
    - **Complexidade:** $O(n^2)$.

- **Pior caso:** Elementos em ordem decrescente.
    - Mesmo número de comparações, mais trocas.
    - **Complexidade:** $O(n^2)$.

#### Inserção (Insertion Sort)

- **Melhor caso:** Vetor já está ordenado ou todos os elementos são iguais.
    - O laço interno nunca executa, pois a condição `v[j] > chave` nunca é satisfeita.
    - Número de comparações: $n-1$.
    - Número de movimentações: $0$.
    - **Complexidade:** $O(n)$.

- **Médio caso:** Elementos em ordem aleatória.
    - O laço interno executa em média metade do tamanho do vetor para cada elemento.
    - **Complexidade:** $O(n^2)$.

- **Pior caso:** Vetor em ordem decrescente.
    - O laço interno executa o máximo possível para cada elemento.
    - **Complexidade:** $O(n^2)$.

#### Resumo

- **Seleção:** Sempre $O(n^2)$, independentemente dos dados.
- **Inserção:** $O(n)$ no melhor caso (vetor já ordenado ou com chaves idênticas), $O(n^2)$ no médio e pior caso.

**Conclusão:**  
Para vetores com chaves idênticas, o método de inserção é muito mais eficiente, pois atinge o melhor caso ($O(n)$), enquanto a seleção permanece em $O(n^2)$.

No contexto de algoritmos de ordenação, o **médio caso** geralmente se refere a um vetor com elementos em ordem aleatória, ou seja, sem nenhuma ordenação prévia ou padrão específico. Também pode ser chamado de "caso típico" ou "caso geral", pois representa a situação mais comum encontrada na prática, quando os dados não estão nem totalmente ordenados nem totalmente invertidos.

Portanto, ao mencionar o médio caso, pode-se dizer:

- **Médio caso:** Vetor com elementos em ordem aleatória, sem padrão específico, representando o comportamento típico do algoritmo para entradas comuns.

Isso é válido tanto para o Selection Sort quanto para o Insertion Sort.

### Análise matemática do melhor, médio e pior caso

#### Selection Sort (Seleção)

- **Melhor caso:** (todos os elementos iguais ou já ordenados)
    - Número de comparações: sempre $\frac{n(n-1)}{2}$, pois o algoritmo compara todos os pares possíveis independentemente da ordem.
    - Número de trocas: $0$ (se não trocar elementos iguais) ou $n$ (se sempre trocar).
    - **Complexidade:** $O(n^2)$.

- **Médio caso:** (elementos em ordem aleatória)
    - Número de comparações: $\frac{n(n-1)}{2}$.
    - Número de trocas: em média, proporcional a $n$.
    - **Complexidade:** $O(n^2)$.

- **Pior caso:** (elementos em ordem inversa)
    - Número de comparações: $\frac{n(n-1)}{2}$ (igual ao melhor caso).
    - Número de trocas: $n-1$ (uma troca por posição).
    - **Complexidade:** $O(n^2)$.

**Resumo matemático Selection Sort:**
$$
C_{comparações}(n) = \frac{n(n-1)}{2}
$$
$$
C_{trocas}(n) \leq n-1
$$

#### Insertion Sort (Inserção)

- **Melhor caso:** (vetor já ordenado ou todos os elementos iguais)
    - Para cada $i$ de $1$ até $n-1$, a condição do laço interno `v[j] > chave` nunca é satisfeita.
    - Número de comparações: $n-1$ (uma por elemento, exceto o primeiro).
    - Número de movimentações: $0$.
    - **Complexidade:** $O(n)$.

- **Médio caso:** (elementos em ordem aleatória)
    - Para cada elemento, o laço interno executa em média metade do número de elementos anteriores.
    - Número de comparações: aproximadamente $\frac{n^2}{4}$.
    - Número de movimentações: aproximadamente $\frac{n^2}{4}$.
    - **Complexidade:** $O(n^2)$.

- **Pior caso:** (vetor em ordem inversa)
    - Para cada $i$, o laço interno percorre todos os elementos anteriores.
    - Número de comparações: $\sum_{i=1}^{n-1} i = \frac{n(n-1)}{2}$
    - Número de movimentações: igual ao número de comparações.
    - **Complexidade:** $O(n^2)$.

**Resumo matemático Insertion Sort:**
- Melhor caso:
$$
C_{comparações}^{melhor}(n) = n-1
$$
$$
C_{movimentações}^{melhor}(n) = 0
$$

- Médio caso:
$$
C_{comparações}^{médio}(n) \approx \frac{n^2}{4}
$$
$$
C_{movimentações}^{médio}(n) \approx \frac{n^2}{4}
$$

- Pior caso:
$$
C_{comparações}^{pior}(n) = \frac{n(n-1)}{2}
$$
$$
C_{movimentações}^{pior}(n) = \frac{n(n-1)}{2}
$$

**Conclusão:**  
- Selection Sort sempre faz $\frac{n(n-1)}{2}$ comparações, independentemente do caso.
- Insertion Sort faz $n-1$ comparações no melhor caso, aproximadamente $\frac{n^2}{4}$ no médio caso e $\frac{n(n-1)}{2}$ no pior caso.
- Para chaves idênticas, Insertion Sort é linear ($O(n)$), enquanto Selection Sort permanece quadrático ($O(n^2)$).

### Explicação detalhada dos algoritmos de Seleção e Inserção

#### Selection Sort (Seleção)

- Percorre o vetor procurando o menor elemento e o coloca na posição correta.
- Para cada posição $i$, busca o menor elemento do subvetor $i+1$ até $n-1$ e troca com $i$.
- O número de comparações é sempre $\frac{n(n-1)}{2}$, independentemente da ordem dos elementos.
- O número de trocas depende da ordem, mas para chaves idênticas, normalmente não há trocas reais.

**Pseudocódigo em Python:**
```python
def selection_sort(v):
    n = len(v)
    for i in range(n-1):
        min_idx = i
        for j in range(i+1, n):
            if v[j] < v[min_idx]:
                min_idx = j
        if min_idx != i:
            v[i], v[min_idx] = v[min_idx], v[i]
```

#### Insertion Sort (Inserção)

- Percorre o vetor da esquerda para a direita, inserindo cada elemento na posição correta do subvetor à esquerda.
- Para cada elemento, compara com os anteriores e desloca os maiores para a direita até encontrar a posição correta.
- No melhor caso (vetor já ordenado ou chaves idênticas), não há movimentações, apenas comparações.
- No pior caso (vetor invertido), cada elemento é comparado e movido até o início.

**Pseudocódigo em Python:**
```python
def insertion_sort(v):
    n = len(v)
    for i in range(1, n):
        chave = v[i]
        j = i - 1
        while j >= 0 and v[j] > chave:
            v[j+1] = v[j]
            j -= 1
        v[j+1] = chave
```

#### Resumo

- **Selection Sort:** Sempre $O(n^2)$ comparações, poucas trocas para chaves idênticas.
- **Insertion Sort:** $O(n)$ no melhor caso (já ordenado ou chaves idênticas), $O(n^2)$ no pior caso (invertido).

**Conclusão:**  
Para vetores com chaves idênticas, o Insertion Sort é mais eficiente, pois faz apenas comparações e nenhuma movimentação.

### Como fica a soma $\sum_{j=i+1}^{n} 1$

A expressão $\sum_{j=i+1}^{n} 1$ representa a quantidade de inteiros de $j$ indo de $i+1$ até $n$.

Para cada valor de $i$, o somatório conta quantos valores $j$ existem nesse intervalo:

- O menor valor de $j$ é $i+1$
- O maior valor de $j$ é $n$
- Portanto, o número de termos é $n - (i+1) + 1 = n - i$

Assim:
$$
\sum_{j=i+1}^{n} 1 = n - i
$$

Se quisermos somar isso para todos os valores de $i$ de $0$ até $n-1$:
$$
\sum_{i=0}^{n-2} \sum_{j=i+1}^{n-1} 1 = \sum_{i=0}^{n-1} (n - i)
$$

Expandindo:
$$
= n + (n-1) + (n-2) + \ldots + 1
$$

Essa soma é igual à soma dos $n$ primeiros inteiros positivos:
$$
\sum_{k=1}^{n} k = \frac{n(n+1)}{2}
$$

Portanto:
$$
\sum_{i=0}^{n-1} \sum_{j=i+1}^{n} 1 = \frac{n(n+1)}{2}
$$

### Soma dupla $\sum_{i=0}^{n-2} \sum_{j=i+1}^{n-1} 1$

A expressão $\sum_{i=0}^{n-2} \sum_{j=i+1}^{n-1} 1$ conta, para cada $i$ de $0$ até $n-2$, quantos valores $j$ existem de $i+1$ até $n-1$.

Para cada $i$, o número de valores de $j$ é:
$$
(n-1) - (i+1) + 1 = n - i - 1
$$

Portanto:
$$
\sum_{i=0}^{n-2} \sum_{j=i+1}^{n-1} 1 = \sum_{i=0}^{n-2} (n - i - 1)
$$

Expandindo:
$$
= (n-1) + (n-2) + (n-3) + \ldots + 1
$$

Essa soma é igual à soma dos $n-1$ primeiros inteiros positivos:
$$
\sum_{k=1}^{n-1} k = \frac{(n-1)n}{2}
$$

Portanto:
$$
\sum_{i=0}^{n-2} \sum_{j=i+1}^{n-1} 1 = \frac{n(n-1)}{2}
$$

### Como fica a soma do laço `for i in range(1, n)`

O laço externo do Insertion Sort:

```python
for i in range(1, n):
    # corpo do laço
```

- Esse laço executa exatamente $n-1$ vezes, pois $i$ vai de $1$ até $n-1$.
- Cada iteração corresponde à tentativa de inserir o elemento da posição $i$ na parte já ordenada do vetor.

Portanto, a soma desse laço é simplesmente:
$$
\sum_{i=1}^{n-1} 1 = n-1
$$

Ou seja, o laço externo é linear em $n$.

### Como fica a soma do laço `for i in range(n-1)`

O laço externo do Selection Sort:

```python
for i in range(n-1):
    min_idx = i
    # corpo do laço
```

- Esse laço executa exatamente $n-1$ vezes, pois $i$ vai de $0$ até $n-2$.
- Cada iteração corresponde à seleção do menor elemento do subvetor restante.

Portanto, a soma desse laço é:
$$
\sum_{i=0}^{n-2} 1 = n-1
$$

Ou seja, o laço externo do Selection Sort também é linear em $n$.

### Como fica a soma do laço `while j >= 0 and v[j] > chave` no Insertion Sort

O laço:

```python
while j >= 0 and v[j] > chave:
    v[j+1] = v[j]
    j -= 1
```

executa, para cada $i$ de $1$ até $n-1$, um número de vezes igual ao número de elementos anteriores a $i$ que são maiores que `chave`.

- **Melhor caso:** O vetor já está ordenado ou todos os elementos são iguais. O laço nunca executa, então a soma é $0$.
- **Pior caso:** O vetor está em ordem decrescente. Para cada $i$, o laço executa $i$ vezes.

A soma total no pior caso é:
$$
\sum_{i=1}^{n-1} i = 1 + 2 + 3 + \ldots + (n-1) = \frac{n(n-1)}{2}
$$

- **Médio caso:** Para elementos em ordem aleatória, o laço executa em média metade do número de elementos anteriores, então a soma é aproximadamente:
$$
\sum_{i=1}^{n-1} \frac{i}{2} = \frac{1}{2} \cdot \frac{n(n-1)}{2} = \frac{n(n-1)}{4}
$$

Portanto, a soma do laço `while` no Insertion Sort é:

- Melhor caso: $0$
- Pior caso: $\frac{n(n-1)}{2}$
- Médio caso: $\approx \frac{n(n-1)}{4}$

### Soma total do laço externo com o interno (`for` com `while`)

No caso médio do Insertion Sort, o laço externo (`for i in range(1, n)`) executa $n-1$ vezes, e o laço interno (`while j >= 0 and v[j] > chave`) executa, em média, metade do número de elementos anteriores a $i$.

Portanto, a soma total de execuções do laço interno ao longo de todas as iterações do laço externo é:

$$
\sum_{i=1}^{n-1} \frac{i}{2} = \frac{1}{2} \cdot \frac{n(n-1)}{2} = \frac{n(n-1)}{4}
$$

Ou seja, no caso médio, o número total de execuções do laço interno (e, portanto, o número de movimentações/comparações relevantes) é aproximadamente $\frac{n(n-1)}{4}$.

### Soma do laço interno no caso médio do Insertion Sort

A soma do número de execuções do laço interno (while) para todos os $i$ é:

$$
\sum_{i=1}^{n-1} \frac{i}{2}
$$

### Desenvolvimento da soma

Primeiro, observe que:

$$
\sum_{i=1}^{n-1} i = \frac{n(n-1)}{2}
$$

Portanto,

$$
\sum_{i=1}^{n-1} \frac{i}{2} = \frac{1}{2} \sum_{i=1}^{n-1} i = \frac{1}{2} \cdot \frac{n(n-1)}{2} = \frac{n(n-1)}{4}
$$

---

Assim, no caso médio do Insertion Sort, o laço interno executa aproximadamente $\frac{n(n-1)}{4}$ vezes.
