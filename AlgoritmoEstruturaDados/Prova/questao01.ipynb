## Questão 01

Mostre a função de complexidade do seguinte método iterativo:

```c
void iterativo (int n) {
    int i, j, a;           // (1)
    a = 0;                 // (2)
    for (i = 0; i < n; i++)    // (3)
        for (j = 0; j < i; j++)   // (4)
            a += i + j;          // (5)
}
```

**Resolução detalhada:**

- O laço externo (`for (i = 0; i < n; i++)`) executa $n$ vezes, com $i$ variando de $0$ até $n-1$.
- Para cada valor de $i$, o laço interno (`for (j = 0; i < i; j`) executa $i$ vezes (pois $j$ vai de $0$ até $i-1$).
- O comando interno `a += i + j;` é executado uma vez para cada iteração do laço interno.

O número total de execuções do comando interno é:

$$
\sum_{i=0}^{n-1} i = 0 + 1 + 2 + \ldots + (n-1) = \frac{n(n-1)}{2}
$$

- Para $n=5$, por exemplo:
  - $i=0$: 0 execuções
  - $i=1$: 1 execução
  - $i=2$: 2 execuções
  - $i=3$: 3 execuções
  - $i=4$: 4 execuções
  - Total: $0+1+2+3+4=10$

- Portanto, a função de complexidade é $O(n^2)$, pois o termo dominante é quadrático.

**Explicação:**  
- O tempo de execução cresce proporcionalmente ao quadrado do tamanho da entrada $n$.
- Para grandes valores de $n$, o termo $\frac{n(n-1)}{2}$ é aproximadamente $\frac{n^2}{2}$, e constantes e termos de menor ordem são desprezados na notação assintótica.

O termo dominante é o das linhas 4 e 5, que é quadrático em $n$.

Vamos detalhar cada operação matemática para o custo de cada linha do algoritmo:

**Linha 1:**  
Declaração de variáveis  
Custo: $1$ (executada uma vez)

**Linha 2:**  
Atribuição  
Custo: $1$ (executada uma vez)

**Linha 3:**  
Controle do laço externo  
O laço `for (i = 0; i &lt; n; i++)` executa $n$ vezes (para $i = 0$ até $i = n-1$).  
Cada iteração corresponde a uma execução do controle do laço, então:  
$$
\sum_{i=0}^{n-1} 1 = n
$$

**Linha 4:**  
Controle do laço interno  
Para cada valor de $i$, o laço interno `for (j = 0; j &lt; i; j++)` executa $i$ vezes (pois $j$ vai de $0$ até $i-1$).  
O total de execuções é:  
$$
\sum_{i=0}^{n-1} \left( \sum_{j=0}^{i-1} 1 \right) = \sum_{i=0}^{n-1} i
$$
Sabemos que:
$$
\sum_{i=0}^{n-1} i = 0 + 1 + 2 + \ldots + (n-1) = \frac{n(n-1)}{2}
$$

**Linha 5:**  
Operação de soma e atribuição  
Executada uma vez para cada iteração do laço interno, ou seja, o mesmo número de vezes que a linha 4:  
$$
\sum_{i=0}^{n-1} i = \frac{n(n-1)}{2}
$$

**Resumo dos cálculos:**

- Linha 1: $1$
- Linha 2: $1$
- Linha 3: $n$
- Linha 4: $\frac{n(n-1)}{2}$
- Linha 5: $\frac{n(n-1)}{2}$

**Total geral:**  
$$
T(n) = 1 + 1 + n + \frac{n(n-1)}{2} + \frac{n(n-1)}{2} = 2 + n + n(n-1)
$$
$$
T(n) = 2 + n + n^2 - n = 2 + n^2
$$

Para grandes valores de $n$, o termo dominante é $n^2$, então:
$$
T(n) = O(n^2)
$$

### Somatório do custo de cada linha

Vamos expressar o custo de cada linha do algoritmo usando somatórios:

- **Linha 1:**  
  Executada uma única vez (declaração de variáveis):  
  $$
  S_1 = 1
  $$

- **Linha 2:**  
  Executada uma única vez (atribuição):  
  $$
  S_2 = 1
  $$

- **Linha 3:**  
  Laço externo:  
  $$
  S_3 = \sum_{i=0}^{n-1} 1 = n
  $$

- **Linha 4:**  
  Laço interno para cada $i$:  
  $$
  S_4 = \sum_{i=0}^{n-1} \sum_{j=0}^{i-1} 1 = \sum_{i=0}^{n-1} i = \frac{n(n-1)}{2}
  $$

- **Linha 5:**  
  Operação dentro do laço interno, executada uma vez para cada iteração do laço interno:  
  $$
  S_5 = \sum_{i=0}^{n-1} \sum_{j=0}^{i-1} 1 = \sum_{i=0}^{n-1} i = \frac{n(n-1)}{2}
  $$

**Resumo dos somatórios:**

- $S_1 = 1$
- $S_2 = 1$
- $S_3 = n$
- $S_4 = \frac{n(n-1)}{2}$
- $S_5 = \frac{n(n-1)}{2}$

O termo dominante é quadrático, vindo das linhas 4 e 5.

**Observação importante sobre os laços:**

No algoritmo apresentado, o comando relevante para análise de complexidade é a linha 5 (`a += i + j;`), pois é ela que é executada repetidamente dentro dos laços aninhados.

- **Linha 3:** Laço externo, controla o valor de $i$ de $0$ até $n-1$.
- **Linha 4:** Laço interno, para cada $i$, controla o valor de $j$ de $0$ até $i-1$.
- **Linha 5:** Executada uma vez para cada iteração do laço interno.

Assim, o número total de execuções da linha 5 já representa o produto dos laços 3 e 4, ou seja:
$$
\sum_{i=0}^{n-1} \sum_{j=0}^{i-1} 1 = \sum_{i=0}^{n-1} i = \frac{n(n-1)}{2}
$$

Portanto, ao analisar a complexidade, basta considerar a quantidade de vezes que a linha 5 é executada, pois ela engloba as repetições dos laços 3 e 4.

### Análise detalhada do custo de cada linha do algoritmo

Vamos analisar cada linha do algoritmo, considerando o número de execuções de cada uma:

- **Linha 1:** `int i, j, a;`  
  - Declaração de variáveis.
  - Executada **1 vez**.
  - $C_1 = 1$

- **Linha 2:** `a = 0;`  
  - Atribuição.
  - Executada **1 vez**.
  - $C_2 = 1$

- **Linha 3:** `for (i = 0; i < n; i++)`  
  - Laço externo.
  - O controle do laço é executado $n$ vezes, mas o que realmente importa para o custo total é o número de vezes que o laço interno (linha 4) é ativado.

- **Linha 4:** `for (j = 0; j < i; j++)`  
  - Laço interno.
  - Para cada valor de $i$, o laço interno executa $i$ vezes (pois $j$ vai de $0$ até $i-1$).
  - O número total de execuções do laço interno é:
    $$
    C_4 = \sum_{i=0}^{n-1} i = \frac{n(n-1)}{2}
    $$

- **Linha 5:** `a += i + j;`  
  - Operação de soma e atribuição.
  - Executada **uma vez para cada iteração do laço interno**.
  - O número total de execuções é exatamente o mesmo da linha 4:
    $$
    C_5 = \sum_{i=0}^{n-1} i = \frac{n(n-1)}{2}
    $$

#### Resumindo:

- Linha 1: $1$
- Linha 2: $1$
- Linhas 3, 4 e 5: O custo relevante é dado pelo duplo somatório das linhas 4 e 5, pois a linha 5 é executada para cada iteração dos dois laços aninhados:
  $$
  \sum_{i=0}^{n-1} \sum_{j=0}^{i-1} 1 = \sum_{i=0}^{n-1} i = \frac{n(n-1)}{2}
  $$

Portanto, a função total de custo é:
$$
T(n) = 1 + 1 + \frac{n(n-1)}{2}
$$

Para grandes valores de $n$, o termo dominante é quadrático:
$$
T(n) = O(n^2)
$$

**Conclusão:**  
O termo dominante da complexidade é o duplo somatório das linhas 4 e 5, pois a linha 5 é executada para cada combinação dos laços aninhados. As linhas 1 e 2 são constantes e não afetam o crescimento assintótico.
