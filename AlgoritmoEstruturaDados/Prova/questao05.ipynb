# Questão 05 – 13,0 pontos

Para o método de ordenação Quicksort, o seu pior caso ocorre quando cada partição divide o arquivo em uma parte de tamanho $n-1$ e outra de tamanho $1$:
$$
T(n) = T(n-1) + n-1
$$

Em contrapartida, o melhor caso ocorre quando cada partição divide o arquivo em duas partes iguais:
$$
T(n) = 2T\left(\frac{n}{2}\right) + n-1
$$

Mostre, a partir da resolução das equações de recorrência, qual a ordem de complexidade para cada um dos casos.

---

## Pior caso: $T(n) = T(n-1) + n-1$

### Passo a passo da expansão

1. **Entendimento da recorrência:**  
   - A cada chamada, o Quicksort resolve um subproblema de tamanho $n-1$ (pois só um elemento é separado pelo pivô) e faz $n-1$ comparações para particionar.
   - Isso acontece, por exemplo, quando o vetor já está ordenado ou invertido e o pivô escolhido é sempre o menor ou maior elemento.

2. **Expandindo a recorrência:**
   - $T(n) = T(n-1) + (n-1)$
   - $T(n-1) = T(n-2) + (n-2)$
   - $T(n-2) = T(n-3) + (n-3)$
   - $\vdots$
   - $T(3) = T(2) + 2$
   - $T(2) = T(1) + 1$

3. **Substituindo recursivamente:**
   - Substitua $T(n-1)$ na equação de $T(n)$:
     $$
     T(n) = [T(n-2) + (n-2)] + (n-1) = T(n-2) + (n-2) + (n-1)
     $$
   - Continue substituindo:
     $$
     T(n) = T(n-3) + (n-3) + (n-2) + (n-1)
     $$
   - Repita até chegar em $T(1)$:
     $$
     T(n) = T(1) + 1 + 2 + 3 + \ldots + (n-1)
     $$

4. **Somando os termos:**
   - A soma $1 + 2 + 3 + \ldots + (n-1)$ é uma progressão aritmética:
     $$
     \sum_{k=1}^{n-1} k = \frac{(n-1)n}{2}
     $$
   - Portanto:
     $$
     T(n) = T(1) + \frac{n(n-1)}{2}
     $$

5. **Conclusão do pior caso:**
   - O termo dominante é quadrático, então:
     $$
     T(n) = O(n^2)
     $$
   - Ou seja, no pior caso, o Quicksort tem complexidade quadrática.

---

## Melhor caso: $T(n) = 2T\left(\frac{n}{2}\right) + n-1$

### Passo a passo da expansão

1. **Entendimento da recorrência:**  
   - A cada chamada, o Quicksort divide o vetor em duas partes iguais (ou quase iguais), resolvendo dois subproblemas de tamanho $n/2$ e faz $n-1$ comparações para particionar.
   - Isso acontece quando o pivô sempre divide o vetor ao meio.

2. **Expandindo a recorrência:**
   - $T(n) = 2T(n/2) + n-1$
   - $T(n/2) = 2T(n/4) + n/2 - 1$
   - $T(n/4) = 2T(n/8) + n/4 - 1$
   - $\vdots$

3. **Substituindo recursivamente:**
   - Substitua $T(n/2)$ na equação de $T(n)$:
     $$
     T(n) = 2[2T(n/4) + n/2 - 1] + n-1 = 4T(n/4) + 2(n/2 - 1) + n-1
     $$
     $$
     = 4T(n/4) + n - 2 + n - 1 = 4T(n/4) + 2n - 3
     $$
   - Substitua $T(n/4)$:
     $$
     T(n) = 4[2T(n/8) + n/4 - 1] + 2n - 3 = 8T(n/8) + 4(n/4 - 1) + 2n - 3
     $$
     $$
     = 8T(n/8) + n - 4 + 2n - 3 = 8T(n/8) + 3n - 7
     $$

4. **Generalizando após $k$ passos:**
   $$
   T(n) = 2^k T\left(\frac{n}{2^k}\right) + k n - (2^k - 1)
   $$
   - Quando $\frac{n}{2^k} = 1 \implies k = \log_2 n$.

5. **Substituindo $k = \log_2 n:**
   $$
   T(n) = n T(1) + n \log_2 n - (n - 1)
   $$
   - Como $T(1)$ é constante, o termo dominante é $n \log n$.

6. **Conclusão do melhor caso:**
   $$
   T(n) = O(n \log n)
   $$
   - Ou seja, no melhor caso, o Quicksort tem complexidade linearítmica.

---

## Resumo final

- **Pior caso:** $O(n^2)$ (quando as partições são extremamente desbalanceadas)
- **Melhor caso:** $O(n \log n)$ (quando as partições são sempre equilibradas)

O Quicksort é eficiente na média, mas pode ser quadrático no pior caso se as partições forem muito desbalanceadas.

## Explicação detalhada das recorrências do Quicksort

### Pior caso: $T(n) = T(n-1) + n-1$

No pior caso do Quicksort, a cada partição, apenas um elemento é separado (por exemplo, o menor ou o maior), e o restante do vetor continua desbalanceado. Isso ocorre, por exemplo, quando o vetor já está ordenado ou invertido e o pivô escolhido é sempre o menor ou maior elemento.

**Expansão da recorrência:**

- $T(n) = T(n-1) + (n-1)$
- $T(n-1) = T(n-2) + (n-2)$
- $T(n-2) = T(n-3) + (n-3)$
- $\vdots$
- $T(2) = T(1) + 1$

Substituindo recursivamente, temos:

$$
T(n) = T(n-1) + (n-1) \\
= [T(n-2) + (n-2)] + (n-1) \\
= [[T(n-3) + (n-3)] + (n-2)] + (n-1) \\
= T(n-3) + (n-3) + (n-2) + (n-1) \\
\vdots \\
= T(1) + 1 + 2 + 3 + \ldots + (n-1)
$$

A soma $1 + 2 + 3 + \ldots + (n-1)$ é uma progressão aritmética:

$$
\sum_{k=1}^{n-1} k = \frac{(n-1)n}{2}
$$

Portanto:

$$
T(n) = T(1) + \frac{n(n-1)}{2}
$$

O termo dominante é quadrático, então:

**No pior caso, a complexidade do Quicksort é $O(n^2)$.**

### Por que $T(2) = T(1) + 1$?

No contexto da recorrência do pior caso do Quicksort:

$$
T(n) = T(n-1) + n-1
$$

Quando $n=2$, temos:

$$
T(2) = T(1) + 1
$$

**Explicação:**  
- Para um vetor de tamanho 2, o Quicksort faz uma partição, compara os dois elementos (1 comparação), e depois resolve o subvetor restante de tamanho 1.
- O custo de ordenar um vetor de tamanho 1 é considerado constante ($T(1)$), pois não há mais comparações ou trocas necessárias.
- O "+1" representa a única comparação feita entre os dois elementos.

Portanto, $T(2) = T(1) + 1$ reflete o custo de resolver o subproblema de tamanho 1 mais a comparação feita para o vetor de tamanho 2.

### Por que escrevemos $T(3) = T(2) + 2$ e $T(2) = T(1) + 1$?

Essas expressões representam a expansão da recorrência do Quicksort para os menores tamanhos de vetor:

- $T(3) = T(2) + 2$ significa: para ordenar um vetor de 3 elementos no pior caso, o Quicksort resolve um subvetor de 2 elementos (custo $T(2)$) e faz 2 comparações (uma para cada partição).
- $T(2) = T(1) + 1$ significa: para ordenar um vetor de 2 elementos, resolve um subvetor de 1 elemento (custo $T(1)$, que é constante) e faz 1 comparação.

Essas equações mostram como a recorrência é construída a partir dos menores casos, permitindo somar todos os custos até chegar ao tamanho $n$. Elas são necessárias para expandir a soma e encontrar a solução fechada da recorrência.
