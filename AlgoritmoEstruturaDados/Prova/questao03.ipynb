# Questão 03

Qual algoritmo de ordenação você usaria para cada um dos seguintes casos? Justifique brevemente.

---

**a) O tempo de execução não deve apresentar grandes variações para nenhum caso.**

O algoritmo recomendado é o **Heapsort** ou o **Mergesort**.

- Ambos possuem complexidade $O(n \log n)$ no melhor, médio e pior caso, ou seja, o tempo de execução é estável e previsível.
- O **Heapsort** não depende da ordem inicial dos dados e não utiliza memória extra significativa.
- O **Mergesort** também é estável em relação ao tempo, mas utiliza memória adicional.

**Justificativa:**  
Algoritmos como QuickSort podem ter pior caso $O(n^2)$, enquanto HeapSort e MergeSort garantem $O(n \log n)$ em todos os casos.

**Pseudocódigo (Heapsort):**
```python
def heapsort(arr):
    n = len(arr)
    # Construir heap máximo
    for i in range(n//2 - 1, -1, -1):
        heapify(arr, n, i)
    # Extrair elementos do heap
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

def heapify(arr, n, i):
    maior = i
    esq = 2*i + 1
    dir = 2*i + 2
    if esq < n and arr[esq] > arr[maior]:
        maior = esq
    if dir < n and arr[dir] > arr[maior]:
        maior = dir
    if maior != i:
        arr[i], arr[maior] = arr[maior], arr[i]
        heapify(arr, n, maior)
```

---

**b) A lista a ser ordenada já está bem próxima da ordem final**

O algoritmo recomendado é o **Insertion Sort**.

- O Insertion Sort tem complexidade $O(n)$ no melhor caso (lista já ordenada ou quase ordenada).
- Para listas quase ordenadas, ele realiza poucas comparações e movimentações.
- É simples, eficiente para listas pequenas ou quase ordenadas e não requer memória extra.

**Justificativa:**  
Quando a lista está quase ordenada, o Insertion Sort é muito eficiente, superando algoritmos mais sofisticados nesse cenário.

**Pseudocódigo (Insertion Sort):**
```python
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        chave = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > chave:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = chave
```
