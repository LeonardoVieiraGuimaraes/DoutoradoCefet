# Leitura de Todos os PDFs em uma Pasta

Este notebook lista todos os arquivos PDF em uma pasta específica e exibe o conteúdo de cada um.

## 1. Importar Bibliotecas Necessárias

````python
%pip install PyPDF2
import os

import PyPDF2
````

## 2. Listar Arquivos PDF na Pasta

Altere o caminho para a pasta desejada.

````python
pasta = r'D:\OneDrive\Pessoais\Doutorado\Cefet\Algoritmos e Estrutura de Dados\Introdução'

arquivos_pdf = [f for f in os.listdir(pasta) if f.lower().endswith('.pdf')]
print("Arquivos PDF encontrados:")
for arquivo in arquivos_pdf:
    print(arquivo)
````

## 3. Ler e Exibir Conteúdo dos Arquivos PDF

Abaixo, o conteúdo dos PDFs será exibido (primeiros 1000 caracteres de cada um).

````python
for arquivo in arquivos_pdf:
    caminho_pdf = os.path.join(pasta, arquivo)
    with open(caminho_pdf, 'rb') as f:
        leitor = PyPDF2.PdfReader(f)
        conteudo = ""
        for pagina in leitor.pages:
            texto = pagina.extract_text()
            if texto:
                conteudo += texto + "\n"
    print(f"\n{'='*40}\nConteúdo de: {arquivo}\n{'='*40}")
    print(conteudo[:1000])
````

## 4. Ler PDFs como Imagem usando OCR

Se os PDFs contêm apenas imagens (scans), é necessário usar OCR para extrair o texto. Vamos utilizar as bibliotecas `pdf2image`, `pytesseract` e `Pillow`. Instale-as se necessário.

````python
# Instale as bibliotecas se necessário:
# %pip install pdf2image pytesseract pillow

import pytesseract
from pdf2image import convert_from_path
from PIL import Image

# Se necessário, configure o caminho do executável do Tesseract:
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
````

### 5. Extrair Texto de PDFs de Imagem

Este código converte cada página do PDF em imagem e aplica OCR para extrair o texto.

````python
for arquivo in arquivos_pdf:
    caminho_pdf = os.path.join(pasta, arquivo)
    print(f"\n{'='*40}\nConteúdo de: {arquivo}\n{'='*40}")
    paginas = convert_from_path(caminho_pdf)
    conteudo = ""
    for i, pagina in enumerate(paginas):
        texto = pytesseract.image_to_string(pagina, lang='por')  # Use 'eng' para inglês, 'por' para português
        conteudo += f"\n--- Página {i+1} ---\n"
        conteudo += texto
    print(conteudo)
````

### Sobre o erro `convert_from_path` e Poppler

A função `convert_from_path` da biblioteca `pdf2image` depende do utilitário externo **Poppler** para converter páginas de PDF em imagens.

**Erro comum:**  
`PDFInfoNotInstalledError: Unable to get page count. Is poppler installed and in PATH?`

**Como resolver:**
- É necessário instalar o Poppler no seu sistema operacional e garantir que o executável esteja disponível no PATH do sistema.
- Para Windows, baixe o Poppler em: [https://github.com/oschwartz10612/poppler-windows/releases/](https://github.com/oschwartz10612/poppler-windows/releases/)
- Extraia o arquivo ZIP e adicione o caminho da pasta `bin` do Poppler nas variáveis de ambiente do Windows (PATH).
- No Linux, use: `sudo apt install poppler-utils`
- No MacOS, use: `brew install poppler`

Após instalar, reinicie o Jupyter e tente novamente.

Se necessário, informe o caminho do Poppler diretamente na função:
```python

```

````python
# Exemplo de OCR em Python para ler PDFs em imagem

import pytesseract
from pdf2image import convert_from_path
from PIL import Image
import os

# Caminho para o Poppler, se necessário (ajuste conforme seu sistema)
poppler_path = r'C:\caminho\para\poppler\bin'  # Altere para o caminho correto

pasta = r'D:\OneDrive\Pessoais\Doutorado\Cefet\Algoritmos e Estrutura de Dados\Introdução'
arquivos_pdf = [f for f in os.listdir(pasta) if f.lower().endswith('.pdf')]

for arquivo in arquivos_pdf:
    caminho_pdf = os.path.join(pasta, arquivo)
    print(f"\n{'='*40}\nConteúdo de: {arquivo}\n{'='*40}")
    # Se necessário, adicione poppler_path=poppler_path
    paginas = convert_from_path(caminho_pdf, poppler_path=poppler_path)
    conteudo = ""
    for i, pagina in enumerate(paginas):
        texto = pytesseract.image_to_string(pagina, lang='por')  # 'por' para português
        conteudo += f"\n--- Página {i+1} ---\n"
        conteudo += texto
    print(conteudo)
````
